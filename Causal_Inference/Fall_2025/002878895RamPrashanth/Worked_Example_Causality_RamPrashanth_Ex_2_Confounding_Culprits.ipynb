# Causal Question-1: Does Poverty cause Crime?

##### Approach to answer this question:
1. Define the causal question.
2. Identify the confounders. (Confounders: Variables that affect both treatment and outcome)
3. Visualize it by DAG
4. Analysis and Visualizations
5. Naive Analysis
6. Causality Methods

In our case, for this question: The treatment is Poverty and Outcome is Crime. We need to find variables which affect both. After that, we will perform some analysis, visualizations to get what the data is saying. Followed by Naive Analysis with crude assumptions and dig deeper with causal methods.

#### Step-2: Identifying the confounders.

Let's use logical reasoning:
**What causes Poverty?**
- Low education ‚Üí harder to get good jobs ‚Üí poverty
- Unemployment ‚Üí no income ‚Üí poverty
- Young population ‚Üí less work experience ‚Üí poverty

**What causes crime?**
- Poverty itself (what we're testing!)
- Low education ‚Üí fewer opportunities ‚Üí crime
- Unemployment ‚Üí desperation ‚Üí crime
- Young males ‚Üí higher crime rates
- Family breakdown ‚Üí less supervision ‚Üí crime

**What affects BOTH? ‚Üê These are our confounders!**
- Education (PctBSorMore, PctNotHSGrad)
- Unemployment (PctUnemployed)
- Age distribution (agePct12t21, agePct16t24) - young people
- Family structure (PctKids2Par) - two-parent households

```python
### Step-3 Drawing DAG

import matplotlib.pyplot as plt
import networkx as nx

# Create a simple DAG visualization
fig, ax = plt.subplots(figsize=(12, 8))

# Define positions for nodes
pos = {
    'Poverty': (2, 2),
    'Crime': (6, 2),
    'Education': (1, 4),
    'Unemployment': (3, 4),
    'Age (Young)': (5, 4),
    'Family Structure': (7, 4)
}

# Create directed graph
G = nx.DiGraph()

# Add edges (arrows showing causal relationships)
edges = [
    # Confounders ‚Üí Treatment
    ('Education', 'Poverty'),
    ('Unemployment', 'Poverty'),
    ('Age (Young)', 'Poverty'),
    ('Family Structure', 'Poverty'),

    # Confounders ‚Üí Outcome
    ('Education', 'Crime'),
    ('Unemployment', 'Crime'),
    ('Age (Young)', 'Crime'),
    ('Family Structure', 'Crime'),

    # Treatment ‚Üí Outcome (what we want to estimate!)
    ('Poverty', 'Crime')
]

G.add_edges_from(edges)

# Draw the graph
nx.draw(G, pos, with_labels=True, node_color='lightblue',
        node_size=3000, font_size=10, font_weight='bold',
        arrows=True, arrowsize=20, edge_color='gray',
        arrowstyle='->', connectionstyle='arc3,rad=0.1', ax=ax)

# Highlight the causal effect we're interested in
nx.draw_networkx_edges(G, pos, [('Poverty', 'Crime')],
                       edge_color='red', width=3, arrows=True,
                       arrowsize=20, ax=ax)

plt.title('Causal Diagram: Does Poverty Cause Crime?\n(Red arrow = causal effect we want to estimate)',
          fontsize=14, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```

**Inference:**
Because of the confounding variables there's **compounding bias** which means you can't exactly point out the crime has caused due to poverty, it can also be affected by differences in our confounding variables. They also create "backdoor paths".


A **backdoor path** is a non-causal pathway that connects the treatment (Poverty) to the outcome (Crime) by going "through the back door" - i.e., through a common cause that affects both variables.

Ex: Poverty ‚Üê Education ‚Üí Crime
- Poverty ‚Üê Unemployment ‚Üí Crime

So, we need to control the backdoor paths so that we can assess the true causal effect (the red arrow)!!



```python
### Selecting the vars for our Analysis

# Variables for this specific analysis
analysis_vars = [
    # Treatment
    'PctPopUnderPov',

    # Outcome
    'ViolentCrimesPerPop',

    # Confounders
    'PctBSorMore',      # Education (% with bachelor's+)
    'PctNotHSGrad',     # Education (% no high school)
    'PctUnemployed',    # Unemployment
    'agePct12t21',      # Young people
    'agePct16t24',      # Young adults
    'PctKids2Par',      # Family structure
]

# Create focused dataset
df_analysis = crime_df_clean[analysis_vars].copy()

print(f"Analysis dataset shape: {df_analysis.shape}")
print(f"\nVariables we're using:")
for var in analysis_vars:
    print(f"  - {var}")
```

### **Naive Analysis**

**Naive analysis** is the simple, unadjusted comparison between groups without controlling for any confounders.

In our example, it would be: Compare crime rates in poor areas vs. non-poor areas.

***Naive Estimate = True Causal Effect + Confounding Bias***

Naive Analyis helps in:
- Establish Basic Association
- Benchmark for comparison
- Transparency and Honesty
- Starting point for Investigation.


```python
from scipy import stats

# Simple correlation (WRONG for causality!)
corr, p_value = stats.pearsonr(df_analysis['PctPopUnderPov'],
                                 df_analysis['ViolentCrimesPerPop'])

print("=== NAIVE ANALYSIS (Ignoring Confounders) ===")
print(f"Correlation: {corr:.3f}")
print(f"P-value: {p_value:.4f}")
print(f"\nInterpretation: {'Significant' if p_value < 0.05 else 'Not significant'} correlation")
print("‚ö†Ô∏è BUT WAIT - this could be due to confounding!")
```

### Causal Analysis

Let's adjust the backdoor paths by:
**1. Simple Stratification:** Divide your data into homogeneous groups (strata) where confounders are held constant, then analyze the poverty-crime relationship within each stratum.

Instead of comparing ALL poor vs. ALL non-poor areas,
compare poor vs. non-poor areas that are SIMILAR in other ways.

Example strata:
- Stratum 1: High unemployment, low education, young population
- Stratum 2: Low unemployment, high education, older population
- Stratum 3: Medium unemployment, medium education, mixed age

**Pros:** Easy to understand and interpret

**Cons:** Requires large sample sizes; gets unwieldy with many confounders

**2. Multiple Regression:** Include all confounders as control variables in a single regression equation.
***Crime = Œ≤‚ÇÄ + Œ≤‚ÇÅ(Poverty) + Œ≤‚ÇÇ(Unemployment) + Œ≤‚ÇÉ(Education)
        + Œ≤‚ÇÑ(Family Structure) + Œ≤‚ÇÖ(Age) + Œµ***

If two areas have the SAME unemployment, education, family structure, and age distribution, but different poverty rates, how much do their crime rates differ?

- **Pros:** Efficient; handles multiple confounders simultaneously; gives you one adjusted coefficient
- **Cons:** Assumes linear relationships; can't easily show you "what's happening" visually

**3. Visualization of Adjusted Effect**: Graphical methods to show how the poverty-crime relationship changes after adjusting for confounders.

**Pros:** Makes the confounding adjustment transparent and intuitive; great for communication

**Cons:** Can only visualize 2-3 dimensions at once; more descriptive than inferential

```python
# Method 1: Stratification by Education Level
# Split into low/medium/high education groups

df_analysis['education_group'] = pd.cut(df_analysis['PctBSorMore'],
                                          bins=3,
                                          labels=['Low Ed', 'Medium Ed', 'High Ed'])

# Look at poverty-crime relationship WITHIN each education group
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, group in enumerate(['Low Ed', 'Medium Ed', 'High Ed']):
    subset = df_analysis[df_analysis['education_group'] == group]

    axes[idx].scatter(subset['PctPopUnderPov'],
                     subset['ViolentCrimesPerPop'],
                     alpha=0.5, s=50)

    # Trend line within group
    if len(subset) > 5:
        z = np.polyfit(subset['PctPopUnderPov'].dropna(),
                      subset['ViolentCrimesPerPop'].dropna(), 1)
        p = np.poly1d(z)
        x_line = np.linspace(subset['PctPopUnderPov'].min(),
                            subset['PctPopUnderPov'].max(), 100)
        axes[idx].plot(x_line, p(x_line), 'r--', linewidth=2)

        # Calculate correlation within group
        corr = subset[['PctPopUnderPov', 'ViolentCrimesPerPop']].corr().iloc[0,1]
        axes[idx].text(0.05, 0.95, f'Correlation: {corr:.3f}',
                      transform=axes[idx].transAxes,
                      bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5),
                      verticalalignment='top')

    axes[idx].set_xlabel('Poverty Rate')
    axes[idx].set_ylabel('Crime Rate')
    axes[idx].set_title(f'{group} Communities', fontweight='bold')
    axes[idx].grid(True, alpha=0.3)

plt.suptitle('Poverty vs Crime AFTER Controlling for Education\n(Stratification Method)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n=== STRATIFICATION RESULTS ===")
for group in ['Low Ed', 'Medium Ed', 'High Ed']:
    subset = df_analysis[df_analysis['education_group'] == group]
    corr = subset[['PctPopUnderPov', 'ViolentCrimesPerPop']].corr().iloc[0,1]
    print(f"{group}: Correlation = {corr:.3f} (n={len(subset)} communities)")
```

**Inference:**

Education is a major confounder:
- Naive (unstratified): 0.523
- Low Ed communities: 0.487
- Medium Ed communities: 0.510
- High Ed communities: 0.242 ‚ö†Ô∏è Much weaker!

The causal effect of poverty on crime might depend on education level! This is called "interaction effect" or "effect modification".

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Prepare data - remove any remaining NaNs
df_regression = df_analysis[['ViolentCrimesPerPop', 'PctPopUnderPov',
                              'PctBSorMore', 'PctNotHSGrad', 'PctUnemployed',
                              'agePct12t21', 'PctKids2Par']].dropna()

print(f"\nRegression dataset: {len(df_regression)} communities")

# Model 1: Naive (no confounders)
X_naive = df_regression[['PctPopUnderPov']]
y = df_regression['ViolentCrimesPerPop']

model_naive = sm.OLS(y, sm.add_constant(X_naive)).fit()

# Model 2: Adjusted (with confounders)
X_adjusted = df_regression[['PctPopUnderPov', 'PctBSorMore', 'PctNotHSGrad',
                            'PctUnemployed', 'agePct12t21', 'PctKids2Par']]
model_adjusted = sm.OLS(y, sm.add_constant(X_adjusted)).fit()

print("\n" + "="*60)
print("CAUSAL EFFECT ESTIMATION: Poverty ‚Üí Crime")
print("="*60)

print("\nüìä MODEL 1: NAIVE (No Confounders)")
print("-" * 60)
print(f"Poverty coefficient: {model_naive.params['PctPopUnderPov']:.4f}")
print(f"P-value: {model_naive.pvalues['PctPopUnderPov']:.4f}")
print(f"R-squared: {model_naive.rsquared:.4f}")
print("\n‚ö†Ô∏è  This is BIASED due to confounding!")

print("\nüìä MODEL 2: ADJUSTED (Controlling for Confounders)")
print("-" * 60)
print(f"Poverty coefficient: {model_adjusted.params['PctPopUnderPov']:.4f}")
print(f"P-value: {model_adjusted.pvalues['PctPopUnderPov']:.4f}")
print(f"R-squared: {model_adjusted.rsquared:.4f}")

print("\nüéØ CAUSAL INTERPRETATION:")
print("-" * 60)
print(f"After adjusting for confounders:")
print(f"A 10% increase in poverty rate ‚Üí")
print(f"{model_adjusted.params['PctPopUnderPov']*0.10:.4f} increase in crime rate")

print("\nüìâ COMPARISON:")
print("-" * 60)
print(f"Naive effect:    {model_naive.params['PctPopUnderPov']:.4f}")
print(f"Adjusted effect: {model_adjusted.params['PctPopUnderPov']:.4f}")
print(f"Difference:      {model_naive.params['PctPopUnderPov'] - model_adjusted.params['PctPopUnderPov']:.4f}")
percent_change = ((model_adjusted.params['PctPopUnderPov'] - model_naive.params['PctPopUnderPov'])
                 / model_naive.params['PctPopUnderPov'] * 100)
print(f"% Change:        {percent_change:.1f}%")

print("\nüîç ALL COEFFICIENTS (Adjusted Model):")
print("-" * 60)
for var in X_adjusted.columns:
    coef = model_adjusted.params[var]
    pval = model_adjusted.pvalues[var]
    sig = "***" if pval < 0.001 else "**" if pval < 0.01 else "*" if pval < 0.05 else ""
    print(f"{var:20s}: {coef:8.4f}  (p={pval:.4f}) {sig}")
```

**Inference:**

This is a case of **"confounding reversal (Simpson's Paradox)"**.

Let's dig into this:

- **Naive Effect**: Poor areas have more crime. Therefore poverty causes crime"
- **Adjusted Effect**: Poor areas have more crime, but NOT because of poverty itself. It's because poor areas also have:
- More unemployment.
- Less education
- Fewer two-parent families ---> This is the big one

Compared to these confounders, poverty is actually associated with SLIGHTLY LOWER crime!

The adjusted poverty coefficient is -0.1528, it could mean:
- **Hyp1: Suppressor Effect**: Once you control for the broken families, unemployment, poverty might proxy for something protective (lower property values, community support networks)
- **Hyp2: Stastical Artificat**: Coefficient is small compared to the noise. It's not the large positive effect we saw in naive analysis.
- **Hyp3: Unmeasured Variables**: There might be omitted variables that we haven't controlled for.

Based on the R-square value: Confounders explain MORE of the variation in crime than poverty does.

```python
# Create a visualization showing the adjustment
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Naive relationship
axes[0].scatter(df_regression['PctPopUnderPov'],
               df_regression['ViolentCrimesPerPop'],
               alpha=0.5, s=50, color='blue')
axes[0].plot(df_regression['PctPopUnderPov'],
            model_naive.fittedvalues,
            'r-', linewidth=3, label=f'Slope = {model_naive.params["PctPopUnderPov"]:.3f}')
axes[0].set_xlabel('Poverty Rate', fontsize=12)
axes[0].set_ylabel('Crime Rate', fontsize=12)
axes[0].set_title('NAIVE: Without Adjusting for Confounders', fontweight='bold', fontsize=13)
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: Show residuals after removing confounder effects
# Calculate residuals after removing confounders (but not poverty)
X_confounders_only = df_regression[['PctBSorMore', 'PctNotHSGrad',
                                    'PctUnemployed', 'agePct12t21', 'PctKids2Par']]
model_confounders = sm.OLS(y, sm.add_constant(X_confounders_only)).fit()

# Residuals = crime after removing confounder effects
crime_residuals = y - model_confounders.fittedvalues + y.mean()

# Also residualize poverty
poverty_model = sm.OLS(df_regression['PctPopUnderPov'],
                       sm.add_constant(X_confounders_only)).fit()
poverty_residuals = (df_regression['PctPopUnderPov'] -
                    poverty_model.fittedvalues +
                    df_regression['PctPopUnderPov'].mean())

axes[1].scatter(poverty_residuals, crime_residuals, alpha=0.5, s=50, color='green')
z = np.polyfit(poverty_residuals, crime_residuals, 1)
p = np.poly1d(z)
axes[1].plot(poverty_residuals, p(poverty_residuals),
            'r-', linewidth=3,
            label=f'Slope = {model_adjusted.params["PctPopUnderPov"]:.3f}')
axes[1].set_xlabel('Poverty Rate (Adjusted)', fontsize=12)
axes[1].set_ylabel('Crime Rate (Adjusted)', fontsize=12)
axes[1].set_title('ADJUSTED: After Controlling for Confounders', fontweight='bold', fontsize=13)
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

***Inference:***

- ‚úì Confounding was massive (slope changed from 0.53 to -0.15)
- ‚úì The true causal effect is near zero (the adjusted plot is essentially flat)
- ‚úì Hypothesis 2 wins: The -0.15 is likely just statistical noise around zero

```python
# Define our analysis variables
treatment = 'PctKids2Par'  # % kids in 2-parent families (HIGH = stable families)
outcome = 'ViolentCrimesPerPop'

# Confounders for family breakdown ‚Üí crime
confounders_b = [
    # Economic factors
    'medIncome',         # Income level
    'PctPopUnderPov',    # Poverty rate
    'PctUnemployed',     # Unemployment

    # Education
    'PctBSorMore',       # High education
    'PctNotHSGrad',      # Low education

    # Demographics
    'agePct12t21',       # Young people
    'agePct16t24',       # Young adults
    'racePctWhite',      # Race/culture
    'racepctblack',      # Race/culture

    # Urban factors
    'PopDens',           # Population density
    'pctUrban',          # Urban percentage
]

# Create focused dataset
analysis_vars_b = [treatment, outcome] + confounders_b
df_b = crime_df_clean[analysis_vars_b].dropna()

print(f"\nüìä Analysis B Dataset: {df_b.shape[0]} communities, {df_b.shape[1]} variables")
print(f"\nVariables:")
print(f"  Treatment: {treatment}")
print(f"  Outcome:   {outcome}")
print(f"  Confounders: {len(confounders_b)}")
```

```python
import statsmodels.api as sm

# Prepare data
X_treatment = df_b[[treatment]]
X_all = df_b[[treatment] + confounders_b]
y = df_b[outcome]

# MODEL 1: Naive (no confounders)
model_naive = sm.OLS(y, sm.add_constant(X_treatment)).fit()

# MODEL 2: Adjusted (with confounders)
model_adjusted = sm.OLS(y, sm.add_constant(X_all)).fit()

print("\n" + "="*70)
print("üéØ CAUSAL EFFECT ESTIMATION: Family Breakdown ‚Üí Crime")
print("="*70)

print("\nüìä MODEL 1: NAIVE (No Confounders)")
print("-" * 70)
print(f"Family Structure coefficient: {model_naive.params[treatment]:.4f}")
print(f"P-value: {model_naive.pvalues[treatment]:.6f}")
print(f"R-squared: {model_naive.rsquared:.4f}")
print("\n‚ö†Ô∏è  This is BIASED - doesn't control for confounders!")

print("\nüìä MODEL 2: ADJUSTED (Controlling for Confounders)")
print("-" * 70)
print(f"Family Structure coefficient: {model_adjusted.params[treatment]:.4f}")
print(f"P-value: {model_adjusted.pvalues[treatment]:.6f}")
print(f"R-squared: {model_adjusted.rsquared:.4f}")

print("\nüéØ CAUSAL INTERPRETATION:")
print("-" * 70)
print(f"After adjusting for confounders:")
print(f"A 10 percentage point increase in 2-parent families ‚Üí")
print(f"{model_adjusted.params[treatment]*0.10:.4f} change in crime rate")
print(f"\nOR: Moving from 50% to 60% two-parent families ‚Üí")
print(f"{model_adjusted.params[treatment]*0.10:.4f} change in crime")

print("\nüìâ COMPARISON:")
print("-" * 70)
print(f"Naive effect:    {model_naive.params[treatment]:.4f}")
print(f"Adjusted effect: {model_adjusted.params[treatment]:.4f}")
print(f"Difference:      {model_naive.params[treatment] - model_adjusted.params[treatment]:.4f}")

percent_change = abs((model_adjusted.params[treatment] - model_naive.params[treatment])
                    / model_naive.params[treatment] * 100)
print(f"% Reduction:     {percent_change:.1f}%")

if abs(model_adjusted.params[treatment]) < abs(model_naive.params[treatment]):
    print("\nüí° The effect is SMALLER after adjustment (confounding inflated it)")
else:
    print("\nüí° The effect is LARGER after adjustment (confounding suppressed it)")

print("\nüîç CONFOUNDING CHECK - Which confounders matter most?")
print("-" * 70)

# Show significant confounders
for var in [treatment] + confounders_b:
    if var in model_adjusted.params:
        coef = model_adjusted.params[var]
        pval = model_adjusted.pvalues[var]
        sig = "***" if pval < 0.001 else "**" if pval < 0.01 else "*" if pval < 0.05 else ""

        if pval < 0.05 or var == treatment:
            print(f"{var:20s}: {coef:8.4f}  (p={pval:.4f}) {sig}")
```

# Causal Question-3: Does Unemployment cause crime?

To answer this, we'll go with the Hybrid approach, where we'll perform Joint Analysis, and then digging deeper into the main cause.

STEP 1: JOINT Analysis
  - ‚Üí Compare unemployment to other factors (education, race, urban, age)
  - ‚Üí Quick ranking: How strong is unemployment compared to others?
  - ‚Üí Joint effect estimate (with simple common confounders)

STEP 2: Identify Winner
 -  ‚Üí Is unemployment the strongest remaining factor?
  - ‚Üí Or does something else (education? race? urban?) dominate?

STEP 3: SEQUENTIAL Deep Dive on Winner
  - ‚Üí Add unemployment-specific confounders
  - ‚Üí Check if effect changes (confounding test)
  - ‚Üí Naive vs Adjusted comparison
  - ‚Üí Heterogeneity analysis (does it vary by community type?)

STEP 4: Synthesis
  - ‚Üí Compare Joint vs Sequential estimates
  - ‚Üí Final causal ranking
  - ‚Üí Policy recommendations

  This helps us to go from case-by-case basis to go with the what matters most in establishing causality.

  

```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("="*70)
print("üéØ HYBRID APPROACH - STEP 1: JOINT ANALYSIS")
print("="*70)

# Define remaining causal factors (excluding poverty and family)
remaining_causes = {
    'Unemployment': 'PctUnemployed',
    'Education (Low - No HS)': 'PctNotHSGrad',
    'Education (High - BS+)': 'PctBSorMore',
    'Race (% White)': 'racePctWhite',
    'Race (% Black)': 'racepctblack',
    'Age (12-21)': 'agePct12t21',
    'Age (16-24)': 'agePct16t24',
    'Urban Density': 'PopDens',
    'Urban %': 'pctUrban',
}

# Common confounders (affect MULTIPLE treatments)
common_confounders = [
    'medIncome',        # Affects unemployment, education, urban
    'PctKids2Par',      # Family structure (we know it matters, keep it as control)
    'PctPopUnderPov',   # Poverty (we know it matters, keep it as control)
]

# Prepare data
outcome = 'ViolentCrimesPerPop'
analysis_vars = list(remaining_causes.values()) + common_confounders + [outcome]
df_joint = crime_df_clean[[v for v in analysis_vars if v in crime_df_clean.columns]].dropna()

print(f"\nüìä Joint Analysis Dataset:")
print(f"   Communities: {len(df_joint)}")
print(f"   Causal factors to compare: {len(remaining_causes)}")
print(f"   Common confounders: {len(common_confounders)}")

# Run joint model
X_joint = df_joint[[v for v in remaining_causes.values() if v in df_joint.columns] +
                   [c for c in common_confounders if c in df_joint.columns]]
y_joint = df_joint[outcome]

model_joint = sm.OLS(y_joint, sm.add_constant(X_joint)).fit()

# Extract results
results = []
for label, var in remaining_causes.items():
    if var in model_joint.params:
        results.append({
            'Factor': label,
            'Variable': var,
            'Effect': model_joint.params[var],
            'Std_Error': model_joint.bse[var],
            'P_value': model_joint.pvalues[var],
            'Significant': model_joint.pvalues[var] < 0.05,
            'Abs_Effect': abs(model_joint.params[var])
        })

results_df = pd.DataFrame(results).sort_values('Abs_Effect', ascending=False)

print("\n" + "="*70)
print("üìä JOINT MODEL RESULTS - Ranked by Effect Size")
print("="*70)
print(f"\nModel R¬≤: {model_joint.rsquared:.4f}")
print(f"Sample size: {len(df_joint)}\n")

print(f"{'Rank':<6}{'Factor':<25}{'Effect':<10}{'P-value':<12}{'Sig':<5}")
print("-" * 70)

for i, (_, row) in enumerate(results_df.iterrows(), 1):
    sig_marker = "***" if row['P_value'] < 0.001 else "**" if row['P_value'] < 0.01 else "*" if row['P_value'] < 0.05 else ""
    print(f"{i:<6}{row['Factor']:<25}{row['Effect']:<10.4f}{row['P_value']:<12.6f}{sig_marker:<5}")

print("\n" + "="*70)
print("üèÜ TOP 3 FACTORS TO INVESTIGATE:")
print("="*70)

top_3 = results_df.head(3)
for i, (_, row) in enumerate(top_3.iterrows(), 1):
    direction = "REDUCES" if row['Effect'] < 0 else "INCREASES"
    print(f"\n#{i}: {row['Factor']}")
    print(f"    Effect: {row['Effect']:.4f} ({direction} crime)")
    print(f"    Significance: {'YES' if row['Significant'] else 'NO'} (p={row['P_value']:.4f})")
```

```python
print("\n" + "="*70)
print(f"üî¨ STEP 3: DEEP DIVE - {top_factor['Factor'].upper()} ‚Üí CRIME")
print("="*70)

# Get the winning variable
treatment_var = top_factor['Variable']
treatment_name = top_factor['Factor']

print(f"\nCausal Question: Does {treatment_name} CAUSE crime?")
print(f"Treatment variable: {treatment_var}")

# Now identify SPECIFIC confounders for this treatment
# These are different from the joint model!

print("\n" + "="*70)
print(f"üéØ IDENTIFYING CONFOUNDERS SPECIFIC TO {treatment_name}")
print("="*70)

# Build factor-specific confounder set based on what the winner is
# I'll create logic for different possible winners

if 'Unemployment' in treatment_name:
    print("\nLogic: What affects BOTH unemployment AND crime?")
    print("  ‚Ä¢ Education ‚Üí affects job prospects ‚Üí unemployment")
    print("  ‚Ä¢ Education ‚Üí affects opportunities ‚Üí crime")
    print("  ‚Ä¢ Age ‚Üí young people have higher unemployment")
    print("  ‚Ä¢ Age ‚Üí young people commit more crime")
    print("  ‚Ä¢ Family structure ‚Üí affects economic stability ‚Üí unemployment")
    print("  ‚Ä¢ Family structure ‚Üí affects supervision ‚Üí crime")
    print("  ‚Ä¢ Race/demographics ‚Üí structural factors")

    confounders_specific = [
        'PctBSorMore', 'PctNotHSGrad',  # Education
        'agePct12t21', 'agePct16t24',     # Age
        'PctKids2Par',                     # Family
        'racePctWhite', 'racepctblack',   # Race
        'medIncome',                       # Income
        'PopDens', 'pctUrban'             # Urban factors
    ]

elif 'Education' in treatment_name:
    print("\nLogic: What affects BOTH education AND crime?")
    print("  ‚Ä¢ Income ‚Üí affects ability to get education")
    print("  ‚Ä¢ Income ‚Üí affects crime through opportunity")
    print("  ‚Ä¢ Family structure ‚Üí affects educational outcomes")
    print("  ‚Ä¢ Family structure ‚Üí affects crime")
    print("  ‚Ä¢ Race/demographics ‚Üí structural educational access")
    print("  ‚Ä¢ Urban vs rural ‚Üí different educational systems")

    confounders_specific = [
        'medIncome', 'PctPopUnderPov',    # Economic
        'PctKids2Par',                     # Family
        'racePctWhite', 'racepctblack',   # Race
        'agePct12t21',                     # Age
        'PopDens', 'pctUrban',            # Urban
        'PctUnemployed'                    # Employment
    ]

elif 'Race' in treatment_name:
    print("\nLogic: What affects demographic composition AND crime?")
    print("  ‚Ä¢ Historical patterns ‚Üí residential segregation")
    print("  ‚Ä¢ Income inequality ‚Üí both demographics and crime")
    print("  ‚Ä¢ Urban vs rural ‚Üí different demographic makeups")
    print("  ‚Ä¢ Regional factors ‚Üí migration patterns")

    confounders_specific = [
        'medIncome', 'PctPopUnderPov',    # Economic
        'PctBSorMore', 'PctNotHSGrad',    # Education
        'PctUnemployed',                   # Employment
        'PctKids2Par',                     # Family
        'PopDens', 'pctUrban',            # Urban
        'agePct12t21'                      # Age
    ]

elif 'Age' in treatment_name:
    print("\nLogic: What affects age structure AND crime?")
    print("  ‚Ä¢ Family structure ‚Üí affects age demographics")
    print("  ‚Ä¢ Urban areas ‚Üí attract young people")
    print("  ‚Ä¢ Economic factors ‚Üí youth migration")

    confounders_specific = [
        'PctKids2Par',                     # Family
        'PctUnemployed',                   # Employment
        'medIncome',                       # Income
        'PopDens', 'pctUrban',            # Urban
        'PctBSorMore'                      # Education
    ]

else:  # Urban density
    print("\nLogic: What affects urban density AND crime?")
    print("  ‚Ä¢ Economic opportunities ‚Üí people move to cities")
    print("  ‚Ä¢ Education ‚Üí cities have universities")
    print("  ‚Ä¢ Demographics ‚Üí cities attract diverse populations")

    confounders_specific = [
        'medIncome',                       # Income
        'PctBSorMore', 'PctNotHSGrad',    # Education
        'racePctWhite', 'racepctblack',   # Demographics
        'PctUnemployed',                   # Employment
        'agePct12t21'                      # Age
    ]

# Remove the treatment variable if it's in confounders
confounders_specific = [c for c in confounders_specific if c != treatment_var]

print(f"\nüìã Confounders for {treatment_name} analysis:")
for conf in confounders_specific:
    print(f"   ‚Ä¢ {conf}")

# Prepare sequential analysis data
sequential_vars = [treatment_var, outcome] + [c for c in confounders_specific if c in crime_df_clean.columns]
df_sequential = crime_df_clean[sequential_vars].dropna()

print(f"\nüìä Sequential Analysis Dataset: {len(df_sequential)} communities")

# Run naive and adjusted models
X_naive_seq = df_sequential[[treatment_var]]
X_adjusted_seq = df_sequential[[treatment_var] + [c for c in confounders_specific if c in df_sequential.columns]]
y_seq = df_sequential[outcome]

model_naive_seq = sm.OLS(y_seq, sm.add_constant(X_naive_seq)).fit()
model_adjusted_seq = sm.OLS(y_seq, sm.add_constant(X_adjusted_seq)).fit()

print("\n" + "="*70)
print(f"üìä SEQUENTIAL ANALYSIS RESULTS: {treatment_name} ‚Üí Crime")
print("="*70)

print("\n1Ô∏è‚É£ NAIVE (No confounders):")
print(f"   Effect: {model_naive_seq.params[treatment_var]:.4f}")
print(f"   P-value: {model_naive_seq.pvalues[treatment_var]:.6f}")
print(f"   R¬≤: {model_naive_seq.rsquared:.4f}")

print("\n2Ô∏è‚É£ ADJUSTED (With confounders):")
print(f"   Effect: {model_adjusted_seq.params[treatment_var]:.4f}")
print(f"   P-value: {model_adjusted_seq.pvalues[treatment_var]:.6f}")
print(f"   R¬≤: {model_adjusted_seq.rsquared:.4f}")

print("\nüìâ COMPARISON:")
effect_change = model_naive_seq.params[treatment_var] - model_adjusted_seq.params[treatment_var]
pct_change = (effect_change / model_naive_seq.params[treatment_var] * 100) if model_naive_seq.params[treatment_var] != 0 else 0

print(f"   Naive effect:    {model_naive_seq.params[treatment_var]:.4f}")
print(f"   Adjusted effect: {model_adjusted_seq.params[treatment_var]:.4f}")
print(f"   Change:          {effect_change:.4f} ({pct_change:.1f}%)")

if abs(pct_change) > 50:
    print("\n   ‚ö†Ô∏è  LARGE CHANGE! Confounding was substantial")
elif abs(pct_change) > 20:
    print("\n   ‚ö†Ô∏è  MODERATE CHANGE! Some confounding present")
else:
    print("\n   ‚úÖ SMALL CHANGE! Effect is robust")

# Check if effect is still significant
if model_adjusted_seq.pvalues[treatment_var] < 0.05:
    print(f"\n   ‚úÖ Still significant after adjustment (p={model_adjusted_seq.pvalues[treatment_var]:.4f})")
    print(f"   ‚Üí {treatment_name} likely has a TRUE CAUSAL effect on crime")
else:
    print(f"\n   ‚ùå No longer significant after adjustment (p={model_adjusted_seq.pvalues[treatment_var]:.4f})")
    print(f"   ‚Üí {treatment_name} effect was likely due to confounding")

# Show which confounders matter
print("\nüîç SIGNIFICANT CONFOUNDERS:")
print("-" * 70)
for var in X_adjusted_seq.columns:
    if var != treatment_var and var in model_adjusted_seq.params:
        coef = model_adjusted_seq.params[var]
        pval = model_adjusted_seq.pvalues[var]
        if pval < 0.05:
            sig = "***" if pval < 0.001 else "**" if pval < 0.01 else "*"
            print(f"   {var:20s}: {coef:7.4f} (p={pval:.4f}) {sig}")
```

This doesn't mean race "causes" crime!

Race is a proxy for these unmeasured factors, not a causal variable itself.
