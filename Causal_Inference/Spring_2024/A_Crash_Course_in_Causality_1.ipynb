<a id='Introduction'></a>
## Introduction

Causality is a fundamental concept that explores the relationship between events where one event, the cause, leads to the effectuation of another event, the effect. 

This principle is essential across various disciplines, including philosophy, statistics, economics, medicine, and social sciences, serving as a cornerstone for understanding the interconnectedness of phenomena in the natural and social worlds.

Let's start with a classic question: Did the rooster's crow cause the sun to rise? Of course, we know the answer is no. The rooster's crow and the sunrise are correlated (they happen around the same time), but one doesn't cause the other. Distinguishing between correlation (things happening together) and causation (one thing actually causing another) is a central challenge in the quest to understand causality.

In realms like medicine, economics, and social sciences, teasing apart cause and effect is crucial. Consider a medical researcher trying to determine whether a new drug can cure a disease. It's not enough to observe that patients who take the drug get better; the improvement must be directly attributable to the drug, not to other factors or mere coincidence. This quest for causality drives the design of experiments, the analysis of data, and the implementation of policies that aim to improve our health, society, and environment.

Despite these challenges, advancements in statistical methods and experimental designs continue to enhance our ability to discern causal relationships, contributing to our knowledge and ability to shape the world around us.

## Concept Overview
<a id='Causality'></a>
### Causality

**Definition:** Causality is an influence by which one event, process, state, or object (a cause) contributes to the production of another event, process, state, or object (an effect) where the cause is partly responsible for the effect, and the effect is partly dependent on the cause.[1]

A process generally has numerous causes, which are often referred to as causal factors, all of which are located in its history. Many more consequences, all of which are future in nature, can have an influence as a cause or a causal component in turn. Some authors have maintained that the concepts of time and space are metaphysically secondary to causality.

**Example:** Exercise directly results in weight loss.

Let's say you choose to begin a fitness program. You exercise frequently, including weight training as well as cardiovascular exercises. You eventually notice a noticeable drop in weight. In this case, weight loss and regular exercise are causally related. Engaging in physical activity boosts your metabolism and burns calories, resulting in weight loss. Here's a simple illustration of causality in action: exercise directly results in weight loss.



```python
import numpy as np
import matplotlib.pyplot as plt

# Setting up a random seed for reproducibility
np.random.seed(42)

# Simulating data for 100 individuals
n = 100

# Causality: Exercise and Weight Loss
exercise_hours = np.random.uniform(1, 10, size=n)  # Hours per week
weight_loss_causal = exercise_hours * -1.5 + np.random.normal(0, 1, size=n)  # kg

# Causality Plot
plt.scatter(exercise_hours, weight_loss_causal)
plt.title('Causality: Exercise vs. Weight Loss')
plt.xlabel('Exercise Hours per Week')
plt.ylabel('Weight Loss (kg)')
```

Causality plot showing a negative relationship between exercise hours and weight loss, indicating that more exercise is associated with more weight loss.

<a id='Correlation'></a>
### Correlation

**Definition:** Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate).[2]

An association between different kinds of variables is described by correlation: as one variable changes, the other also changes. A statistical measure of the link between two variables is called a correlation. Together, these variables fluctuate: they are covariant. However, there may not be a direct or indirect causal relationship behind this covariation.

**Example:** Let us consider our example from before, now suppose you also made changes to your diet around the same time you started exercising, opting for healthier, lower-calorie foods. As you track your progress, you notice that your weight loss correlates with your improved dietary habits. However, without isolating the effects of diet from exercise, you can't definitively say your diet alone caused the weight loss. It's possible that both your diet and your exercise regimen contributed to your weight loss, demonstrating a correlation between diet and weight loss without establishing a direct cause-and-effect relationship.


```python
# Setting up a random seed for reproducibility
np.random.seed(42)

# Simulating data for 100 individuals
n = 100

# Correlation: Diet Quality and Weight Loss
diet_quality = np.random.uniform(5, 10, size=n)  # Scale from 1 to 10
# Adding noise to introduce correlation without direct causation
weight_loss_correlated = diet_quality * -1 + np.random.normal(0, 2, size=n)  # kg

# Correlation Plot
plt.scatter(diet_quality, weight_loss_correlated)
plt.title('Correlation: Diet Quality vs. Weight Loss')
plt.xlabel('Diet Quality (Scale 1-10)')
plt.ylabel('Weight Loss (kg)')

plt.tight_layout()
plt.show()
```

Correlation plot showing a negative relationship between diet quality and weight loss, suggesting that higher diet quality is associated with more weight loss, but without implying direct causation.

Correlation is represented by r. Correlation coefficients range from -1 to 1. 

A correlation of 1 indicates a perfect positive linear relationship, where as one variable increases, the other variable also increases in a linear fashion. 

A correlation of -1 indicates a perfect negative linear relationship, where as one variable increases, the other variable decreases linearly. 

A correlation of 0 indicates no linear relationship between the variables.

**Pearson Correlation:** The Pearson correlation coefficient is used to measure linear relationships between continuous variables. It assumes that the variables are normally distributed and have a linear relationship.

**Spearman Correlation:** Spearman correlation coefficient, is a non-parametric measure of correlation that assesses how well the relationship between two variables can be described using a monotonic function. It is more appropriate when the relationship between variables is non-linear or when the variables are not normally distributed.


<a id='Confounding'></a>
### Confounding

**Definition:** Confounding variables are those that affect other variables in a way that produces spurious or distorted associations between two variables. They confound the "true" relationship between two variables.[3]

A confounding variable is a third variable in a study that affects both the independent and dependent variables, making it appear as though there is a causal relationship between them when none exists or masking the true relationship. Confounding variables can lead to inaccurate conclusions if they are not properly accounted for in the study design or analysis.

**Example:** Now let's add stress as a third variable. Imagine that along with your desire to eat better and exercise more, you experienced a period of lower stress as a result of learning to meditate or improving your work-life balance. Reduction of stress is a confusing element in this case. It's feasible that less stress helped people lose weight on their own, maybe by lowering cortisol levels, which affect whether people gain or lose weight. It is more difficult to link weight loss purely to food or exercise because of this confusing interaction. This third element, stress reduction, which impacts weight independently, may have an impact on the observed link between exercise and weight loss.

```python
# Setting up a random seed for reproducibility
np.random.seed(42)

# Simulating data for 100 individuals
n = 100

# Confounding: Stress levels affecting both exercise and weight loss
stress_levels = np.random.uniform(1, 10, size=n)  # Lower is better
exercise_hours_confounded = 10 - stress_levels + np.random.normal(0, 1, size=n)  # Reduced exercise by stress
weight_loss_confounded = exercise_hours_confounded * -1.2 + (10 - stress_levels) * -0.5 + np.random.normal(0, 1, size=n)

# Confounding Plot
scatter = plt.scatter(stress_levels, weight_loss_confounded, c=exercise_hours_confounded, cmap='viridis')
plt.title('Confounding: Stress Levels, Exercise, and Weight Loss')
plt.xlabel('Stress Levels (Lower is Better)')
plt.ylabel('Weight Loss (kg)')
cb = plt.colorbar(scatter, orientation='vertical')
cb.set_label('Exercise Hours per Week')

plt.show()
```

Confounding plot where stress levels influence both exercise hours (color intensity) and weight loss, demonstrating how stress can confound the relationship between exercise and weight loss.


<a id='Potential Outcomes'></a>
### Potential Outcomes

**Definition:** The potential outcomes framework provides a way to quantify causal effects. For a hypothetical intervention, it defines the causal effect for an individual as the difference between the outcomes that would be observed for that individual with versus without the exposure or intervention under consideration. 

**Example:**

In our context of exercise, diet, and weight loss:

Treatment: Engaging in a regular exercise regimen.

Control: Not engaging in a regular exercise regimen.

Outcome: Amount of weight loss.

For each individual, there are two potential outcomes: the amount of weight loss if they follow a regular exercise regimen (treatment) and the amount of weight loss if they do not (control). However, we can only observe one of these outcomes for each personthe one corresponding to the action they actually took.

```python
# Random seed for reproducibility
np.random.seed(42)

# Simulating data
people = np.arange(1, 101)  # 100 people
exercise_group = np.random.choice([True, False], size=100)  # Randomly assigned to exercise or not
weight_loss_if_exercise = np.random.normal(loc=5, scale=1.5, size=100)  # Potential outcome with exercise
weight_loss_if_no_exercise = np.random.normal(loc=2, scale=1.5, size=100)  # Potential outcome without exercise

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(people[exercise_group], weight_loss_if_exercise[exercise_group], label='Weight Loss if Exercise', color='blue')
plt.scatter(people[~exercise_group], weight_loss_if_no_exercise[~exercise_group], label='Weight Loss if No Exercise', color='red')
plt.title('Potential Outcomes: Exercise Effect on Weight Loss')
plt.xlabel('Individuals')
plt.ylabel('Weight Loss (kg)')
plt.legend()
plt.grid(True)
plt.show()
```

<a id='Directed Acyclic Graphs (DAGs)'></a>
### Directed Acyclic Graphs (DAGs)

**Definition:** Directed Acyclic Graphs (DAGs) are graphical tools used to represent and reason about causal relationships. 

A DAG is a diagram consisting of nodes (representing variables, such as exercise, diet, stress levels, and weight loss) and directed edges (arrows) that indicate causal relationships between these variables. The "acyclic" part means that you can't follow the arrows in a way that brings you back to a starting pointthere are no loops.

DAGs help identify:

Direct causal paths (e.g., exercise directly leading to weight loss).
Indirect paths (e.g., exercise affecting stress levels, which in turn affect weight loss).
Confounding paths (e.g., stress influencing both exercise and weight loss, potentially misleading the observed relationship between exercise and weight loss).


```python
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes with labels
G.add_nodes_from(["Exercise", "Stress", "Diet", "Weight Loss"])

# Add edges to define causal relationships
G.add_edges_from([("Exercise", "Weight Loss"), ("Stress", "Exercise"), ("Stress", "Weight Loss"), ("Diet", "Weight Loss")])

# Drawing the graph
pos = nx.spring_layout(G, seed=42)  # For consistent layout
nx.draw(G, pos, with_labels=True, node_color='orange', edge_color='black', node_size=3700, font_size=10, arrowsize=20)
plt.title('Directed Acyclic Graph (DAG) for Exercise, Stress, Diet, and Weight Loss')
plt.show()
```

### Randomized Control Trial (RCT)

**Definition:** A randomized controlled trial (RCT) is a form of scientific experiment used to control factors not under direct experimental control.

Randomized Controlled Trials (RCTs) are considered the gold standard in research design for determining the causal effects of interventions.

Participants are randomly assigned to either the treatment group, which receives the intervention, or the control group, which does not receive the intervention (and may receive a placebo or standard care instead).

The treatment group receives the intervention, while the control group does not. Outcomes are measured in both groups after the intervention period. 

The outcomes of the treatment group are compared to the outcomes of the control group. If the study is well-designed and executed, any significant difference in outcomes between the groups can be attributed to the intervention, since randomization should have balanced all other potential confounding factors.

**Applications:** Some common applications are clinical trials that compare the effects of drugs, surgical techniques, medical devices, diagnostic procedures, diets or other medical treatments. It is also widely used in psychology, education and in social science to test the efficacy of treatments, educational programs, and policy interventions respectively. 


Lets illustrate using the weight loss example. Let us conduct a trial on a new weight loss diet and random assort participants into two groups, the treament and control group. Simulating the baseline weights for participants and grouping and performing the T-Test to test the significance of the new diet. 

```python
from scipy import stats

np.random.seed(42)

n_participants = 100

# Simulate baseline weights for the participants
baseline_weights = np.random.normal(80, 10, n_participants)

# Randomly assign participants to either the treatment (new diet) or control group
groups = np.random.choice(['Treatment', 'Control'], size=n_participants)

# Assume the new diet leads to an average of 5 kg weight loss, and the control group loses 2 kg on average
weight_loss_treatment = np.random.normal(-5, 1.5, (groups == 'Treatment').sum())
weight_loss_control = np.random.normal(-2, 1.5, (groups == 'Control').sum())

# Calculate the final weights
final_weights_treatment = baseline_weights[groups == 'Treatment'] + weight_loss_treatment
final_weights_control = baseline_weights[groups == 'Control'] + weight_loss_control

# Perform a t-test to compare the weight loss between the two groups
t_stat, p_value = stats.ttest_ind(weight_loss_treatment, weight_loss_control)

# Output the results
print(f"Average weight loss - Treatment group: {-np.mean(weight_loss_treatment):.2f} kg")
print(f"Average weight loss - Control group: {-np.mean(weight_loss_control):.2f} kg")
print(f"T-test results: t = {t_stat:.2f}, p = {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("The difference in weight loss between the treatment and control groups is statistically significant.")
else:
    print("There is no statistically significant difference in weight loss between the treatment and control groups.")
```

<a id='Observational Studies'></a>
### Observational Studies

**Definition:** Observational studies are a category of research designs used to infer the possible effect of a treatment on subjects where the researcher does not have control over the assignment of the treatment.

Unlike randomized controlled trials (RCTs), where participants are randomly assigned to treatment or control groups, observational studies observe outcomes without manipulating the study environment or the assignment of interventions. 

These studies are particularly useful in situations where RCTs are impractical, unethical, or impossible to conduct.

Observational studies often take place in natural settings, making their findings potentially more generalizable to real-world scenarios.

**Applications:** Observational studies are invaluable in epidemiology, economics, psychology, and other fields where experimental studies are not feasible. They play a crucial role in hypothesis generation, policy analysis, and the preliminary investigation of causal relationships.

<a id='Instrumental Variable'></a>
### Instrumental Variable

**Definition:** An instrumental variable is a variable that is correlated with the explanatory variable of interest (the one causing endogeneity issues) but is not directly related to the dependent variable, except through its correlation with the explanatory variable. 

An IV influences the dependent variable only through its effect on the independent variable that is causing endogeneity problems.

**Application:** It's widely used in economics, epidemiology, and social sciences to study the effects of policies, treatments, and interventions when randomized controlled trials are not possible.

**Requirements:** 

1. Relevance: The instrumental variable must be correlated with the endogenous explanatory variable. This ensures that the instrument can explain some of the variations in the explanatory variable.

2. Exogeneity: The instrument must not be correlated with the error term in the regression model. This means it should not directly affect the dependent variable outside of its effect through the endogenous explanatory variable.

**Example:**
Imagine researching the effect of education on earnings. However, education might be endogenous due to factors like ability biaspeople with higher earning potential might also be more likely to obtain more education. An instrumental variable could be the proximity to a college when someone was 18 years old. Proximity may encourage higher education due to lower costs (relevance), but it's unlikely to affect future earnings directly, except through education (exogeneity).

```python
# Simulate data
np.random.seed(42)
proximity_to_college = np.random.normal(10, 2, 100)  # Proximity score
years_of_education = 12 + 0.5 * proximity_to_college + np.random.normal(0, 1, 100)  # Education

plt.figure(figsize=(8, 5))
plt.scatter(proximity_to_college, years_of_education, alpha=0.7)
plt.title('IV (Proximity to College) vs. Endogenous Variable (Years of Education)')
plt.xlabel('Proximity to College')
plt.ylabel('Years of Education')
plt.show()
```

```python
# Simulate data for hourly wage
hourly_wage = 15 + 1.5 * years_of_education + np.random.normal(0, 2, 100)

plt.figure(figsize=(8, 5))
plt.scatter(years_of_education, hourly_wage, alpha=0.7)
plt.title('Observed Relationship: Years of Education vs. Hourly Wage')
plt.xlabel('Years of Education')
plt.ylabel('Hourly Wage')
plt.show()
```

The plots illustrate two key steps in understanding and addressing endogeneity with an instrumental variable:

**IV (Proximity to College) vs. Endogenous Variable (Years of Education):** The first plot shows a positive relationship between the proximity to college (the instrumental variable) and years of education (the endogenous variable). This visualizes the relevance condition of the instrumental variableproximity to college is associated with higher education levels, suggesting it can serve as a tool to isolate the exogenous variation in education.

**Observed Relationship: Years of Education vs. Hourly Wage:** The second plot displays the observed relationship between years of education and hourly wage. This relationship might be confounded by unobserved factors like innate ability or family background, making education an endogenous variable in the context of wage determination.

## Importance of A/B Testing in Causality

**Definition:** A/B testing (also known as bucket testing, split-run testing, or split testing) is a user experience research methodology. A/B tests consist of a randomized experiment that usually involves two variants (A and B),although the concept can be also extended to multiple variants of the same variable.

A/B testing is one of the most straightforward and practical methods to empirically test hypotheses about causal relationships, especially in tech industries, marketing, product development, and web design.

It serves as a real-world application of the potential outcomes framework, illustrating how randomized controlled trials (RCTs) can be conducted outside of traditional research settings.

**Applications:** A/B Testing is widely used in website optimization, email marketing, digital optimization.

**Why is A/B Testing important:** A/B testing, also known as split testing, is crucial in the study of causality because it provides a rigorous methodological framework for establishing causal relationships between variables, especially in real-world settings where numerous confounding factors can obscure the effects of interventions. 

**Illustration:**

Let us consider a fictitous drug trial that claims to improves IQ of people. To test the efficiency of the drug let us employ A/B Testing.


The trail compares the IQ of the people who were given the drug (treatment group) and the control group which was administered with placebo. IQ afrom 90-110 is considered average and IQ above 120 is considered exceptional.

```python
import pandas as pd
drug = pd.DataFrame(dict(iq=(101,100,102,104,102,97,105,105,98,101,100,123,105,103,100,95,102,106,
        109,102,82,102,100,102,102,101,102,102,103,103,97,97,103,101,97,104,
        96,103,124,101,101,100,101,101,104,100,101),
                         group='drug'))
placebo = pd.DataFrame(dict(iq=(99,101,100,101,102,100,97,101,104,101,102,102,100,105,88,101,100,
           104,100,100,100,101,102,103,97,101,101,100,101,99,101,100,100,
           101,100,99,101,100,102,99,100,99),
                            group='placebo'))
```

Let us visualize the data.

```python
# Let's visualize the data with seaborn, which also gives us a kde estimate
import matplotlib as plt
%matplotlib inline
import seaborn as sns
sns.set(style="ticks")

bins = np.arange(70,150,6)

A = drug.iq.values
B = placebo.iq.values

# Show the results of a linear regression within each dataset
ax1 = sns.histplot(A,bins=bins,label='Drug Group')
ax2 = sns.histplot(B,bins=bins,label='Placebo group')

plt.pyplot.xlabel('IQ level')
plt.pyplot.ylabel('Distribution of IQ levels')
plt.pyplot.legend(bbox_to_anchor=(0.45, 0.95), loc=2, borderaxespad=0.)

plt.pyplot.xlim((80,120))
plt.pyplot.show()
```

```python
import scipy.stats as stats
exp_df = pd.DataFrame(dict(group=[0]*len(drug.iq)+[1]*len(placebo.iq), score=np.r_[drug.iq,placebo.iq]))


combined_iq = np.concatenate((drug.iq, placebo.iq))
print(len(drug.iq))
print(len(placebo.iq))
print( drug.iq.mean() )
print( placebo.iq.mean() )
print( drug.iq.var() )
print( placebo.iq.var() )
print( combined_iq.mean() )
stats.ttest_ind(a= drug.iq,
                b= placebo.iq,
                equal_var=False) 
```

The results show that the t-value for the mean difference is 1.622. The p-value is 0.11 which is greater than the traditional "alpha" cut off at p=0.05. Therefore this test is *not significant*: The null hypothesis is in effect.

The test yields a p-value of 0.11 means ***there is a 11% chance we'd see sample data this far apart statistically if the two groups tested are actually identical***. 

If we were using a 95% confidence level, we would **fail** to reject the null hypothesis, since the p-value is greater than the corresponding significance level of 0.05. And so we conclude that the drug is as effective as placebo at a 95% confidence level: there is nothing interesting going on. At a significance level of 0.05, our drug outcome on IQ is statistically **nonsignificant**.

<a id='Guide To Selection of Tools'></a>
## Guide To Selection of Tools

Main considerations to ponder over before selecting the right tools for causal inference are related to the research question, the nature of the data, and the assumptions underlying different methods.

Let us understand them further:

1. **Nature of the Research Question:** Determine whether the aim is to understand the causal mechanism (how and why something happens) or merely to predict outcomes. Consider whether you're interested in average treatment effects, conditional effects (effects that vary with covariates), or the effects of treatments that vary in intensity or type.



2. **Data Structure and Quality:** When rich covariate data is available, propensity score matching or weighting can help simulate a randomized experiment by balancing treatment and control groups.



3. **Assumptions and Robustness:** Every causal inference method relies on assumptions (e.g., no unmeasured confounding, linearity, etc.). It's crucial to assess the plausibility of these assumptions in the context of your study. Consider methods that allow for sensitivity analysis to test how robust your findings are to violations of assumptions or to changes in model specifications.

### Logistic Regression

Conducting logistic regression to estimate the effect of race and gender on callback rates, controlling for other covariates. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Select relevant features and target variable
features = ['race', 'gender', 'years_college', 'college_degree', 'years_experience', 'job_city', 'resume_quality']
target = 'received_callback'

# Convert categorical variables to numerical using one-hot encoding

resume_encoded = pd.get_dummies(resume[features + [target]], drop_first=True)  # drop_first=True to avoid dummy variable trap

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(resume_encoded.drop(target, axis=1), resume_encoded[target], test_size=0.2, random_state=42)

# Train a logistic regression model
logreg = LogisticRegression(max_iter=1000)  # Increased max_iter to ensure convergence if necessary
logreg.fit(X_train, y_train)

# Get feature importance
# In logistic regression, coefficients represent the log odds of the feature importance which can be positive or negative
feature_importance = logreg.coef_[0]
feature_names = X_train.columns

# Sort features by their absolute importance values for clarity
sorted_features = sorted(zip(np.abs(feature_importance), feature_names), reverse=True)

# Print the feature importance
for importance, feature in sorted_features:
    print(f"{feature}: {importance:.4f}")  # Added formatting for readability

# Visualize feature importance
# Adjusted to correctly match features with their importance values
plt.figure(figsize=(8, 6))
plt.barh([feature for _, feature in sorted_features], [importance for importance, _ in sorted_features])
plt.xlabel('Absolute Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance for Callback Rates')
plt.gca().invert_yaxis()  # Invert y-axis to have the most important feature on top
plt.show()
```

The graph suggest potential biases in callback rates based on white race has significant impact on the callback rate. The gender male also has impact on the callback rate.

Education credentials (having a college degree and years spent in college) are influential in callback rates, but their relative importance is lower than race or gender.

Work experience has the least impact among the shown factors but is still a contributing factor in the prediction model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score



# Convert categorical variables to numerical using one-hot encoding
resume_encoded = pd.get_dummies(resume[features + [target]], drop_first=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    resume_encoded.drop(target, axis=1), 
    resume_encoded[target], 
    test_size=0.2, 
    random_state=42
)

# Logistic Regression
logreg = LogisticRegression(max_iter=1000)  # Increase max_iter if the algorithm has convergence issues
logreg.fit(X_train, y_train)
logreg_preds = logreg.predict(X_test)
logreg_probs = logreg.predict_proba(X_test)[:, 1]

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
rf_probs = rf.predict_proba(X_test)[:, 1]

# Support Vector Machines (SVM)
svm = SVC(probability=True, kernel='rbf')  # 'rbf' is the default but is specified here for clarity
svm.fit(X_train, y_train)
svm_preds = svm.predict(X_test)
svm_probs = svm.predict_proba(X_test)[:, 1]

# Evaluate model performance
def evaluate_model(y_true, y_pred, y_probs, pos_label=1):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, pos_label=pos_label)
    recall = recall_score(y_true, y_pred, pos_label=pos_label)
    auc_roc = roc_auc_score(y_true, y_probs)
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'AUC-ROC: {auc_roc:.4f}')

print('Logistic Regression:')
evaluate_model(y_test, logreg_preds, logreg_probs)
print('\nRandom Forest:')
evaluate_model(y_test, rf_preds, rf_probs)
print('\nSupport Vector Machines (SVM):')
evaluate_model(y_test, svm_preds, svm_probs)
```

The prediction suggests that while there is high accuracy, the precision and recall is poor which suggests that the model has significant class imbalance. The model recall and precision can be improved as shown below.

```python
logreg_probs = logreg.predict_proba(X_test)[:, 1]  # Probabilities of the positive class

from sklearn.metrics import roc_curve

# Calculate the ROC curve points
fpr, tpr, thresholds = roc_curve(y_test, logreg_probs)

# Find the optimal threshold: This is a simplification; you might choose the threshold that maximizes tpr-fpr, for example
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]


from sklearn.preprocessing import binarize

# Apply the new threshold to get the class predictions
# Note: binarize expects a 2D array so we need to reshape logreg_probs
logreg_preds_adjusted = binarize(logreg_probs.reshape(-1, 1), threshold=optimal_threshold).flatten()

print('Adjusted Logistic Regression:')
evaluate_model(y_test, logreg_preds_adjusted, logreg_probs)
```

<a id='Logistic Regression'></a>
### Logistic Regression
Let us perform Logistic Regression on the data frame and calculate the Propensity Score. 

Propensity Score Matching is a statistical matching technique that aims to identify the effect of treament by accounting for the covariates that predict receiving the treatment.

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


covariates = ['Age', 'GPA', 'WorkExperience', 'Gender', 'Internship']  
target = 'Participated'

# Split data into features (X) and target variable (y)
X = pd.get_dummies(df[covariates], drop_first=True)  # One-hot encoding for categorical variables
y = df[target]

# Logistic regression to estimate propensity scores
logistic = LogisticRegression(max_iter=1000)
logistic.fit(X, y)

# Save the propensity scores in the DataFrame
df['propensity_score'] = logistic.predict_proba(X)[:, 1]
```

<a id='Propensity Score Matching'></a>
### Propensity Score Matching
Performing Propensity Score Matching:

```python
from sklearn.neighbors import NearestNeighbors

# Separate the treatment and control groups
treatment = df[df[target] == 1]
control = df[df[target] == 0]

# Use Nearest Neighbors to find the closest match in propensity score
nn = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(control[['propensity_score']])
distances, indices = nn.kneighbors(treatment[['propensity_score']])

# Extract the matched control samples
matched_control = control.iloc[indices.flatten()]

# Combine matched treatment and control samples for analysis
matched_df = pd.concat([treatment, matched_control]).reset_index(drop=True)


treatment_effect = matched_df.groupby(target)['post_training'].mean().diff().iloc[-1]

print(f"The estimated effect of job training on employment status is: {treatment_effect}")
```

An estimated effect of job training on employment status of approximately 0.0905 suggests that, according to your Propensity Score Matching (PSM) analysis, participation in job training increases the likelihood of being employed post-training by about 9.05 percentage points compared to those who did not participate, holding other observed covariates constant.
