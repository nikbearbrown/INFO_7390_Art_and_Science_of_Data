1. Which of the following are key assumptions of causal inference methods?

   A) Ignorability
   B) Consistency
   C) Positivity
   D) Exchangeability

   **Answers: A, C, D**

   Explanation:
   - Ignorability assumption implies that all relevant variables are included in the model.
   - Positivity ensures that there is a non-zero probability of receiving all levels of treatment.
   - Exchangeability assumes that treatment assignment is independent of potential outcomes, given covariates.

2. What are common causal inference methods used in observational studies?

   A) Propensity score matching
   B) Instrumental variable analysis
   C) Difference-in-differences
   D) Structural equation modeling

   **Answers: A, B, C**

   Explanation:
   - Propensity score matching aims to balance covariates between treated and control groups.
   - Instrumental variable analysis helps to address endogeneity by using an instrument that affects treatment but not outcome directly.
   - Difference-in-differences compares changes in outcomes before and after treatment between treatment and control groups.

3. Which of the following conditions are necessary for establishing causality?

   A) Temporal precedence
   B) Association between variables
   C) Absence of confounding variables
   D) Statistical significance

   **Answers: A, B, C**

   Explanation:
   - Temporal precedence asserts that the cause must precede the effect.
   - Association between variables means changes in one variable are related to changes in another.
   - Absence of confounding variables ensures that observed effects are not due to other variables.

4. In causal inference, what is the purpose of conducting sensitivity analysis?

   A) To assess the robustness of causal estimates
   B) To determine the statistical significance of the findings
   C) To identify potential confounding variables
   D) To estimate the average treatment effect

   **Answers: A, C**

   Explanation:
   - Sensitivity analysis examines how changes in assumptions or model specifications affect the results, testing the robustness of causal estimates.
   - It also helps in identifying potential confounding variables that could influence the causal inference.

5. Which of the following statements are true regarding randomized controlled trials (RCTs)?

   A) RCTs are the gold standard for establishing causality.
   B) RCTs are always feasible and ethical.
   C) RCTs are immune to confounding.
   D) RCTs may not be generalizable to the entire population.

   **Answers: A, D**

   Explanation:
   - RCTs are often considered the gold standard because randomization minimizes bias.
   - However, they may not always be feasible or ethical, and results may not be generalizable to the entire population due to specific inclusion criteria or other factors.

6. Which of the following factors affect the validity of causal inference from observational studies?

   A) Selection bias
   B) Measurement error
   C) Sample size
   D) Randomization

   **Answers: A, B**

   Explanation:
   - Selection bias occurs when characteristics of the sample differ systematically from the population.
   - Measurement error introduces inaccuracies in data collection, potentially biasing results.

7. What is the purpose of propensity score weighting in causal inference?

   A) To balance covariates between treatment groups
   B) To assess the statistical significance of causal estimates
   C) To control for random variation in the data
   D) To estimate the effect size of the treatment

   **Answer: A**

   Explanation:
   - Propensity score weighting aims to balance covariates between treatment groups, mimicking the balance achieved in a randomized controlled trial.

8. Which of the following strategies can help address confounding in observational studies?

   A) Matching
   B) Stratification
   C) Sensitivity analysis
   D) Cross-sectional analysis

   **Answers: A, B, C**

   Explanation:
   - Matching and stratification aim to create comparable groups based on covariates.
   - Sensitivity analysis assesses the impact of potential unmeasured confounders on the results.

9. What distinguishes causal inference from association in observational studies?

   A) Causal inference establishes a cause-and-effect relationship.
   B) Association only identifies correlations between variables.
   C) Causal inference relies on experimental data only.
   D) Association analysis considers temporal precedence.

   **Answers: A, B**

   Explanation:
   - Causal inference goes beyond identifying correlations, aiming to establish a cause-and-effect relationship between variables.
   - Association analysis focuses solely on identifying correlations between variables.

10. In causal inference, what is the role of instrumental variables?

   A) To control for confounding variables
   B) To establish temporal precedence
   C) To identify spurious correlations
   D) To address endogeneity

   **Answer: D**

   Explanation:
   - Instrumental variables are used to address endogeneity by providing variation in the treatment that is independent of potential confounders, helping to identify causal effects.
