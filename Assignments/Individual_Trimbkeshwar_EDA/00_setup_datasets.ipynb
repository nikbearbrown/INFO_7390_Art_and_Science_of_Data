```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
import os

print("✅ All libraries imported!")
```

```python
# Create folders
os.makedirs('datasets', exist_ok=True)
os.makedirs('Example1_Titanic', exist_ok=True)
os.makedirs('Example2_Housing', exist_ok=True)
os.makedirs('images', exist_ok=True)

print("✅ Folders created!")
```

```python
# Download Titanic dataset
titanic = sns.load_dataset('titanic')
titanic.to_csv('datasets/titanic.csv', index=False)
print(f"✅ Titanic downloaded: {titanic.shape[0]} rows × {titanic.shape[1]} columns")
titanic.head()
```

```python
# Download Housing dataset
housing = fetch_california_housing(as_frame=True)
housing_df = housing.frame
housing_df.to_csv('datasets/housing.csv', index=False)
print(f"✅ Housing downloaded: {housing_df.shape[0]} rows × {housing_df.shape[1]} columns")
housing_df.head()
```
