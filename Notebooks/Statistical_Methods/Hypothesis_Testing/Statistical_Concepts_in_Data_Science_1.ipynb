## Introduction

The study of data gathering, organisation, analysis, interpretation, and presentation is known as statistics. 

It is customary to start with a statistical population or statistical model to be examined when applying statistics to a scientific, industrial, or social problem. "All people living in a country" or "every atom composing a crystal" are examples of various groups of people or objects that make up a population. Every facet of data is covered by statistics, including the organisation of data gathering through the creation of surveys and trials.

When analysing data, two primary statistical techniques are employed: inferential statistics, which derive conclusions from data that are susceptible to random variation (e.g., observational errors, sampling variation), and descriptive statistics, which use indexes like the mean or standard deviation to summarise data from a sample. Two sets of distributional properties—central tendency, also known as location—are typically the focus of descriptive statistics. Central tendency aims to characterise the distribution's typical or centre value, while dispersion, also known as variability, characterises the degree to which distribution members deviate from the distribution's centre and from one another. Based on the investigation of random processes, probability theory serves as the foundation for inferences regarding mathematical statistics.

Probability distributions and transformations are vital tools in data science, playing a central role in understanding, processing, and extracting meaningful insights from data. Distributions provide a framework for understanding the inherent characteristics of data, such as its central tendency, variability, and the likelihood of different outcomes. This understanding is crucial for choosing appropriate modeling techniques, making accurate predictions, and conducting reliable statistical inferences. Transformations, on the other hand, are used to manipulate data into a more suitable form for analysis, helping to reveal hidden patterns, reduce skewness, and meet the assumptions of various statistical and machine learning algorithms. 

The application of these concepts extends across numerous fields, from risk assessment in finance to forecasting in meteorology, making them not only essential but also incredibly fascinating. They allow data scientists to make sense of complex, real-world phenomena and turn raw data into actionable insights, driving decision-making and innovation. The ability to effectively use probability distributions and transformations is what empowers data scientists to solve complex problems and deliver solutions that can significantly impact businesses, policy-making, and scientific research.

## Classic Distribution

One of the first statistical techniques applied to a dataset include mean, median and mode. 

1. **Mean**: The mean, often referred to as the average, is calculated by adding up all the values in a dataset and then dividing by the number of values.
    If you have `n` values, $x_1, x_2, ..., x_n$, the mean (often represented as $\bar{x}$) is $\bar{x}$ = $\frac{x_1 + x_2 + ... + x_n}{n}$ . 
     - The mean is sensitive to outliers or extremely high or low values, which can skew the mean.
     - It is most appropriate for interval or ratio data.
     - It provides a useful measure of the central tendency when the data distribution is symmetric.

2. **Median**: The median is the middle value in a dataset when the values are arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle numbers.
     - The median is not affected by outliers or skewed data, making it a better measure of central tendency for skewed distributions.
     - It is appropriate for ordinal, interval, or ratio data.
     - It divides a dataset into two equal halves.

3. **Mode**: The mode is the value that appears most frequently in a dataset. A dataset can have one mode (unimodal), two modes (bimodal), or more (multimodal).
     - The mode is the only measure of central tendency that can be used with nominal data.
     - It is not affected by outliers.
     - In some distributions, there may be no mode if no number repeats, or multiple modes if there are several numbers that appear with equal frequency.


A box (or whisker) plot perfectly illustrates what basic statistical features can explain:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_csv("500_Person_Gender_Height_Weight_Index.csv")

# Create the subplots
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (0.2, 1)})

# Calculate the statistics
mean = df['Height'].mean()
median = df['Height'].median()
mode = df['Height'].mode()[0]

# Create the boxplot
sns.boxplot(x=df["Height"], ax=ax_box)
ax_box.axvline(mean, color='r', linestyle='--')
ax_box.axvline(median, color='g', linestyle='-')
ax_box.axvline(mode, color='b', linestyle='-')

# Create the histogram
sns.histplot(df["Height"], ax=ax_hist, kde=True) # Use histplot and kde for newer versions of seaborn
ax_hist.axvline(mean, color='r', linestyle='--')
ax_hist.axvline(median, color='g', linestyle='-')
ax_hist.axvline(mode, color='b', linestyle='-')

ax_hist.set_xticks(range(100, 251, 20)) 
plt.legend({'Mean':mean,'Median':median,'Mode':mode})
ax_box.set(xlabel='')

# Show the plot
plt.show()
```

The first plot is **whiskers plot** . 

The data's median value is represented by the centre line. Due to its greater resilience to outlier values, the median is frequently chosen over the mean. The most commonly occurring number in your dataset is the mode, and the intermediate point is the median. 

In essence, the first quartile represents the 25th percentile, meaning that 25% of the data points fall below that threshold. The 75th percentile, or 75% of the data points, is represented by the third quartile. The upper and lower limits of our data range are represented by the min and max numbers.

When the box plot is short it implies that much of the data points are similar, since there are many values in a small range

When the box plot is tall it implies that much of the data points are quite different, since the values are spread over a wide range

If the median value is closer to the bottom then most of the data has lower values. If the median value is closer to the top then  most of the data has higher values. This means that, if the median line is not in the middle of the box then it is an indication of skewed data.

If the whiskers are very long means the data has a high standard deviation and variance i.e the values are spread out and highly varying. If there are long whiskers on one side of the box but not the other, then the data may be highly varying only in one direction.



The second plot is **histogram** . A histogram is a graphical representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable and was first introduced by Karl Pearson.

<a id='Bernoulli Distribution'></a>
### Bernoulli Distribution

**Definition**: The Bernoulli distribution is a fundamental probability distribution in statistics, particularly applicable to scenarios with two distinct outcomes, commonly referred to as "success" and "failure". It's characterized as a discrete distribution with only two possible outcomes, where the probability of success is denoted as $p$ and the probability of failure as $1 - p$. 

This distribution is a special case of the binomial distribution where only a single trial is considered (n=1). In terms of statistical parameters, the mean or expected value of a Bernoulli distribution is  $p$, and its variance is $p(1 - p)$. It's widely used in various fields to model binary outcomes, such as flipping a coin (head or tail), conducting a quality control check (pass or fail), or responding to a survey question (yes or no). 

**Probability Mass Function**: The probability mass function (PMF) of a Bernoulli distributed random variable is straightforward, assigning probability $p$ to the outcome 1 and $p(1 - p)$ to the outcome 0. Graphically, this distribution can be represented by two bars at the points 0 and 1 on the x-axis, with their heights corresponding to $p(1 - p)$ and $p$ respectively. 

The simplicity and foundational nature of the Bernoulli distribution make it a cornerstone in the field of probability and statistics, forming the basis for more complex models and statistical methods.

**Application**: Some of the common applications are in the field of Medicine and Healthcare, Quality Control and Testing, Finance and Risk Assessment etc.

**Methods:** 
- Parameter: p (probability of success)
- First Moment (Mean): μ = p
- Estimate p by the sample mean, p̂ = x̄.

```python
import scipy.stats as stats
# size = number of experiments
# p = probability of success
bern = stats.bernoulli.rvs(size=10000, p = 0.5)
```

```python
ax= sns.distplot(bern,
                 kde=False)
ax.set(xlabel='Bernoulli Distribution', ylabel='Frequency')
```

<a id='Binomial Distribution'></a>
### Binomial Distribution

**Definition**: The binomial distribution is a fundamental probability distribution in statistics, commonly used for modeling the number of successes in a fixed number of independent trials, each with the same probability of success. This distribution is particularly relevant in scenarios where there are exactly two mutually exclusive outcomes, often labeled as "success" and "failure".

The binomial distribution is based on a fixed number of trials, denoted as $n$. Each trial is independent, meaning the outcome of one trial does not affect the outcome of another.

In each trial, there is a constant probability of success, denoted as $p$. Correspondingly, the probability of failure in each trial is $1−p$.

It is a discrete distribution, with possible values ranging from 0 (no successes in any trial) to $n$ (success in every trial).

The probability of getting exactly k successes in n independent Bernoulli trials is given by: 

$$p(k,n,p) = (^n_k) p^k (1-p)^{n-k}$$

where

$$(^n_k) = \frac{n!}{k!\;(n-k)!}$$

**Applications**: Some common applications are in the domain of sports, educational testing, polling etc.

**Methods**: 
- Parameters: p (probability of success), n (number of trials)
- First Moment (Mean): μ = np
- Estimate p by p̂ = x̄ / n, where x̄ is the sample mean.


```python
# size = number of experiments
# n = number of trials (number of trials for each experiment)
# p = probability that one of two events is going to take place
binomial_data = stats.binom.rvs(size=10000, n=10, p=0.5, random_state=0)
pd.Series(binomial_data).plot(kind="hist", bins = 50)
```

<a id='Uniform Distribution'></a>
### Uniform Distribution
**Definition**: The Uniform distribution is one of the simplest probability distributions, characterized by equal probability for all outcomes within a defined range.

It can be both continuous (over an interval) and discrete (over a set of outcomes).

All outcomes in the specified range are equally likely.

In its continuous form, it is defined by two parameters, a and b, which are the minimum and maximum values of the distribution, respectively.

**Applications**: It is ideal for modeling scenarios where each outcome is equally likely, such as rolling a fair die (discrete) or random number generation within a range (continuous). Often used in simulations and as a basis for generating random numbers.

**Methods**: 
- Parameters: a (minimum), b (maximum)
- First Moment (Mean): μ = (a + b) / 2
- Second Moment (Variance): σ² = (b - a)² / 12
- Estimate a and b by solving the equations μ̂ = (â + b̂) / 2 and σ̂² = (b̂ - â)² / 12 where μ̂ and σ̂² are the sample mean and variance.


```python
import pymc as pm
import matplotlib.pyplot as plt
import seaborn as sns

# Define a model context
with pm.Model() as model:
    # Define a Uniform distribution
    y = pm.Uniform("y", lower=0, upper=1)

    # Sample from the prior distribution
    prior = pm.sample_prior_predictive(samples=1000)

# Extract samples
samples = prior.prior["y"].values.flatten()

# Plotting
sns.histplot(samples, bins=30, kde=False)
plt.title('Histogram of Samples from a Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

<a id='Poisson Distribution'></a>
### Poisson Distribution
**Definition**: The Poisson distribution is a discrete distribution used to model the number of times an event occurs in a fixed interval of time or space.

It models events that occur independently. Rate of Occurrence is defined by a single parameter λ (lambda), the average rate (or frequency) at which events occur.
It is typically right-skewed, especially for small values of λ.


**Probability *mass* function**: 

$$Pr(X=x) = f(x\;|\; \lambda)$$

<div style="font-size: 120%;">  
$$Pr(X=x)=\frac{e^{-\lambda}\lambda^x}{x!}$$
</div>

* $X=\{0,1,2,\ldots\}$ 
* $\lambda > 0$

x is the number of occurance.

**Applications**: It is widely used for modeling count-based data, such as the number of emails received in an hour or the number of cars passing through a checkpoint. It is very useful in various fields like telecommunication, astronomy, and biology.

**Methods:** 
- Parameter: λ (rate parameter)
- First Moment (Mean): μ = λ
- Estimate λ by the sample mean, λ̂ = x̄.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set the lambda parameter for the Poisson distribution
lambda_param = 3  # This can be any non-negative number

# Generate a sample of 1000 points from a Poisson distribution
samples = np.random.poisson(lambda_param, 1000)

# Plotting
plt.hist(samples, bins=range(min(samples), max(samples) + 1, 1), align='left')
plt.title('Histogram of Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.xticks(range(min(samples), max(samples) + 1, 1))
plt.show()
```

<a id='Gamma Distribution'></a>
### Gamma Distribution
**Definition**: The Gamma distribution is a continuous probability distribution with wide applications, particularly in scenarios involving waiting times or lifespan modeling.

It can take on various shapes, making it versatile.It is defined by two parameters: a shape parameter k and a scale parameter θ.
It generalizes other distributions like exponential and chi-squared distributions.
Applications:

**Applications**: Commonly used in modeling waiting times, such as the time until the next earthquake or the lifespan of a machine component.
Useful in finance for modeling risk and returns over time, and in hydrology for modeling rainfall events.
Probability Density Function:

**Probability Distribution Function**: Its PDF is more complex, involving the gamma function. It varies significantly based on the shape and scale parameters, allowing it to model a wide range of phenomena.

Gamma Distribution:

<div style="font-size: 120%;">  
$$f(x) = \frac{\beta^\alpha x^{\alpha -1} e^{- \beta x}}{\Gamma (\alpha)}$$
</div>

where $\Gamma()$ interpolates the factorial function to non-integer values. 
The gamma distribution is a two-parameter ($\alpha, \beta$) family of continuous probability distributions.

**Methods:**
- Parameters: k (shape), θ (scale)
- First Moment (Mean): μ = kθ
- Second Moment (Variance): σ² = kθ²
- Solve the system of equations μ̂ = k̂θ̂ and σ̂² = k̂θ̂² for k̂ and θ̂ using the sample mean μ̂ and variance σ̂².


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parameters for the gamma distribution
shape_parameter = 2  # Shape parameter (k)
scale_parameter = 2  # Scale parameter (theta)
size = 1000  # Number of random variables to draw

# Generate random variables from the gamma distribution
rvs = gamma.rvs(a=shape_parameter, scale=scale_parameter, size=size)

# Plotting
plt.figure(figsize=(8, 6))
plt.hist(rvs, bins=30, density=True)
plt.title('Histogram of Gamma Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()

```

<a id='Normal Distribution'></a>
### Normal Distribution
**Definition:** The Normal distribution, also known as the Gaussian distribution, is a fundamental probability distribution in statistics and is often considered the most important distribution in the field. It's characterized by its bell-shaped curve and is fully described by two parameters: the mean (μ) and the standard deviation (σ).

The Normal distribution is symmetric around its mean. This means that the left half of the distribution is a mirror image of the right half.

In a Normal distribution, the mean, median, and mode are all the same and are located at the center of the distribution.

The distribution has a distinctive bell-shaped curve, where the probability density is highest at the mean and decreases as you move away from the mean.

The mean (μ) is the central tendency of the distribution.
The standard deviation (σ) measures the spread or dispersion of the distribution. A larger standard deviation means a wider spread.

**Application:** Application of Normal distribution vary over a wide domain. Many natural phenomena are normally distributed, especially those that are the result of many small, independent effects, such as measurement errors or heights of people. The Normal distribution is used extensively in hypothesis testing, confidence interval estimation, and regression analysis.

**Probability Density Function:**
The probability density function (PDF) of the Normal distribution is given by the formula:

Its probability density function (pdf): 

For continuous $X$,

$$Pr(x \le X \le x + dx) = f(x|\theta)dx \, \text{ as } \, dx \rightarrow 0$$

<div style="font-size: 110%;">  
$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp\left[-\frac{(x-\mu)^2}{2\sigma^2}\right]$$
</div>

* $X \in \mathbf{R}$
* $\mu \in \mathbf{R}$
* $\sigma>0$

**Methods:**
- Parameters: μ (mean), σ² (variance)
- First Moment (Mean): μ
- Second Moment (Variance): σ²
- Estimate μ by the sample mean μ̂ and σ² by the sample variance σ̂².



```python
# size = number of data points
# loc = mean
# scale = standard deviation
normal_dist = stats.norm.rvs(size=100000, loc=200, scale=9, 
                             random_state=0) 

pd.Series(normal_dist).plot(kind="hist", bins=100)
```

Lets plot normal distribution for varying means

```python
for mean in range(400, 900, 200):
    normal_dist = stats.norm.rvs(size=10000, loc=mean, scale=90, 
                             random_state=0) 
    pd.Series(normal_dist).plot(kind="hist", bins=100, 
                                label="Mean = {:f}".format(mean))
    plt.xlabel('Distribution', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title("Normal Distribution varying the mean")
    plt.legend()
```

## Distribution Transformation

Transformation of data is a critical step in data preprocessing, especially in statistical modeling and analysis. Applying transformations can be essential for various reasons, and it can significantly impact the outcome and effectiveness of statistical methods and machine learning models. Here are some of the primary reasons why transformation should be applied:

1. **Normality Assumption**:
   Many statistical techniques assume that the data follows a normal distribution. Transformations can help in making skewed data more symmetric, thereby approximating a normal distribution. This is important for methods like ANOVA, linear regression, and t-tests, which assume normality.

2. **Variance Stabilization**:
   Some data exhibit heteroscedasticity, where the variance is not constant across the range of data. Transformations can stabilize the variance, making it more uniform across the data range, which is a key assumption in many statistical models.

3. **Linear Relationships**:
    Linear regression and some other techniques assume a linear relationship between variables. Transforming variables can sometimes reveal a linear relationship where the original variables show a non-linear relationship.

4. **Handling Outliers**:
    Data with significant outliers can skew the results of statistical analysis. Transformations can reduce the impact of outliers by decreasing the variability or spread of problematic data points.

5. **Improving Model Fit and Performance**:
   Transformations can lead to improvements in the performance of machine learning models by reshaping data distribution, making it easier for the models to learn and make predictions.

6. **Scale Compatibility**:
     In datasets with variables of different scales, transformations can bring variables to a comparable scale, which is crucial for methods like K-nearest neighbors (KNN) or Principal Component Analysis (PCA) that are sensitive to the scale of data.

## Types of Transformation

There are various transformation techniques, lets discuss few of them with the help of a dataset.
The dataset used for this exercise is Boston Housing Prices. The dataset can be imported directly from `Scikit-learn`. 
Lets try to apply different transformation techniques to the AveRooms column:

```python
from sklearn.datasets import fetch_california_housing

cali = fetch_california_housing()
x = cali['data']
y = cali['target']
cols = cali['feature_names']

df = pd.DataFrame(x, columns = cols)
df['Price'] = y
df.head()
```

```python
df['AveRooms'].skew()
```

This value indicates the AveRooms is heavily right skewed. Lets plot the histogram to visualize.

```python
sns.distplot(df['AveRooms'])
```

Let's apply different transformation to handle skewed data.
<a id='Log Transformation'></a>
### Log Transform
This is used to transform skewed data into a more normal distribution. It is particularly useful for data with exponential growth or multiplicative noise. It can be easily done via Numpy, just by calling the log() function on the desired column

```python
AveRooms_log = np.log(df['AveRooms'])
AveRooms_log.skew()
```

```python
sns.distplot(AveRooms_log)
```

The distribution is now normalised and now the various statistical techniques which require normalised data, can be applied to it!

Let's try other methods of transformation for comparision.

<a id='Square-Root Transformation'></a>
### Square-Root Transformation

The square root is a transformation with a moderate effect on distribution shape: it is weaker than the logarithm and the cube root. It is also used for reducing right skewness, and also has the advantage that it can be applied to zero values.

```python
AveRooms_sqr = np.sqrt(df['AveRooms'])
AveRooms_sqr.skew()
```

```python
sns.distplot(AveRooms_sqr)
```

It is somewhat normalised but still skewed as here the range is just the square-root of the values therefore resulting in smaller but not completely normalised variables.

<a id='Box-Cox Transformation'></a>
### Box-Cox Transformation

The Box-Cox transform , named after statisticians George Box and Sir David Roxbee Cox who collaborated on a 1964 paper and developed the technique, is the classical way for handling skewed data. To use it, the data must be positive.

At the core of the Box Cox transformation is an exponent, $λ$, which varies from -5 to 5. All values of $λ$ are considered and the optimal value for your data is selected; The *optimal value* is the one which results in the best approximation of a normal distribution curve.

The Box-Cox transform can be imported from `Scipy`.

```python
from scipy import stats

AveRooms_boco = stats.boxcox(df['AveRooms'])[0]
pd.Series(AveRooms_boco).skew()
```

```python
sns.distplot(AveRooms_boco)
```

This is a better transformation as the skewness measure is closest to 0. The value, however, is negative, but compared to the log transformation(0.8680) the Box-Cox transformation results in a more symmetrical, more normalised plot. 

<a id='Central Limit Theorem'></a>
## Central Limit Theorem

**Definition**: The Central Limit Theorem states that if you take a sufficiently large sample size from a population with a finite level of variance, the distribution of the sample means will approximate a normal distribution (Gaussian distribution), regardless of the shape of the population distribution.

The larger the sample size, the closer the distribution of the sample mean will be to a normal distribution. The general rule of thumb is that a sample size of 30 or more is considered sufficient for the CLT to hold, but this can vary depending on the original distribution of the data.

The samples must be independent of each other. This means the selection of one sample should not influence the selection of other samples.

**Implications:**

1. Predictibility: The CLT allows for making predictions about the behavior of sample means, even when the underlying population distribution is not normal.

2. Statistical Analysis: It is foundational for many statistical methods, including hypothesis testing and confidence intervals. The assumption of normality for sample means (under the CLT) is crucial for these techniques.

3. Error Reduction: It explains why averaging several measurements of the same quantity reduces variation.

**Applications:**

 Sampling: When taking random samples from a population and calculating means or sums, the CLT provides a basis for assuming that these calculations will follow a normal distribution.

 Hypothesis Testing: In hypothesis testing, especially concerning means, the CLT is used to justify the assumption of normality, which underpins the calculation of confidence intervals and the execution of significance tests.

 Quality Control: In manufacturing and quality control, the CLT is used to understand the distribution of sample means for monitoring product quality.


The Central Limit Theorem is a cornerstone of statistics because it justifies the use of normal probability models in many scenarios, even when the underlying data are not normally distributed. This theorem is vital for understanding sampling distributions and for making inferences about population parameters based on sample statistics.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the simulation
population_size = 10000  # Size of the population
sample_sizes = [1, 10, 30, 50, 100]  # Different sample sizes to illustrate CLT
num_samples = 1000  # Number of samples to take for each sample size

# Create a population with a non-normal distribution (e.g., exponential)
population = np.random.exponential(scale=1.0, size=population_size)

# Function to take samples and calculate means
def sample_means(sample_size, num_samples):
    return [np.mean(np.random.choice(population, size=sample_size, replace=True)) for _ in range(num_samples)]

# Plotting
plt.figure(figsize=(15, 8))

for i, sample_size in enumerate(sample_sizes, 1):
    means = sample_means(sample_size, num_samples)
    plt.subplot(2, 3, i)
    plt.hist(means, bins=20, color='blue', edgecolor='black', density=True)
    plt.title(f'Sample size: {sample_size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```

**Visualisation:**
To visualise the CLT, let us consider the above plots:
Sample Size 1: With a sample size of just 1, the distribution of sample means is the same as the population distribution, which is exponential in this case.

Increasing Sample Sizes (10, 30, 50, 100): As the sample size increases, the distribution of sample means becomes more symmetric and starts to resemble a normal distribution, even though the underlying population distribution is not normal. This is the essence of the Central Limit Theorem.

Trend Towards Normality: You can observe that with larger sample sizes, the histograms become more bell-shaped and centered around the population mean. This trend towards a normal distribution is a key aspect of the CLT.

# Worked Example 1

The dataset nashville_precip.txt contains NOAA precipitation data for the city of Nashville, Tennessee, measured since 1871. 

## Data Exploration

Let's plot the histogram of precipitation of each month to visualise the data.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

precip = pd.read_table("nashville_precip.txt", index_col=0, na_values='NA', delim_whitespace=True)
precip.head()
```

```python
_ = precip.hist(sharex=True, sharey=True, grid=False)
plt.tight_layout()
```

From the graphs above we can draw the following conclusions:
The first step is recognizing what sort of distribution to fit our data to. A couple of observations:

1. The data is skewed, with a longer tail to the right than to the left
2. The data is positive-valued, since they are measuring rainfall
3. The data is continuous

This indicates that the data follows a Gamma distribution


## Data Cleansing

The next step is to clean the data for any null values or any inconsistent values. Since this is a precipitation dataset, it's highly unlikely that the rainfall for a particular month would be 0. 

The missing value can be filled by using average of the months.

```python
precip.fillna(value={'Oct': precip.Oct.mean()}, inplace=True)
precip
```

Lets now perform Parameter Estimation

```python
precip_mean = precip.mean()
precip_var = precip.var()
precip_mean , precip_var
```

Calculate the alpha and beta parameters and plot the parameters

```python
alpha_mom = (precip_mean**2)/precip_var
beta_mom = precip_var/precip_mean
alpha_mom, beta_mom
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats.distributions import gamma

precip.Jan.hist(density=True, bins=20)
plt.plot(np.linspace(0, 10), gamma.pdf(np.linspace(0, 10), alpha_mom[0], beta_mom[0]))
```

```python
axs = precip.hist(density=True, figsize=(12, 8), sharex=True, sharey=True, bins=15, grid=True)

for ax in axs.ravel():
    
    # Get month
    m = ax.get_title()
    
    # Plot fitted distribution
    x = np.linspace(*ax.get_xlim())
    ax.plot(x, gamma.pdf(x, alpha_mom[m], beta_mom[m]))
    
    # Annotate with parameter estimates
    label = 'alpha = {0:.2f}\nbeta = {1:.2f}'.format(alpha_mom[m], beta_mom[m])
    ax.annotate(label, xy=(10, 0.2))
    
plt.tight_layout()
```

# Worked Example 2

Lets perform the parameter estimation for the California Housing Data set above.

## Applying Box-Cox Transformation

Lets first transform the data to normalised form to reduce the heavy skewness



```python
from sklearn.datasets import fetch_california_housing
from scipy import stats

cali = fetch_california_housing()
x = cali['data']
y = cali['target']
cols = cali['feature_names']

df = pd.DataFrame(x, columns = cols)
df['Price'] = y
df.head()


AveRooms_boco = stats.boxcox(df['AveRooms'])[0]
pd.Series(AveRooms_boco).skew()
```

```python
import matplotlib.pyplot as plt

plt.hist(AveRooms_boco, bins=30)  
plt.xlabel('AveRooms')  
plt.ylabel('Density')
plt.title('Histogram of AveRooms_boco')  
plt.show()
```

Since this distribution is normal, for parameter estimation we need mean and standard deviation.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  


plt.hist(AveRooms_boco, bins=30, density=True)  

# Now plotting the normal distribution
mean = np.mean(AveRooms_boco)
std_dev = np.std(AveRooms_boco)
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100) 
plt.plot(x, norm.pdf(x, mean, std_dev))

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of AveRooms_boco with Normal Distribution Overlay')
plt.show()
```
