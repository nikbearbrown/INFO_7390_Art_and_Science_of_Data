### Explanation of Key AI Concepts for Week 1 Lab 1

1. **Pre-training Models and Its Importance vs. Fine-tuning**

   * **Model Pre-training**: This process involves training a model on a large dataset to gain a broad understanding of the language or domain before it is fine-tuned for specific applications.
   
   * **Value of Extended Pre-training**: By continuing the pre-training phase, the model can better adapt to the subtle characteristics of a particular dataset or domain, improving its efficacy and depth of understanding.
   
   * **Fine-tuning**: Refers to the adjustment of a pre-trained model to perform specific tasks or to process particular data types. This phase requires considerably less data and computational power compared to the pre-training phase.
   
   * **Pre-training vs. Fine-tuning**: While pre-training equips the model with a comprehensive base of knowledge, fine-tuning tailors this knowledge to meet specific needs. The decision to continue pre-training or to commence fine-tuning depends on the task's specificity and the resources available.

2. **Terminology and Explanations**

   * **Generative AI**: This AI discipline specializes in generating new content, including text, images, and music, by learning from existing data to produce novel outputs that mimic the style and structure of the learned data.
   
   * **Large Language Models (LLMs)**: These substantial models in natural language processing (NLP) are capable of understanding, processing, and generating text that closely resembles human communication. They are termed 'large' due to their extensive parameter counts and the significant data volumes required for training.
   
   * **Prompt**: In LLMs, a prompt is the initial input used to provoke a response from the model. It could be a question, a statement, or any textual element that sets the context for the model's output.
   
   * **Transformer Architecture**: Central to many LLMs, this architecture processes sequential data like text, not sequentially, but by considering all parts of the data simultaneously (parallel processing) and employing attention mechanisms to prioritize different segments of the input.

3. **Lifecycle of LLM-Based Generative AI Models**

   * **Data Collection**: Assembling extensive datasets pertinent to the models proposed utility.
   
   * **Model Pre-training**: Developing a general comprehension of the language or task by training on a broad dataset.
   
   * **Fine-tuning**: Specializing the pre-trained model for particular tasks or narrower datasets.
   
   * **Evaluation**: Measuring the models effectiveness and implementing necessary refinements.
   
   * **Deployment**: Applying the model in practical, real-world scenarios.
   
   * **Limiting Factors**: These encompass data availability and quality, computational resources, ethical concerns, and the specific demands of the application.

4. **Computational Obstacles in Model Pre-training**

   * **Extensive Datasets and Large Models**: The pre-training phase demands substantial computational resources and memory due to the size of the data and the complexity of the model architectures.
   
   * **Reduction of Memory Footprint**: Techniques include model pruning (eliminating non-critical connections), quantization (using lower precision for numerical representations), and optimizing data handling and processing.

5. **Scaling Laws in LLMs**

   * **Definition**: These laws describe the relationship between the scale of the training data, the computational resources (e.g., model parameters), and the performance of the model.
   
   * **Observed Trends**: Studies indicate that increasing the dataset size and model parameters generally enhances performance, though with reduced marginal gains. This insight is crucial for allocating resources during model development, balancing dataset size, model complexity, and computational budgets.
   
   * **Inference Demands**: Larger models also necessitate greater computational power for inference, which influences their deployment in practical settings.

These sections provide a structured overview of Generative AI, focusing on Large Language Models, highlighting their foundational principles, developmental stages, and the dynamics influencing their efficiency and application.

### In-Depth Exploration of AI Principles for Week 2 Lab 2

1. **Refinement through Instructional Prompt Datasets**

   - **Concept**: Refinement through instructional prompts involves enhancing a pre-trained model by using datasets specifically curated with prompts or directives. These are structured to steer the model toward mastering certain tasks or response styles.
   
   - **Performance Enhancement**: Training with these targeted datasets sharpens the model's ability to interpret and react to similar prompts in practical situations. This focused training significantly bolsters the model's task-specific performance, boosting both its efficiency and accuracy.

2. **Catastrophic Forgetting in Neural Networks**

   - **Definition**: Catastrophic forgetting is a challenge in neural networks where a model loses previously acquired knowledge from its initial training after being fine-tuned with new data or tasks. This issue is a major hurdle in the ongoing learning journey of AI systems.
   
   - **Mitigation Techniques**:
     - **Elastic Weight Consolidation (EWC)**: This method involves impeding the learning process on certain neural weights that are crucial for past tasks.
     - **Experience Replay**: This technique periodically reintroduces a mix of old and new data during training to help retain earlier knowledge.
     - **Incremental Learning**: This strategy introduces new information gradually, enabling the model to preserve former knowledge while adapting to new inputs.

3. **Parameter-Efficient Fine Tuning (PEFT)**

   - **Definition**: PEFT is a strategy for optimizing fine-tuning by adjusting only a limited set of the models parameters. This method aims to enhance computational efficiency during the fine-tuning phase.
   
   - **Advantages**: PEFT lessens the computational burden and memory usage compared to conventional fine-tuning methods that modify a broader range of parameters. This makes it especially beneficial for deploying large models in resource-constrained environments.

4. **PEFTs Role in Minimizing Computational Costs and Counteracting Catastrophic Forgetting**

   - **Reducing Computational Costs**: By altering fewer parameters, PEFT demands less computational power and memory, streamlining the fine-tuning process to be more resource-efficient.
   
   - **Addressing Catastrophic Forgetting**: PEFT's selective parameter modification helps maintain the general knowledge acquired during pre-training, thus reducing the impact of catastrophic forgetting.

5. **Enhancing LLM Performance through Fine-Tuning with Instructional Prompt Datasets**

   - **Concept**: This method employs datasets equipped with specific prompts or tasks during the fine-tuning phase to refine the models capabilities.
   
   - **Boosting Performance**: By training on prompts similar to those expected in real-world applications, the model becomes adept at handling specific queries or tasks, significantly enhancing its effectiveness in these areas.
   
   - **Practical Uses**: This specialized fine-tuning is invaluable in scenarios where precise responses are crucial, such as in automated customer support, creative content generation, or targeted data analysis tools.

These segments elucidate advanced strategies for enhancing AI model capabilities, particularly in Large Language Models, while also tackling major challenges like catastrophic forgetting and managing computational demands.

### Explanation of Advanced AI Concepts for Week 3 Lab 3

1. **Reinforcement Learning from Human Feedback (RLHF) in Large Language Models**

   - **Concept**: RLHF is a technique that enhances the capabilities and alignment of Large Language Models (LLMs) by integrating human feedback into the training regimen.
   
   - **Enhancing Performance and Alignment**: Through RLHF, human feedback is utilized to either reward or penalize the model depending on the relevance and quality of its outputs. This process helps the model to better comprehend and meet human expectations, thereby boosting its ability to generate responses that are accurate and contextually suitable.

2. **Incorporating Human Labeler Data in RLHF**

   - **Data Collection**: Outputs produced by the LLM are presented to human labelers, who assess them based on criteria such as accuracy, relevance, and appropriateness.
   
   - **Developing a Reward Model**: The assessments from human labelers are used to develop a reward model. This model is trained to assess the quality of LLM outputs using human feedback as a benchmark.
   
   - **Utilization in RLHF**: The reward model that has been trained is then employed to provide feedback to the LLM during its training phase, directing it to produce outputs that are more likely to receive positive evaluations from humans.

3. **Chain-of-Thought Prompting**

   - **Definition**: Chain-of-thought prompting is a technique where LLMs are given prompts that encourage them to follow a sequential reasoning process rather than producing an immediate response.
   
   - **Enhancing Reasoning and Planning**: This method teaches the model to tackle problems methodically, mimicking human-like logical progression. This significantly boosts the models capability in handling complex reasoning and planning tasks, resulting in outputs that are more logical and coherent.

4. **Addressing Knowledge Cut-offs in LLMs**

   - **The Challenge**: Knowledge cut-offs are limitations faced by LLMs related to the use of information beyond their training data. Since LLMs are trained on static datasets, they cannot access or integrate new information post-training.
   
   - **Mitigating Strategies**:
     - **Information Retrieval Techniques**: Implementing systems that enable LLMs to pull in information from external sources can ensure they remain updated with the latest knowledge and facts.
     - **Data Augmentation**: Continuously refreshing the training datasets with new information, or employing methods to emulate fresh data, helps maintain the relevancy and informed status of the LLMs.
     - **Hybrid Approaches**: Integrating LLMs with other technologies, such as search engines or databases, to access real-time information can significantly improve their efficacy in tasks that demand contemporary knowledge.

These discussions delve into sophisticated methods and strategies within AI, specifically aimed at advancing the functionality, precision, and contextual relevance of Large Language Models in varied and evolving environments.

### 2.2 **Methods**

````python
def print_number_of_trainable_model_parameters(model):
    trainable_model_params = 0
    all_model_params = 0
    for _, param in model.named_parameters():
        all_model_params += param.numel()
        if param.requires_grad:
            trainable_model_params += param.numel()
    return f"trainable model parameters: {trainable_model_params}\nall model parameters: {all_model_params}\npercentage of trainable model parameters: {100 * trainable_model_params / all_model_params:.2f}%"

def tokenize_function(example):
    start_prompt = 'Summarize the following conversation.\n\n'
    end_prompt = '\n\nSummary: '
    prompt = [start_prompt + dialogue + end_prompt for dialogue in example["dialogue"]]
    example['input_ids'] = tokenizer(prompt, padding="max_length", truncation=True, return_tensors="pt").input_ids
    example['labels'] = tokenizer(example["summary"], padding="max_length", truncation=True, return_tensors="pt").input_ids

    return example
````

## 6. Dataset Preprocessing: Dialogue and Summary

We will explore the concept of Parameter Efficient Fine-Tuning (PEFT), a more resource-efficient alternative to traditional full fine-tuning. PEFT includes various methods such as Low-Rank Adaptation (LoRA) and prompt tuning, which is different from prompt engineering.

PEFT often utilizes Low-Rank Adaptation (LoRA).

LoRA allows for the fine-tuning of a model using considerably fewer computational resources, often manageable with just a single GPU. After fine-tuning for a specific task, application, or client using LoRA, the core Language Model (LLM) remains intact, while a new "LoRA adapter" is developed. This adapter is significantly smaller than the full LLM, typically only a fraction of the original model's size (measured in megabytes rather than gigabytes).

During inference, however, this LoRA adapter must be combined back with the original LLM to process the inference requests effectively. The benefit here is that multiple LoRA adapters can share the same base LLM, which can substantially reduce the memory footprint when managing multiple tasks or applications.

### 6.1 PEFT/LoRA Model Configuration for Fine-Tuning

To set up the PEFT/LoRA model for fine-tuning with a new parameter adapter, we undertake the following steps:

1. **PEFT/LoRA Initialization**:
   - We utilize PEFT/LoRA, which entails freezing the base Language Model (LLM) and concentrating on training the adapter alone.

2. **Adapter Setup**:
   - Configure the LoRA adapter by setting the `rank (r)` hyper-parameter, which defines the rank or dimensionality of the adapter to be trained.

By implementing PEFT/LoRA, we maintain the integrity of the original LLM while introducing a dedicated parameterized layer tailored for a specific task or application. The `rank (r)` hyper-parameter is pivotal in shaping the adapters complexity and its effectiveness for the designated task.

````python
lora_config = LoraConfig(
    r=32, # Rank
    lora_alpha=32,
    target_modules=["q", "v"],
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM # FLAN-T5
)
````

Integrate LoRA adapter layers and parameters into the base Language Model (LLM) for the training process.

````python
peft_model = get_peft_model(original_model,
                            lora_config)
print(print_number_of_trainable_model_parameters(peft_model))
````

### 6.2 Training the PEFT/LoRA Adapter

````python
output_dir = f'./peft-dialogue-summary-training'

peft_training_args = TrainingArguments(
    output_dir=output_dir,
    auto_find_batch_size=True,
    learning_rate=1e-3, # Higher learning rate than full fine-tuning.
    num_train_epochs=1,
    logging_steps=1,
    max_steps=1
)

peft_trainer = Trainer(
    model=peft_model,
    args=peft_training_args,
    train_dataset=tokenized_datasets["train"],
)
````

````python
peft_trainer.train()

peft_model_path="./peft-dialogue-summary-checkpoint-local"

peft_trainer.model.save_pretrained(peft_model_path)
tokenizer.save_pretrained(peft_model_path)
````

````python
peft_model_base = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base", torch_dtype=torch.bfloat16)
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")

peft_model = PeftModel.from_pretrained(peft_model_base,
                                       peft_model_path,
                                       torch_dtype=torch.bfloat16,
                                       is_trainable=False)
````

````python
def print_number_of_trainable_model_parameters(model):
    trainable_model_params = 0
    all_model_params = 0
    for _, param in model.named_parameters():
        all_model_params += param.numel()
        if param.requires_grad:
            trainable_model_params += param.numel()
    return f"\ntrainable model parameters: {trainable_model_params}\nall model parameters: {all_model_params}\npercentage of trainable model parameters: {100 * trainable_model_params / all_model_params:.2f}%"
````

## 3. Data Tokenization


The subsequent step is to preprocess the dataset. We will choose a subset of the data, filter the dialogues to a specific length to balance readability with the retention of significant content, and then prepend each dialogue with an instruction prior to tokenization. The token IDs will be stored under the input_ids field, and the decoded prompts will be saved in the query field.

To facilitate this process, it's recommended to implement a function named build_dataset. This function can be outlined as follows:

````python
def build_dataset(model_name,
                  dataset_name,
                  input_min_text_length,
                  input_max_text_length):

    """
    Preprocess the dataset and split it into train and test parts.

    Parameters:
    - model_name (str): Tokenizer model name.
    - dataset_name (str): Name of the dataset to load.
    - input_min_text_length (int): Minimum length of the dialogues.
    - input_max_text_length (int): Maximum length of the dialogues.

    Returns:
    - dataset_splits (datasets.dataset_dict.DatasetDict): Preprocessed dataset containing train and test parts.
    """

    # load dataset (only "train" part will be enough for this lab).
    dataset = load_dataset(dataset_name, split="train")

    # Filter the dialogues of length between input_min_text_length and input_max_text_length characters.
    dataset = dataset.filter(lambda x: len(x["dialogue"]) > input_min_text_length and len(x["dialogue"]) <= input_max_text_length, batched=False)

    # Prepare tokenizer. Setting device_map="auto" allows to switch between GPU and CPU automatically.
    tokenizer = AutoTokenizer.from_pretrained(model_name, device_map="auto")

    def tokenize(sample):

        # Wrap each dialogue with the instruction.
        prompt = f"""
Summarize the following conversation.

{sample["dialogue"]}

Summary:
"""
        sample["input_ids"] = tokenizer.encode(prompt)

        # This must be called "query", which is a requirement of our PPO library.
        sample["query"] = tokenizer.decode(sample["input_ids"])
        return sample

    # Tokenize each dialogue.
    dataset = dataset.map(tokenize, batched=False)
    dataset.set_format(type="torch")

    # Split the dataset into train and test parts.
    dataset_splits = dataset.train_test_split(test_size=0.2, shuffle=False, seed=42)

    return dataset_splits

dataset = build_dataset(model_name=model_name,
                        dataset_name=huggingface_dataset_name,
                        input_min_text_length=200,
                        input_max_text_length=1000)

print(dataset)
````

## 5. Developing a Reward Model for Reinforcement Learning

**Reinforcement Learning (RL)** is a crucial area of machine learning where agents operate within an environment to maximize cumulative rewards. These agents are directed by a **policy** that governs their decision-making, with the primary goal of RL being to develop an optimal or nearly optimal policy that enhances the **reward function**.

Originally, the policy was based on the instruct PEFT model, which is the Language Model (LLM) prior to any detoxification efforts. One method previously used involved engaging human labelers to assess the toxicity levels of the models outputs, a strategy that can be excessively expensive if extended throughout the entire fine-tuning process. A cost-effective alternative is to establish a reward model that motivates the agent to generate detoxified dialogue summaries.

A practical strategy is to conduct sentiment analysis on the outputs of the model, sorting them into two categories: `nothate` and `hate`. A higher reward is granted when there is a greater probability of an output being classified as `nothate`.

For this purpose, we will utilize [Meta AI's RoBERTa-based hate speech model](https://huggingface.co/facebook/roberta-hate-speech-dynabench-r4-target) as our reward model. This evaluator can efficiently compute the toxicity levels of the dialogues. To achieve this, you'll need to supply several crucial components: the test dataset (`dataset["test"]`), the tokenizer utilized earlier, the previously frozen PEFT model, and the toxicity evaluator. For a more organized and systematic approach, it's advisable to encapsulate these essential steps within a dedicated function called `evaluate_toxicity`.

````python
def evaluate_toxicity(model,
                      toxicity_evaluator,
                      tokenizer,
                      dataset,
                      num_samples):

    """
    Preprocess the dataset and split it into train and test parts.

    Parameters:
    - model (trl model): Model to be evaluated.
    - toxicity_evaluator (evaluate_modules toxicity metrics): Toxicity evaluator.
    - tokenizer (transformers tokenizer): Tokenizer to be used.
    - dataset (dataset): Input dataset for the evaluation.
    - num_samples (int): Maximum number of samples for the evaluation.

    Returns:
    tuple: A tuple containing two numpy.float64 values:
    - mean (numpy.float64): Mean of the samples toxicity.
    - std (numpy.float64): Standard deviation of the samples toxicity.
    """

    max_new_tokens=100

    toxicities = []
    input_texts = []
    for i, sample in tqdm(enumerate(dataset)):
        input_text = sample["query"]

        if i > num_samples:
            break

        input_ids = tokenizer(input_text, return_tensors="pt", padding=True).input_ids

        generation_config = GenerationConfig(max_new_tokens=max_new_tokens,
                                             top_k=0.0,
                                             top_p=1.0,
                                             do_sample=True)

        response_token_ids = model.generate(input_ids=input_ids,
                                            generation_config=generation_config)

        generated_text = tokenizer.decode(response_token_ids[0], skip_special_tokens=True)

        toxicity_score = toxicity_evaluator.compute(predictions=[(input_text + " " + generated_text)])

        toxicities.extend(toxicity_score["toxicity"])

    # Compute mean & std using np.
    mean = np.mean(toxicities)
    std = np.std(toxicities)

    return mean, std
````

### 6.2 Model Fine-Tuning

The fine-tuning loop consists of several pivotal steps:

1. Obtain responses to queries from the Policy Language Model (PEFT model).
2. Assess the sentiment associated with both the queries and responses using the hate speech RoBERTa model.
3. Employ Proximal Policy Optimization (PPO) to optimize the policy with a triplet of inputs, including the query, response, and the corresponding reward.

You can ensure the smooth operation of the process by monitoring key metrics, including:

- `objective/kl`: Aim to minimize the Kullback-Leibler (KL) divergence.
- `ppo/returns/mean`: Strive to maximize the mean returns.
- `ppo/policy/advantages_mean`: Seek to maximize the mean advantages.

These metrics provide valuable insights into the progress of the training process and the attainment of specific objectives within the fine-tuning loop.

````python
output_min_length = 100
output_max_length = 400
output_length_sampler = LengthSampler(output_min_length, output_max_length)

generation_kwargs = {
    "min_length": 5,
    "top_k": 0.0,
    "top_p": 1.0,
    "do_sample": True
}

reward_kwargs = {
    "top_k": None, # Return all scores.
    "function_to_apply": "none", # You want the raw logits without softmax.
    "batch_size": 16
}

max_ppo_steps = 10

for step, batch in tqdm(enumerate(ppo_trainer.dataloader)):
    # Break when you reach max_steps.
    if step >= max_ppo_steps:
        break

    prompt_tensors = batch["input_ids"]

    # Get response from FLAN-T5/PEFT LLM.
    summary_tensors = []

    for prompt_tensor in prompt_tensors:
        max_new_tokens = output_length_sampler()

        generation_kwargs["max_new_tokens"] = max_new_tokens
        summary = ppo_trainer.generate(prompt_tensor, **generation_kwargs)

        summary_tensors.append(summary.squeeze()[-max_new_tokens:])

    # This needs to be called "response".
    batch["response"] = [tokenizer.decode(r.squeeze()) for r in summary_tensors]

    # Compute reward outputs.
    query_response_pairs = [q + r for q, r in zip(batch["query"], batch["response"])]
    rewards = sentiment_pipe(query_response_pairs, **reward_kwargs)

    # You use the `nothate` item because this is the score for the positive `nothate` class.
    reward_tensors = [torch.tensor(reward[not_hate_index]["score"]) for reward in rewards]

    # Run PPO step.
    stats = ppo_trainer.step(prompt_tensors, summary_tensors, reward_tensors)
    ppo_trainer.log_stats(stats, batch, reward_tensors)

    print(f'objective/kl: {stats["objective/kl"]}')
    print(f'ppo/returns/mean: {stats["ppo/returns/mean"]}')
    print(f'ppo/policy/advantages_mean: {stats["ppo/policy/advantages_mean"]}')
    print('-'.join('' for x in range(100)))
````

### 6.3 Quantitative Model Evaluation

````python
mean_after_detoxification, std_after_detoxification = evaluate_toxicity(model=ppo_model,
                                                                        toxicity_evaluator=toxicity_evaluator,
                                                                        tokenizer=tokenizer,
                                                                        dataset=dataset["test"],
                                                                        num_samples=10)
print(f'toxicity [mean, std] after detox: [{mean_after_detoxification}, {std_after_detoxification}]')
````

Compare the toxicity scores between the reference model (before detoxification) and the fine-tuned model (after detoxification).

### 6.4 Evaluate the Model Qualitatively

Examine sample examples from the test dataset to compare the initial `ref_model` with the fine-tuned/detoxified `ppo_model` using the toxicity evaluator.

````python
batch_size = 20
compare_results = {}

df_batch = dataset["test"][0:batch_size]

compare_results["query"] = df_batch["query"]
prompt_tensors = df_batch["input_ids"]

summary_tensors_ref = []
summary_tensors = []

# Get response from ppo and base model.
for i in tqdm(range(batch_size)):
    gen_len = output_length_sampler()
    generation_kwargs["max_new_tokens"] = gen_len

    summary = ref_model.generate(
        input_ids=torch.as_tensor(prompt_tensors[i]).unsqueeze(dim=0).to(device),
        **generation_kwargs
    ).squeeze()[-gen_len:]
    summary_tensors_ref.append(summary)

    summary = ppo_model.generate(
        input_ids=torch.as_tensor(prompt_tensors[i]).unsqueeze(dim=0).to(device),
        **generation_kwargs
    ).squeeze()[-gen_len:]
    summary_tensors.append(summary)

# Decode responses.
compare_results["response_before"] = [tokenizer.decode(summary_tensors_ref[i]) for i in range(batch_size)]
compare_results["response_after"] = [tokenizer.decode(summary_tensors[i]) for i in range(batch_size)]

# Sentiment analysis of query/response pairs before/after.
texts_before = [d + s for d, s in zip(compare_results["query"], compare_results["response_before"])]
rewards_before = sentiment_pipe(texts_before, **reward_kwargs)
compare_results["reward_before"] = [reward[not_hate_index]["score"] for reward in rewards_before]

texts_after = [d + s for d, s in zip(compare_results["query"], compare_results["response_after"])]
rewards_after = sentiment_pipe(texts_after, **reward_kwargs)
compare_results["reward_after"] = [reward[not_hate_index]["score"] for reward in rewards_after]
````
