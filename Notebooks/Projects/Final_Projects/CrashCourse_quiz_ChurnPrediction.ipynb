## "Understanding Data" Assignment 1 Crash Course in Data Quiz Questions

1. What is the primary objective of a churn prediction model in a subscription-based service?

a. Maximize revenue
b. Minimize customer acquisition costs
c. Identify potential customers for upselling
d. Retain customers by predicting likely churn

**Correct Answer: d. Retain customers by predicting likely churn**

2. Which of the following is a crucial consideration when preprocessing data for churn prediction?

a. Data anonymization
b. Feature scaling
c. Removing all missing values
d. Increasing dataset size through duplication

**Correct Answer: b. Feature scaling**

3. In churn prediction, what type of machine learning algorithm is commonly used for binary classification tasks?

a. Linear Regression
b. Decision Trees
c. K-Means Clustering
d. Naive Bayes Classifier

**Correct Answer: b. Decision Trees**

4. What metric is often used to evaluate the performance of a churn prediction model on imbalanced data?

a. Precision
b. Accuracy
c. F1 Score
d. Mean Absolute Error

**Correct Answer: c. F1 Score**

5. How does a time-series component enhance a churn prediction model?

a. It increases feature dimensionality
b. It captures evolving patterns over time
c. It reduces training time
d. It simplifies model interpretation

**Correct Answer: b. It captures evolving patterns over time**

6. What challenge does imbalanced data pose in churn prediction?

a. Slows down model training
b. May lead to biased predictions
c. Reduces the need for feature engineering
d. Improves model generalization

**Correct Answer: b. May lead to biased predictions**

7. Which of the following is an essential consideration for deploying a churn prediction model in a production environment?

a. Model interpretability
b. Using a large training dataset
c. Ignoring data privacy concerns
d. Exclusively relying on offline evaluation metrics

**Correct Answer: a. Model interpretability**

8. What role do retention interventions play in the context of a churn prediction model?

a. They are irrelevant to the model's predictions
b. They are implemented before model training
c. They are strategies to prevent customer churn
d. They are part of the feature engineering process

**Correct Answer: c. They are strategies to prevent customer churn**

9. How can feature importance in a churn prediction model be determined?

a. By randomly selecting features
b. Using correlation coefficients
c. Through model-specific techniques like permutation importance
d. Ignoring feature importance altogether

**Correct Answer: c. Through model-specific techniques like permutation importance**

10. What potential challenge should be addressed when dealing with privacy concerns in churn prediction?

a. Share customer data openly with third parties
b. Use anonymized data without any precautions
c. Implement data encryption and access controls
d. Ignore privacy concerns for better model performance

**Correct Answer: c. Implement data encryption and access controls**
