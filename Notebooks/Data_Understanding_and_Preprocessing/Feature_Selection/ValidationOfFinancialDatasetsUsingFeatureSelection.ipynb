# **The Theory behind it -**

The theory behind validating financial datasets to ensure their consistency and integrity involves several key concepts and principles:

1. **Data Validation and Quality:**
   - **Data Quality Frameworks:** Theoretical frameworks like DAMA (Data Management Association) or the Data Quality Assessment Framework provide guidelines for assessing and improving data quality.
   - **Data Validation Techniques:** Various techniques, including range checks, consistency checks, and cross-validation with external sources, are used to verify the accuracy and completeness of data.

2. **Financial Data Analysis:**
   - **Financial Ratios:** Theoretical concepts related to financial ratios, such as liquidity ratios, profitability ratios, and leverage ratios, underpin financial analysis. Validating data ensures that these ratios are calculated correctly.
   - **Time Series Analysis:** Financial data often involves time-series analysis, where theories of time-series modeling, stationarity, and autocorrelation play a role in validating data for trends and patterns.

3. **Data Governance and Documentation:**
   - **Data Governance:** Theoretical principles of data governance, including data ownership, data stewardship, and data lineage, are essential for maintaining data integrity.
   - **Metadata Management:** Metadata, which includes information about the dataset's structure and origin, is critical for understanding and validating data.

4. **Anomaly Detection:**
   - **Statistical Methods:** Theoretical foundations of statistical methods, such as the z-score, are used for identifying anomalies in financial data.
   - **Machine Learning:** Theoretical principles of machine learning, including supervised and unsupervised learning algorithms, are applied for automated anomaly detection.

5. **Regulatory Compliance:**
   - **Regulatory Knowledge:** Understanding financial regulations and accounting standards (e.g., GAAP, IFRS) is crucial for validating financial datasets to ensure compliance.

6. **Financial Domain Knowledge:**
   - **Accounting Principles:** Knowledge of accounting principles, such as double-entry bookkeeping, is fundamental for ensuring data consistency in financial datasets.
   - **Business Rules:** Theoretical understanding of business rules specific to the financial industry informs the validation process.

7. **Error Identification and Resolution:**
   - **Error Handling Theories:** Theoretical concepts related to error handling and resolution strategies, such as error codes and reconciliation processes, are applied when identifying and correcting data errors.

8. **Data Auditing and Logging:**
   - **Auditing Theories:** Theoretical principles of auditing and logging are relevant for tracking changes to financial datasets and ensuring their traceability.

In summary, the theory behind validating financial datasets draws from a diverse set of principles and concepts, encompassing data quality, financial analysis, data governance, statistical methods, regulatory compliance, domain knowledge, error handling, and auditing. Combining these theoretical foundations helps researchers and practitioners ensure the integrity and consistency of financial data, ultimately supporting sound financial analysis and decision-making.

#**Features Selection And PCA**

Feature selection is an important part of predictive modelling. Features compression is often done using Principle Component Analysis (PCA). PCA is widely used in quantitative finance.

**Examples in Financial Datasets-**


1- An investment portfolio of bonds with future cash flows is sensitive to changes in interest rates for different maturities. If we desire to estimate portfolio risk using a smaller number of factors we can use PCA. By performing PCA on historical interest rate moves for the relevant set of maturities, one can select the first n factors explaining most of the variation in the data.



2- Another example is in foreign exchange. The pricing and risk management of foreign exchange derivatives uses volatility surface. A volatility surface of a currency pair shows how implied volatilities vary by moneyness/profitability and maturities.



3- A good example of using PCA in financial risk is to reduce a volatility surface structure in the maturity dimension to a single factor that is most responsible for variation in profit and loss. Value-at-risk is a form of market risk capitalization.


**What is Principal Component Analysis(PCA)**

This is feature extraction technique meant to reduce the dimensions of our dataset.

The main idea of principal component analysis (PCA) is to reduce the dimensionality of a data set consisting of many variables correlated with each other,while retaining the variation present in the dataset, up to the maximum extent.


PCA helps to identify the correlation and dependencies among the features in a data set. A covariance matrix expresses the correlation between the different variables in the data set. It is essential to identify heavily dependent variables because they contain biased and redundant information which reduces the overall performance of the model.

#**Feature Selection Using PCA**

Principal Component Analysis (PCA) can be a valuable tool in validating our data and gaining insights into its quality and structure in several ways:

1. **Dimensionality Reduction:** PCA reduces the dimensionality of dataset while retaining most of the variance. By visualizing data in a lower-dimensional space (e.g., two or three dimensions)

2. **Outlier Detection:** Outliers are data points that deviate significantly from the norm. PCA can help identify outliers by showing data points that are far from the main cluster in the PCA plot.

3. **Data Consistency:** Inconsistencies in data collection or recording can lead to high variability. PCA can highlight inconsistencies by revealing patterns or clusters in the data.

4. **Feature Importance:** By examining the loading values of each original variable in the principal components, we can identify which variables contribute most to the variation in the data. This can help us prioritize variables for further analysis or feature selection.

6. **Visualizing Relationships:** PCA allows us to visualize relationships between data points and variables in a more interpretable way.

7. **Identifying Redundant Variables:** If certain principal components capture little variance or have low loading values for all original variables, it may suggest that some of our variables are redundant or carry little information. Removing redundant variables can simplify your analysis and potentially improve model performance.

8. **Exploratory Data Analysis (EDA):** PCA can be an essential part of our exploratory data analysis toolkit, helping us understand the structure and characteristics of our data before diving into more advanced analyses.

In summary, PCA is a powerful technique for data validation and exploration. By reducing dimensionality and revealing patterns and relationships in your data, it can assist in identifying issues, assessing data quality, and providing insights that guide further analysis and decision-making.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_columns)

num_components = 2
pca = PCA(n_components=num_components)

pca_result = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
explained_variance_ratio = pca.explained_variance_ratio_
print(" Variance Ratios:", explained_variance_ratio)

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Results')
plt.grid()
plt.show()
```

The variance ratios we provided indicate how much of the total variance in our data is explained by each of the principal components. In this case:


**The first principal component (PC1) explains approximately 23.00% of the total variance.**

**The second principal component (PC2) explains approximately 12.37% of the total variance.**


These ratios can help us to decide how many principal components to retain.

If we want to retain a certain amount of variance (e.g., 95% of the variance), we can calculate the cumulative explained variance and choose the number of components that achieve that level of explained variance.

```python
cumulative_variance_ratio = explained_variance_ratio.cumsum()
print("Cumulative Explained Variance Ratios:", cumulative_variance_ratio)
target_variance_ratio = 0.95
num_components_needed = len(cumulative_variance_ratio[cumulative_variance_ratio < target_variance_ratio]) + 1
print("Number of Components to Explain 95% Variance:", num_components_needed)
```

This code calculates the cumulative explained variance ratios and finds the number of components needed to explain 95% of the variance. We can adjust the target_variance_ratio to your desired level of explained variance.






#Selecting Features

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_columns)

num_components = 2
pca = PCA(n_components=num_components)
pca.fit(scaled_data)
loadings = np.abs(pca.components_)
loadings_df = pd.DataFrame(loadings, columns=numeric_columns.columns)
print("PCA Loadings:")
print(loadings_df)

loadings = np.abs(pca.components_)
feature_importance = loadings[0] + loadings[1]
sorted_features = np.argsort(feature_importance)[::-1]
k = 5
selected_features = numeric_columns.columns[sorted_features[:k]]
print("Selected Features:", selected_features)
```

```python
X_pca = df[selected_features]
y = df['52WeekHigh'].astype(float)
```

#Calculating Mean Squared Error (MSE)and R-squared (R2) value

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
model = DecisionTreeRegressor(random_state=42)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

#**Feature Selection Using RFE**

RFE - Recursive Feature Elimination (RFE) is a feature selection technique that recursively removes the least important features from the dataset based on a chosen machine learning model's feature importance scores.

Why Is RFE USED?

In practice, RFE is often used in combination with various machine learning algorithms and models to optimize the feature set and improve model performance.

We employ RFE to gain insights into the most important variables, refine their models, and make them more robust for real-world applications.


```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
X = numeric_columns.drop(columns=['52WeekHigh'])
y = df['52WeekHigh']
model = LinearRegression()

num_features_to_select = 5
rfe = RFE(estimator=model, n_features_to_select=num_features_to_select)
rfe.fit(X, y)
selected_features = X.columns[rfe.support_]
print("Selected Features:")
print(selected_features)
```

#**How do you ensure data quality when working with complex financial datasets?**

How do we ensure data quality when working with complex financial datasets?

Here are some tips and best practices to follow.

**Define data quality criteria**

Before we start working with any dataset, we need to define what data quality means for our specific project and context. Data quality criteria may include aspects such as accuracy, completeness, validity, consistency, timeliness, and relevance.

**Perform data profiling and assessment**

Data profiling is the process of examining the structure, content, and metadata of a dataset to understand its characteristics, quality, and potential issues.


Data assessment is the process of measuring the actual quality of a dataset against the predefined criteria and standards. Data assessment can help us quantify and prioritize data quality issues, and determine their root causes and impacts.

**Implement data cleaning and transformation**

Data cleaning is the process of detecting and correcting data quality issues, such as errors, duplicates, inconsistencies, or missing values.

**Use data quality tools and automation**

Data quality tools are software applications that can help us automate and streamline data quality tasks, such as data profiling, assessment, cleaning, and transformation.

**Apply data quality checks and controls**

Data quality checks and controls are methods and procedures that can help us ensure and maintain data quality throughout the data lifecycle.

Data quality checks and controls can include aspects such as data quality policies, roles, responsibilities, rules, standards, guidelines, processes, workflows, documentation, reviews, approvals, feedback, testing, verification, validation, monitoring, reporting, and improvement.



**Learn from data quality issues and feedback**

Data quality is not a one-time activity, but a continuous process that requires constant learning and improvement. We should always track and analyze data quality issues and feedback, and use them as opportunities to enhance our data quality skills and practices.

**We perform anomaly detection using two approaches:**

Statistical Methods: We calculate z-scores for the 'Amount' column and flag data points as anomalies if their z-score exceeds a predefined threshold.
Machine Learning (Isolation Forest): We use the Isolation Forest algorithm to detect anomalies in the 'Amount' column. Adjust the contamination parameter based on our dataset's characteristics.
We display the detected anomalies for both methods.

Optionally, we visualize the anomalies detected using the Isolation Forest algorithm.

Z- Scores Method

```python
mean_amount = df['Credit amount'].mean()
std_amount = df['Credit amount'].std()
threshold = 2.0
df['Z-Score'] = (df['Credit amount'] - mean_amount) / std_amount
df['Anomaly'] = df['Z-Score'].abs() > threshold
anomalies = df[df['Anomaly']]
print("Anomalies:")
print(anomalies)
```

Isolation Method

```python
from sklearn.ensemble import IsolationForest
contamination = 0.05
X = df[['Credit amount']]
model = IsolationForest(contamination=contamination, random_state=42)
model.fit(X)
df['Anomaly'] = model.predict(X)
anomalies = df[df['Anomaly'] == -1]
print("Anomalies:")
print(anomalies)
```

```python
def get_outliers(df, feature, iqr_threshold=1.5):
    q1 = np.percentile(df[feature], 25)
    q3 = np.percentile(df[feature], 75)
    iqr = q3 - q1
    return df[(df[feature] < (q1 - iqr_threshold * iqr)) | (df[feature] > (q3 + iqr_threshold * iqr))]

numeric_features = df_pre.select_dtypes(include=[int, float]).columns.to_list()
for feature in numeric_features:
    nouts = len(get_outliers(df_pre, feature))
    if nouts > 0:
        print("%s: %d" % (feature, nouts))
```

6.**Reducing Data Collection Costs:** If certain features are expensive or time-consuming to collect, feature selection can help you focus resources on collecting only the most critical data, potentially saving costs.

7.**Risk Assessment:** In the context of financial risk assessment, feature selection can help identify factors that are strongly associated with high-risk individuals or transactions. This information is valuable for making lending decisions and managing financial risk.
