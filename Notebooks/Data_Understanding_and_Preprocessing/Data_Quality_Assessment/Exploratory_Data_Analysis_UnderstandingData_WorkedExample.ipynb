# **Worked Example**



The Iris flower data set or Fisher's Iris data set is a multivariate data set used and made famous by the British statistician and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis. It is sometimes called Anderson's Iris data set because Edgar Anderson collected the data to quantify the morphologic variation of Iris flowers of three related species. Two of the three species were collected in the Gasp√© Peninsula "all from the same pasture, and picked on the same day and measured at the same time by the same person with the same apparatus".

The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters. Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species from each other. Fisher's paper was published in the Annals of Eugenics (today the Annals of Human Genetics)

# **Handling Outliers**

Outliers are data points that exhibit significant deviations from the overall pattern of a dataset, often lying at a considerable distance from the majority of the data. Identifying and addressing outliers is essential in both statistical analysis and machine learning, as they can distort results and potentially lead to inaccurate interpretations. Outliers may arise due to various factors, including measurement errors, data entry mistakes, or genuinely rare occurrences within the population.

One widely used method for outlier detection is based on the interquartile range (IQR). The IQR is the range between the first quartile (Q1) and the third quartile (Q3) of the dataset. Outliers are typically identified as values falling below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR. The formula for calculating the

IQR is given by IQR = Q3 - Q1.

````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
sns.set(style="whitegrid")

# Create box plots for each feature
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(2, 6, i+1)
    sns.boxplot(y=df[column], color='skyblue')

plt.suptitle('Box Plots for Outlier Detection', y=1.02, fontsize=16)
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=1.0)
plt.show()
````

# **Removing Outliers**

To address outliers, the removal process involves identifying the specific entries within the dataset that meet the outlier criteria based on the detection method employed. It is imperative to note that each outlier detection method generates a list of data items that fulfill the outlier definition. Subsequently, the identified outliers can be excluded from the dataset by deleting or filtering out the corresponding entries

````python
import numpy as np
import pandas as pd
import seaborn as sns
# Function to remove outliers based on IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (df[column] < lower_bound) | (df[column] > upper_bound)
    return df[~outliers]

# Specify the column for outlier removal
column_to_check = 'sepal_width'

# Remove outliers
df_no_outliers = remove_outliers(df, column_to_check)

# Print the shape before and after removing outliers
print("Old Shape: ", df.shape)
print("New Shape: ", df_no_outliers.shape)

# Create a boxplot to visualize the distribution after outlier removal
sns.boxplot(x=column_to_check, data=df_no_outliers)
````

**Violin Plot for Distribution Analysis**

````python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

features = ["petal_width", "petal_length", "sepal_length", "sepal_width"]
for feature, ax in zip(features, axes.flatten()):
    sns.violinplot(y=feature, x="species", data=df, inner="quartile", ax=ax)

plt.show()
````

**Observations from Violin Plots:**

- Setosa exhibits a narrower distribution with lower density in both petal length and width.

- Versicolor displays an average distribution and features for both petal length and width.

- Virginica showcases a broader distribution with numerous values and features, particularly in sepal length and width.

- High-density regions in the plots correspond to mean/median values. For instance, Iris Setosa has the highest density around 5.0 cm for sepal length, which aligns with the median value (5.0) from the dataset table.


**Histogram and Probability Density Function (PDF) Plot**

````python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
plt.figure(figsize=(15, 10))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=df, x=feature, hue="species", kde=True)
    plt.title(f'Distribution of {feature} by Species')
    plt.xlabel(feature)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
````

# Observations from the Plots:

1. **Plot 1:**
   - Indicates a substantial overlap among species concerning sepal length, making it an ineffective classification feature.

2. **Plot 2:**
   - Reveals even higher overlap in species with respect to sepal width, rendering it an ineffective classification feature.

3. **Plot 3:**
   - Highlights petal length as a robust classification feature, exhibiting clear separation among species.
   - Minimal overlap is observed, particularly between Versicolor and Virginica, while Setosa is distinctly separated from the other two.

4. **Plot 4:**
   - Underscores petal width as a valuable classification feature.
   - Shows significantly less overlap, particularly between Versicolor and Virginica, with Setosa distinctly separated from the other two.

**These observations emphasize the importance of petal length and width over sepal length and width in effectively classifying the species.**
