**Handling Missing Values in Penguin Dataset: A Strategy Guide**

You have missing values in critical features of your penguin dataset: culmen length, depth, mass, and the categorical feature sex. Let's explore options for addressing these gaps through various techniques and their justifications:

Numerical Features (Culmen Length, Depth, Mass):
Mean Imputation:

Justification: Suitable for normally distributed features, replacing missing values with the average maintains central tendency.
Example: If culmen length follows a normal distribution, mean imputation offers a reasonable estimate based on typical beak length.
Drawback: May distort data if outliers are present and introduce bias towards the mean.
Median Imputation:

Justification: Robust to outliers and skewed distributions, replacing missing values with the middle value minimizes the influence of extreme values.
Example: For body mass with outliers, median imputation provides a safer estimate than mean.
Drawback: May not accurately represent central tendency with highly skewed distributions.
K-Nearest Neighbors (KNN):

Justification: A more sophisticated method predicting missing values based on similar data points in the K nearest neighbors.
Example: Captures complex relationships between features for accurate imputation, especially if distributions are non-normal.
Drawback: Requires choosing an appropriate K value and can be computationally expensive for large datasets.
Model-based Imputation:

Justification: Uses machine learning models like decision trees to predict missing values based on relationships with other features.
Example: Leverages additional predictors for more informed missing value estimates.
Drawback: Requires model training and may be susceptible to overfitting, especially with limited data.
Choosing the Right Technique:
Consider feature distribution (normality, skewness, outliers) to choose between mean, median, or advanced techniques.
Factor in data size; KNN and model-based imputation might be computationally expensive for large datasets.
Ensure compatibility with chosen analysis models.
Categorical Feature (Sex):
Mode Imputation:

Justification: Replaces missing values with the most frequent category, preserving the original distribution.
Example: Maintains observed gender proportions if "Male" is 80% and "Female" is 20%.
Drawback: May introduce bias towards the majority category.
"Missing" Category:

Justification: Creates a new category for missing values, highlighting missingness and avoiding specific gender assignment when information is absent.
Example: Allows differentiation between unknown sex and identified "Male" or "Female," useful for understanding missingness patterns.
Drawback: Increases the number of categories and requires handling the "Missing" category during analysis.
Choosing the Right Technique:
Consider analysis goals; use mode imputation if preserving gender distribution is crucial, and "Missing" category if understanding missingness is essential.
Ensure model compatibility; some algorithms may not handle missing categories directly.
Remember:
Explore reasons for missingness to inform your handling strategy.
Document your chosen technique for transparency and reproducibility.
Evaluate the impact of different techniques on analysis results.
By carefully considering these options and the specific context of your penguin data, you can effectively handle missing values and unlock reliable insights from your analysis.

```python
# before filling null values, we need to check for outliers for numerical features
# using boxplot for it
df = penguins
fig = make_subplots(rows = 2, cols = 3)
row_no = col_no = 1

for col in df.columns:
    # only numerical fields are allowed and not "Object" or "Datatime" datatype
    if(df[col].dtype != "O"):
        fig.add_trace(
            go.Box(y=df[col], name = col),
            row = row_no,
            col = col_no,
        )
        col_no += 1
        if col_no % 4 == 0: # necessary calculations for correctly presenting subplots
            row_no = 2
            col_no = 1

fig.update_layout(title_text = "Outlier detection through boxplot")
fig.show()
```

**Analysis**

No outliers were identified, enabling the use of mean imputation for numerical columns.

For categorical features, where outliers are also absent, the imputation method of choice is mode.

```python
# fill null values
for col in df.columns:
    if df[col].dtype != 'O':
        df[col].fillna(df[col].mean(), inplace = True) # using mean to fill null values because there are no outliers
    else:
        df[col].fillna(df[col].mode()[0], inplace = True)

# check whether null values have been removed or not
df.info()
```

```python
item = pd.read_csv("olist_order_items_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
orders = pd.read_csv("olist_orders_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
geolocation = pd.read_csv("olist_geolocation_dataset.csv")
sellers = pd.read_csv("olist_sellers_dataset.csv")
order_pay = pd.read_csv("olist_order_payments_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
category = pd.read_csv("product_category_name_translation.csv")
```

```python
df = orders.merge(item, on='order_id', how='left')
df = df.merge(order_pay, on='order_id', how='outer', validate='m:m')
df = df.merge(reviews, on='order_id', how='outer')
df = df.merge(products, on='product_id', how='outer')
df = df.merge(customers, on='customer_id', how='outer')
df = df.merge(sellers, on='seller_id', how='outer')

print(df.shape)
```

```python
def resumetable(df):
    print(f"Dataset Shape: {df.shape}")
    summary = pd.DataFrame(df.dtypes,columns=['dtypes'])
    summary = summary.reset_index()
    summary['Name'] = summary['index']
    summary = summary[['Name','dtypes']]
    summary['Missing'] = df.isnull().sum().values    
    summary['Uniques'] = df.nunique().values
    summary['First Value'] = df.loc[0].values
    summary['Second Value'] = df.loc[1].values
    summary['Third Value'] = df.loc[2].values

    for name in summary['Name'].value_counts().index:
        summary.loc[summary['Name'] == name, 'Entropy'] = round(stats.entropy(df[name].value_counts(normalize=True), base=2),2) 

    return summary

def cross_heatmap(df, cols, normalize=False, values=None, aggfunc=None):
    temp = cols
    cm = sns.light_palette("green", as_cmap=True)
    return pd.crosstab(df[temp[0]], df[temp[1]], 
                       normalize=normalize, values=values, aggfunc=aggfunc).style.background_gradient(cmap = cm)
resumetable
```

```python
id_cols = ['order_id', 'seller_id', 'customer_id', 'order_item_id', 'product_id', 
           'review_id', 'customer_unique_id', 'seller_zip_code_prefix']
#categorical columns
cat_cols = df.nunique()[df.nunique() <= 27].keys().tolist()
# cat_cols = [x for x in cat_cols if x not in target_col]

#numerical columns
num_cols = [x for x in df.columns if x not in cat_cols + id_cols]

#Binary columns with 2 values
bin_cols = df.nunique()[df.nunique() == 2].keys().tolist()

#Columns more than 2 values
multi_cols = [i for i in cat_cols if i not in bin_cols]
```
