# Understanding Data

## Is the data biased?

The question of whether data is biased is a critical and increasingly relevant concern in today's data-driven world. Data, often considered objective and factual, can carry hidden biases that reflect historical, societal, or methodological influences. These biases, if left unexamined, can lead to unfair or inaccurate outcomes in areas ranging from artificial intelligence and machine learning to decision-making processes in fields like healthcare, finance, and criminal justice. Understanding and addressing data bias is essential for ensuring equitable and just applications of data-driven technologies and fostering a more inclusive and unbiased future.

## Why is it intriguing?

1. <strong>Impact on Fairness:</strong> Data-driven systems and algorithms are increasingly used in critical decision-making processes, including  hiring, lending, and criminal justice. Data bias can result in unfair and discriminatory outcomes, exacerbating societal inequalities.

2. <strong>Ethical Considerations:</strong> Data bias raises ethical questions about the fairness, transparency, and accountability of algorithms and systems that rely on biased data. Understanding and addressing these biases is crucial for responsible data use.

3. <strong>AI and Machine Learning:</strong> In the realm of artificial intelligence and machine learning, data bias can significantly affect the performance and accuracy of models. Researchers and practitioners are continuously exploring ways to mitigate bias to build more equitable AI systems.

4. <strong>Social Implications:</strong> Data bias can reinforce stereotypes and perpetuate historical injustices. Recognizing and rectifying these biases are essential steps toward building a more just and inclusive society.

5. <strong>Challenges and Solutions:</strong> The complexity of data bias presents interesting challenges in data collection, preprocessing, and model development. Researchers and data scientists are actively working on innovative techniques and approaches to detect and mitigate bias in data.

6. <strong>Legal and Regulatory Frameworks:</strong> Governments and regulatory bodies are beginning to address data bias through legislation and guidelines, making it a legally relevant and evolving field of study.

7. <strong>Intersectionality:</strong> Data bias can affect different demographic groups in unique ways, making it a multifaceted issue that intersects with other social and technical considerations.

## What is the theory behind it? 

1. <strong>Sources of Bias:</strong>

    1. Selection Bias: Data may not accurately represent the entire population due to issues in the sampling process. Biases can  arise if certain groups are overrepresented or underrepresented.
    2. Measurement Bias: Biases can result from errors or inaccuracies in data collection methods or instruments, leading to skewed or incorrect data.
    3. Historical Bias: Data can inherit biases from historical practices, cultural norms, or systemic inequalities, which persist over time.
    4. Algorithmic Bias: Machine learning algorithms can perpetuate or exacerbate biases present in the training data, leading to unfair or discriminatory outcomes.
    5. Data Imbalance: In classification tasks, data imbalance occurs when one class significantly outnumbers the others. This can lead to models that perform poorly on minority classes.
    
2. <strong>Types of Bias:</strong>

    1. Sampling Bias: When the sample used for data collection is not representative of the larger population.
    2. Non-response Bias: Arises when certain individuals or groups are less likely to respond to surveys or data collection efforts.
    3. Observer Bias: Occurs when the person or system collecting data has preconceived notions or expectations that influence data collection.
    4. Reporting Bias: Involves the selective reporting of data, often unintentional, where some information is omitted or emphasized more than others.
    
3. <strong>Consequences of Bias:</strong>

    1. Unfair Outcomes: Biased data can lead to unfair and discriminatory decisions, particularly in applications like lending, hiring, and criminal justice.
    2. Inaccurate Predictions: Biased data can result in inaccurate predictions or classifications in machine learning models.
    3. Reinforcing Stereotypes: Biased data can reinforce existing stereotypes and perpetuate societal inequalities.
    4. Ethical and Legal Issues: Data bias can raise ethical concerns and legal challenges related to discrimination and fairness.
    
4. <strong>Bias Mitigation:</strong>

    1. Various techniques and approaches can be employed to mitigate bias in data, including re-sampling, re-weighting, fairness-aware algorithms, and bias-aware preprocessing methods.
    2. Ethical guidelines, transparency, and accountability measures can also help reduce the impact of bias in decision-making processes.
    
5. <strong>Intersectionality:</strong>

    1. Data bias is often complex and can intersect with multiple dimensions, including race, gender, age, and socioeconomic status. Understanding how these biases interact is crucial for comprehensive bias mitigation.
    
6. <strong>Continual Learning:</strong>

    1. The theory behind addressing data bias is an evolving field as new challenges and techniques emerge. Researchers and practitioners continually work on advancing the theory and practice of bias detection and mitigation.

## How to determine data bias?

1. Examine Data Sources: Start by understanding the sources of your data. Bias can be introduced at the data collection stage, so knowing where the data comes from is crucial.

2. Conduct Exploratory Data Analysis (EDA): Use statistical and visualization techniques to explore your dataset. Look for any patterns, imbalances, or anomalies that could indicate bias. Pay attention to the distribution of data across different groups or categories.

3. Compare to External Benchmarks: Compare your dataset to external benchmarks or reference data to see if it aligns with expectations. This can help identify discrepancies that might be indicative of bias.

4. Consult Domain Experts: Seek input from domain experts who can provide insights into potential sources of bias specific to your dataset and application.

5. Check for Missing Data: Analyze missing data patterns. Biases can arise if certain groups are more likely to have missing data, which can affect the representativeness of your dataset.

6. Assess Data Collection Methods: Evaluate the methods used to collect the data. Consider the survey questions, sampling techniques, and data collection tools for potential sources of bias.

7. Consider Historical and Social Context: Recognize historical and societal factors that might have influenced the data. Historical biases can persist in datasets, especially if they reflect past inequalities.

8. Utilize Bias Metrics: Use quantitative metrics and fairness measures to assess bias systematically. Common metrics include disparate impact, equal opportunity, and demographic parity.

9. Visualize Bias: Create visualizations that highlight disparities or biases within the data. Heatmaps, bar charts, and histograms can be useful for visualizing group-specific patterns.

10. Test Model Predictions: If your data is used to train machine learning models, test the models for bias in their predictions. Evaluate whether the models make consistent errors across different groups.

11. Run Controlled Experiments: In some cases, running controlled experiments with randomized data can help reveal biases by comparing outcomes under different conditions.

12. Collect Diverse Data: If you suspect that your dataset is biased, consider collecting additional data from underrepresented groups to improve representation and reduce bias.

13. External Audits: Engage third-party auditors or experts to conduct an external audit of your data and processes to identify bias.

14. Continuous Monitoring: Data bias is not a one-time issue; it should be continuously monitored as new data is collected and used. Implement ongoing monitoring and auditing processes.

15. Document Bias Findings: Keep a detailed record of your bias assessment findings, including the methods used and any actions taken to address bias. This documentation can be crucial for transparency and accountability.
Remember that data bias can be complex and multifaceted, and it may require a combination of approaches to fully understand and mitigate. It's essential to approach bias assessment systematically and collaboratively, involving domain experts and stakeholders throughout the process.

## Data Generation Code

The code generates a synthetic dataset that represents customer demographics for different locations using numpy.
It defines various locations, such as "University Area," "Factory Area," and more, as categories for customer data.
For each location, the code simulates customer data by generating random ages and genders.
The mean age and gender distribution for each location are predefined to introduce variability.
Customer data, including location, age, and gender, is collected and organized into a Pandas DataFrame.
The resulting dataset is saved to a CSV file for further analysis or experimentation.

```python
import pandas as pd
import numpy as np

# Define locations
locations = ["University Area", "Female Clothing Store", "Factory Area", "Shopping Mall", "Residency", "Hospital", "Park"]

# Generate synthetic data with defined probabilities of male and female, and defining the mean age wrt the location
data = []

for location in locations:
    num_customers = np.random.randint(50, 200)
    
    if location == "University Area":
        mean_age = 25
        gender_bias = {"Male": 0.6, "Female": 0.4}
    elif location == "Female Clothing Store":
        mean_age = 35
        gender_bias = {"Male": 0.05, "Female": 0.95}
    elif location == "Factory Area":
        mean_age = 45
        gender_bias = {"Male": 0.7, "Female": 0.3}
    elif location == "Shopping Mall":
        mean_age = 30
        gender_bias = {"Male": 0.4, "Female": 0.6}
    elif location == "Residency":
        mean_age = 55
        gender_bias = {"Male": 0.4, "Female": 0.6}
    elif location == "Hospital":
        mean_age = 45
        gender_bias = {"Male": 0.45, "Female": 0.55}
    elif location == "Park":
        mean_age = 30
        gender_bias = {"Male": 0.5, "Female": 0.5}
    
    ages = np.random.normal(mean_age, 5, num_customers)
    genders = np.random.choice(list(gender_bias.keys()), num_customers, p=list(gender_bias.values()))
    
    for i in range(num_customers):
        data.append([location, ages[i], genders[i]])

# Create a DataFrame
df = pd.DataFrame(data, columns=["Location", "Age", "Gender"])

# Save the dataset to a CSV file
df.to_csv("customer_data.csv", index=False)
```

### Code to determine and visualize data bias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("customer_data.csv")

# Define expected gender distribution for each location
expected_gender_distribution = {
    "University Area": {"Male": 0.6, "Female": 0.4},
    "Female Clothing Store": {"Male": 0.05, "Female": 0.95},
    "Factory Area": {"Male": 0.7, "Female": 0.3},
    "Shopping Mall": {"Male": 0.4, "Female": 0.6},
    "Residency": {"Male": 0.4, "Female": 0.6},
    "Hospital": {"Male": 0.45, "Female": 0.55},
    "Park": {"Male": 0.5, "Female": 0.5}
}

# Define expected age mean for each location
expected_age_mean = {
    "University Area": 25,
    "Female Clothing Store": 35,
    "Factory Area": 45,
    "Shopping Mall": 30,
    "Residency": 55,
    "Hospital": 45,
    "Park": 30
}

# Calculate gender distribution bias
gender_bias = df.groupby("Location")["Gender"].value_counts(normalize=True).unstack().fillna(0)
gender_bias = gender_bias.divide(gender_bias.sum(axis=1), axis=0)

# Calculate age distribution bias
age_bias = df.groupby("Location")["Age"].mean()

# Visualize gender bias
# Define colors for male and female
colors = ['b', 'g']  # Blue for Male, Green for Female

# Visualize gender bias
# Bar chart for gender distribution
# Define colors for male and female
colors = ['b', 'g']  # Blue for Male, Green for Female

# Visualize gender bias
plt.figure(figsize=(12, 6))
x = np.arange(len(locations))
width = 0.2

for i, location in enumerate(locations):
    expected_ratios = expected_gender_distribution[location]
    observed_ratios = gender_bias.loc[location]
    
    for j, (gender, expected_ratio) in enumerate(expected_ratios.items()):
        observed_ratio = observed_ratios[gender]
        color = colors[j]  # Assign color based on gender
        plt.bar(x[i] + j * width, observed_ratio, width, label=f"Observed {gender} Ratio ({location})", alpha=0.7, color=color)
        plt.text(x[i] + j * width - 0.02, observed_ratio + 0.01, f"{observed_ratio:.2f}", fontsize=8)

plt.xticks(x + width, locations)
plt.title("Gender Distribution by Location (Bar Chart)")
plt.xlabel("Location")
plt.ylabel("Gender Ratio")

plt.show()


# Visualize age bias
plt.figure(figsize=(12, 6))
plt.bar(age_bias.index, age_bias.values, label="Observed Age Mean")
plt.plot(expected_age_mean.keys(), list(expected_age_mean.values()), 'ro-', label="Expected Age Mean")

plt.title("Age Distribution Bias by Location")
plt.xlabel("Location")
plt.ylabel("Age Mean")
plt.legend()
plt.show()

```

### Using Machine Learning to calculate accuracies and bias

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.utils import resample

# Load the dataset
df = pd.read_csv("customer_data.csv")

# Define expected gender distribution for each location
expected_gender_distribution = {
    "University Area": {"Male": 0.6, "Female": 0.4},
    "Female Clothing Store": {"Male": 0.05, "Female": 0.95},
    "Factory Area": {"Male": 0.7, "Female": 0.3},
    "Shopping Mall": {"Male": 0.4, "Female": 0.6},
    "Residency": {"Male": 0.4, "Female": 0.6},
    "Hospital": {"Male": 0.45, "Female": 0.55},
    "Park": {"Male": 0.5, "Female": 0.5}
}

# Initialize variables to store accuracy values
accuracy_biased = 0.0
accuracy_balanced = 0.0

# Loop through each location for analysis
for location in expected_gender_distribution.keys():
    # Subset the data for the current location
    location_data = df[df["Location"] == location]

    # Extract features and target
    X = location_data[["Age"]]
    y = location_data["Gender"]

    # Train a logistic regression model (biased)
    model = LogisticRegression()
    model.fit(X, y)

    # Test the model
    y_pred = model.predict(X)
    accuracy_biased += accuracy_score(y, y_pred)

    # Apply bias mitigation (Re-sampling to balance gender representation)
    # We'll balance the dataset by resampling the minority class (Female) to match the majority class (Male)
    minority_class = location_data[location_data["Gender"] == "Female"]
    majority_class = location_data[location_data["Gender"] == "Male"]

    # Upsample the minority class
    minority_upsampled = resample(minority_class, replace=True, n_samples=len(majority_class), random_state=0)

    # Combine the upsampled minority class with the majority class
    balanced_data = pd.concat([majority_class, minority_upsampled])

    X_balanced = balanced_data[["Age"]]
    y_balanced = balanced_data["Gender"]

    # Train a logistic regression model on the balanced data (bias mitigated)
    model_balanced = LogisticRegression()
    model_balanced.fit(X_balanced, y_balanced)

    # Test the mitigated model
    y_pred_balanced = model_balanced.predict(X)
    accuracy_balanced += accuracy_score(y, y_pred_balanced)

# Calculate average accuracy over all locations
num_locations = len(expected_gender_distribution)
accuracy_biased /= num_locations
accuracy_balanced /= num_locations

print(f"Average Accuracy (Biased Model): {accuracy_biased:.2f}")
print(f"Average Accuracy (Bias Mitigated Model): {accuracy_balanced:.2f}")
```

This code applies the bias mitigation technique by resampling the minority class (Female) to balance the dataset and then trains a logistic regression model on the balanced data. It calculates the average accuracy for both the biased and mitigated models across all locations.
