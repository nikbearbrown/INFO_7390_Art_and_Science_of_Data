1. Which of the following are key components of establishing causality in a study?

   a) Random assignment  
   b) Correlation  
   c) Control group  
   d) Manipulation of variables  

   **Correct Answer:** a) Random assignment, c) Control group, d) Manipulation of variables  
   **Explanation:** Random assignment helps ensure that participants have an equal chance of being assigned to different experimental conditions, control group allows comparison between treatment and non-treatment conditions, and manipulation of variables ensures that changes in the independent variable cause changes in the dependent variable.

2. What is the purpose of a control group in experimental research?

   a) To provide a baseline for comparison  
   b) To manipulate independent variables  
   c) To ensure correlation  
   d) To eliminate random assignment  

   **Correct Answer:** a) To provide a baseline for comparison  
   **Explanation:** A control group is essential in experimental research to provide a baseline against which the effects of the treatment group can be compared.

3. Which of the following are threats to internal validity in an experiment?

   a) Selection bias  
   b) Observer bias  
   c) History  
   d) Maturation  

   **Correct Answer:** a) Selection bias, c) History, d) Maturation  
   **Explanation:** Selection bias occurs when participants in different groups are not equivalent at the outset, history refers to external events that may influence the results, and maturation refers to changes within participants over time.

4. Which of the following are criteria for establishing causality according to Hill's criteria?

   a) Consistency  
   b) Specificity  
   c) Temporality  
   d) Coherence  

   **Correct Answer:** a) Consistency, b) Specificity, c) Temporality  
   **Explanation:** Hill's criteria include consistency (reproducibility of findings), specificity (specific cause leads to specific effect), and temporality (cause precedes effect).

5. In an observational study, which of the following techniques can help control for confounding variables?

   a) Randomization  
   b) Matching  
   c) Placebo control  
   d) Double-blind procedure  

   **Correct Answer:** b) Matching  
   **Explanation:** Matching involves selecting participants in such a way that they are similar on potential confounding variables, thus reducing the impact of these variables on the outcome.

6. Which of the following statements accurately describes the difference between correlation and causation?

   a) Correlation implies causation  
   b) Causation implies correlation  
   c) Correlation indicates a relationship, but causation implies a directional influence  
   d) Causation indicates a relationship, but correlation implies a directional influence  

   **Correct Answer:** c) Correlation indicates a relationship, but causation implies a directional influence  
   **Explanation:** Correlation merely indicates a relationship between variables, whereas causation implies that changes in one variable directly cause changes in another.

7. Which of the following experimental designs involve testing multiple levels of an independent variable?

   a) Between-subjects design  
   b) Within-subjects design  
   c) Mixed-design  
   d) Factorial design  

   **Correct Answer:** d) Factorial design  
   **Explanation:** Factorial design involves manipulating two or more independent variables simultaneously to study their effects on the dependent variable.

8. Which of the following types of studies are most susceptible to confounding variables?

   a) Observational studies  
   b) Experimental studies  
   c) Longitudinal studies  
   d) Cross-sectional studies  

   **Correct Answer:** a) Observational studies  
   **Explanation:** Observational studies, which involve observing subjects in their natural environment without interference, are more susceptible to confounding variables compared to experimental studies where variables are controlled.

9. What is the purpose of random assignment in experimental research?

   a) To ensure that participants are similar across conditions  
   b) To ensure that participants are randomly selected  
   c) To eliminate all biases  
   d) To establish causality  

   **Correct Answer:** a) To ensure that participants are similar across conditions  
   **Explanation:** Random assignment helps ensure that participants in different experimental conditions are equivalent at the outset, reducing the impact of pre-existing differences between groups.

10. Which of the following best describes the concept of spurious correlation?

   a) A correlation that is statistically significant  
   b) A correlation that disappears after controlling for confounding variables  
   c) A correlation that is not statistically significant  
   d) A correlation that is particularly strong  

   **Correct Answer:** b) A correlation that disappears after controlling for confounding variables  
   **Explanation:** Spurious correlation refers to a relationship between two variables that appears to be statistically significant but disappears after controlling for other variables that may be influencing the relationship.
