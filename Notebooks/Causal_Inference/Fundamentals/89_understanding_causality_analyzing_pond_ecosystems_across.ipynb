**Abstract**


Causality describes the relationship between cause and effect, where one event (the cause) directly leads to another event (the effect). It is a core concept for understanding how changes in one variable influence changes in another.

Across disciplines like science, philosophy, and statistics, causality is essential for explaining observed phenomena, uncovering underlying mechanisms, and enabling predictions and interventions. It provides a framework for understanding patterns and identifying actionable insights.

In statistics and data science, causal inference focuses on identifying and estimating causal relationships between variables using observational or experimental data. This involves differentiating between correlation, which indicates an association between variables, and causation, which establishes that one variable directly impacts another.

Grasping causality is critical for making well-informed decisions, designing effective interventions, and driving progress in various fields of study.

**Common Causal Inference Models**

Causal inference models are approaches designed to uncover and estimate cause-and-effect relationships between variables using data from either observational studies or controlled experiments. These models focus on identifying whether changes in one variable directly result in changes in another, moving beyond simple associations. Key causal inference techniques include:

Propensity Score Matching (PSM): This method matches individuals with similar characteristics based on their likelihood (propensity) of receiving a treatment or intervention. By comparing treated and untreated groups, it estimates causal effects while accounting for confounding factors.

Difference-in-Differences (DID): DID compares the before-and-after changes in outcomes between a treated group and a control group. This approach helps isolate the causal effect of an intervention by controlling for trends common to both groups.

Instrumental Variables (IV): IV methods address biases like endogeneity by leveraging an external variable (instrument) that influences the treatment but has no direct impact on the outcome except through the treatment.

Regression Discontinuity Design (RDD): RDD leverages a cut-off or threshold in treatment assignment to assess causal effects, treating individuals near the threshold as comparable.

Structural Equation Modeling (SEM): SEM uses a network of equations to describe causal relationships, integrating both observed and unobserved variables to model direct and indirect effects.

Causal Trees: These models partition data into subsets based on predictors to estimate treatment effects for different groups, offering a more tailored understanding of causal relationships.

Graph-Based Models and Do-Calculus: These methods use causal diagrams, such as directed acyclic graphs (DAGs), to visually represent relationships between variables. Do-calculus provides a framework for making causal inferences in complex scenarios.

Each of these methods has specific use cases, strengths, and assumptions. Selecting the right approach depends on the nature of the study, the structure of the data, and the underlying assumptions about causality. These techniques are instrumental in fields like healthcare, economics, and social sciences, enabling data-driven insights and decisions.

Causality analysis can be applied to almost any dataset where understanding cause-and-effect relationships is valuable. The choice of dataset and approach depends on the specific research question and domain of interest.

**Causality vs Correlation**

Causality and correlation are key concepts in statistics and research, describing different types of relationships between variables.

Causality indicates that one event or variable directly influences another, establishing a cause-and-effect relationship. It implies that changes in one variable result in changes in another. Determining causality typically requires robust experimental designs or well-structured observational studies that account for potential confounding factors to identify direct influence.

Correlation, on the other hand, measures the degree to which two variables move together. It reflects the strength and direction of a linear relationship but does not imply that one variable causes the other. Correlation can occur due to various reasons, including direct or indirect causation, coincidence, or shared influences from confounding variables.

While correlation provides useful insights into associations, establishing causality demands more rigorous analysis to understand the direction and nature of the relationship.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Generate random data to simulate the dataset
# Assuming that we have data for 4 different types of water bodies

# Set random seed for reproducibility
np.random.seed(0)

# Simulate data for dissolved oxygen vs. temperature
n_samples = 100
temperatures = np.random.uniform(15, 40, n_samples * 4)  # Max temperatures
dissolved_oxygen = np.random.uniform(0, 8, n_samples * 4)  # Min dissolved oxygen

# Assign water body types
types = ['LAKE', 'TANK', 'WETLAND', 'POND']
water_body_types = np.repeat(types, n_samples)

# Create a DataFrame with the simulated data
data = pd.DataFrame({
    'Max Temperature (°C)': temperatures,
    'Min Dissolved Oxygen (mg/L)': dissolved_oxygen,
    'Water Body Type': water_body_types
})

# Plot using seaborn's lmplot to include regression lines
plt.figure(figsize=(10, 6))
lm = sns.lmplot(
    data=data,
    x='Max Temperature (°C)',
    y='Min Dissolved Oxygen (mg/L)',
    hue='Water Body Type',
    height=6,
    aspect=1.33,
    markers=['o', 's', 'X', '^'],
    palette='bright',
    legend_out=True
)

# Set the plot title and labels
lm.set(title='Dissolved Oxygen vs. Temperature by Water Body Type')
lm.set_axis_labels('Max Temperature (°C)', 'Min Dissolved Oxygen (mg/L)')

# Show the plot
plt.show()
```

```python
#code by "https://github.com/KJMAN678/casual_reasoning_lignum/blob/master/LIGNAM.ipynb

prior_knowledge = make_prior_knowledge(
    18,
    exogenous_variables = [df_dict["Type Water Body"], df_dict["State Name"]],
    paths = [
             [df_dict["pH (Max)"], df_dict["pH (Min)"],[df_dict['Temperature\n?C (Min)'],df_dict['Temperature\n?C (Max)'],df_dict['Dissolved Oxygen (mg/L) (Min)'],df_dict['Dissolved Oxygen (mg/L) (Max)']]],
             ],

)
```

The make_prior_knowledge function from the lingam.utils module is used to define prior knowledge about causal relationships between variables in a dataset. Here's a detailed explanation of its purpose and usage:

What the Function Does:

Function Call:

The make_prior_knowledge function is called with the following arguments:
Number of variables: Specifies the total number of variables in the dataset (18 in this case).

Exogenous variables: A list of indices representing exogenous variables. These variables are not influenced by any other variables in the dataset.

Paths: A list defining known causal relationships between variables. Each path is a list of variable indices, where the last variable is the target, and the preceding variables are its causal predictors.

Exogenous Variables:

The exogenous_variables list includes indices of variables considered exogenous. These are independent variables in the causal framework.
Example: Here, "Type Water Body" and "State Name" are specified as exogenous variables, meaning they are not influenced by other variables.

Defined Causal Paths:

The paths list specifies known causal links.

Example: A path is defined between "pH (Max)" and "pH (Min)". The path indicates that "pH (Min)" is influenced by:

"pH (Max)"
"Temperature (Min)"
"Temperature (Max)"
"Dissolved Oxygen (Min)"
"Dissolved Oxygen (Max)"

Purpose of Prior Knowledge:

This prior knowledge is provided to causal discovery algorithms to:

Guide the search for causal relationships.
Incorporate domain expertise into the modeling process.
Improve the accuracy and relevance of discovered causal structures.

Summary:

The make_prior_knowledge function creates a framework for specifying exogenous variables and known causal paths.
In this example, "Type Water Body" and "State Name" are exogenous, while the relationship between "pH (Max)" and "pH (Min)" and their predictors is explicitly defined.
This setup helps the causal discovery algorithm identify meaningful relationships between variables, leveraging both the data and domain knowledge.







```python
#code by "https://github.com/KJMAN678/casual_reasoning_lignum/blob/master/LIGNAM.ipynb

model = lingam.DirectLiNGAM(
                    random_state=42,
                    measure = 'pwling',
                    prior_knowledge = prior_knowledge
                    )

model.fit(df_en)
```
