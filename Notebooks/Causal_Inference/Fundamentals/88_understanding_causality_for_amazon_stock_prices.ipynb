# **Causality**

Understanding causality is foundational in data science, especially when the goal is to make reliable predictions or to understand the dynamics of how one event influences another. Causality plays a pivotal role in data science, guiding us in understanding the relationships where one event influences or causes another. It stretches beyond mere correlation, aiming to decipher how and why certain connections exist, providing a foundation for making informed predictions and crafting strategies to steer outcomes in favorable directions. Establishing causality involves rigorous methods, ranging from Randomized Controlled Trials, seen as the gold standard, to sophisticated causal inference techniques and counterfactual reasoning. This endeavor is not just academic; it's practical, underpinning decision-making processes in various fields like healthcare, policy-making, and beyond, moving us from mere predictions to actionable insights that can significantly impact our world.

## **Methods for Investigating Causality**



## **1. Randomized Controlled Trials (RCTs)**

Randomized Controlled Trials (RCTs) are a cornerstone of clinical and experimental research, particularly in the medical, psychological, and social sciences. Their main strength lies in their ability to demonstrate causality, meaning they can show that a specific intervention directly causes an outcome. This is achieved through a rigorous design that involves several key components:

**1.Randomization:** Randomization is the process of assigning trial participants to either the treatment group (which receives the intervention) or the control group (which does not receive the intervention or receives a placebo) in a completely random manner. This randomness is crucial because it ensures that each participant has an equal chance of being assigned to either group, thereby minimizing selection bias. Selection bias occurs when the characteristics of the participants in each group affect the outcome independently of the intervention.

**2. Control Group:**
The control group serves as a benchmark to measure the effects of the treatment. In drug trials, for example, this group might receive a placebo—an inert substance that has no therapeutic effect. Comparing outcomes between the treatment and control groups allows researchers to determine if the intervention has a significant effect beyond that of a placebo or standard treatment.

**3. Blinding:**
Blinding (or masking) is a method used to prevent participants, and sometimes the researchers themselves, from knowing which group (control or treatment) participants have been assigned to. This helps to prevent bias in reporting and assessing outcomes. Single-blind trials keep the participant unaware, while double-blind trials keep both participants and researchers in the dark regarding group assignments.

**4.Confounding Variables:**
Confounding variables are factors other than the intervention that might affect the outcome. Because participants are randomly assigned to groups, these variables should be equally distributed among the groups, reducing the risk that they could skew the results. This distribution helps to isolate the effect of the intervention from other factors.

### **Strengths of RCTs**
**Causality:** By carefully controlling variables and through randomization, RCTs can provide strong evidence of causal relationships.

**Bias Reduction:** Randomization and blinding help to minimize selection bias and measurement bias.

### **Limitations of RCTs**

Despite their strengths, RCTs also have limitations:

**Cost and Time:** RCTs can be expensive and time-consuming to design, conduct, and analyze.

**Generalizability:** The strict conditions of RCTs may not always reflect real-world scenarios, potentially limiting the applicability of the results to broader populations.

**Ethical Considerations:** Not all interventions or hypotheses can be tested with RCTs due to ethical concerns. For instance, it would be unethical to assign participants to a harmful condition or withhold a treatment that is known to be effective.

This image illustrates a flowchart for the design of a randomized control trial (RCT), which is a type of scientific experiment frequently used to test the efficacy of new treatments or interventions.

2. **Sample Population**: From the general population, a larger group of individuals who are the focus of the research is identified.

4. **Randomised Allocation**: The study sample is then randomly assigned to one of two groups to ensure that each participant has an equal chance of being placed in either group. This randomization is critical for reducing selection bias.

7. **Outcomes**: The final step is to compare the outcomes of both groups. This comparison helps researchers determine if there is a significant difference between the two groups that can be attributed to the treatment or intervention.

## **2. Causal Inference Models**
Causal inference models, such as Directed Acyclic Graphs (DAGs) and counterfactual reasoning, provide a framework for understanding complex causal relationships. DAGs help visualize and analyze the relationships among variables, identifying potential confounding factors and clarifying the direction of causality. Counterfactual reasoning, on the other hand, involves considering what would happen to the outcome under alternative scenarios. This approach is closely linked to the concept of potential outcomes, a fundamental element in causal inference, which compares the observed outcome to the hypothetical outcome that would have occurred had the treatment been different.

### **1. Directed Acyclic Graphs (DAGs)**

**Definition:** A Directed Acyclic Graph is a graphical representation of a set of variables and the causal relationships between them. Each node in the graph represents a variable, and each directed edge (arrow) represents a causal influence from one variable to another. The "acyclic" part means that the graph does not contain any cycles, ensuring that you can't start at one variable and follow a series of causal influences that lead back to the original variable. This property helps clarify the direction of causality.

**Uses:** DAGs are particularly useful for identifying potential confounding factors that could bias the estimation of causal effects. A confounder is a variable that influences both the treatment (or exposure) and the outcome, potentially misleading the observed association between them. By visually mapping out the relationships, researchers can design studies that better account for these factors, either through experimental design or statistical adjustment.

**Advantages:** The visual simplicity of DAGs allows researchers to communicate complex relationships and assumptions underlying their causal inference models clearly. They help in distinguishing between direct and indirect causes, guiding the selection of variables to be controlled for in analytical models.

### **2. Counterfactual Reasoning**

Counterfactual reasoning is a foundational concept in causal inference and machine learning, aimed at understanding the causal relationships between actions (or interventions) and outcomes. It allows us to answer "what-if" questions by imagining alternative scenarios and evaluating their potential outcomes, even when these scenarios never actually took place. This approach is crucial for assessing the impact of specific actions, policies, or treatments in a wide range of fields, from economics and healthcare to social sciences and artificial intelligence.

### **Key Components**
1. **Causal Models:** These are frameworks or mathematical models that describe the causal relationships between variables. They enable us to represent and reason about the system under study, including how variables influence one another.

2. **Potential Outcomes:** This concept refers to the different outcomes that could occur under various interventions or actions. For each unit (e.g., an individual in a study), we consider the outcome that would occur if a specific intervention were applied, even if that intervention did not actually happen.

3. **Interventions:** Actions or changes introduced into the system, which can be hypothetical. In counterfactual reasoning, we consider the effects of these interventions to understand their causal impact.

### **How It Works**

Counterfactual reasoning involves three main steps:

1. **Define the Causal Model:** Establish the variables and the causal relationships between them. This includes identifying which variables are causes and which are effects.

2. **Consider Counterfactual Scenarios:** For a given event, imagine alternative scenarios by asking "what-if" questions. For example, "What would have been the outcome if intervention X had been applied instead of Y?"

3. **Compare Outcomes:** Analyze the difference between the actual outcome and the counterfactual outcome (what would have happened under a different scenario). This comparison reveals the causal effect of the intervention.

##**Challenges in Establishing Causality**


Despite these methods, establishing causality remains challenging due to several factors:

###**1. Confounding Factors:**

Confounding factors are variables that are related to both the independent variable (the cause or intervention) and the dependent variable (the effect or outcome). These factors can create a false impression of a relationship between the independent and dependent variables. For example, in a study exploring the relationship between exercise and health outcomes, diet might be a confounding factor if it influences both exercise habits and health outcomes but is not accounted for.

**Addressing Confounders:**

 Statistical methods like regression analysis can control for confounding variables by including them as covariates in the model. Another sophisticated approach is propensity score matching, a statistical technique that attempts to estimate the effect of a treatment, policy, or other intervention by accounting for the covariates that predict receiving the treatment. This method involves matching units (e.g., individuals) with similar propensity scores (i.e., probabilities of being treated, given the observed covariates) and comparing the outcomes within these matched pairs.

### **2. Reverse Causation:**

Reverse causation refers to the difficulty in determining the direction of causality between two variables. It's a scenario where it's unclear whether A causes B or B causes A. This often arises in observational studies where the temporal sequence of events is not clear.

**Addressing Reverse Causation:**

 Longitudinal studies, where researchers follow subjects over time and record changes in both the independent and dependent variables, can help address reverse causation by establishing a clear timeline of events. Another approach is using instrumental variables that are related to the independent variable but not directly to the dependent variable, except through the independent variable. This can help isolate the direction of the causal relationship.

### **3. Conducting Controlled Experiments:**

Controlled experiments are considered the gold standard for establishing causality. However, they may not always be practical or ethical. For example, it's not feasible to assign participants to smoke for a study on lung cancer risk due to ethical concerns.

**Addressing the Challenge:** Quasi-experimental designs, such as natural experiments and difference-in-differences analyses, offer alternatives. These methods leverage natural variations or policy changes as "experiments" to infer causality. For instance, comparing the health outcomes of populations before and after a smoking ban can provide insights into the effects of smoking on health without conducting an unethical experiment.

## **Fix missing dates**
The dataset sourced from Yahoo Finance does not include entries for dates on which the stock market was closed and no trading occurred. This absence of data for non-trading days could pose challenges for subsequent time series forecasting models used in the analysis. Forecasting models often rely on continuous data sequences to predict future trends accurately. Gaps in the dataset, such as weekends, public holidays, or any other days the stock market remains closed, disrupt these sequences, potentially leading to inaccuracies in the forecasted results. Addressing these missing dates is crucial for maintaining the integrity of the time series analysis, ensuring that models can account for the entire timeline without being misled by unintended data gaps.

```python
# Create a dataframe with the complete date range
full_dates = pd.date_range(start = pre_period[0], end = post_period[1])
full_dates = pd.DataFrame(full_dates, columns=['full_dates'])

# Left join the stock data to the dataframe with the complete date range
adjusted_close = full_dates.merge(adjusted_close, left_on='full_dates', right_index=True, how='left')

# Put the date in the index
adjusted_close.set_index('full_dates', inplace=True)

print('Shape of data after fixing missing dates:', adjusted_close.shape)
```

Here we are addressing the issue of missing dates in a stock data dataset sourced from Yahoo Finance by ensuring the dataset covers a complete date range, including days when the stock market was not trading. This process ensures that the dataset now includes a continuous sequence of dates, crucial for accurate time series forecasting by eliminating gaps in the data due to non-trading days.

Facebook Prophet decomposes time series data into its components (daily, weekly, yearly, etc.). I'll use Prophet to identify seasonal trends in the pre period for the Amazon stock price, and use that seasonal trend as another predictor for building the counterfactual.

To use Facebook Prophet for identifying seasonal trends in Amazon stock price pre-period and then use those trends for building the counterfactual, follow these steps:

**Prepare Data:** Collect Amazon's historical stock price data.

**Prophet Model:** Use Prophet to forecast the stock prices, allowing it to decompose the time series into trend, seasonality (daily, weekly, yearly), and holiday effects.

**Identify Seasonal Trends:** Analyze the decomposed components from the Prophet model to identify key seasonal patterns in the stock price.

**Build the Counterfactual:** Incorporate the identified seasonal trends as predictors or adjustments in your counterfactual model, which aims to estimate what would have happened to the stock price in the absence of a specific event or intervention.

**Predict and Analyze:** Use the counterfactual model to predict stock prices, adjusting for seasonal effects to get a more accurate estimation of the event's impact.

```python
# Filter for AMZN and reset the date index into a column
prophet_data = adjusted_close['AMZN'].reset_index()

# Columns need to be named 'ds' and 'y' for Prophet
prophet_data.columns = ['ds', 'y']

# Forecasts future values using Prophet model based on pre-period data
prophet_data_pre = prophet_data[prophet_data['ds'] <= pre_period[1]]
prophet_data_post = prophet_data[prophet_data['ds'] > pre_period[1]]
ph = Prophet()
ph.fit(prophet_data_pre)
forecast_days = 1 + (post_period[1] - post_period[0]).days
future_prediction = ph.make_future_dataframe(periods=103)
forecast = ph.predict(future_prediction)

# Extracts the trend component from the forecast for further analysis
prophet_trend = forecast[['ds', 'trend']]
```

The stock data for Amazon (AMZN) for forecasting has been prepared using the Prophet model, a tool designed for time series forecasting. It first resets the DataFrame index to turn dates into a column and renames the columns to 'ds' (datestamp) and 'y' (the variable to be forecasted, in this case, Amazon's stock price). The data is then split into pre- and post-announcement periods. Using only the pre-announcement data, the Prophet model is trained to predict future stock prices. The forecast is generated for a specified number of days beyond the pre-period, and the trend component of this forecast is extracted for further analysis.

# **Google CausalImpact**

The Google CausalImpact model is a sophisticated analytical tool used in the field of time series analysis, particularly beneficial for assessing the effect of a specific intervention or event on a target variable over time. The model operates within the Bayesian statistical framework, which allows for incorporating prior knowledge and uncertainty into the analysis. Let's break down the key components and process of how the CausalImpact model works, especially in the context of the example provided regarding Volkswagen stock prices and the company's emissions scandal.

## **Key Concepts**

**1. Time Series Data**

A time series is a sequence of data points collected or recorded at successive time intervals. The data points in a time series can represent anything that changes over time, such as stock prices, weather patterns, or sales figures.

**2. Intervention**

An intervention refers to an event or action that is hypothesized to cause a change in the observed time series. In the Volkswagen example, the intervention is the announcement that Volkswagen lied about their vehicle emissions.

**3. Covariates**
Covariates are other variables that might influence the target variable but are not the primary focus of the study. In the context of stock prices, covariates could include the stock prices of other companies, economic indicators, or market indices. It's crucial that these covariates are not impacted by the intervention, ensuring they serve as a reliable basis for comparison.

**4. Pre-Period and Post-Period**
The analysis is divided into two main periods: the pre-period (before the intervention) and the post-period (after the intervention). The model learns the relationship between the target variable and covariates during the pre-period to make predictions about the post-period.

```python
# Displays the date range, pre-period and post-period for the Causal Impact
print("Data date range:", adjusted_close.index.min(), "-", adjusted_close.index.max())
print("Pre-period:", pre_period)
print("Post-period:", post_period)
```

```python
# Implements CausalImpact analysis to understand the impact of an event on AMZN stock
causal_impact_data = adjusted_close[['AMZN', 'UPS', 'ADBE', 'META', 'ISRG', 'CRM', 'TXN', 'NOW', 'AAPL', 'NEE', 'NKE', 'DHR', 'SBUX', 'BLK', 'LIN', 'WMT', 'NFLX', 'HON', 'CMCSA', 'SNPS']]

# Fit the CausalImpact model
causal_impact = CausalImpact(causal_impact_data, pre_period, post_period)
```

Here, the causal_impact_data DataFrame is created by selecting Amazon's stock along with other stocks, such as UPS, ADBE, META, etc., possibly to use as covariates in the analysis. The CausalImpact model is then fitted using this data, along with predefined pre_period and post_period date ranges, to assess the effect of the event on Amazon's stock by comparing its actual performance against a counterfactual scenario (what would have happened without the event).








```python
# Check model accuracy (MAPE) in the pre period
y_pred_pre = causal_impact.inferences.loc[:pre_period[1]]['preds']
y_pre = causal_impact_data.loc[:pre_period[1]]['AMZN']

mape = MAPE(y_pre, y_pred_pre)
print(f'The mean absolute percentage error in the pre period is {mape:.2f}%')
```

Now we are evaluating the accuracy of a CausalImpact model's predictions for Amazon's stock price before an event, using the Mean Absolute Percentage Error (MAPE). The resulting MAPE of 0.01% indicates an exceptionally high accuracy, with the model's predictions deviating from the actual stock prices by a mere 0.01% on average during the pre-event period. This high level of precision suggests the model is well-suited for analyzing the impact of the event on Amazon's stock by providing a reliable baseline for comparison

```python
# Model summary
print(causal_impact.summary())
```

The output of the CausalImpact model summary indicates a significant positive effect of the event on Amazon's stock price. The actual post-event average and cumulative values (153.27 and 13181.63, respectively) are notably higher than the predicted values had the event not occurred (127.7 and 10982.27). The model estimates an absolute effect of a 20.03% increase in stock price, with a 100% posterior probability of a causal effect. This strong evidence suggests that the event had a positive and statistically significant impact on Amazon's stock value.

```python
# This graph shows the empirical and counterfactual data
causal_impact.plot(panels=['original'], figsize=(15,5))
```

## **Observations:**

The graph depicts the actual observed data for a target variable (solid black line) and its predicted counterfactual values (dashed blue line) over time. The vertical dashed line indicates when an intervention (such as an event or announcement) occurred. Before the intervention, the actual data closely follows the predicted line, suggesting the model's predictions are well-fitted to the pre-intervention reality. After the intervention, the actual data deviates sharply from the predicted counterfactual, suggesting a negative impact from the intervention. The shaded blue area represents the prediction uncertainty, which seems to increase post-intervention, indicating greater volatility in the observed data.

```python
# This graph shows the difference between the empirical and counterfactual data
causal_impact.plot(panels=['pointwise'], figsize=(15,5))
```

## **Observations:**

This graph shows the point effects of an intervention over time. The dashed blue line represents the estimated effect of the intervention on the target variable at each point in time. The shaded blue area indicates the confidence interval around the point estimates, reflecting the statistical uncertainty of these estimates.

Before the vertical dashed line (the intervention time), the point effects hover around zero, which is expected since the intervention hasn't occurred yet. After the intervention, there's a noticeable variation, with the point effects showing a general upward trend, especially towards the right end of the graph. This suggests that the intervention had an increasing impact on the target variable over time.

The note about "the first 1 observations were removed due to approximate diffuse initialization" indicates that the model excluded the initial data point(s) to improve the stability and accuracy of the estimates. This is a common practice in time series analysis when the initial conditions are not well known or when the model is sensitive to the starting values.

This graph represents the cumulative effect of an intervention over time.
