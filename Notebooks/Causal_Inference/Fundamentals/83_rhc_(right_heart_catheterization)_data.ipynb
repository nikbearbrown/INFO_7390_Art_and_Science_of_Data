####**Propensity Score Matching**

Propensity score matching (PSM) aims to control for confounding variables that could affect the treatment assignment (in this case, medical conditions) and outcomes (mortality).

Propensity Score Matching (PSM) is a statistical technique that attempts to estimate the effect of a treatment, policy, or other intervention by accounting for the covariates that predict receiving the treatment. PSM is used in observational studies where random assignment to treatment and control groups is not feasible.

In PSM, the treatment variable is the factor that we are interested in investigating, which in the context of our dataset could be a specific medical history variable (like having a cardiac disease or not). The control variable is the absence of this condition. That is comparing outcomes (mortality) between patients with the condition (treatment group) and without the condition (control group).



**Calculate the propensity scores for each individual, then match individuals from the treatment and control groups based on these scores**

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from causalml.match import NearestNeighborMatch
import matplotlib.pyplot as plt

# Define the features (covariates) and treatment groups
X = data.drop(['cardiohx', 'death'], axis=1)  # Covariates
treatment = data['cardiohx']  # Treatment assignment based on cardiac history

# Calculate propensity scores with logistic regression
logit = LogisticRegression(solver='liblinear')
logit.fit(X, treatment)
propensity_scores = logit.predict_proba(X)[:, 1]

# Add propensity scores to the original data
data['propensity_score'] = propensity_scores

# Define the treatment and control groups based on the treatment variable
treatment_data = data[data['cardiohx'] == 1]
control_data = data[data['cardiohx'] == 0]
```

**Match the treatment and control groups using the nearest neighbor matching method**

We will use Nearest Neighbor Matching to match individuals from the treatment group (those with a cardiac history, cardiohx = 1) with those from the control group (without cardiac history, cardiohx = 0), based on their propensity scores.

```python
from sklearn.neighbors import NearestNeighbors


# Extract propensity scores
propensity_scores = data['propensity_score'].values

# Split the dataset into treatment and control groups
treatment_data = data[data['cardiohx'] == 1]
control_data = data[data['cardiohx'] == 0]

# Extract propensity scores for treatment and control groups
treatment_scores = treatment_data['propensity_score'].values
control_scores = control_data['propensity_score'].values

# Use Nearest Neighbors for matching
nn = NearestNeighbors(n_neighbors=1, algorithm='auto').fit(control_scores.reshape(-1, 1))

# For each treatment case, find the closest control case
_, indices = nn.kneighbors(treatment_scores.reshape(-1, 1))

# Extract the matched control cases
matched_control_data = control_data.iloc[indices.flatten()]

# Optional: Combine the treatment and matched control data for analysis
matched_data = pd.concat([treatment_data, matched_control_data])

# Display the first few rows of the matched data
matched_data.head()
```

####**Covariate Balance Checking**

Check if the matching process has balanced the covariates between the treatment and control groups effectively.

This involves comparing the distribution of covariates and propensity scores in both groups before and after matching, using methods Standardized Mean Differences (SMD).

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

def calculate_smd(df, treatment_col, group_col, covariate_cols):
    """
    Calculate the Standardized Mean Differences (SMD) for covariates
    between treatment and control groups.
    """
    treatment_group = df[df[group_col] == 'treatment']
    control_group = df[df[group_col] == 'control']

    smds = {}
    scaler = StandardScaler()

    for col in covariate_cols:
        mean_treatment = treatment_group[col].mean()
        mean_control = control_group[col].mean()
        pooled_std = np.sqrt((treatment_group[col].std()**2 + control_group[col].std()**2) / 2)
        smd = (mean_treatment - mean_control) / pooled_std
        smds[col] = smd

    return smds

# Define the list of covariates to check balance for
covariate_cols = ['chfhx', 'dementhx', 'psychhx', 'chrpulhx', 'renalhx', 'liverhx', 'gibledhx', 'malighx', 'immunhx', 'transhx', 'amihx']  # Add or remove columns as necessary

# Calculate SMD for each covariate
smd_values = calculate_smd(matched_data, 'cardiohx', 'group', covariate_cols)

# Print the SMD values
for covariate, smd in smd_values.items():
    print(f"{covariate}: SMD = {smd:.3f}")
```

The output values provided show the Standardized Mean Differences (SMD) for each covariate in the matched dataset.

Results Interpretation:

SMD Values: SMD is a measure of the balance of covariates between the treatment and control groups. It is the difference in means between the two groups, divided by the pooled standard deviation. SMD values close to 0 suggest good balance.

Interpretation:

Values Close to 0: An SMD value less than 0.1 is generally considered indicative of a good balance. Based on the output, all covariates have SMD values well below 0.1, which suggests that the matching process has been effective in balancing these covariates between the treatment and control groups.

Negative Values:

The negative values for malighx and transhx indicate that the mean for these covariates is slightly higher in the control group than in the treatment group. However, since the absolute values are very small, this difference is not likely to be of practical significance.

####**Estimate the effect of the treatment (having a cardiac history) on the outcome (death) using statistical methods: logistic regression**

```python
import statsmodels.api as sm

# Convert the 'death' column to a binary variable if it's not already
# Assuming 'Yes' is 1 and 'No' is 0
matched_data['death'] = matched_data['death'].apply(lambda x: 1 if x == 'Yes' else 0)

# Define the independent variables (covariates) and the dependent variable (outcome)
# Here, 'propensity_score' is included to adjust for any residual imbalances
independent_vars = ['cardiohx', 'propensity_score'] + ['chfhx', 'dementhx', 'psychhx', 'chrpulhx', 'renalhx', 'liverhx', 'gibledhx', 'malighx', 'immunhx', 'transhx', 'amihx']
X = matched_data[independent_vars]
y = matched_data['death']

# Add a constant to the independent variables matrix
X = sm.add_constant(X)

# Fit the logistic regression model
model = sm.Logit(y, X)
result = model.fit()

# Print the summary of the logistic regression model
print(result.summary())
```

The results from your logistic regression analysis provide valuable insights into the impact of various covariates, including cardiac history (cardiohx), on the likelihood of death. Here's a breakdown of the key findings:

Model Fit and Significance:

Log-Likelihood:

The log-likelihood value is -1199.8, which is a measure of the fit of the model.

LLR (Likelihood Ratio Test) p-value:

The LLR p-value is significantly less than 0.05, indicating that the model is a better fit than the null model (a model with no predictors).

Coefficients and Their Significance:

cardiohx (Cardiac History):

The coefficient is 0.2988 with a p-value of 0.002, suggesting that having a cardiac history significantly increases the likelihood of death. The positive coefficient indicates a higher odds ratio.

Other Covariates:

Several other covariates like chfhx, dementhx, renalhx, immunhx, and amihx have significant coefficients, indicating their respective influences on the likelihood of death.

propensity_score:

Its coefficient is -14.8847 with a significant p-value, suggesting a strong effect of the propensity score in adjusting the model. The negative coefficient indicates an inverse relationship with the likelihood of death, which is expected as it's controlling for the probability of receiving the treatment.

Interpreting Coefficients:

The coefficients in a logistic regression model represent the log odds. For example, for cardiohx, a one-unit increase (from 0 to 1, i.e., not having to having a cardiac history) is associated with an increase in the log odds of death by 0.2988.

For continuous variables, the coefficient represents the change in the log odds of the outcome for a one-unit increase in the predictor variable.

Cautions:

Some variables, like liverhx and gibledhx, are not statistically significant (high p-values), indicating that their effect on the outcome is not statistically discernable from zero in this model.

The large negative coefficient for propensity_score indicates its strong control in the model, which aligns with its purpose to adjust for potential confounding factors.

Conclusion:

This analysis provides evidence that cardiac history and other significant health conditions have a measurable impact on the likelihood of death.
