## **Introduction to Causality**


Grasping the concept of causality is a cornerstone in the realm of data science, particularly when the objective is to forge reliable forecasts or to grasp the nuances of how one occurrence impacts another. The significance of causality in data science cannot be overstated; it serves as a compass for identifying and understanding the connections where one event leads to or affects another. This understanding delves deeper than simple correlations, striving to unravel the mechanisms and reasons behind specific relationships, thereby laying the groundwork for well-informed prognostications and the development of strategies aimed at guiding outcomes towards desired directions. The process of determining causality requires stringent methodologies, from the highly regarded Randomized Controlled Trials to advanced methods of causal inference and the use of counterfactual thinking. This quest for understanding causality transcends academic interest, playing a crucial role in the decision-making processes across various sectors such as healthcare and policy-making. It propels us from the realm of speculative predictions to the domain of practical insights, paving the way for interventions that have the potential to reshape our reality significantly.

**The Intricacies of Causal Relationships**

Causality transcends simple correlation by probing into the 'why' and 'how'—it seeks to uncover the underlying mechanisms driving relationships within data. This is critical because correlation alone can be misleading; it may imply a relationship where none exists or mask deeper, more complex interactions. Causality clarifies these connections, paving the way for predictions and strategies that are both informed and effective.

**The Methodological Landscape**

This journey into causality is navigated using a suite of robust methodologies. The gold standard of Randomized Controlled Trials sits alongside sophisticated causal inference techniques and the strategic use of counterfactual reasoning. These methods bring rigor and precision to our analysis, ensuring that our conclusions about causal effects are drawn from a place of strength and reliability.

**The Power of Counterfactuals**

A significant aspect of our analysis is counterfactual thinking—considering what would have happened if a particular event had not occurred. It allows us to isolate the effect of specific interventions and assess their impact with greater clarity. In the realm of stock prices, this can differentiate between normal market fluctuations and the true effect of a corporate announcement, regulatory change, or market shock.

**Impact on Decision-Making**

Causal insights gleaned from this analysis have far-reaching implications. For businesses, they inform strategic decisions and investment strategies. For policymakers, they offer a basis for crafting regulations that aim to stabilize markets. In healthcare, similar approaches to causality can mean the difference between an effective treatment and an ineffective one.




![maxresdefault](https://github.com/UtkarshaShirke/DataScience/assets/114371417/bc3dfc62-6ac6-4da4-85e4-b1f40252f1d6)



The above image shows the framework of the Causal Inference. The impartnat aspects related to Causal nference are explained below:

 **Treatment**:
   - Refers to an intervention or exposure whose effect is being studied.
   - Can be any factor whose influence on an outcome is of interest, such as a new drug, a policy change, or an educational program.

**Outcome**:
   - The variable that is hypothesized to be affected by the treatment.
   - This could be a health outcome, economic metric, educational achievement, etc.

**Confounders**:
   - Variables that have an influence on both the treatment and the outcome.
   - Confounders can lead to a spurious association between the treatment and the outcome if not properly controlled.
   - Examples include socioeconomic status, which might affect both healthcare treatment (the treatment) and health outcomes (the outcome).

**Unknown Confounders**:
   - Factors that affect the outcome but are not measured or included in the analysis.
   - They introduce bias into the causal estimation, as they may be correlated with both the treatment and the outcome.
   - The presence of unknown confounders can undermine the validity of the causal inference.

**Causal Pathways**:
   - The direct arrows from treatment to outcome represent the hypothesized causal pathways.
   - Understanding these pathways is key to developing theories about how and why a treatment affects an outcome.

**Controlling for Confounders**:
   - Strategies to address confounding include matching, stratification, and regression techniques.
   - Advanced methods like propensity score matching and instrumental variable analysis are also employed to mimic randomization.

**Causal Inference Challenges**:
   - Establishing causality in observational studies is complex due to the potential for confounding.
   - Randomized Controlled Trials (RCTs) are the gold standard for causal inference as they randomly assign treatments, helping to balance both known and unknown confounders across treatment groups.

**Impact of Unknown Confounders**:
   - They represent a major challenge in observational studies and can significantly bias results.
   - Sensitivity analysis is often conducted to assess how robust the causal conclusions are to the presence of such unmeasured confounding.

**Estimating Causal Effect**:
   - The ultimate goal of causal inference is to estimate the true effect of the treatment on the outcome, independent of confounding factors.
   - This involves comparing the observed outcome with the estimated counterfactual outcome—what would have happened in the absence of the treatment.

## **Fix missing dates**
The dataset from Yahoo Finance omits data for non-trading days, including weekends and public holidays, when the stock market is closed. This lack of entries for these days introduces challenges for time series forecasting models, which depend on uninterrupted data sequences to accurately predict future trends. The absence of data for these periods can disrupt the continuity needed for precise forecasting, potentially affecting the accuracy of predictions. Therefore, it's essential to fill in these missing dates to preserve the consistency of the time series analysis, allowing models to accurately represent the entire timeline without being compromised by gaps in the data.

```python
# Create a dataframe with the complete date range
full_dates = pd.date_range(start = pre_period[0], end = post_period[1])
full_dates = pd.DataFrame(full_dates, columns=['full_dates'])

# Left join the stock data to the dataframe with the complete date range
adjusted_close = full_dates.merge(adjusted_close, left_on='full_dates', right_index=True, how='left')

# Put the date in the index
adjusted_close.set_index('full_dates', inplace=True)

print('Shape of data after fixing missing dates:', adjusted_close.shape)
```

We've addressed the issue of missing dates in the stock data from Yahoo Finance by ensuring the dataset now spans a complete date range, encompassing even non-trading days. This adjustment creates a continuous sequence of dates, essential for accurate time series forecasting, by removing data gaps caused by the stock market's closure on weekends and public holidays.

###**Facebook Prophet**
Predicting stock market trends is vital for investors, and META Platforms Inc. is no exception. Facebook's Prophet model offers a sophisticated yet accessible approach to understanding and forecasting META's stock prices. By breaking down historical data into trends and seasonal patterns, Prophet provides clear insights into the stock's behavior over time. Analysts can identify how weekly, monthly, and yearly cycles affect prices and build models to predict future movements, accounting for regular fluctuations and special events. This technique allows for creating informed scenarios that estimate META's stock performance in various conditions, equipping investors with a strategic edge in the market.



```python
# Filter for META and reset the date index into a column
prophet_data = adjusted_close['META'].reset_index()

# Columns need to be named 'ds' and 'y' for Prophet
prophet_data.columns = ['ds', 'y']

# Forecasts future values using Prophet model based on pre-period data
prophet_data_pre = prophet_data[prophet_data['ds'] <= pre_period[1]]
prophet_data_post = prophet_data[prophet_data['ds'] > pre_period[1]]
ph = Prophet()
ph.fit(prophet_data_pre)
forecast_days = 1 + (post_period[1] - post_period[0]).days
future_prediction = ph.make_future_dataframe(periods=103)
forecast = ph.predict(future_prediction)

# Extracts the trend component from the forecast for further analysis
prophet_trend = forecast[['ds', 'trend']]
```

Here, we are using Facebook's Prophet model to analyze and forecast the stock price of META. It begins by preparing the data, renaming columns for compatibility with Prophet, fitting the model to historical data before a specified date (pre_period), and then using the model to predict future stock prices. It generates a forecast for 103 days beyond the known data (post_period) and extracts the 'trend' component from the forecast for further analysis, likely to assist in understanding the general direction in which the stock price is heading.

# **Google CausalImpact**

The Google CausalImpact model is a tool for time series analysis, useful for assessing the impact of events on a specific variable over time, using Bayesian statistics. It involves:

1. **Time Series Data**: Sequences of data points over time, like stock prices.
2. **Intervention**: An event thought to affect the time series, e.g., Volkswagen's emissions scandal.
3. **Covariates**: Additional variables that might influence the target but aren't the main focus, ensuring they're unaffected by the intervention.
4. **Pre-Period and Post-Period**: The analysis compares before and after the intervention to predict what would have happened without the event.

CausalImpact essentially examines the difference between actual outcomes and what was expected without the intervention, considering other influencing factors.

```python
# Displays the date range, pre-period and post-period for the Causal Impact
print("Data date range:", adjusted_close.index.min(), "-", adjusted_close.index.max())
print("Pre-period:", pre_period)
print("Post-period:", post_period)
```

```python
# Implements CausalImpact analysis to understand the impact of an event on META stock
causal_impact_data = adjusted_close[['META', 'GOOG', 'AMT', 'ISRG', 'GOOGL', 'BLK', 'ORCL', 'DHR', 'PM', 'SPGI', 'ADBE']]

# Fit the CausalImpact model
causal_impact = CausalImpact(causal_impact_data, pre_period, post_period)
```

Here we are making use of CausalImpact package to assess how a particular event affected META's stock price. It compares the stock's performance before and after the event, using other related stocks as a control group to understand what META's performance might have been like if the event hadn't occurred.

```python
causal_impact_data
```

```python
# Check model accuracy (MAPE) in the pre period
y_pred_pre = causal_impact.inferences.loc[:pre_period[1]]['preds']
y_pre = causal_impact_data.loc[:pre_period[1]]['META']

mape = MAPE(y_pre, y_pred_pre)
print(f'The mean absolute percentage error in the pre period is {mape:.2f}%')
```

The Mean Absolute Percentage Error (MAPE) is a measure of prediction accuracy in a forecasting model. It represents the average absolute percent difference between predicted and actual values for a variable—in this case, META's stock prices during the pre-period.

A MAPE of 0.02% indicates a very high level of accuracy for the CausalImpact model's predictions compared to the actual stock prices. In practical terms, it means the model's predictions were, on average, only 0.02% away from the actual stock prices. Such a low error rate suggests that the model can reliably estimate the expected behavior of META's stock price in the absence of the event under study, giving analysts confidence in the model's estimates of the event's impact in the post-period.

```python
# Model summary
print(causal_impact.summary())
```

The Causal Impact summary indicates with high confidence that the event in question had a significant negative impact on META's stock prices, leading to an average decrease of about 29.69% compared to what was predicted without the event. The model is quite certain (100% probability) that the event caused this decrease.

The graph compares the actual META stock prices with the predicted prices if a specific event hadn't occurred. Before the event, actual and predicted prices are similar. The graph illustrates the pointwise differences between the observed and predicted values from a CausalImpact analysis. Here's a concise interpretation:

- The dashed line represents the difference, or the 'point effects,' between actual and predicted META stock prices over time.
- Before the vertical dashed line, representing the event, the differences fluctuate around zero, indicating the predictions were fairly accurate.
- After the event, the point effects drop significantly below zero, showing that the actual stock prices were much lower than predicted.
- The shaded area represents the confidence interval, which gets wider after the event, indicating increased uncertainty.

```python
# This graph shows the cumulative difference between the empirical and counterfactual data in the post period
causal_impact.plot(panels=['cumulative'], figsize=(15,5))
```

### **Observation:**

The graph shows the cumulative effect from a CausalImpact analysis. The dashed line indicates the cumulative difference between the actual observed data and the counterfactual prediction—that is, what the model estimates would have happened if the event of interest had not occurred.


- The graph starts at zero, indicating the point at which the event occurs.
- As time progresses, the cumulative effect becomes increasingly negative, suggesting that the actual stock prices are consistently lower than what would have been expected if the event had not taken place.
- The shaded area around the dashed line suggests the confidence interval, which appears to grow over time, indicating increasing uncertainty in the cumulative effect as you move further from the event.
- The steady downward slope of the cumulative difference indicates that the impact of the event is sustained over time and continues to accumulate.
