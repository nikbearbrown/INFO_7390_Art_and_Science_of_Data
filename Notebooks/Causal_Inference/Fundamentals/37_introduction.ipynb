# Introduction

# **Abstract:**



Many of the process skills needed for mathematics are similar to reading skills and, when taught together, would reinforce each other. Examples of common skills are predicting, inferring, communicating, comparing and contrasting, and recognizing cause and effect relationships. Teachers who recognize the interrelatedness of mathematics and literacy processes can design instruction that reflects these similarities. Becoming a Nation of Readers suggests that the most logical place for instruction in most reading and thinking strategies is in the content areas rather than in separate lessons about reading.

The importance of writing in the mathematics classroom cannot be overemphasized. In the process of writing, students clarify their own understanding of mathematics and hone their communication skills. They must organize their ideas and thoughts more logically and structure their conclusions in a more coherent way. Competency in writing can only be accomplished through active practice; solving mathematics problems is a natural vehicle for increasing studentsâ€™ writing competence.

**So the fact is that :**

* mathematics and reading have mutually casual relation
* mathematics cause writing skill

```python
from IPython.display import Image
import category_encoders as ce

import graphviz
import lingam
from lingam.utils import make_dot, make_prior_knowledge
```

**2) Data Preprocessing**

```python
oe = ce.ordinal.OrdinalEncoder(
    cols=df.select_dtypes(include="object")
    )

df_en = oe.fit_transform(df)
```

```python
df_list = df.columns.to_list()
df_dict = {}

for i, column in zip(range(len(df_list)), df_list):
    df_dict[column] = i

print(len(df_list))
print(df_dict)
```

**3) Setting prior knowledge**

```python
prior_knowledge = make_prior_knowledge(
    8,
    paths = [
             [df_dict["math score"], df_dict["reading score"],df_dict['writing score']
             ]],

)
```

**4) LiNGAM Modeling**

```python
model = lingam.DirectLiNGAM(
                    random_state=42,
                    measure = 'pwling',
                    prior_knowledge = prior_knowledge
                    )

model.fit(df_en)
```

```python
dot = make_dot(
    model.adjacency_matrix_,
     labels=df_en.columns.to_list(),
    )


dot.format = 'png'
dot.render('dag')

Image("dag.png")
```

**5) LiNGAM with Linear Regression (Target Variable : "writing score")**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression(fit_intercept=True)
reg.fit(df_en.drop(["writing score"], axis=1), df_en["writing score"])
```

```python
features = [i for i in range(df_en.drop(["writing score"], axis=1).shape[1])]
print("features: ", features)

dot = make_dot(
    model.adjacency_matrix_,
    labels=df_en.columns.to_list(),
    prediction_feature_indices=features,
    prediction_target_label='writing score',
    prediction_coefs=reg.coef_
    )

dot.format = 'png'
dot.render('dag')

Image("dag.png")
dot = make_dot(
    model.adjacency_matrix_,
    labels=df_en.columns.to_list(),
    prediction_feature_indices=features,
    prediction_target_label='writing score',
    prediction_coefs=reg.coef_
    )

dot.format = 'png'
dot.render('dag')

Image("dag.png")
```

# Insight

**How we should interpret adjacency_matrix of lingam?**

In causal inference and the Lingam framework, an adjacency matrix is used to represent the causal relationships among variables in a system. Specifically, the adjacency matrix is a binary matrix that indicates whether there is a directed edge (i.e., causal relationship) between each pair of variables in the system.

The LiNGAM (Linear Non-Gaussian Acyclic Model) analysis conducted on the dataset sought to uncover the causal relationships among various academic scores (math, reading, and writing) and other demographic factors. The primary hypotheses were twofold:

1. Mathematics and reading have a mutually causal relationship.
2. Mathematics skills causally influence writing skills.


However, the LiNGAM results revealed insights that both confirm and challenge these hypotheses:

**1. Causal Influence of Mathematics on Reading and Writing:**
The causal influence from math score to reading score was found to be relatively small (0.15), suggesting that the direct impact of mathematics ability on reading ability is minimal in this dataset.

Contrary to the initial hypothesis, the analysis showed a stronger causal influence from writing score to math score (0.95), indicating that writing skills might play a significant role in enhancing mathematics skills, rather than the other way around.

**2. Dominant Role of Writing Skills:**
The analysis also revealed a strong causal influence from writing score to reading score (0.84), suggesting that writing skills have a significant impact on reading skills.
This finding implies that writing skills may serve as a foundational component that supports and enhances both math and reading abilities.
