#**What is Causality???**
"Causality" refers to the relationship between causes and effects. It's a fundamental concept in many fields, including philosophy, physics, and statistics.

In a causal relationship, one event (the cause) is understood to be responsible for the occurrence of another event (the effect). Understanding causality is crucial for scientific inquiry and practical decision-making, as it helps to predict and explain the outcomes of various phenomena.

For example, in medicine, understanding the causal relationship between a virus and a disease helps in developing treatments. In physics, causality is important for understanding why certain physical events occur.

In everyday life, causality helps us understand the consequences of our actions and plan accordingly.


#**Why is it Important???**
Causality is important for several key reasons across various domains:

**Understanding and Predicting Outcomes:** In science and everyday life, understanding causality allows us to predict the consequences of actions, events, or conditions.

**Problem Solving and Decision Making:** Knowing the causal relationships between different factors helps in solving problems and making informed decisions.

**Scientific Research:** In scientific research, establishing causality is essential for developing theories and models that accurately describe how the world works.


**Legal and Ethical Implications:** In the legal system, establishing causality can be critical in cases of liability and responsibility. Determining whether a certain action caused harm is key to many legal decisions.

**Technological Development:** In technology and engineering, understanding the causal mechanisms of systems is essential for designing and improving technology.

**Preventing Unintended Consequences:** By understanding causal relationships, individuals and organizations can avoid actions that might have unintended and undesirable consequences.

**Philosophical and Theoretical Importance:** Causality is a core concept in many philosophical debates and theories about the nature of reality, free will, and the structure of the universe.

In summary, causality is foundational to our understanding of the world and our ability to interact with it effectively and responsibly.

# Advanced Causal Inference:
 Propensity Score Matching (PSM):
For a more nuanced approach to causal inference, we can use Propensity Score Matching (PSM).

PSM attempts to estimate the effect of a treatment (in this case, obtaining a bachelor's degree or higher) by accounting for the covariates that predict receiving the treatment. Here, we simplify the concept, focusing on education as a treatment and its impact on income levels.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
import statsmodels.api as sm

# Assuming 'Bachelors_degree_or_higher' as treatment, we need a binary indicator (0 or 1)
# For simplicity, let's categorize based on median as a threshold
median_education = merged_df['Bachelors_degree_or_higher'].median()
merged_df['Education_Treatment'] = (merged_df['Bachelors_degree_or_higher'] > median_education).astype(int)

# Standardize the continuous variables for logistic regression
scaler = StandardScaler()
merged_df['Standardized_Education'] = scaler.fit_transform(merged_df[['Bachelors_degree_or_higher']])

# Fit logistic regression to estimate propensity scores
logit = LogisticRegression()
X = merged_df[['Standardized_Education']]
y = merged_df['Education_Treatment']
logit.fit(X, y)
merged_df['Propensity_Score'] = logit.predict_proba(X)[:, 1]


# Viewing the first few rows to see the propensity scores
print(merged_df[['Bachelors_degree_or_higher', 'Education_Treatment', 'Propensity_Score']])
```

From the logistic regression model used to calculate propensity scores for regions based on their level of Bachelor's degrees or higher, we can conclude:

1. **Identification of Treatment Groups**: The model successfully distinguishes between regions with above-median and below-median levels of higher education. This is evident from the binary `Education_Treatment` indicator.

2. **Propensity Score Distribution**: The wide range of propensity scores, from near 0 to 1, indicates variability in how likely regions are to have above-median levels of Bachelor's degrees, based on the standardized number of degrees.

3. **Match Quality for Causal Analysis**: The variability in propensity scores suggests potential for matching regions (for causal analysis) based on their likelihood of having higher educational attainment. This matching aims to compare similar regions to estimate the effect of higher education levels on outcomes like income.

In summary, the propensity scores derived from our logistic regression model set the stage for more sophisticated causal inference analyses by enabling the matching of treated and untreated units in an observational study setting.

# **This example includes a Granger causality test for whether changes in AAPL's prices can predict changes in MSFT's prices**.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller

# Override pandas_datareader data method to use yfinance
yf.pdr_override()

sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")


# Defining the tech stocks and the time frame for data retrieval
tech_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN']
end = datetime.now()
start = datetime(end.year - 1, end.month, end.day)


all_data = {ticker: yf.download(ticker, start, end) for ticker in tech_list}

# Create a single DataFrame with the adjusted close prices
prices = pd.DataFrame({ticker: data['Adj Close'] for ticker, data in all_data.items()})


# Check for stationarity of each series
def check_stationarity(data):
    for column in data.columns:
        result = adfuller(data[column])
        print(f'Stationarity test for {column}:')
        print('p-value:', result[1])
        if result[1] > 0.05:
            print(f"{column} is not stationary")
        else:
            print(f"{column} is stationary")
        print('---' * 10)

check_stationarity(prices)


# Prepare for VAR
model = VAR(prices)
results = model.fit(maxlags=15, ic='aic')
print(results.summary())



granger_test = results.test_causality('MSFT', ['AAPL'], kind='f')
print(granger_test.summary())



# Plot the prices to visualize
prices.plot(figsize=(15, 7))
plt.title("Stock Prices")
plt.show()
```

#**Understanding the VAR Model Summary**

The VAR model summary provides coefficients for each variable at each lag for every equation in the system. Here's what to infer from it:

- **Coefficients**: For each stock (AAPL, GOOG, MSFT, AMZN), the model estimates coefficients for its own lagged values and the lagged values of other stocks. A coefficient measures the change in the dependent variable for a one-unit change in the predictor variable, holding other predictors constant.
- **Significance (p-value)**: Tells you whether the relationships identified by the coefficients are statistically significant. Typically, a p-value below 0.05 is considered significant, suggesting that there's a less than 5% chance that the observed relationship is due to random variation.
- **Stationarity Warning**: The warning about no associated frequency information indicates that the time series data didn't have a defined frequency (e.g., daily, monthly). This doesn't affect the VAR model fit but would matter for forecasting.

For example, the coefficient for `L1.AAPL` in the equation for `AAPL` is 0.973487 with a p-value close to 0, indicating a strong and significant relationship between AAPL's lagged price and its current price, as expected for a stock's price series.

#**Granger Causality Test**

The Granger causality test checks if past values of one variable (e.g., AAPL's price) can predict the future value of another variable (e.g., MSFT's price) better than the past values of the second variable alone.

- **Test Result**: The conclusion "fail to reject H_0 at 5% significance level" for the test where H_0 posits that "AAPL does not Granger-cause MSFT" means there's not enough evidence to say that past values of AAPL's prices can predict MSFT's prices.
- **p-value**: At 0.094, it's above the typical 0.05 threshold, reinforcing the conclusion that there's insufficient evidence to claim a Granger-causal relationship between AAPL and MSFT prices in this context.

### How Causality is Used

Causality, in the context of this analysis, is explored through the lens of prediction rather than true causal influence. Granger causality tests if knowing the past values of one variable helps predict future values of another variable better than not knowing those past values.

### Limitations and Implications

1. **Granger Causality**: It's important to note that Granger causality isn't about true causality in the philosophical or physical sense. It's about predictability.

2. **Financial Data Complexity**: The relationships between stock prices can be influenced by myriad factors not captured in this model, such as macroeconomic indicators, company news, and market sentiment.

3. **Model Selection and Stationarity**: The need to ensure stationarity and the choice of lag length based on information criteria like AIC are critical steps in the modeling process. Non-stationary data or poorly chosen lag lengths can lead to misleading results.



In summary, while the VAR model and Granger causality tests provide valuable tools for exploring interdependencies and predictability among time series, their findings, especially in complex domains like stock markets, should be viewed as part of a broader analysis rather than definitive evidence of causal relationships.

```python
# Create Total Employment if it doesn't exist
if 'Total_Employment' not in industry_df.columns:
    industry_df['Total_Employment'] = industry_df['Male'] + industry_df['Female']

# Normalize data by total employment to get proportions
industry_cols = [col for col in industry_df.columns if col.startswith('Total_')]
for col in industry_cols:
    industry_df[f'{col}_prop'] = industry_df[col] / industry_df['Total_Employment']

# Log transformation if necessary to reduce skewness
industry_df['log_Total_Construction'] = np.log1p(industry_df['Total_Construction'])
```

3. **Running the IV Regression**: We conducted a two-stage regression analysis. In the first stage, we tried to predict construction employment using the year. 4. **Checking the Results**:
