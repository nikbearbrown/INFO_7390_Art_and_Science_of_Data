## **What is Causality?**

<div align="center">
  <img src="https://raw.githubusercontent.com/sanikadhayabar/ADS/refs/heads/main/CrashCourse/Images/causalityimg.jpeg" alt="Causality" style="width:50%;"/>

Causality refers to the relationship between two events or variables where one directly influences or causes the other. It goes beyond merely observing patterns or associations (correlation) to establish a definitive cause-and-effect connection. In simpler terms, causality answers the question: "Did one event directly cause another to happen?" For example, while higher temperatures and increased ice cream sales may be correlated, causality establishes that it is the rise in temperature that causes more people to buy ice cream, rather than the two events simply happening together by coincidence.

Causality requires specific conditions to be met. First, **the cause must precede the effect in time**—temperature must rise before ice cream sales increase for causality to hold. Second, there should be **no confounding factors**, meaning other variables should not explain the observed relationship. For instance, if holidays influence both temperature-related outings and ice cream sales, holidays are a confounder that must be accounted for before confirming causality. Lastly, there must be a **mechanism or logical pathway** linking the cause and the effect, such as heat making people desire cold treats.

Understanding causality is crucial because it enables actionable decision-making. While correlation might suggest a relationship, causality confirms it, allowing businesses, scientists, and policymakers to design interventions that address root causes rather than symptoms. Establishing causality often requires controlled experiments or advanced statistical methods, making it a cornerstone of robust analysis in fields ranging from medicine to economics and beyond.


## **Is Correlation Not the Same as Causation?**

<div align="center">
  <img src="https://raw.githubusercontent.com/sanikadhayabar/ADS/refs/heads/main/CrashCourse/Images/cvc.png" alt="IcecreamSun" style="width:50%;"/>
</div>


No, correlation and causation are not the same. While they are related concepts, they serve distinct purposes in understanding relationships between variables. Correlation refers to a statistical measure that quantifies the extent to which two variables move together. For instance, when temperatures rise, ice cream sales often increase, which suggests a correlation between the two variables. However, correlation merely tells us that there is an association—it does not explain why the association exists. It leaves unanswered questions about whether one variable directly affects the other or if there is another factor influencing both.

Causation, on the other hand, establishes a direct cause-and-effect relationship between variables. In a causal relationship, a change in one variable is responsible for a change in another. Using the same example, causation would mean that higher temperatures directly cause people to buy more ice cream because they seek to cool down. Unlike correlation, causation provides a deeper understanding by explaining the "why" behind the observed relationship. This distinction is crucial because correlation alone can sometimes be misleading. For instance, shark attacks and ice cream sales are positively correlated during summer, but eating ice cream does not cause shark attacks. The underlying cause is temperature, which increases both beach attendance (leading to more shark encounters) and the desire for ice cream. This is an example of a confounder—an external factor influencing both variables.

The risks of confusing correlation with causation are significant. Acting on correlation without proving causation can lead to flawed conclusions and decisions. Imagine a business observing a correlation between advertising spend and revenue. Without further investigation, they might assume that increasing advertising budgets will always lead to higher sales. However, causation would require proving that the increased spend, and not another factor like seasonal demand or product improvements, is directly responsible for the higher revenue. Misinterpreting correlation as causation can waste resources, lead to ineffective strategies, and result in missed opportunities to address the actual drivers of change.

Establishing causation often requires rigorous analysis. Researchers rely on methods like controlled experiments, where one variable is manipulated to observe its direct effect on another. For observational data, advanced statistical techniques such as regression analysis, instrumental variables, or propensity score matching help control for confounders and identify causal relationships. By distinguishing correlation from causation, we can go beyond surface-level observations and make decisions grounded in a true understanding of how variables interact. This distinction empowers researchers, businesses, and policymakers to design solutions that address the root causes of challenges rather than their symptoms.

## **But What Are the Requirements for Causality?**


![Requirement](https://raw.githubusercontent.com/sanikadhayabar/ADS/refs/heads/main/CrashCourse/Images/Requirement.png)


Causality refers to a direct cause-and-effect relationship between two variables. For a relationship to be considered causal, certain criteria must be met to ensure that the observed connection is not coincidental or influenced by external factors. Two fundamental requirements form the foundation of establishing causality:

#### **1. Temporal Precedence**
The cause must occur before the effect. This ensures the logical sequence of events, where changes in the independent variable lead to changes in the dependent variable, rather than the other way around. Without temporal precedence, it is impossible to determine the direction of the relationship. For instance, if variable A and variable B are related, we must prove that changes in A consistently precede changes in B to argue that A causes B.

#### **2. No Confounding Variables**
A confounder is a third variable that influences both the cause and the effect, creating a false or misleading relationship. To establish causality, we must account for these confounders, ensuring they do not bias the observed relationship. For example, if variable A and variable B are correlated, but a third variable C independently affects both, the relationship between A and B might not be causal. Controlling for confounders helps isolate the true effect of the independent variable on the dependent variable.

Meeting these two requirements—**temporal precedence** and eliminating confounding variables—is essential for establishing causality. By fulfilling these criteria, researchers can move beyond surface-level correlations and identify true cause-and-effect relationships, leading to more accurate and actionable insights.

### **Model : Propensity Score Matching** 

This method (Propensity Score Matching) is a robust way to estimate causal effects in observational data, where randomization (like in experiments) is not possible. By matching treated and control units with similar characteristics, you reduce bias caused by confounding variables and isolate the effect of treatment (temperature > 25°C) on revenue.

```python
# Define treatment and outcome
data['Treatment'] = (data['Temperature'] > 25).astype(int)

# Step 1: Estimate propensity scores using logistic regression
propensity_model = LogisticRegression()
propensity_model.fit(data[['Temperature']], data['Treatment'])
data['Propensity_Score'] = propensity_model.predict_proba(data[['Temperature']])[:, 1]

# Step 2: Match treated and control units based on propensity scores
treated = data[data['Treatment'] == 1]
control = data[data['Treatment'] == 0]

# Use nearest neighbors to find matches
nn = NearestNeighbors(n_neighbors=1)
nn.fit(control[['Propensity_Score']])
distances, indices = nn.kneighbors(treated[['Propensity_Score']])

# Extract matched control group
matched_control = control.iloc[indices.flatten()]

# Step 3: Compare outcomes between treated and matched control groups
treatment_effect = treated['Revenue'].mean() - matched_control['Revenue'].mean()
print(f"Average Treatment Effect (Temperature > 25°C): {treatment_effect:.2f}")
```

### **Model: Granger Causality Test**

The Granger Causality Test checks whether changes in one variable **(like Temperature** can predict future changes in another variable **(like Revenue)**. By setting maxlag=2, the test examines whether temperature from up to 2 time periods earlier provides useful information for forecasting revenue. The test works by fitting models that include and exclude lagged values of temperature and compares their performance. If including past temperature values significantly improves the prediction of revenue, it suggests that temperature Granger-causes revenue. This test helps to identify directional relationships between variables in time-series data, rather than just a correlation.

```python
# Granger Causality Test with lag=2
granger_test = grangercausalitytests(data[['Revenue', 'Temperature']], maxlag=2, verbose=True)
```

The Granger Causality test results show that past temperature values (up to 2 lags) do not significantly predict revenue. The p-values are above 0.05 for both lag 1 and lag 2, meaning there is no evidence to reject the null hypothesis. This indicates that temperature and revenue may not have a time-dependent relationship, and any connection between them is likely immediate or due to direct correlation. The low F-statistics further confirm that adding lagged temperature values does not improve the prediction of revenue. Overall, the test suggests that past temperatures do not causally impact revenue in this dataset.

### **Model: Regression Discontinuity Analysis**

A Regression Discontinuity Analysis is to check if crossing a temperature threshold (25°C) causes a significant change in revenue. It creates a treatment variable that separates data into two groups: one for temperatures above 25°C (treated) and another for temperatures 25°C or below (control). The regression model then estimates the effect of temperature and treatment on revenue, specifically looking at whether being in the treated group (higher temperatures) significantly increases revenue. The key output is the coefficient for the treatment variable, which shows the causal impact of crossing the 25°C threshold on revenue. 

This method helps isolate the effect of the treatment while accounting for other variables like temperature itself.

```python
# Define a treatment variable (e.g., high-temperature threshold)
data['Treatment'] = (data['Temperature'] > 25).astype(int)

# Fit a Regression Discontinuity Model
rd_model = smf.ols('Revenue ~ Temperature + Treatment', data=data).fit()

# Summarize results
print("\nRegression Discontinuity Analysis Results:")
print(rd_model.summary())
```

The **Regression Discontinuity Analysis (RDA)** results show how crossing the temperature threshold (25°C) impacts revenue. The coefficient for **Treatment** (-2.0693) is small and not statistically significant (p-value = 0.572), meaning there is no evidence that crossing the threshold causes a sudden change in revenue. 

However, the coefficient for **Temperature** (21.5386) is highly significant (p-value < 0.000), showing that temperature itself has a strong positive effect on revenue regardless of the threshold. 
The **R-squared** value (0.980) indicates that the model explains 98% of the variability in revenue, suggesting temperature is a key driver of sales. 

Overall, this analysis suggests that while temperature strongly influences revenue, the specific threshold of 25°C does not cause a significant discontinuity.

### **Model: DoWhy Causal Model**

#### **Okay, But What's Do Why?**

**DoWhy** is a tool that helps us figure out cause-and-effect relationships in data rather than just identifying correlations. It works by creating **a clear framework where we can define what we think is causing something** (like temperature affecting revenue) and test those assumptions. It uses causal graphs to map out the relationships between variables and applies methods like the backdoor criterion to control for other factors that might influence the results. What’s great about DoWhy is that **it doesn’t stop at estimating the causal effect**—it also includes refutation tests to check if the effect is real or just a fluke. This makes it a powerful way to dig deeper into data and ensure we’re getting accurate, trustworthy results.

#### **How are we proceeding with this ?**

We are proceeding with DoWhy by first identifying the **treatment variable** (Temperature) and the **outcome variable** (Revenue) to define the causal relationship we want to study. Then, we create a **causal graph** to visualize the connections between variables and make assumptions about potential confounders. Once the graph is established, we use the **backdoor criterion** to estimate the causal effect, ensuring it accounts for any confounding influences. After estimating the effect, we perform **refutation tests**, such as placebo treatment and bootstrap sampling, to validate the robustness and reliability of the results. This process helps ensure that our findings are not only statistically significant but also truly causal.

```python

# Step 1: Build the causal model
causal_model = CausalModel(
    data=data,
    treatment=treatment_variable,
    outcome=outcome_variable,
    common_causes=None  # If you have confounders, list them here as column names
)

# Visualize the causal graph
causal_model.view_model()

# Step 2: Identify the causal effect
identified_estimand = causal_model.identify_effect(proceed_when_unidentifiable=True)
print("Identified Estimand:")
print(identified_estimand)

# Step 3: Estimate the causal effect
causal_estimate = causal_model.estimate_effect(
    identified_estimand,
    method_name="backdoor.linear_regression"  # Uses linear regression for estimation
)
print("Causal Estimate:")
print(causal_estimate)

# Step 4: Refute the causal estimate
refutation = causal_model.refute_estimate(
    identified_estimand,
    causal_estimate,
    method_name="placebo_treatment_refuter"  # Placebo test for robustness
)

print("Refutation Test 1 (Placebo Treatment):")
print(refutation)

# Refutation Test 2: Random subset refuter
refutation_subset = causal_model.refute_estimate(
    identified_estimand,
    causal_estimate,
    method_name="data_subset_refuter"  # Randomly subsets the data
)
print("\nRefutation Test 2 (Data Subset):")
print(refutation_subset)

# Refutation Test 3: Bootstrap Refuter
refute_bootstrap = causal_model.refute_estimate(
    identified_estimand,
    causal_estimate,
    method_name="bootstrap_refuter"
)
print("\nRefutation Test 3 (Bootstrap Refuter):")
print(refute_bootstrap)
```

The analysis confirms a robust causal relationship between **Temperature** and **Revenue**, with an Average Treatment Effect (ATE) of **21.44**, meaning that each 1-unit increase in temperature leads to a 21.44-unit rise in revenue. The **Placebo Treatment Refuter** yielded an effect of **0.0**, confirming that the observed causal effect is not due to random noise. Both the **Data Subset Refuter** and the **Bootstrap Refuter** produced nearly identical effects to the original estimate, demonstrating the consistency and stability of the causal inference across different subsets and resampled datasets. The high **p-values** from all refutation tests reinforce the reliability of the model and the estimated causal effect. Overall, the results provide strong evidence that temperature has a significant and consistent causal impact on revenue.
