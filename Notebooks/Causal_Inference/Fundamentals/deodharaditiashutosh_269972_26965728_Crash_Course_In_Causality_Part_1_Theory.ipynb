### So, what is causality, at all? [The formal definition]
Causality refers to the relationship between cause and effect, where one event (the cause) influences the occurrence of another event (the effect). 

### What is it's significance in Data Science?

In the context of data science, causality seeks to understand and quantify the direct impact of one variable on another within a system. This understanding allows us to make predictions, intervene effectively, and infer causal relationships from observational data i.e.:

1. **Informed Decision Making**: Understanding causal relationships enables us to make informed decisions and take appropriate actions based on the expected outcomes.

2. **Predictive Modeling**: Incorporating causal factors into predictive models enhances their accuracy and reliability by capturing the underlying mechanisms driving observed phenomena.

3. **Policy and Intervention Design**: Causal inference guides the design and evaluation of policies, interventions, and treatments aimed at achieving desired outcomes in various domains, including healthcare, economics, and social sciences.

4. **Avoiding Misinterpretation**: Distinguishing between correlation and causation helps prevent erroneous conclusions and misleading interpretations of data.

### Is Causality same as Correlation?




**Correlation** refers to a statistical association between two variables, indicating that changes in one variable are related to changes in another. However, correlation does not imply causation. Just because two variables are correlated does not mean that one causes the other. There could be other factors influencing both variables or a spurious relationship due to chance.

**Causation**, on the other hand, implies a direct cause-and-effect relationship between variables. Establishing causation requires demonstrating that changes in one variable directly lead to changes in another, controlling for other factors that might influence the relationship.

### Randomized Controlled Trials (RCTs) 
Randomized Controlled Trials (RCTs) are research studies where participants are randomly assigned to either receive an intervention (treatment group) or not (control group). 
This random assignment helps ensure groups are comparable, minimizing bias. 
RCTs are widely used to evaluate the effectiveness of interventions and treatments. They involve blinding, where participants and researchers are unaware of group assignments to minimize bias. 
Outcome measures are assessed, and statistical analysis is conducted to compare outcomes between groups. RCTs provide strong evidence for causality but can be costly and have ethical considerations. Overall, they are a gold standard for evaluating interventions in various fields

### What are some fundamental concepts that we need to know?

1. **Confounders:** <br>
A confounding variable (also known as confounder) is a variable that influences both the treatment and outcome, causing a _spurious association_. Simply put, variables that are not in our control but do have effects on the inference.
<br>
<br>
2. **Counterfactuals:** <br>
Counterfactuals are hypothetical statements about what would have happened if a particular event or action had not occurred. In the context of causality, counterfactuals help us compare the observed outcome with what would have happened under alternative scenarios, allowing us to assess the causal impact of an intervention or treatment.
<br>
<br>
3. **Potential Outcome:**<br>
The potential outcomes framework formalizes the concept of causality by defining potential outcomes for each unit under different treatment conditions. It posits that each unit has potential outcomes corresponding to each possible treatment level, and causal effects are estimated by comparing observed outcomes to the counterfactual outcomes that would have occurred under alternative treatments.
<br>
<br>
4. **Selection bias:**<br>
Selection bias is a phenomenon in which the distribution of the observed group is not representative of the group we are interested in.
<br>
<br>
5. **Causal Graphs:**<br>
Causal graphs, also known as Directed Acyclic Graphs (DAGs), visually represent causal relationships between variables in a system. Nodes in the graph represent variables, and directed edges indicate causal relationships. Causal graphs help clarify the directionality of causal effects and identify confounding variables that need to be controlled for in causal inference analyses.

### Typical Assumptions

1. **SUTVA (Stable Unit Treatment Value Assumption):** <br>

   The Stable Unit Treatment Value Assumption (SUTVA) is a fundamental principle in causal inference. 
 - It posits that the potential outcomes for any unit or sample should not vary with the treatment assigned to other units.  Additionally, it assumes that there is only one version of each treatment, ensuring consistency in the treatment application across all units. 
- Adhering to SUTVA is crucial for ensuring the validity of causal inference methods, particularly in experimental studies.
- Violations of SUTVA can lead to biases and compromise the accuracy of causal effect estimates derived from empirical data. Therefore, researchers must carefully consider and assess SUTVA assumptions when conducting and analyzing studies to draw valid causal conclusions.

2. **Ignorability:** <br>

   The principle of ignorability, also known as the "no unmeasured confounding" assumption or "conditional exchangeability," is a fundamental concept in causal inference. It states that given the observed covariates, the treatment assignment is independent of potential outcomes. 
   In simpler terms, it means that after accounting for the measured covariates, there are no hidden or unmeasured factors that affect both the treatment assignment and the outcome.

    In mathematical terms, the principle of ignorability can be expressed as:

    Y(1), Y(0)  T | X 

    Where:
    -   Y(1) represents the potential outcome under treatment.
    -   Y(0) represents the potential outcome under no treatment.
    -    T represents the treatment assignment.
    -    X  represents the covariates or observed factors.

    This equation signifies that the potential outcomes  Y(1)  and  Y(0) are independent of the treatment assignment T , given the observed covariates X. In other words, once we condition on (or account for) the observed covariates, there should be no association between the treatment assignment and the potential outcomes.

    The principle of ignorability is essential for making causal inferences from observational data, especially when using methods like propensity score matching, inverse probability weighting, or regression adjustment. It allows researchers to estimate causal effects while controlling for observed confounding variables.

    _However, it's important to note that the principle of ignorability relies on the assumption that all relevant confounding factors are measured and included in the analysis. If there are unmeasured confounders or if the measured covariates are insufficient to capture all confounding effects, the principle of ignorability may be violated, leading to biased causal inference results.)_ 

### Types of treatment effects

Our objective for causal inference is to estimate the treatment effect from the observational data. The treatment effect can be measured at the population, treated group, subgroup, and individual levels. Here, we define the treatment effect under binary treatment, but it can be easily extended to multiple treatment cases. The definition for each type follows (see A Survey on Causal inference for more information).

1. **Average treatment effect (ATE):** At the population level, the treatment effect is called the average treatment effect (ATE), which is defined as:

    ATE = E[Y(T = 1)  Y(T = 0)]

    where T denotes the treatment, and Y(T = 1) and Y(T = 0) are the potential treated and control outcomes of the whole population, respectively.

2. **Average treatment effect on the treated (ATT):** For the treated group, the treatment effect is called the average treatment effect on the treated group (ATT), and is defined as:

    ATT = E[Y(T = 1)|T = 1]  E[Y(T = 0)|T = 1]

    where Y(T = 1)|T = 1 and Y(T = 0)|T = 1 are the potential treated and control outcomes of the treated group, respectively. ATT can also be called local average treatment effect (LATE).

3. **Conditional average treatment effect (CATE):** At the subgroup level, the treatment effect is called conditional average treatment effect (CATE), which is defined as:

    CATE = E[Y(T = 1)|X = x]  E[Y(T = 0)|X = x]

    where Y(T = 1)|X = x and Y(T = 0)|X are the potential treated and control outcomes of the subgroup with X = x respectively. CATE is also known as the heterogeneous treatment effect.

4. **Individual treatment effect (ITE):** At the individual level, the treatment effect is called individual treatment effect (ITE), and the ITE of unit i is defined as:

    ITE = Y (T = 1)  Y (T = 0)

    where Y (T = 1) and Y (T = 0) are the potential treated and control outcomes of unit i, , respectively. In related work from Kunzel et al. (2019), estimators that minimize the mean square error (MSE) of the individual treatment effect (ITE) of i also minimize the MSE for the CATE at x and so therefore the ITE is viewed equivalent to the CATE.

_Please note that ATE and ATT are estimated at population level, but CATE and ITE are estimated at the subgroup or individual level. CATE (as mentioned, also known as heterogeneous treatment effect) is where the treatment effect varies across different subgroups. When the goal is to understand program-level treatment effects, we focus on ATE/ATT. When the goal is to estimate the individual level treatment effect, the heterogeneous research design is employed to estimate CATE._

### Worked Example

Now let's say,

In an educational research study, we aim to evaluate the effectiveness of a new teaching method on student performance. 
The study seeks to address the following questions:

 - What is the causal impact of the new teaching method on student test scores?
 - How does socioeconomic status (SES) influence the relationship between the teaching method and test scores?
 - Are there any potential biases or confounding factors that need to be accounted for in the analysis?
 
To address these questions, we will:

 - Simulate student data including variables such as socioeconomic status (SES), teaching method (Traditional or New), and test scores.
 - Introduce confounding by artificially inflating the test scores of students with higher SES levels, reflecting real-world scenarios.
  -  Address potential selection bias by randomly assigning students to treatment (New teaching method) and control (Traditional teaching method) groups.
  - Employ counterfactual reasoning to compare the observed outcome (student test scores with the new teaching method) with what would have happened under alternative scenarios (counterfactual test scores without the new teaching method).
 - Calculate the treatment effect using both traditional methods and counterfactual reasoning to assess the causal impact of the new teaching method while accounting for potential biases and confounding variables.
 
Through this research study, we aim to provide insights into the effectiveness of the new teaching method and the influence of socioeconomic status on student performance, thus informing educational practices and policies.

```python
# Importing necesssary modules
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Simulating data: Student performance and socioeconomic status
np.random.seed(42)
n_students = 1000
```

```python
# Generating student data
student_data = pd.DataFrame({
    'SES': np.random.randint(1, 4, n_students),  # SES levels (1: low, 2: medium, 3: high)
    'TeachingMethod': np.random.choice(['Traditional', 'New'], n_students),  # Teaching method
    'TestScore': np.random.normal(loc=70, scale=10, size=n_students),  # Test scores
})
```

```python
# Introducing confounding: Higher SES students tend to perform better
student_data.loc[student_data['SES'] == 3, 'TestScore'] += 5
```

```python
# Add Treatment column: Randomly assign treatment to each student
student_data['Treatment'] = np.random.choice([True, False], n_students)
```

```python
# Defining function to simulate counterfactual scenarios
def simulate_counterfactual(data, treatment):
    counterfactual_data = data.copy()
    counterfactual_data['TeachingMethod'] = np.where(counterfactual_data['Treatment'] == treatment,
                                                     'Traditional', 'New')
    return counterfactual_data
```

We define a function simulate_counterfactual to simulate counterfactual scenarios by flipping the treatment variable for a given group (e.g., treatment group becomes control group and vice versa).

```python
# Defining function to calculate treatment effect using counterfactuals
def calculate_treatment_effect_counterfactual(data):
    treated_data = data[data['Treatment']]
    control_data = data[~data['Treatment']]
    treated_scores = treated_data['TestScore'].mean()
    counterfactual_data = simulate_counterfactual(data, True)
    counterfactual_scores = counterfactual_data['TestScore'].mean()
    return treated_scores - counterfactual_scores
```

We define a function calculate_treatment_effect_counterfactual to calculate the treatment effect using counterfactuals. It computes the difference in mean test scores between the actual treated group and the counterfactual scenario where the treated group received the alternative treatment.

```python
# Calculating treatment effect using counterfactuals
counterfactual_treatment_effect = calculate_treatment_effect_counterfactual(student_data)
print("Counterfactual Treatment Effect:", counterfactual_treatment_effect)
```

We calculate the treatment effect using counterfactual reasoning based on the simulated data.

Now, we want to model the causal relationship between socioeconomic status, education level, and income. Thus, we'll create a causal graph to represent the causal relationships between these variables.

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Creating a directed graph
G = nx.DiGraph()

# Adding nodes (variables)
G.add_nodes_from(['Socioeconomic Status', 'Education Level', 'Income'])

# Adding directed edges (causal relationships)
G.add_edge('Socioeconomic Status', 'Education Level')
G.add_edge('Education Level', 'Income')

# Drawing the graph
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='skyblue', font_size=12, font_weight='bold')
plt.title("Causal Graph: Socioeconomic Status, Education Level, and Income")
plt.show()
```
