## Introduction

Causality is a fundamental concept in understanding the relationships between variables in data science. It aims to answer questions like "What causes what?" or "What effect does one variable have on another?" Understanding causality is crucial for making informed decisions, predicting outcomes, and designing effective interventions. In this crash course, we will explore key concepts and methods related to causality in the context of data science.

## Key Concepts

1. **Causal Inference**:
    Causal inference is the process of drawing conclusions about causal relationships based on observed data. It aims to determine whether a change in one variable directly influences the change in another variable. This process is essential for understanding cause-and-effect relationships, predicting outcomes, and making informed decisions.

2. **Causal Graphs**:
    Causal graphs, also known as causal diagrams or directed acyclic graphs (DAGs), are graphical representations of causal relationships between variables. In a causal graph, nodes represent variables, and directed edges indicate causal relationships between variables. By visually representing causal relationships, causal graphs help in identifying confounding variables and determining the causal effects of interventions or actions.

3. **Counterfactuals**:
    Counterfactuals are hypothetical scenarios that describe what would have happened if a certain intervention or action had taken place. In causal inference, counterfactuals are used to compare observed outcomes with the outcomes that would have occurred under different conditions. By comparing actual outcomes with counterfactual outcomes, researchers can estimate the causal effects of interventions or actions.

4. **Confounding Variables**:
    Confounding variables are extraneous factors that are associated with both the treatment variable and the outcome variable in a study. These variables can distort the true relationship between the variables of interest, leading to biased estimates of causal effects. Proper control of confounding variables is essential for accurate causal inference and can be achieved through study design, statistical methods, or randomization.

5. **Randomized Controlled Trials (RCTs)**:
    Randomized controlled trials (RCTs) are experimental designs where participants are randomly assigned to treatment and control groups. In RCTs, the treatment group receives the intervention or treatment being studied, while the control group does not. RCTs are considered the gold standard for establishing causal relationships because they help minimize confounding and other sources of bias. By randomly assigning participants to groups, RCTs ensure that any differences between the groups are due to the intervention or treatment being studied.

## Methods for Causal Inference:

1. **Propensity Score Matching (PSM)**:

    Propensity Score Matching is a statistical technique used in observational studies to balance covariates between treatment and control groups. In observational studies, treatment assignment is not randomized, leading to potential confounding biases. PSM addresses this issue by creating a propensity score, which is the probability of receiving treatment given observed covariates.

     - Estimating Propensity Score: First, a model (e.g., logistic regression) is used to estimate the propensity score for each individual in the dataset. The propensity score represents the likelihood of being treated based on observed covariates.

     - Matching Individuals: Next, individuals in the treatment group are matched with individuals in the control group who have similar propensity scores. This matching process balances the distribution of covariates between the two groups, mimicking the randomization achieved in randomized controlled trials.

     - Analysis: After matching, causal effects can be estimated by comparing outcomes between the matched treatment and control groups, accounting for any remaining differences in observed covariates.
<br>

2. **Instrumental Variables (IV)**:

    Instrumental Variables are used to estimate causal effects in situations where there may be unobserved confounding variables that affect both the treatment variable and the outcome variable. An instrumental variable is a variable that affects the treatment variable but is unrelated to the outcome except through its effect on the treatment.

    - Identification: To use instrumental variables, researchers must identify variables that meet two criteria:

        * Firstly, the instrumental variable must be correlated with the treatment variable.
        * Secondly, the instrumental variable must be unrelated to the outcome variable except through its influence on the treatment variable.
   
   - Estimation: Once suitable instrumental variables are identified, causal effects can be estimated using instrumental variable estimation techniques, such as Two-Stage Least Squares (2SLS) regression.
<br>

3. **Regression Discontinuity Design (RDD)**:

    Regression Discontinuity Design is a quasi-experimental design used to estimate causal effects when treatment assignment is determined by a cutoff point or threshold in a continuous variable.

    - Design: In RDD, individuals are assigned to the treatment or control group based on whether they fall above or below a predetermined threshold in a continuous variable. This assignment is assumed to be as good as random near the cutoff point.

    - Estimation: Causal effects are estimated by comparing outcomes just above and below the cutoff point. By focusing on individuals close to the threshold, RDD controls for observed and unobserved factors that vary continuously across the threshold.
<br>

4. **Difference-in-Differences (DID)**:

    Difference-in-Differences is a method used to estimate causal effects by comparing changes in outcomes between a treatment group and a control group before and after an intervention.

    - Design: DID requires data from both treatment and control groups, observed over multiple time periods. The treatment group experiences the intervention, while the control group does not.

    - Estimation: Causal effects are estimated by comparing the difference in outcomes between the treatment and control groups before and after the intervention. DID helps control for time-invariant confounders by differencing out their effects.

## Introducing Causal Inference: Linear Regression

Now, we'll perform the linear regression to explore the relationship between the Starting Grid position and the Position_numeric (final race position). This analysis will help us understand if and how the starting grid position impacts the final race outcome.

```python
import statsmodels.api as sm

# Prepare the data for regression
X = sm.add_constant(df_filtered['Starting Grid'])  # Adds a constant term to the predictors
y = df_filtered['Position_numeric']

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression
model_summary = model.summary()
model_summary
```

The linear regression results provide insights into the relationship between the starting grid position and the final race position:

- **Coefficient for Starting Grid**: The coefficient for Starting Grid is approximately 0.5995. This means that, on average, each one-unit improvement in starting grid position is associated with a 0.5995 unit improvement in final race position. This supports the hypothesis that starting closer to the front of the grid is associated with finishing in a better position.

- **R-squared**: The R-squared value is 0.459, indicating that approximately 45.9% of the variance in the final race positions can be explained by the starting grid positions. While this is a significant portion, it also suggests that over half of the variance in race outcomes is determined by other factors not included in this model.

- **P-value**: The P-value for the Starting Grid coefficient is very close to 0, indicating that the effect of starting grid position on final race position is statistically significant.

### Linear Regression Analysis
The linear regression analysis revealed a statistically significant relationship between the starting grid position and the final race position of drivers in Formula 1 races. Specifically, the model indicated that an improvement in the starting grid position (i.e., a lower starting number) is associated with an improvement in the final race position, on average. The coefficient for the starting grid position was positive, suggesting that each unit decrease in the starting grid (moving closer to the front) is associated with a decrease (improvement) in the final race position. The statistical significance of this relationship, indicated by a very low p-value, supports the hypothesis that starting grid position causally affects race outcomes.

This analysis provides valuable insights into the dynamics of Formula 1 racing, highlighting the importance of qualifying performance as a determinant of race success. The starting grid position, determined by qualifying times, not only reflects the performance potential of the car and driver but also directly influences their strategic positioning and opportunities during the race. However, it's important to note that while linear regression can indicate relationships and their significance, establishing true causality in observational data often requires further methods to address potential confounding factors.
