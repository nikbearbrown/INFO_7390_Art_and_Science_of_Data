#CAUSALITY

Causality encompasses the connection between cause and effect, elucidating how an event or variable influences another, leading to observable changes. This principle is at the core of understanding the reasons behind specific occurrences and unraveling the underlying mechanisms governing observed behaviors or patterns. Essentially, causality seeks to answer the pivotal question: "What instigates a particular outcome?"

Exploring causality involves distinguishing between correlation and causation. Correlation denotes a statistical connection between two variables, implying that changes in one variable coincide with changes in another. However, correlation alone doesn't establish causation; it merely suggests a relationship without confirming that changes in one variable cause changes in the other.

Validating causation requires in-depth analysis and compelling evidence to showcase a direct cause-and-effect association. This process entails demonstrating that alterations in one variable directly lead to changes in another, often supported by empirical evidence, controlled experiments, or logical reasoning. For instance, understanding the relationship between smoking and health issues necessitates more than recognizing a correlation; it requires empirical studies and evidence demonstrating the direct causal link between smoking and specific health ailments.

**Understanding Causality:**

Comprehending causality involves untangling the connections between cause and effect, exploring the intricate links that clarify the reasons behind specific occurrences. It aims to distinguish between simple associations and genuine causal relationships, necessitating thorough scrutiny to establish direct connections between factors. This pursuit delves into understanding how alterations in one variable impact another, relying on empirical evidence, experiments, or logical analysis. Profound comprehension of causality empowers us to make informed decisions, foresee outcomes accurately, and unravel the complexities of our surroundings, bridging the divide between observation and understanding.

**Counterfactuals and Potential Outcomes:**

Potential Outcome Framework: This framework deals with understanding causal effects by comparing potential outcomes under different treatment conditions. It helps infer causal relationships by evaluating what would have happened under various scenarios.

**Counterfactuals:**

These are theoretical constructs imagining what would have occurred if a different action or condition had taken place. They provide a way to reason about causality by comparing actual outcomes to potential alternatives.

#PROPENSITY SCORE MATCHING

Propensity score matching is a statistical method used in observational studies to estimate the causal effect of a treatment, intervention, or exposure on an outcome. It helps to reduce selection bias by matching treated and untreated subjects based on their propensity scores, which are the estimated probabilities of receiving the treatment given observed covariates.

Propensity score matching is a statistical technique used in observational studies to reduce selection bias when estimating the causal effect of a treatment, intervention, or exposure. It aims to emulate the randomization process of controlled experiments by creating comparable groups from observational data.

In observational studies, treatment assignment might be influenced by various factors, leading to confounding variables that affect both treatment selection and outcomes. Propensity score matching addresses this by estimating the propensity score of receiving a treatment based on observed covariates. The goal is to balance these covariates between treated and untreated groups, making them more comparable and reducing the influence of confounding factors.



**Propensity Score Estimation:**

First, a logistic regression model is often used to estimate the propensity score for each individual. This model predicts the probability of receiving the treatment based on observed covariates.

**Matching:**

Individuals who received the treatment (treated group) are then matched with similar individuals from the control (untreated) group based on their propensity scores. Various matching algorithms (e.g., nearest neighbor, kernel, exact matching) pair individuals with similar or closest propensity scores.

**Assessing Balance:**

 After matching, the balance of covariates between treated and untreated groups is evaluated using statistical tests or visualization methods. The aim is to ensure that the matched groups are comparable in terms of observed characteristics.

**Estimating Treatment Effect:**

 Finally, the treatment effect is estimated by comparing the outcome of interest between the matched treated and untreated groups. This estimation is more robust against confounding, assuming the propensity score model adequately captures the treatment selection process.

**Advantages:**

Reduces Bias:

 Helps control for observed covariates, reducing bias due to confounding variables.
Applicability:

 Useful when random assignment to treatment is not feasible, common in observational studies.
Flexibility:

 Allows for various matching algorithms and sensitivity analyses.
Limitations:
Assumptions:

 Relies on the assumption that all relevant covariates are observed and correctly included in the propensity score model.

Covariate Balance:

 Matching can be challenging, especially in studies with limited sample sizes or when dealing with high-dimensional covariates.
Generalizability:

Matching might limit the generalizability of results if not done carefully.
Propensity score matching is a powerful tool in causal inference, enabling researchers to approximate the randomized controlled trial conditions in observational data. Careful consideration of model specification, matching methods, and assessing balance is crucial to derive valid causal inferences from observational studies.

Explanation:
Propensity Score Estimation:

The propensity score is the probability of receiving the treatment based on observed covariates.
Method: It is typically estimated using logistic regression, where the treatment assignment is regressed on observed covariates.
Matching:

Once the propensity scores are estimated, treated and untreated subjects are matched based on these scores.
Method: Several matching algorithms (e.g., nearest neighbor matching, exact matching, kernel-based matching) are used to pair treated and untreated subjects with similar or close propensity scores.
Balance Assessment:

After matching, it's crucial to check whether the covariates are balanced between the treated and untreated groups.
Method: Statistical tests or graphical representations are used to assess the balance of covariates. Balanced covariates suggest a successful matching process.
Causal Effect Estimation:

Once balance is achieved, the causal effect of the treatment on the outcome is estimated.

Method: Various statistical techniques like difference in means, regression analysis are employed to calculate the average treatment effect on the treated  or other effect measures.

Let's say,


After obtaining the propensity scores, matching is performed using various methods to pair treated and untreated subjects based on these scores. For instance, if using nearest neighbor matching, for each treated subject, an untreated subject with the closest propensity score is selected as a match.

Once matching is completed, the balance of covariates between the treated and untreated groups is assessed. If the covariates are balanced, the causal effect can be estimated using various statistical techniques, such as calculating the difference in means of the outcomes between the matched treated and untreated groups.

Propensity score matching helps to mitigate bias in observational studies by mimicking some aspects of a randomized controlled trial, providing a more reliable estimation of causal effects from observational data.

Unlike RCTs, where treatment allocation is random, in observational studies, treatment groups might not entirely overlap, or certain characteristics may not be equally distributed between treatment groups. A pipeline in machine learning is a sequential set of data processing steps that are chained together to streamline the workflow. Each step in the pipeline performs a specific transformation or operation on the data and passes the output to the next step.

In the context, a pipeline is designed to build an estimator or a predictive model for the treatment variable ('T') based on the features ('X_encoded').

**Data Preprocessing:**

**Standardization:** The pipeline starts with a data preprocessing step using a StandardScaler(). This scaler transforms the input features ('X_encoded') by standardizing them, ensuring that each feature has a mean of 0 and a standard deviation of 1. It helps in bringing all features to a similar scale, preventing certain features from dominating the modeling process due to their larger magnitudes.

**Model Building:**

**Classifier :** After the data preprocessing step, the pipeline incorporates a machine learning classifier, such as logistic regression (lr()). This classifier is used to learn the patterns or relationships between the standardized features and the treatment variable ('T'). Logistic regression is commonly used for binary classification tasks, predicting a binary outcome based on input features.

**Training the Pipeline:**

**Fitting the Pipeline:** The pipeline is trained or 'fit' using the .fit() method, where the standardized features ('X_encoded') and the treatment variable ('T') are provided as inputs. This fitting process involves applying the data transformations defined in the pipeline (such as standard scaling) and training the logistic regression classifier on the processed data.

```python
# Pipeline to build the treatment estimator
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic_classifier', lr())
])

pipe.fit(X_encoded, T)
```

```python
predictions = pipe.predict_proba(X_encoded)
predictions_binary = pipe.predict(X_encoded)
```
