# **Correlation vs. Causation </br>**

Before delving deeper, it's crucial to distinguish between correlation and causation. Correlation refers to a statistical connection between two variables, indicating that changes in one variable coincide with changes in another. However, correlation alone does not imply causation; it merely suggests a relationship without confirming a direct cause-and-effect link.


Understanding the distinction between correlation and causation is crucial for drawing accurate inferences from data.

Correlation refers to a statistical association between two variables. It indicates that the values of the variables tend to change together, but it doesn't establish a cause-and-effect relationship.

Causation, on the other hand, implies that one variable (the cause) directly influences the other variable (the effect). A change in the cause variable leads to a change in the effect variable.


**Why Correlation Does Not Equal Causation**


There are several reasons why observing a correlation between two variables doesn't necessarily imply causation:

**Reverse Causation:**  It's possible that the seemingly causal variable is actually the effect. For instance, people who frequently get sunburned might seek relief by consuming ice cream, leading to a correlation between sunburn rates and ice cream sales, but not a causal relationship.

**Lurking Variables**: A third, unaccounted-for variable might be influencing both of the observed variables. In the ice cream and sunburn example, the lurking variable could be sunny weather. Sunny weather prompts people to buy ice cream and also increases UV exposure, leading to sunburns.

**Coincidence: ** In some cases, the observed correlation might be purely coincidental. Random fluctuations can create spurious correlations that lack a true underlying connection.

**Importance of Distinguishing Correlation and Causation**

Misinterpreting correlation as causation can lead to flawed decision-making.  For instance, a company might mistakenly invest in marketing ice cream based on the correlation between ice cream sales and sunburns, neglecting the true causal factor (sunny weather) as shown below in the image.

By carefully considering the potential explanations for a correlation, we can make more informed inferences from data and avoid drawing misleading conclusions.

**Additional Considerations**

Statistical methods can help assess the strength and direction of a correlation, but they cannot definitively establish causation.

Controlled experiments, where one variable is manipulated while others are held constant, are a powerful tool for investigating causal relationships.

## **3. Causal Inference Methods:**

Various methods have been developed for causal inference using graphical models. Two widely used criteria are the Back-Door Criterion and the Front-Door Criterion.

Back-Door Criterion: This criterion helps identify causal relationships between variables by identifying and blocking all back-door paths from the cause to the effect. By conditioning on the appropriate variables, we can isolate the causal effect of interest.

Front-Door Criterion: In cases where there is an unobserved confounder between the cause and effect, the front-door criterion provides a method for estimating causal effects by considering a mediator variable that lies on the causal pathway between the cause and effect.

These criteria provide principled ways to identify causal relationships from observational data, even in the presence of confounding variables.

# **Understanding Propensity Score Matching**

# **Introduction**:</br>
Propensity score matching is a statistical method used in research studies to understand the impact of a treatment or intervention on an outcome. It's like finding twins in a non-twin world â€“ we're trying to match individuals who received a treatment with those who didn't, but are very similar in other respects.

# **Why Propensity Score Matching?**</br>

In real life, people choose whether or not to undergo treatment, and these choices might be influenced by various factors. This can make it hard to tell if the treatment itself caused any changes or if it was something else. Propensity score matching helps us deal with this issue by creating fair comparisons between treated and untreated groups.


# **How Does it Work?**</br>
First, we estimate the likelihood (propensity score) of someone getting the treatment based on factors like age, gender, and health status. Then, we pair up individuals who got the treatment with similar ones who didn't, based on their propensity scores. This creates balanced groups where the only difference is whether they got the treatment or not.
 Assessing the Match:
After pairing people up, we check if they're really similar in terms of their characteristics. If they are, it suggests our matching worked well. This step is crucial because we want to be confident that any differences in outcomes between the two groups are due to the treatment, not other factors.

# **Estimating the Treatment Effect:**</br>
Once we've matched everyone up nicely, we can compare the outcomes between the treated and untreated groups. By looking at how things turned out for both groups, we can get a good idea of how effective the treatment really is.

# **Pros and Cons:**</br>
Propensity score matching helps reduce bias and is handy when we can't randomly assign treatments. But it's not foolproof â€“ we have to make sure we're comparing apples to apples and that we're not missing any important details.


Propensity score matching is like playing matchmaker between treated and untreated individuals in research studies. By creating balanced groups, it helps us tease out the true effects of treatments, even in situations where random assignment isn't possible. Just like matchmaking, it takes some care and attention to get it right, but when done well, it can lead to valuable insights into cause and effect.

```python
def perform_match_exact(row, df , *args):
    # row is the the item that we want to match
    # df is the source Pandas dataframe that we want to match it with other items
    # print('Start matching')
    sub_set = df

    for arg in args:
        sub_set = sub_set.loc[sub_set[arg] == row[arg]]
        # print(sub_set)

    return sub_set.index


def logit(p):
    logit_value = math.log(p / (1-p))
    return logit_value

def hasCabin(x):
    if pd.isna(x):
        return 0
    else:
        return 1

def cohenD (tmp, metricName):
    treated_metric = tmp[tmp.treatment == 1][metricName]
    untreated_metric = tmp[tmp.treatment == 0][metricName]

    d = ( treated_metric.mean() - untreated_metric.mean() ) / math.sqrt(((treated_metric.count()-1)*treated_metric.std()**2 + (untreated_metric.count()-1)*untreated_metric.std()**2) / (treated_metric.count() + untreated_metric.count()-2))
    return d
```

The above functions provide tools for data manipulation (perform_match_exact), mathematical transformation (logit), feature engineering (hasCabin), and effect size calculation (cohenD). These are used together to analyze and compare data from the Titanic dataset, especially in the context of survival analysis.

# **Machine Learning Pipelines**: Automating the Workflow
Imagine you're building a house. You wouldn't just dump all the materials in a pile and hope for the best, right? Instead, you'd follow a series of steps: foundation, framing, plumbing, electrical, and so on. In machine learning, a pipeline is similar. It's a structured approach that automates the process of preparing and using data to build a model.

**The Titanic Example: Predicting Cabin Class**
Let's say you're analyzing the Titanic dataset and want to predict whether a passenger had a cabin (treatment) based on factors like sex, age, and passenger class. Here's a simplified pipeline for this task:

**Data Preparation:**

We'll focus on specific features like sex, age, and passenger class that might influence cabin assignment.

**Categorical features** (like sex) need special treatment. We'll use a technique called one-hot encoding to convert them into numerical features that machine learning models can understand.

**Feature Scaling (Standardization):**

Imagine features like age or fare. Values can range widely. We wouldn't want one feature to dominate the model just because it has larger numbers. Standardization ensures all features are on a similar scale (usually with a mean of 0 and a standard deviation of 1). This creates a fairer training ground for the model.

**Model Building:**

We'll use a logistic regression model as our building block. This is a common choice for predicting binary outcomes (like having a cabin or not) based on several features.

**Training the Pipeline:**

Finally, we train the entire pipeline in one go. This involves feeding the prepared data (encoded features and the target variable - cabin presence) into the pipeline. The pipeline performs the steps we defined (encoding, scaling, training the model) and learns the relationship between features and cabin presence.

**Benefits of Pipelines:**

*Efficiency:* Pipelines automate repetitive tasks, saving time and reducing errors.
*Reproducibility:* By keeping all steps together, it's easier to re-run the analysis or share it with others.
*Flexibility:* You can swap different preprocessing steps or models within the pipeline for experimentation.

This is a basic example, but hopefully, it gives you a taste of how machine learning pipelines can streamline the model building process!

```python

# Pipeline to build the treatment estimator
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic_classifier', lr())
])

pipe.fit(X_encoded, T)
```

```python

predictions = pipe.predict_proba(X_encoded)
predictions_binary = pipe.predict(X_encoded)
```

# **Propensity Score Distribution**

**Balancing the Scales: Understanding Propensity Scores** </br>
Imagine you're studying the effects of a new medication. Ideally, you'd give it to everyone and see the results. But in reality, you might only be able to test it on a specific group. Here's where propensity scores come in.

**What are Propensity Scores?**

Propensity scores are like personalized scores that estimate how likely someone was to receive a particular treatment (like the new medication) based on their background (age, health history, etc.). They're kind of like predicting who would have gotten the medication even if they weren't part of the test group.

**Why are Propensity Scores Important?**

**Fair Comparisons:** If the treatment group is very different from the non-treatment group (e.g., mostly young and healthy people received the medication), it's hard to say if the medication itself caused the results. Propensity scores help create a fairer comparison by considering these background differences.

**Visualizing Propensity Scores:**

**Distribution Plots**: We can plot the distribution of propensity scores for both the treatment group (who got the medication) and the non-treatment group (who didn't). This helps us see how the scores are spread out and if there's any overlap between the groups.

**Logit Transformation:**  Sometimes, raw scores can be bunched up at the extremes (0 or 1). The logit transformation stretches these scores out, making it easier to see if there's a clear separation between the treatment and non-treatment groups.

**The Goal:**

The ideal scenario is to see separate peaks in the distributions for the treated and untreated groups. This suggests the model effectively distinguished who received the treatment based on their characteristics. The logit transformation might help visualize these differences more clearly.

The above code demonstrates a common technique in causal inference analysis called propensity score matching. </br>
By creating propensity scores and using a matching algorithm like KNN, the aim is to create comparable groups of treated and untreated individuals, allowing for a more fair assessment of the treatment effect.

```python
# Common support distances and indexes
distances , indexes = knn.kneighbors(
    df_data[['propensity_score_logit']].to_numpy(), \
    n_neighbors=10)

print('For item 0, the 4 closest distances are (first item is self):')
for idx, df in enumerate(distances[0,0:4]):
    print('Element distance {}: {:4f}'.format(idx, df))
print('xyz')
```

```python

print('For item 0, the 4 closest indexes are (first item is self):')
for idx in indexes[0,0:4]:
    print('Element index: {}'.format(idx))
print('xyz')
```

The code is identifying similar individuals based on their propensity scores using a KNN approach.</br>
The distances indicate how close different individuals are in terms of their predicted probabilities of receiving the treatment (e.g., passengers with similar survival probabilities on the Titanic).</br>
The indexes provide a way to locate those similar individuals within the dataset for further analysis or matching.

```python
def perfom_matching_v2(row, indexes, df_data):
    current_index = int(row['index']) # Obtain value from index-named column, not the actual DF index.
    prop_score_logit = row['propensity_score_logit']
    for idx in indexes[current_index,:]:
        if (current_index != idx) and (row.treatment == 1) and (df_data.loc[idx].treatment == 0):
            return int(idx)

df_data['matched_element'] = df_data.reset_index().apply(perfom_matching_v2, axis = 1, args = (indexes, df_data))
```

The function performs the following steps:

**Extracts Information:**

It retrieves the current data point's index (current_index) from a column named "index" (assuming it exists) in the row.
It obtains the data point's logit-transformed propensity score (prop_score_logit) from the row.
Iterates Through Nearest Neighbors:

It loops through the nearest neighbor indexes (indexes) corresponding to the current data point (current_index).

Matching Criteria:

Inside the loop, it checks for three conditions:</br>
current_index != idx: Ensures the neighbor is not the data point itself.
row.treatment == 1: Checks if the current data point belongs to the treatment group.
df_data.loc[idx].treatment == 0: Checks if the neighbor belongs to the control group.

Match Found (Return):

If all conditions are met, it means a suitable match has been found. The function returns the index (idx) of the matched data point (neighbor) within the DataFrame.

This code snippet implements a matching strategy based on propensity scores using a KNN approach. The perfom_matching_v2 function finds suitable control group members for treated individuals based on their propensity scores.

# **Average Treatment Effect (ATT)**

Imagine a science experiment: You want to see if giving vitamin C to people helps prevent them from getting a cold. You randomly assign half the people (treatment group) to take vitamin C daily, while the other half (control group) don't get any. Ideally, the two groups would be identical in every way except for taking vitamin C.

However, in the real world, it's often difficult to achieve perfect randomization. Maybe people who are more health-conscious are more likely to volunteer for the experiment, and they might also be less likely to get colds anyway. This is called confounding â€“ a lurking variable that affects both getting a cold and being assigned to take vitamin C.

**Enter ATT (Average Treatment Effect):**

ATT is a statistical measure that tries to estimate the causal effect of a treatment (vitamin C) by comparing what happens in the treatment group (people taking vitamin C) to what would have happened to them if they hadn't received the treatment (their counterfactual outcome).

Propensity Score Matching to the Rescue!

Since we can't rewind time and make everyone not take vitamin C, propensity score matching helps us create a better comparison group. It does this by finding people in the control group (who didn't take vitamin C) who are similar to the people in the treatment group (who did) based on various characteristics (age, health status, etc.). This reduces the influence of confounding variables and allows for a fairer estimate of the treatment effect.

**ATT = 0.1049:**

In the context of the Titanic data, the "treatment" could be being in a lifeboat (increased chance of survival).  ATT would then try to estimate the average difference in survival rates between those who were in lifeboats and those who weren't, accounting for factors that might have influenced survival such as age, gender, and class.

An ATT of 0.1049 suggests that, on average, being in a lifeboat increased a person's chance of survival on the Titanic by approximately 10.49%. However, it's important to consider limitations:

Propensity score matching isn't perfect, and some confounding factors might remain.
ATT is an average effect â€“ the actual impact of being in a lifeboat could vary depending on individual characteristics.

This study investigates the causal effect of being in a lifeboat on the survival rate of passengers aboard the RMS Titanic.  While a randomized controlled trial assigning passengers to lifeboats would provide the most robust estimate of this effect, such an experiment is not ethically feasible in this historical context.  Observational data, however, is readily available, prompting the use of alternative methods to account for potential confounding variables.

**Propensity score matching (PSM)** is a statistical technique employed to reduce bias in observational studies.  Here, we outline the steps involved in applying PSM to the Titanic survival data and discuss the estimated causal effect of being in a lifeboat.

1. **Identifying Confounding Variables** </br>
The first step involves identifying characteristics that might influence both the likelihood of being assigned to a lifeboat (treatment) and the probability of survival (outcome).  In the context of the Titanic disaster, these confounding variables might include passenger age, gender, and socioeconomic status (as reflected by travel class).

2. **Estimating Propensity Scores**</br>
Logistic regression is employed to estimate the propensity score for each passenger.  This score represents the predicted probability of being assigned to a lifeboat based on the passenger's characteristics.  Essentially, the propensity score creates a measure of how "similar" a passenger in the control group (not assigned to a lifeboat) is to a passenger in the treatment group (assigned to a lifeboat), excluding the fact that one was in a lifeboat and the other was not.

3. **Matching Treatment and Control Groups** </br>
Passengers in the control group (not assigned to a lifeboat) are matched with passengers in the treatment group (assigned to a lifeboat) based on their propensity scores.  The goal is to create comparable groups with respect to all measured characteristics except for the treatment itself (being in a lifeboat).  This matching process helps to mitigate the influence of confounding variables on the estimated treatment effect.

4. **Assessing Balance** </br>
Several graphical techniques, such as jitter plots and density plots, can be used to assess the effectiveness of the matching process.  Ideally, these plots should reveal similar distributions of covariates (age, gender, class) between the matched treatment and control groups after matching.  This visual confirmation helps to ensure that the matching procedure has successfully balanced the groups on relevant characteristics.

5. **Estimating the Average Treatment Effect (ATT)** </br>
Once the matching process is complete, the ATT can be estimated.  The ATT represents the average difference in the outcome variable (survival in this case) between the treatment and control groups after accounting for confounding variables through matching.  In the context of the Titanic, the ATT would indicate the average increase (or decrease) in survival probability for a passenger due to being assigned to a lifeboat.



It is important to acknowledge that PSM is not a perfect solution.  Unobserved confounding variables may still influence the estimated treatment effect.  Additionally, the ATT represents an average effect, and the impact of being in a lifeboat could vary depending on individual passenger characteristics.
