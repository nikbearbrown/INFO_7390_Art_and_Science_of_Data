# Theory
**Foundational Concepts of Causality Principles**

Causality lies at the heart of understanding relationships in data, distinguishing between correlation and causation. While correlation measures the degree to which two variables move together, causality seeks to determine whether one variable directly influences another. This distinction is crucial for making informed decisions, as relying on correlations alone can lead to spurious conclusions. For example, while there may be a correlation between higher education and increased income, causality requires establishing that education directly impacts earning potential, controlling for confounding factors like socioeconomic background or work experience.

Causal analysis often begins with the formulation of causal questions and the creation of causal graphs, also known as directed acyclic graphs (DAGs). These graphs visually represent the relationships between variables, highlighting potential confounders, mediators, and colliders. They guide the identification of the variables to control for when estimating causal effects. The process of causal inference relies on assumptions such as unconfoundedness (no unmeasured confounders) and the presence of sufficient variability in the treatment variable. By integrating domain knowledge and statistical rigor, causal analysis enables researchers to uncover deeper insights into the mechanisms driving observed outcomes.

Causality: A Comprehensive Overview
Causality is the study of cause-and-effect relationships, where the goal is to determine how one variable influences another directly. Understanding causality is critical for decision-making across various fields, including healthcare, economics, and social sciences. Unlike correlation, which merely identifies patterns or associations, causality establishes a directional and mechanistic connection between variables.

Key Concepts of Causality
Treatment Variable: The variable whose effect we want to measure (e.g., education level in the Adult Census dataset).
Outcome Variable: The variable influenced by the treatment (e.g., income).
Confounders: Variables that affect both the treatment and the outcome, potentially distorting the true causal relationship.
Estimand: The formal mathematical expression of the causal effect we aim to identify (e.g., Average Treatment Effect, ATE).
Unobserved Confounding: When hidden variables influence both treatment and outcome, introducing bias into causal estimates.
By systematically addressing these concepts, causal analysis enables researchers to move beyond superficial patterns and uncover actionable insights.

# Causal Trials and Randomized Controlled Experiments (RCTs)
One of the most rigorous ways to establish causality is through Randomized Controlled Trials (RCTs), often considered the gold standard in experimental design. RCTs eliminate bias by randomly assigning participants into treatment and control groups, ensuring any observed differences in outcomes are due to the treatment rather than confounding variables.

Key Features of RCTs:
Randomization:

Participants are randomly assigned to either the treatment or control group.
Randomization ensures that confounders are equally distributed between the groups, minimizing bias.
Control Group:

Serves as a baseline for comparison.
This group does not receive the treatment, allowing researchers to isolate the treatment's effect.
Blinding:

Single-Blind: Only the participants are unaware of their group assignment.
Double-Blind: Both participants and researchers are blinded to group assignments to prevent bias in administering or assessing treatments.
Outcome Measurement:

Measured consistently across both groups to ensure comparability.
Reduces variability unrelated to the treatment.
Advantages of RCTs:
Provides strong evidence of causality due to controlled conditions.
Reduces bias and confounding through randomization and blinding.
Establishes a clear timeline for treatment and outcome.
Limitations of RCTs:
Ethical Constraints:

Not all treatments can be randomized due to ethical concerns. For example, assigning harmful conditions is not permissible.
Cost and Time:

RCTs are resource-intensive and may not be feasible for large-scale studies or rare outcomes.
Generalizability:

Strict experimental conditions may not reflect real-world scenarios, limiting the applicability of results.
Example of an RCT:
In healthcare, a trial might test the effectiveness of a new drug:

Treatment Group: Receives the drug.
Control Group: Receives a placebo.
Outcome: Improvement in symptoms is compared between the two groups to determine the drug's efficacy.


# Observational Techniques in Causality
When randomized controlled trials (RCTs) are impractical due to ethical, financial, or logistical constraints, researchers rely on observational data to infer causality. Observational techniques aim to identify causal relationships from existing data where treatments or interventions are not randomly assigned.

Key Challenges in Observational Data:
Confounding Variables:

Variables that influence both the treatment and the outcome can introduce bias.
For example, in analyzing the effect of education on income, confounders like socioeconomic status or parental education need to be controlled.
Selection Bias:

Occurs when the treatment group differs systematically from the control group.
Example: People who work longer hours may inherently have better access to higher-paying jobs, confounding the analysis.
Unobserved Confounding:

Hidden variables that are not recorded in the dataset but influence the treatment and outcome.
Common Observational Techniques:
Propensity Score Matching (PSM):

A statistical technique that creates a balanced dataset by matching treated and untreated individuals based on similar covariates.
Steps:
Estimate the probability of receiving treatment (propensity score) for each individual.
Match individuals with similar scores from the treatment and control groups.
Compare outcomes between matched pairs to estimate the treatment effect.
Difference-in-Differences (DiD):

Compares the change in outcomes over time between a treatment group and a control group.
Example:
Analyzing the effect of a minimum wage increase on employment rates by comparing before-and-after changes in affected and unaffected regions.
Assumes that both groups would have followed similar trends in the absence of treatment.
Instrumental Variable (IV) Analysis:

Uses an external variable (instrument) that affects the treatment but not the outcome directly (except through the treatment).
Example:
Using distance to the nearest college as an instrument for analyzing the effect of education on income. Distance influences education decisions but does not directly affect income.
Regression Discontinuity Design (RDD):

Exploits cutoff-based policies to identify causal effects.
Example:
If a scholarship is granted to students scoring above a threshold, compare outcomes (e.g., future income) for students just above and below the threshold.
Structural Equation Modeling (SEM):

Combines statistical modeling and causal graphs to estimate direct and indirect effects in complex systems.
Useful when multiple variables interact in a web of causal relationships.
Advantages of Observational Techniques:
Leverages readily available data, reducing cost and time.
Allows for the study of treatments that cannot be randomized due to ethical or logistical constraints.
Limitations:
Reliance on strong assumptions (e.g., no unmeasured confounders).
Results are often sensitive to model specifications and variable inclusion.

# Data Preprocessing Techniques

Data preprocessing is a crucial step in preparing datasets for causal analysis and machine learning tasks. It ensures that the dataset is clean, structured, and suitable for robust analysis. Below, we discuss key preprocessing techniques:

1. Handling Missing Data
Missing data can significantly affect the quality of the analysis and lead to biased or unreliable results. Techniques to handle missing data include:

Dropping Missing Values: This is suitable when the number of missing values is minimal compared to the dataset's size. However, this method may lead to a loss of valuable information.
Imputation: Imputation techniques replace missing values with estimated ones, such as:
Mean/Median Imputation: For numerical features, replace missing values with the mean or median.
Mode Imputation: For categorical features, replace missing values with the most frequent value.
Advanced Imputation Methods: Use machine learning models or statistical methods to predict and fill missing values.
In the Adult Census dataset, placeholders like ? in columns such as workclass and occupation were replaced with the most frequent value using mode imputation, ensuring no rows were dropped unnecessarily.

2. Encoding Categorical Variables
Many machine learning models require numerical inputs, necessitating the transformation of categorical variables into numerical formats. Common techniques include:

Label Encoding: Assigns a unique integer to each category (e.g., Male = 0, Female = 1).
One-Hot Encoding: Creates binary columns for each category in a feature (e.g., for education, columns like education_Bachelors, education_HS-grad).
Target Encoding: Maps categories to their corresponding mean of the target variable, often used in specific machine learning tasks.
For the Adult Census dataset, categorical variables like workclass, education, and marital.status were transformed using a combination of label encoding and one-hot encoding. This ensured that the model could interpret categorical data accurately while avoiding multicollinearity.

3. Feature Scaling
Feature scaling ensures that numerical features have a uniform scale, which is particularly important for algorithms sensitive to feature magnitude (e.g., SVM, k-NN). Techniques include:

Normalization: Scales features to a range of [0, 1].
Standardization: Centers features by subtracting the mean and dividing by the standard deviation.
In our case, scaling was not critical since the causal analysis methods used (e.g., DoWhy's estimators) are scale-invariant. However, for predictive tasks, this step is recommended.

4. Feature Selection
Feature selection is essential for simplifying models, improving interpretability, and reducing overfitting. Strategies include:

Domain Knowledge: Select features based on their relevance to the causal question.
Correlation Analysis: Use heatmaps to identify and eliminate highly correlated features to prevent redundancy.
Statistical Methods: Use techniques like mutual information or recursive feature elimination (RFE) to identify impactful features.
In the Adult Census dataset, features such as age, hours.per.week, and categorical variables related to work and education were chosen as predictors for income. Correlation heatmaps helped identify relationships between numerical features and guided feature selection.

5. Outlier Detection
Outliers can distort model performance and causal estimations. Common methods for outlier detection include:

Visualization: Use box plots and scatterplots to spot anomalies.
Statistical Methods: Identify outliers using z-scores, the interquartile range (IQR), or robust statistical techniques.
Domain-Specific Rules: For example, in the Adult Census dataset, detecting and handling outliers in hours.per.week ensures realistic data interpretation.

# Data Encoding:
    
Preparing the Dataset for Machine Learning

To ensure the dataset is machine-learning ready, categorical variables were encoded using two techniques: binary encoding and one-hot encoding. Binary categorical variables, such as sex and income, were label encoded to map categories to numerical values. This encoding is efficient for binary attributes and avoids increasing dimensionality unnecessarily.

For multi-category categorical variables, such as workclass, education, and marital status, one-hot encoding was applied. The drop_first=True parameter was used to avoid multicollinearity by dropping the first category as a baseline.

The encoded DataFrame shows an increase in the number of columns due to one-hot encoding, reflecting the transformed categorical variables. This step is crucial for machine learning models that require numerical input and ensures all categorical data is represented in a format suitable for analysis.



```python
# Create a copy of the DataFrame to work on
df_encoded = df.copy()

# Encode binary categorical variables
binary_columns = ['sex', 'income']  # Update with other binary variables if any
encoder = LabelEncoder()

for col in binary_columns:
    df_encoded[col] = encoder.fit_transform(df_encoded[col])

# Apply One-Hot Encoding to multi-category variables
multi_category_columns = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'native.country']
df_encoded = pd.get_dummies(df_encoded, columns=multi_category_columns, drop_first=True)

# Display the first few rows of the encoded DataFrame
print("Encoded DataFrame:")
print(df_encoded.head())

# Check updated column count
print("\nNumber of columns after encoding:", len(df_encoded.columns))
```

# Defining the Causal Model

In the initial cell, the CausalModel from the dowhy library is instantiated to define the causal framework. This involves specifying:

Data: The preprocessed and encoded dataset (df_encoded).
Treatment Variable: education_Bachelors, representing whether an individual holds a Bachelor's degree.
Outcome Variable: income, indicating the income category (above or below 50K).
Confounders: A set of variables like workclass_Private, age, hours.per.week, and relationship_Not-in-family that are potential common causes affecting both treatment and outcome.
Visualizing the Causal Graph
The causal graph is generated using the view_model method, which provides a graphical representation of the relationships between variables. Nodes represent variables, and directed edges indicate potential causal effects. This graph is essential for understanding the assumptions and guiding further causal inference steps.

Observations from the Graph
The treatment variable education_Bachelors is directly linked to the outcome income.
Confounders like hours.per.week and relationship_Not-in-family influence both the treatment and outcome, emphasizing the need for their control during causal effect estimation.
This process sets the foundation for identifying causal estimands and quantifying the impact of the treatment variable on the outcome while accounting for confounding factors.








```python
# Define the causal model
causal_model = CausalModel(
    data=df_encoded,
    treatment="education_Bachelors",  # Example treatment variable
    outcome="income",  # Outcome variable
    common_causes=["workclass_Private", "age", "hours.per.week", "relationship_Not-in-family"]  # Confounders
)

# Visualize the causal graph
causal_model.view_model()
plt.show()
```

Following step focuses on identifying and estimating the causal effect of the treatment variable (education_Bachelors) on the outcome variable (income) using the causal model defined earlier.

Identifying the Causal Effect
Using the causal_model.identify_effect() method, the algorithm identifies the causal estimand based on the backdoor criterion. The backdoor estimand represents the mathematical formulation required to estimate the causal effect, accounting for confounding variables such as workclass_Private, age, hours.per.week, and relationship_Not-in-family. This step ensures that the causal effect is properly adjusted for confounding factors to derive accurate results.

Estimating the Causal Effect
With the causal estimand identified, the effect is estimated using linear regression via the causal_model.estimate_effect() method. Linear regression is employed as the statistical technique to calculate the average treatment effect (ATE) of education_Bachelors on income. The result quantifies the direct influence of holding a Bachelor's degree on income, controlling for the confounding variables.

This process demonstrates the practical application of causal inference, moving from theoretical identification to empirical estimation, ensuring robust and interpretable insights.

```python
# Identify the causal effect
identified_estimand = causal_model.identify_effect()
print("Identified Estimand:")
print(identified_estimand)
```

In this step, the causal effect of the treatment variable (education_Bachelors) on the outcome variable (income) is quantitatively estimated. The estimation process leverages the identified causal estimand and applies a linear regression model through the method "backdoor.linear_regression".

Key Components:
Estimated Causal Effect:

The backdoor criterion is used to adjust for confounders like workclass_Private, age, hours.per.week, and relationship_Not-in-family. This ensures the causal effect is isolated from spurious relationships.
Realized Estimand:

The mathematical relationship incorporates the treatment and confounders, producing a reliable representation of how education_Bachelors affects income.
Mean Value:

The calculated mean value of 0.1993 indicates the average causal effect of obtaining a Bachelor's degree on the probability of earning above a certain income threshold, after controlling for the identified confounders.
This result provides actionable insights into the direct influence of educational attainment on economic outcomes, emphasizing the importance of causality in deriving meaningful conclusions from data.

```python
# Estimate the causal effect using linear regression
causal_effect = causal_model.estimate_effect(
    identified_estimand,
    method_name="backdoor.linear_regression"
)
print("\nEstimated Causal Effect:")
print(causal_effect)
```

In this segment of the analysis, we focus on validating the causal estimates using a series of refutation methods provided by DoWhy. These methods aim to ensure the robustness and reliability of the estimated causal effects by challenging the assumptions and introducing perturbations to the dataset.

Placebo Treatment Refuter
The placebo treatment refuter introduces a hypothetical "placebo" treatment that is unrelated to the outcome. This method tests if the estimated causal effect can be attributed to the actual treatment or if it could occur by chance. The results showed that the placebo treatment yielded an effect close to zero, indicating the reliability of the causal effect associated with the actual treatment.

Data Subset Refuter
The data subset refuter evaluates the stability of causal estimates by analyzing subsets of the data. It randomly selects a subset of the dataset and re-estimates the causal effect. The results remained consistent across subsets, demonstrating the robustness of the causal effect estimate even when parts of the data are removed.

Bootstrap Refuter
The bootstrap refuter employs a resampling technique to generate new datasets by sampling with replacement. This method re-estimates the causal effect for each bootstrap sample to check for variations. The bootstrap results aligned closely with the original estimates, reinforcing the reliability of the causal conclusions.

```python
# Refute the estimated causal effect using a placebo treatment refuter
refutation_placebo = causal_model.refute_estimate(
    identified_estimand,
    causal_effect,
    method_name="placebo_treatment_refuter"
)
print("\nPlacebo Treatment Refuter Result:")
print(refutation_placebo)
```

```python
# Refute the estimated causal effect using a data subset refuter
refutation_subset = causal_model.refute_estimate(
    identified_estimand,
    causal_effect,
    method_name="data_subset_refuter"
)
print("\nData Subset Refuter Result:")
print(refutation_subset)
```

```python
# Refute the estimated causal effect using a bootstrap refuter
refutation_bootstrap = causal_model.refute_estimate(
    identified_estimand,
    causal_effect,
    method_name="bootstrap_refuter"
)
print("\nBootstrap Refuter Result:")
print(refutation_bootstrap)
```

```python
# Add Bootstrap Effect to the bar chart
effects = ["Original Effect", "Placebo Effect", "Subset Effect", "Bootstrap Effect"]
values = [
    causal_effect.value,
    refutation_placebo.new_effect,
    refutation_subset.new_effect,
    refutation_bootstrap.new_effect,
]

plt.bar(effects, values, color=["blue", "orange", "green", "red"])
plt.title("Comparison of Original, Placebo, Subset, and Bootstrap Effects")
plt.ylabel("Effect Value")
plt.show()
```

To visualize and compare the effects estimated through the original model and the refutation techniques, a bar chart was generated. The chart highlights the similarity between the original effect and the effects derived from placebo, subset, and bootstrap refuters. This consistency underscores the robustness of the estimated causal relationships in the model.

Overall, these refutation methods validate the causal analysis, providing confidence in the derived causal effect and affirming the model's reliability for drawing causal inferences.
