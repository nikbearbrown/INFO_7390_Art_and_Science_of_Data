####**Introduction to Causality**

**Causality**

Causality, refers to the relationship between cause and effect. It implies that a particular event (the cause) directly influences another event (the effect). In more technical terms, causality is the study of how actions, occurrences, or states can bring about changes in other actions, occurrences, or states.

In data science, understanding causality is crucial for making informed decisions based on data. While data can provide insights into patterns and relationships, causality goes a step further by explaining why those patterns exist and how changes in one variable might impact another. This understanding is vital in fields like healthcare, economics, and social sciences, where interventions based on causal understanding can lead to significant impacts.

**Differences between Correlation and Causation**

**Correlation:** This is a statistical measure that describes the extent to which two variables change together. However, correlation does not imply causation. It simply indicates that there is a relationship between the two variables.

**Causation:** Causation implies that changes in one variable bring about changes in another. It indicates a direct relationship where one event is the result of the occurrence of the other event.

**Examples**

**Correlation without Causation:** Suppose a study finds a correlation between ice cream sales and drowning incidents. While these two variables appear to be related, ice cream sales do not cause drowning incidents. Instead, a third variable, such as hot weather, is influencing both.

**Causation Example:** Consider a clinical trial where a new drug is tested for lowering blood pressure. If the study finds that patients taking the drug have significantly lower blood pressure compared to those not taking the drug, and other factors are controlled, it can be inferred that the drug causes a reduction in blood pressure.

**Spurious Correlation:** Sometimes, two variables may appear to be causally related but are actually influenced by an external factor.

For example, the number of people who drowned by falling into a pool correlates with films Nicolas Cage appeared in. While these two variables show a correlation, they certainly do not have a causal relationship. This is an example of a spurious correlation where the relationship is coincidental or due to a third variable.

####**Fundamentals of Causal Inference**

Causal inference is a central task in data science, aimed at understanding the effect of one variable (or set of variables) on another. Fundamental concepts:

**Counterfactuals and Potential Outcomes**

The concept of counterfactuals lies at the heart of causal inference. It involves considering what would have happened to an individual or a system under different circumstances. The term "potential outcomes" refers to the different possible states that could result from different treatments or interventions.

**Example:** Imagine a medical study aiming to evaluate the effect of a new drug on reducing blood pressure. For each patient, the potential outcome if they receive the drug (the treatment effect) and if they don't (the control effect) are considered. The difference between these outcomes is the causal effect of the drug.

However, in reality, we can only observe one of these outcomes for each individual, leading to the fundamental problem of causal inference we never observe the counterfactual outcome.

**Rubin Causal Model (RCM)**

**Overview:** Also known as the potential outcomes framework, the Rubin Causal Model provides a structured approach to thinking about causal inference. It posits that the causal effect of a treatment is the difference between the potential outcomes under the treatment and control conditions.

**Key Concepts:** The RCM emphasizes the importance of the assignment mechanism (how subjects are assigned to treatment or control groups) and the concept of "ignorability," which suggests that if the treatment assignment is independent of the potential outcomes, then causal effects can be estimated unbiasedly.

**Example:** In a randomized controlled trial (RCT), subjects are randomly assigned to treatment or control groups. This randomization helps ensure that the treatment assignment is independent of potential outcomes, allowing for an unbiased estimate of the causal effect.

**Causal Diagrams and Directed Acyclic Graphs (DAGs)**

**Causal Diagrams:** These are graphical representations used to express assumptions about the causal structure of the data. They help in visualizing and understanding the relationships between variables.

**Directed Acyclic Graphs (DAGs):** A DAG is a type of causal diagram where edges represent causal relationships and the graph contains no cycles (i.e., it is acyclic). Each node represents a variable, and the absence of an edge between two nodes indicates the assumption of no direct causal effect.

**Example:** Consider a study examining the impact of exercise on weight loss. A simple DAG might include nodes for exercise, weight loss, and a third confounding variable like diet. The arrows in the DAG would represent the assumed causal relationships for instance, an arrow from exercise to weight loss indicating the direct effect of exercise on weight loss, and another from diet to both exercise and weight loss, indicating that diet is a confounder in this relationship.







####**Statistical Methods in Causality**

Statistical methods play a pivotal role in causal inference, particularly in situations where randomized controlled trials are not feasible.

**Regression Analysis**

**Overview:** Regression analysis is a statistical method used to examine the relationship between a dependent variable and one or more independent variables. In the context of causality, it aims to isolate the effect of an independent variable on the outcome while controlling for other variables.

**Example:** Imagine a study investigating the impact of education level on income. A linear regression model can be used to estimate the causal effect of education on income, controlling for other factors like age, work experience, and location. The regression coefficient for education represents the average difference in income for each additional year of education, assuming other factors are held constant.

**Propensity Score Matching**

Propensity score matching (PSM) is a statistical technique used to reduce selection bias by matching individuals in a treatment group with similar individuals in a control group based on a set of observed characteristics.

**Process:** First, a propensity score (the probability of receiving the treatment given observed covariates) is estimated for each individual, usually using logistic regression. Then, individuals in the treatment and control groups are matched based on their propensity scores.

**Example:** In a study assessing the effectiveness of a job training program, participants who chose to enroll in the program (treatment group) are compared with those who did not (control group). PSM is used to match each participant in the treatment group with a non-participant who has a similar propensity score based on characteristics like age, education level, and previous employment history. This matching helps to simulate a randomized experiment and allows for a more accurate estimation of the program's causal effect on employment outcomes.

**Instrumental Variables**

**Concept:**

Instrumental variables (IV) are used in situations where the treatment is not randomly assigned and is correlated with unobserved confounders. An IV is a variable that influences the treatment but has no direct effect on the outcome, except through the treatment.

**Requirements:** For a variable to be a valid instrumental variable, it must satisfy two conditions: (1) it must be correlated with the treatment (relevance condition), and (2) it must not be correlated with the outcome except through the treatment (exclusion restriction).

**Example:** Consider a study on the effect of education on earnings where education is self-selected (and hence correlated with unobserved factors like motivation). An instrumental variable could be the distance to the nearest college. The assumption is that living closer to a college increases the likelihood of obtaining more education (relevance) and does not directly affect earnings (exclusion restriction). By using this IV, we can estimate the causal effect of education on earnings, despite the initial selection bias in educational attainment.

Each of these methods addresses different challenges in causal inference, especially in observational studies where random assignment to treatment and control groups is not possible.

####**Experimental Design**

Experimental design is critical in determining causal relationships. It involves setting up studies to test hypotheses about these relationships.

**Randomized Controlled Trials (RCTs)**

RCTs are considered the gold standard for experimental research. They involve randomly assigning participants into treatment and control groups to assess the effect of an intervention.

**Process:** Randomization ensures that both known and unknown factors are evenly distributed across groups, minimizing selection bias. The treatment group receives the intervention, while the control group does not (or receives a placebo).

**Example:** In a pharmaceutical study, researchers might test the effectiveness of a new drug by randomly assigning patients to either receive the drug (treatment group) or a placebo (control group). The outcome (e.g., improvement in health condition) is then compared between these groups.


**Quasi-Experiments:** These occur when random assignment is not feasible. Instead, researchers rely on existing groups or conditions. The lack of randomization increases the risk of confounding variables affecting the results.

**Natural Experiments:** These take advantage of external events or situations as 'natural' forms of random assignment. They are particularly useful when ethical or practical constraints prevent controlled experiments.

**Example of Quasi-Experiment:** Suppose a new educational curriculum is introduced in some schools but not others due to budget constraints. Researchers can compare outcomes between students in schools with the new curriculum and those without, though differences other than the curriculum might affect the results.

**Example of Natural Experiment:** An example is the study of the impact of a new law or policy where the 'treatment' (new law) is applied to some regions or groups but not others by virtue of policy decisions, not researcher intervention.



####**Observational Data and Causal Discovery**

Observational data, which are collected without any experimental manipulation, are common in many fields. Causal discovery from such data is challenging but essential for understanding complex systems and phenomena.

**Challenges with Observational Data**

**Confounding Variables:** These are variables that influence both the independent (cause) and dependent (effect) variables, leading to spurious associations. Identifying and controlling for confounders is a major challenge.

**Reverse Causation:** In observational studies, determining the direction of causality can be difficult. It's often unclear whether A causes B, B causes A, or if a third factor influences both.

**Selection Bias:** This occurs when the sample is not representative of the population, leading to biased results. For instance, a health study that only includes hospital patients may not be generalizable to the broader population.

**Measurement Error:** Imperfect measurement tools or techniques can lead to incorrect conclusions about relationships between variables.

**Example:** Suppose a study finds that people who drink coffee are less likely to develop heart disease. This could be due to a confounding variable like physical activity levels perhaps those who drink coffee are also more likely to engage in physical exercise, which is the actual protective factor.

**Methods for Causal Discovery from Observational Data**

**Matching:** This involves selecting control subjects who are similar to treatment subjects on certain observed characteristics. It helps to mimic a randomized experiment by balancing these characteristics across groups.

**Instrumental Variables (IV):** IVs are used when a causal variable is correlated with unobserved confounders. A good IV influences the dependent variable only through the independent variable.

**Regression Discontinuity Design (RDD):** This method exploits a cutoff or threshold for assigning treatment, assuming that subjects near the cutoff are comparable. It can reveal the causal effect at the cutoff point.

**Difference-in-Differences (DiD):** Compares the before-and-after differences in outcomes between a treatment group and a control group, helping to control for confounders that are constant over time.

**Causal Diagrams and Structural Equations:** These methods involve creating a visual or mathematical model of the assumed causal structure, which can then be tested against the data.

**Example with IV:** In studying the effect of education on earnings, the distance to the nearest college can serve as an instrumental variable. The assumption is that proximity to college affects education level (but not earnings directly), allowing for the isolation of the education's causal effect on earnings.

**Example with RDD:** A study evaluating the impact of a scholarship program might use students' test scores as the cutoff. By comparing students who scored just above and just below the threshold, researchers can estimate the causal effect of receiving the scholarship.

####**Individual-Level Outcome Analysis**
conducted an analysis at the individual level to compare the mean achievement scores between the treatment (intervention) and control groups.
* Performing a two-sample t-test to test the statistical significance of the difference in means.


```python
# Hypothesis 1: Does the intervention have a statistically significant impact on achievement scores?
# Split the data into intervention and non-intervention groups
intervention_group = data[data['intervention'] == 1]['achievement_score']
non_intervention_group = data[data['intervention'] == 0]['achievement_score']
```

```python
# Perform a t-test between the two groups
t_stat, p_value = stats.ttest_ind(intervention_group, non_intervention_group)
print(f'T-test results: t-statistic = {t_stat}, p-value = {p_value}')
```

The t-test results  provided show a t-statistic of 23.13 and a p-value of approximately 1.74e-115, which indicates a highly statistically significant difference in achievement scores between the groups that did and did not receive the intervention. Given the very low p-value, the null hypothesis that there is no difference between the two groups can be confidently rejected. The positive t-statistic suggests that the intervention group has a higher mean achievement score compared to the non-intervention group.

```python
# Hypothesis 2: Is the intervention equally distributed across different ethnic groups?
# Perform a chi-square test to determine if the distribution of the intervention is independent of ethnicity
contingency_table = pd.crosstab(data['intervention'], data['ethnicity'])
chi2_stat, p_val_chi2, dof, ex = stats.chi2_contingency(contingency_table)
print(f'Chi-square test results: chi2-statistic = {chi2_stat}, p-value = {p_val_chi2}')
```

The chi-square test results with a chi-square statistic of 19.806 and a p-value of 0.136 indicate that there is no statistically significant association between the intervention and ethnicity at the conventional 0.05 alpha level. The p-value is higher than 0.05, so you we would not reject the null hypothesis that the intervention is distributed independently of ethnicity based on this test.

This means that, according to the chi-square test and the data, the intervention appears to be equally distributed across different ethnic groups.

```python
# Hypothesis 3: Are there significant differences in achievement scores across different ethnic groups or between genders?
# ANOVA for achievement scores across different ethnic groups
f_stat_ethnicity, p_value_ethnicity = stats.f_oneway(
    *[data[data['ethnicity'] == e]['achievement_score'] for e in data['ethnicity'].unique()]
)
```

```python
# ANOVA for achievement scores between genders
f_stat_gender, p_value_gender = stats.f_oneway(
    data[data['gender'] == 1]['achievement_score'],
    data[data['gender'] == 2]['achievement_score']
)

print(f'ANOVA for ethnicities results: F-statistic = {f_stat_ethnicity}, p-value = {p_value_ethnicity}')
print(f'ANOVA for genders results: F-statistic = {f_stat_gender}, p-value = {p_value_gender}')
```

In this dataset, the intervention variable fits this description as it indicates whether a particular intervention was applied to the students or the school, coded as 0 for no intervention and 1 for intervention.

Outcome Variable:

####**Propensity Scores calculation using logistic regression**

Propensity scores estimate the probability of each unit (e.g., student, school) receiving the treatment (in this case, the intervention) given a set of observed covariates. This method is widely used in observational studies to reduce selection bias by equating groups based on these covariates.


Trained a logistic regression model using the potential confounders to predict the likelihood of each unit receiving the intervention.

After fitting the model, calculated propensity scores to estimate the likelihood of receiving the intervention, taking into account potential confounding variables such as ethnicity, gender, school characteristics, and others.

Finally added the propensity scores to the dataset, which will allow to adjust for the potential confounders when estimating the effect of the intervention.

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Define the confounders - variables that may influence both the treatment and the outcome
confounders_columns = ['success_expect', 'ethnicity', 'gender', 'frst_in_family',
                       'school_urbanicity', 'school_mindset', 'school_achievement',
                       'school_ethnic_minority', 'school_poverty', 'school_size']
confounders = data[confounders_columns]

# The treatment variable is 'intervention'
treatment = data['intervention']

#outcome variable

outcome = data['school_achievement']

# Initialize logistic regression model
# Note: solver 'liblinear' is good for small datasets and binary classification
logistic_model = LogisticRegression(solver='liblinear', random_state=42)

# Fit the model to compute the propensity scores
logistic_model.fit(confounders, treatment)

# Compute the propensity scores and add them as a new column in the DataFrame
data['propensity_score'] = logistic_model.predict_proba(confounders)[:, 1]

# Display the first few rows to ensure the propensity score column has been added
print(data[['intervention', 'propensity_score']].head())

```

```python
from sklearn.neighbors import NearestNeighbors

# Separate the treatment and control groups
treatment_df = data[data['intervention'] == 1]
control_df = data[data['intervention'] == 0]

# Initialize the NearestNeighbors model
nn = NearestNeighbors(n_neighbors=1, algorithm='ball_tree')

# Fit the model using the control group
nn.fit(control_df[['propensity_score']])

# Find the nearest neighbor in the control group for each treatment group member
distances, indices = nn.kneighbors(treatment_df[['propensity_score']])

# Create the matched dataset
matched_control_df = control_df.iloc[indices.reshape(-1)]

# Concatenate the matched treatment and control datasets
matched_df = pd.concat([treatment_df, matched_control_df])

# Show the head of the matched dataset
matched_df.head()
```

####**Calculating the weighted average treatment effect (ATE) using Inverse Probability of Treatment Weighting (IPTW)**

Using propensity scores to estimate causal effects. This approach reweights the data to create a synthetic sample in which the distribution of measured baseline covariates is independent of treatment assignment.



For each unit, we have calculated the weight as the inverse of the propensity score for the treated units and as the inverse of one minus the propensity score for the control units. This aims to adjust for the fact that units with higher/lower propensity scores are overrepresented/underrepresented in the treated/control groups

```python
# Calculate the weights for each observation
data['weight'] = data.apply(
    lambda x: 1 / x['propensity_score'] if x['intervention'] == 1 else 1 / (1 - x['propensity_score']),
    axis=1
)

# You can now use these weights in subsequent analyses to estimate the causal effect
# For example, to calculate the weighted average treatment effect:
weighted_ate = (data['weight'] * data['intervention'] * data['achievement_score']).sum() / (
    data['weight'] * data['intervention']).sum() - (data['weight'] * (1 - data['intervention']) * data['achievement_score']).sum() / (
    data['weight'] * (1 - data['intervention'])).sum()

weighted_ate
```

The output 0.39357386408396633 represents the estimated average treatment effect (ATE) using Inverse Probability of Treatment Weighting (IPTW). This value indicates the difference in the average achievement scores between the treatment group (those who received the intervention) and the control group (those who did not), after adjusting for the confounders using the propensity scores.

To interpret this value:

It's important to note that this estimate assumes that:

All Confounders are Observed:

The IPTW method assumes that all the relevant confounders that could influence both the treatment assignment and the outcome are included in the model. If there are unobserved confounders, the estimate may be biased.

Correct Model Specification:

The outcome for any individual is not affected by the treatment status of other individuals.

**Performing a causal inference analysis using nearest neighbor matching based on propensity scores**

This is a standard approach in causal inference for estimating the effect of an intervention when working with observational data, where random assignment to treatment and control groups is not possible. The use of propensity score matching helps in creating a balanced comparison group, reducing the impact of confounding variables.



```python
from causalml.match import NearestNeighborMatch
from sklearn.linear_model import LinearRegression

# Define the treatment and outcome variables
treatment = 'intervention'
outcome = 'achievement_score'

# Selecting variables for the propensity score model
predictors = ['ethnicity', 'gender', 'school_mindset', 'school_achievement', 'school_ethnic_minority', 'school_poverty']
X = data[predictors]
y = data[treatment]

# Fitting a logistic regression model to estimate propensity scores
ps_model = LogisticRegression()
ps_model.fit(X, y)
data['propensity_score'] = ps_model.predict_proba(X)[:, 1]

# Performing matching
nn_match = NearestNeighborMatch(replace=False, ratio=1, random_state=123)
matched_data = nn_match.match(data, treatment_col=treatment, score_cols=['propensity_score'])

# Perform regression analysis on the matched dataset
reg_model = LinearRegression()
X_matched = matched_data[[treatment]]  # Add other predictors if necessary
y_matched = matched_data[outcome]
reg_model.fit(X_matched, y_matched)

# Print regression results
print("Intercept:", reg_model.intercept_)
print("Coefficients:", reg_model.coef_)
```

####**G-computation for causal effect estimation**

 Implementing G-computation, for estimating causal effects in observational data. This involves fitting a statistical model to the data and then using that model to simulate what would happen to each individual's achievement score under both treatment and control conditions. The average difference between these simulated outcomes will provide an estimate of the causal effect of the intervention.

1. Prepare the data: Select the relevant variables and prepare the variables for the analysis.
2. Fit the model: Fit a linear regression model to estimate the relationship between the intervention, confounders, and the outcome.
3. Perform G-computation: Create scenarios where everyone receives the treatment and where no one does, and then predict outcomes under both scenarios.
4. Calculate the average causal effect: Compute the average difference between the predicted outcomes under treatment and control

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Selecting the relevant variables for the model
X = data.drop(columns=['achievement_score', 'schoolid'])  # Assuming these are the columns to drop
y = data['achievement_score']

# Fitting a linear regression model
model = LinearRegression()
model.fit(X, y)

# Creating scenarios for G-computation
X_treatment = X.copy()
X_control = X.copy()
X_treatment['intervention'] = 1
X_control['intervention'] = 0

# Predicting outcomes under both scenarios
y_pred_treatment = model.predict(X_treatment)
y_pred_control = model.predict(X_control)

# Calculating the average causal effect
average_causal_effect = np.mean(y_pred_treatment - y_pred_control)
print("Average Causal Effect: ", average_causal_effect)
```

The output indicates an "Average Causal Effect" of approximately 0.565, is derived from a process known as G-computation. This method is used in causal inference to estimate the effect of a treatment or intervention on an outcome.

Dataset, which includes a mix of variables: a treatment/intervention indicator, an outcome variable (achievement score), and other variables that might be confounders.

Variable Selection:

Identify the treatment variable (intervention) and the outcome variable (achievement_score).

Other variables in the dataset are treated as potential confounders.

Model Fitting (Linear Regression):

Used a linear regression model, where the achievement score is the dependent variable and other variables (including the intervention) are independent variables.

The model aims to capture the relationship between the intervention and the achievement score, controlling for other factors.
G-Computation

create two hypothetical scenarios:

In the first scenario (X_treatment), every individual receives the intervention (treatment is set to 1 for all).

In the second scenario (X_control), no one receives the intervention (treatment is set to 0 for all).

Used the fitted model to predict outcomes (achievement scores) under both scenarios.

Estimating the Average Causal Effect:

The difference in predicted outcomes (achievement scores) between the treatment and control scenarios is calculated for each individual.

The average of these differences across all individuals in the dataset is computed, resulting in the estimated average causal effect.

Interpretation of the Result:

Magnitude of the Effect:

The estimated effect size of 0.565 suggests that the intervention leads to an average increase of 0.565 units in the achievement score.
The practical significance of this effect depends on the context and scale of the achievement scores.

Causal Inference Assumptions:

This estimate assumes that all relevant confounders were included in the model. If important confounders are omitted or measured inaccurately, this could bias the estimate.

The linear regression model assumes a linear relationship between the variables. If the true relationship is non-linear, the estimate may be misleading.
Model Reliance

The accuracy of the estimate is reliant on the correctness of the model specification, including the functional form and the interaction between variables.
