#  Introduction to Causality
Causality is a fundamental concept that explores the relationship between events where one event is understood to be a consequence of the occurrence of another event. It is the backbone of how we understand, predict, and influence the world around us. In simpler terms, causality answers the "why" and "how" something happens. Its importance spans across various fields such as medicine, economics, environmental science, and public policy, where understanding the causal relationship between factors can lead to better decision-making, policy development, and scientific advancements.

## Difference Between Causality and Correlation

To grasp the distinction between causality and correlation, it's essential to understand that correlation simply means that two things happen together or show some pattern of behavior simultaneously, without necessarily causing each other. Causality, on the other hand, implies that one event directly influences the occurrence of another.

## Simple Example:

* Correlation: Ice cream sales and the number of sunburn incidents increase during the summer months. This does not mean buying ice cream causes sunburn incidents or vice versa. They are correlated because they both happen to increase during the same time due to the higher temperatures.
* Causality: Higher temperatures lead to an increase in ice cream sales. In this case, the rise in temperature is a cause that directly affects the increase in ice cream sales.
The ice cream sales example illustrates how correlation might suggest a relationship between two variables, but it doesn't prove that one causes the other. Many mistakenly interpret correlations as causations, leading to incorrect assumptions and decisions.

Let's consider a non-technical, relatable example to further clarify the concept:

# Basic Concepts in Causality
Understanding causality requires familiarity with several key concepts and terminologies. These concepts help us navigate the complexities of determining cause and effect relationships, especially in fields where experimentation is difficult. Here's an in-depth look at three foundational concepts: confounding variables, Randomized Controlled Trials (RCTs), and observational studies.

## Confounding Variables

A confounding variable is an outside influence that changes the effect of a dependent and independent variable. This third variable can falsely suggest a strong correlation between the studied variables, leading to misleading conclusions about their causal relationship.

confound2.png

**Example**: Imagine a study aiming to assess the impact of exercise on weight loss. Age could be a confounding variable. Younger individuals might lose weight faster due to higher metabolic rates, not just because they exercise more. If age is not accounted for, one might wrongly conclude that exercise alone leads to weight loss, ignoring the effect of metabolism influenced by age.

## Randomized Controlled Trials (RCTs)

RCTs are considered the gold standard for establishing causality. In an RCT, participants are randomly assigned to either a treatment group or a control group. The treatment group receives the intervention (e.g., a new drug), while the control group receives a placebo or standard treatment. This random assignment helps ensure that any differences observed between the groups can be attributed to the treatment effect, minimizing the impact of confounding variables.

Randomized-Controlled-Trial-RCTs-Features-Principle-Steps-Bias copy.png

**Example**: To test a new medication's effectiveness, researchers might conduct an RCT where half the participants, chosen randomly, receive the medication, and the other half receive a placebo. If significant health improvements are observed in the treatment group compared to the control group, researchers can confidently attribute those improvements to the medication, assuming the study is well-designed and confounding factors are controlled.

## Observational Studies and the Challenges They Face

When RCTs are not feasible due to ethical, logistical, or financial reasons, researchers rely on observational studies. In these studies, researchers observe subjects in their natural settings without manipulating the study environment. However, observational studies are more susceptible to confounding variables, making it harder to establish causality.

**Challenges**: The primary challenge of observational studies is the difficulty in ruling out the influence of confounding variables. Without the ability to control the assignment of treatments or interventions, it's challenging to determine whether the observed effects are due to the variable of interest or some other factor.

**Example**: Consider a study examining the relationship between dietary habits and heart disease. Researchers might observe that individuals who eat more fruits and vegetables have a lower incidence of heart disease. However, those individuals might also engage in other healthy behaviors, such as exercising more or smoking less, which could influence heart disease risk. These factors are confounding variables that could skew the results, making it difficult to ascertain the causal impact of diet alone.

To address these challenges, researchers employ statistical methods to adjust for confounding variables, attempting to isolate the effect of the variable of interest. Despite these efforts, the potential for unmeasured confounders means that observational studies can seldom provide as strong evidence of causality as RCTs.

In summary, understanding the roles of confounding variables, the methodology of RCTs, and the challenges of observational studies is crucial in dissecting causal relationships from complex data. These concepts are foundational in the pursuit of genuine causal insights, guiding researchers in making informed and accurate conclusions.

# Tools and Techniques for Causal Inference
To navigate the complexities of establishing causality, especially in observational studies where randomized controlled trials (RCTs) are not possible, researchers employ a variety of statistical methods and conceptual tools. Two prominent approaches include regression analysis and propensity score matching. Additionally, causal diagrams, specifically Directed Acyclic Graphs (DAGs), play a crucial role in visualizing and understanding causal relationships. Heres a closer look at these tools and techniques.

## Regression Analysis

Regression analysis is a statistical method used to understand the relationship between a dependent variable and one or more independent variables. By adjusting for various confounding variables, regression can help estimate the effect of an independent variable on the outcome, offering insights into potential causal relationships.

**Example**: If researchers want to study the effect of exercise on heart health, they might use regression analysis, where heart health is the dependent variable and exercise is the independent variable. By including age, diet, and genetic predispositions as additional variables in the analysis, researchers can isolate the effect of exercise on heart health, accounting for these potential confounders.

## Propensity Score Matching (PSM)

Propensity score matching is a statistical technique that attempts to estimate the effect of a treatment, policy, or other intervention by accounting for the covariates that predict receiving the treatment. PSM involves matching individuals who received the treatment with those who did not but had a similar likelihood (propensity score) of receiving the treatment based on observed characteristics.

**Example**: In studying the impact of a job training program on employment outcomes, researchers might use PSM to compare participants in the program with those who did not participate but had similar demographics and employment histories. By matching these groups based on their propensity scores, researchers can more reliably estimate the programs causal effect on employment.

## Causal Diagrams (DAGs)

Directed Acyclic Graphs (DAGs) are graphical representations of the assumed causal relationships between variables. DAGs help researchers visually identify potential confounding variables, direct effects, and indirect effects, facilitating a clearer understanding of the causal structure underlying their data.

In a DAG, nodes represent variables, and arrows represent causal relationships, pointing from cause to effect. A key feature of DAGs is that they do not contain cycles (i.e., a path that starts and ends at the same variable), ensuring that the causality flow is unidirectional.

**Example**: This DAG illustrates the relationship between smoking, chronic respiratory disease, and death from COVID-19. Smoking is shown to directly influence the development of chronic respiratory disease, which in turn increases the risk of death from COVID-19, indicating an indirect effect of smoking on COVID-19 mortality through respiratory disease. Additionally, an unmeasured variable "U" affects both smoking and COVID-19 deaths, suggesting potential confounding factors in the analysis.

These tools and techniques are essential for researchers looking to draw robust causal inferences from their studies.

Propensity score methods are designed to estimate the effect of a treatment or intervention in observational studies where the assignment to a treatment group is not random. They can help in creating a scenario similar to a randomized controlled trial by matching or balancing the dataset so that the treatment and control groups have similar observed characteristics.

**Here are some methods that use propensity scores:**

* Propensity Score Matching (PSM): This method matches each treated unit with one or more control units with a similar propensity score, which is the probability of receiving the treatment given the observed covariates.
* Stratification on Propensity Scores: This involves dividing the range of propensity scores into intervals and ensuring that each interval has both treated and control units. The comparison is then made within each stratum.
* Covariate Balancing Propensity Score (CBPS): This method optimizes the propensity score model to achieve better balance on the observed covariates between treated and control groups.
* Inverse Probability of Treatment Weighting (IPTW): This uses the propensity scores to create weights for each unit such that the weighted dataset mimics a randomized controlled trial.
* Propensity Score Analysis with Covariate Adjustment: This is a two-step approach where you first calculate propensity scores and then use them as covariates in a regression model to adjust for potential confounding.

Each of these methods requires careful implementation and an understanding of the assumptions underlying the propensity score approach. To conduct propensity score analysis, follow these general steps:

1. Estimate Propensity Scores: Fit a logistic regression model (or another appropriate model) where the treatment is regressed on observed covariates.
2. Match or Weight: Depending on the chosen method, use the propensity scores to match or weight the treatment and control units.
3. Check Balance: Assess whether the matching or weighting process has achieved balance in covariates between the treatment and control groups.
4. Estimate Treatment Effects: Once balance is achieved, the treatment effect can be estimated by comparing outcomes between the treated and control units.
5. Sensitivity Analysis: Conduct sensitivity analyses to check how the estimated treatment effect might change with different model specifications or under different assumptions about unobserved confounders.


```python
# Logistic regression is commonly used to estimate propensity scores.
lr = LogisticRegression()
X_covariates = df_encoded[covariates]  # The covariates are the observed characteristics.
y_treatment = df_encoded[treatment]    # This is the treatment indicator.
lr.fit(X_covariates, y_treatment)
# Predict_proba gives the probability of each unit being treated or not.
df_encoded['propensity_score'] = lr.predict_proba(X_covariates)[:, 1]
```

**Estimating Propensity Scores:**

A logistic regression model is fitted with the treatment variable as the dependent variable and other covariates as independent variables.

predict_proba is then used to calculate the propensity score, which is the probability of each unit receiving the treatment given their covariates.

```python
from sklearn.neighbors import NearestNeighbors

# Initialize the NearestNeighbors model to find the closest propensity score.
nn = NearestNeighbors(n_neighbors=1, algorithm='ball_tree')
nn.fit(df_encoded.loc[df_encoded[treatment] == 0, 'propensity_score'].values.reshape(-1, 1))

# Create a function to find the closest match for a given propensity score
def find_match(ps):
    return nn.kneighbors([[ps]], return_distance=False)[0][0]

# For each treated unit, find the closest match among the control units.
treated = df_encoded[df_encoded[treatment] == 1].copy()
treated['match'] = treated['propensity_score'].apply(find_match)
```

**Matching Treated and Control Units:**

NearestNeighbors is used to find the closest control unit for each treated unit based on the propensity score.

The find_match function finds the closest match for each treated unit's propensity score from the control group.

```python
control_matches = df_encoded.loc[df_encoded[treatment] == 0].iloc[treated['match'].values]

# Reset the index to enable a direct comparison
control_matches.reset_index(drop=True, inplace=True)
treated.reset_index(drop=True, inplace=True)

# Now, combine the treated and control_matches to create df_matched
df_matched = treated.append(control_matches).reset_index(drop=True)
```

**Creating Matched Dataset:**

The indices of the matched control units are used to create a new DataFrame control_matches.

This DataFrame is combined with the treated DataFrame to form a matched dataset df_matched.

```python
def standardized_mean_difference(treated, control):
    mean_diff = treated.mean() - control.mean()
    pooled_std = np.sqrt((treated.std()**2 + control.std()**2) / 2)
    smd = mean_diff / pooled_std
    return smd

balance_results = []

for covariate in covariates:
    treated_group = df_encoded.loc[treated.index, covariate]
    control_group = df_encoded.loc[treated['match'], covariate]

    smd = standardized_mean_difference(treated_group, control_group)
    balance_results.append({'covariate': covariate, 'smd': smd})

balance_df = pd.DataFrame(balance_results)
print(balance_df)
```

If any covariate has an SMD above this threshold, you may need to refine your matching algorithm or include additional covariates in your propensity score model.

## Caliper Matching

To improve the balance between treated and control groups in your propensity score matching, you can consider various strategies. Here, we'll focus on two commonly used approaches: caliper matching and nearest-neighbor matching with replacement. Both methods can help achieve a better balance on the covariates between the matched groups.

Caliper matching restricts the match to pairs of treated and control units whose propensity scores are within a predefined distance, called the "caliper." This ensures that matched units are similar in terms of their propensity scores, leading to better balance on the covariates.

```python
# Define the caliper value as a proportion of the standard deviation of the propensity scores
caliper_value = 0.05 * df_encoded['propensity_score'].std()

# Fit NearestNeighbors for control group
nn = NearestNeighbors(n_neighbors=1)
control_ps = df_encoded.loc[df_encoded[treatment] == 0, 'propensity_score'].values.reshape(-1, 1)
nn.fit(control_ps)

# Match with caliper
treated_ps = df_encoded.loc[df_encoded[treatment] == 1, 'propensity_score'].values.reshape(-1, 1)
distances, indices = nn.kneighbors(treated_ps)

# Apply the caliper: only keep matches within the caliper value
matches = indices[distances.flatten() <= caliper_value].flatten()
valid_distances = distances[distances.flatten() <= caliper_value].flatten()

# Create the matched DataFrame
matched_control = df_encoded.loc[df_encoded[treatment] == 0].iloc[matches]
matched_treated = df_encoded.loc[df_encoded[treatment] == 1][distances.flatten() <= caliper_value]

# Remember to check the balance after matching
```

## Nearest-Neighbor Matching with Replacement

Matching with replacement allows a control unit to be matched to more than one treated unit if it's the nearest neighbor, potentially improving balance by reusing control units that are very similar to multiple treated units.

```python
from sklearn.neighbors import NearestNeighbors

# Initialize the NearestNeighbors model for the control group with propensity scores
nn_with_replacement = NearestNeighbors(n_neighbors=1, algorithm='ball_tree')
control_ps_with_replacement = df_encoded.loc[df_encoded[treatment] == 0, 'propensity_score'].values.reshape(-1, 1)
nn_with_replacement.fit(control_ps_with_replacement)

# For each treated unit, find the closest control unit, allowing for replacements
_, indices_with_replacement = nn_with_replacement.kneighbors(treated_ps)

# Create the matched DataFrame with replacements
matched_control_with_replacement = df_encoded.loc[df_encoded[treatment] == 0].iloc[indices_with_replacement.flatten()]
matched_treated_with_replacement = df_encoded.loc[df_encoded[treatment] == 1]

# Again, check the balance after matching
```

**Implications**

These results suggest that the caliper matching process has been successful in creating a control group that is comparable to the treated group in terms of the covariates measured. This balance is crucial for causal inference, as it reduces the likelihood that observed differences in outcomes between the treated and control groups are due to differences in these covariates rather than the treatment itself.
