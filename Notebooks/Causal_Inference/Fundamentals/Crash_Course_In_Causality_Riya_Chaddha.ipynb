## Theory 

#### Causal Inference Overview:
Causal inference is a critical area of statistics and data science that focuses on determining whether a relationship between two variables is causal rather than merely correlational. Traditional statistical methods, such as regression analysis, often identify associations but do not provide evidence of causation. Causal inference seeks to answer questions like "Does X cause Y?" by employing various methodologies that account for confounding variables and other complexities inherent in observational data.

#### Key Concepts:
1. Causality vs. Correlation
- Correlation indicates a statistical association between two variables, meaning they tend to vary together. However, correlation does not imply causation; two variables can be correlated due to a third variable influencing both (confounding).
- Causation implies that changes in one variable (the cause) directly result in changes in another variable (the effect). Establishing causation requires rigorous analysis to rule out confounding factors.

2. Sleep Duration and Quality
- Sleep Duration: This refers to the total number of hours a person sleeps in a given period. It is a critical factor in overall health, with both insufficient and excessive sleep linked to various health issues.
- Sleep Quality: This encompasses subjective and objective measures of how restful and restorative sleep is. Factors affecting sleep quality include sleep disturbances, sleep architecture, and individual perceptions of sleep satisfaction. Both sleep duration and quality are essential for understanding their impact on health outcomes, particularly mental health.

3. Mental Health Outcomes
- Mental health outcomes, such as depression and anxiety, are often quantified using standardized scales like the PHQ-9 (Patient Health Questionnaire-9) for depression and the GAD-7 (Generalized Anxiety Disorder-7) for anxiety. These scales provide a systematic way to assess the severity of symptoms and track changes over time.
- Understanding the relationship between sleep and mental health is crucial, as poor sleep can exacerbate mental health issues, creating a feedback loop that can be difficult to break.

4. Mediation in Causal Inference
- Mediation occurs when a third variable (the mediator) explains the relationship between an independent variable (X) and a dependent variable (Y). For example, stress may mediate the relationship between sleep quality and depression. In this case, poor sleep quality leads to increased stress, which in turn contributes to higher depression scores.
- Mediation analysis helps researchers understand the mechanisms through which an independent variable affects a dependent variable, providing deeper insights into causal pathways.

5. Directed Acyclic Graphs (DAGs): DAGs represent causal relationships in a clear graphical form. Each node represents a variable, and the directed edges (arrows) represent the presumed direction of causality.

6. Propensity Score Matching: This technique is used to reduce bias when estimating treatment effects from observational data by matching individuals based on similar characteristics.


**Directed Acyclic Graphs (DAGs) in Causal Inference**
Directed Acyclic Graphs (DAGs) are powerful tools used in causal inference to visually represent the relationships between variables. Each node in a DAG represents a variable, while directed edges (arrows) indicate the presumed direction of causality. DAGs help clarify assumptions about causal relationships, identify potential confounders, and guide the analysis of observational data.

**Key Concepts of DAGs**
1. Paths in DAGs:

In a DAG, two variables can be connected by what is called a “path.” The nature of these paths can be classified into three main types: directed paths, backdoor paths, and closed (or blocked) paths. The openness or closedness of these paths reflects the statistical associations between the variables.
2. Open and Closed Paths:

- Open Paths: Represent statistical associations between two variables. An open path indicates that there is a potential for a causal relationship or correlation.
- Closed Paths: Represent the absence of statistical associations. A closed path indicates that the relationship between the two variables is blocked by a collider or other factors.
**Types of Paths**
1. Directed Paths:

A directed path consists of arrows that all point in the same direction, indicating a causal relationship between the variables. For example, consider the following scenario:
Example: Increased screen time leads directly to reduced physical activity, which in turn leads to an increased risk of obesity. In this case, the path from screen time to obesity is open and reflects a causal chain.
Directed Path Example


2. Backdoor Paths:

A backdoor path occurs when two variables share a common cause. This path connects the exposure and outcome through a third variable, which can introduce confounding.
Example: Increased screen time and childhood obesity may both be influenced by low parental education. In this case, screen time and childhood adiposity are connected by a backdoor path through low parental education. This path is open and depicts a statistical association between screen time and adiposity, but the association is non-causal and represents confounding.
Backdoor Path Example

3. Closed (or Blocked) Paths:

A closed path occurs when two variables have a common effect, known as a collider. In this case, the path is blocked, and there is no association between the two variables through the collider.
Example: Both screen time and obesity may increase the risk of low self-esteem, self-harm, and suicidal ideation in adolescents. Here, self-harm acts as a collider between physical activity and obesity. Since the path is closed, there is no association between screen time and adiposity transmitted through self-harm.
Closed Path Example

#### Step 1: Import Libraries and Load Dataset

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.impute import SimpleImputer

# Load the dataset
df = pd.read_csv('C://Users//riyac//Downloads//Sleep_health_and_lifestyle_dataset.csv')

# Display the first few rows of the dataset
df.head()
```

#### Explanation:
We start by importing the necessary libraries. Pandas is used for data manipulation, Matplotlib and Seaborn are used for visualization, and Statsmodels is used for statistical analysis. We load the dataset using pd.read_csv() and display the first few rows to examine the structure.

#### Feature Engineering

```python
# Splitting the 'Blood Pressure' column into 'Systolic' and 'Diastolic'
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True)

# Convert the new columns to numeric
df['Systolic'] = pd.to_numeric(df['Systolic'], errors='coerce')
df['Diastolic'] = pd.to_numeric(df['Diastolic'], errors='coerce')

# Drop the original 'Blood Pressure' column if it's no longer needed
df.drop('Blood Pressure', axis=1, inplace=True)

# Display the updated DataFrame
df.head()
```

#### Step 2: Data Exploration and Preprocessing

#### 1. Convert Categorical Variables to Numeric
We'll use LabelEncoder from sklearn to convert categorical variables into numeric values.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

# Encoding categorical variables
encoder = LabelEncoder()

# Encode Gender
df['Gender'] = encoder.fit_transform(df['Gender'])

# Encode Occupation
df['Occupation'] = encoder.fit_transform(df['Occupation'])

# Encode BMI Category
df['BMI Category'] = encoder.fit_transform(df['BMI Category'])

# Encode Sleep Disorder (NaN will remain NaN for now)
df['Sleep Disorder'] = encoder.fit_transform(df['Sleep Disorder'].astype(str))

# Display the dataframe after encoding
print("Data after encoding categorical variables:")
print(df)
```

#### 2. Handle Missing Values
Now that all categorical variables are numeric, we'll impute missing values using appropriate strategies:

1. Mean for continuous numerical features.
2. Most Frequent for categorical features.

```python
# Define imputers
mean_imputer = SimpleImputer(strategy='mean')
most_frequent_imputer = SimpleImputer(strategy='most_frequent')

# Impute numerical features
num_cols = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Systolic','Diastolic','Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']
df[num_cols] = mean_imputer.fit_transform(df[num_cols])

# Impute categorical features
cat_cols = ['BMI Category', 'Sleep Disorder', 'Occupation']
df[cat_cols] = most_frequent_imputer.fit_transform(df[cat_cols])

# Display the dataframe after imputing missing values
print("\nData after imputing missing values:")
print(df)
```

Ensure that there are no missing values left:

```python
print("\nCheck for missing values in the dataset:")
print(df.isnull().sum())
```

#### Step 4: Causal Analysis (Modeling)
Simple Linear Regression: Sleep Duration and Stress Level

```python
# Independent and dependent variables
X = df[['Sleep Duration']]
y = df['Stress Level']

# Add constant to X
X = sm.add_constant(X)

# Fit regression model
model = sm.OLS(y, X).fit()

# Model summary
print(model.summary())
```

- **Adjusted R-squared = 0.657:**

After adjusting for the model's complexity, the explanatory power remains consistent, suggesting that sleep duration is a significant factor influencing stress levels.

#### Mediation Analysis: Physical Activity as a Mediator

Concept:
Mediation analysis is used to determine if the effect of sleep duration on stress is mediated through physical activity.

```python
# Step 1: Sleep Duration -> Physical Activity
X1 = df[['Sleep Duration']]
y1 = df['Physical Activity Level']
X1 = sm.add_constant(X1)
model1 = sm.OLS(y1, X1).fit()

# Step 2: Physical Activity + Sleep Duration -> Stress Level
df['Predicted Activity'] = model1.predict(X1)
X2 = df[['Sleep Duration', 'Predicted Activity']]
y2 = df['Stress Level']
X2 = sm.add_constant(X2)
model2 = sm.OLS(y2, X2).fit()

# Summarize results
print(model1.summary())
print(model2.summary())
```

This mediation analysis explores whether physical activity serves as a mediator in the relationship between sleep duration (independent variable) and stress level (dependent variable). The process involves two regression steps:

### Step 1: Effect of Sleep Duration on Physical Activity (Mediator)
**Regression Equation:** 

### Physical Activity Level=19.52+5.56⋅Sleep Duration

- **R-squared = 0.045:**

Only 4.5% of the variance in physical activity level is explained by sleep duration, suggesting a weak linear relationship.
- **Sleep Duration Coefficient (5.56, p < 0.001):**

Sleep duration has a statistically significant positive effect on physical activity. For every additional hour of sleep, physical activity level increases by approximately 5.56 units.
- **Interpretation:**

Longer sleep duration is associated with higher physical activity levels, but the weak R-squared value indicates other factors contribute significantly to physical activity.

Only 4.5% of the variance in physical activity level is explained by sleep duration, suggesting a weak linear relationship. Sleep duration has a statistically significant positive effect on physical activity. For every additional hour of sleep, physical activity level increases by approximately 5.56 units. Longer sleep duration is associated with higher physical activity levels, but the weak R-squared value indicates other factors contribute significantly to physical activity. 


### Step 2: Effect of Sleep Duration and Physical Activity on Stress Level
**Regression Equation:**
### Stress Level=1.89−6.48⋅Sleep Duration+0.84⋅Predicted Activity

- **R-squared = 0.658:**

Approximately 65.8% of the variability in stress levels is explained by sleep duration and predicted physical activity together.
- **Sleep Duration Coefficient (-6.48, p < 0.001):**

Sleep duration has a strong negative effect on stress level, even when accounting for physical activity. This coefficient remains highly significant.
- **Predicted Activity Coefficient (0.84, p < 0.001):**

Physical activity also shows a significant positive association with stress levels. For every unit increase in physical activity, stress levels increase by 0.84 units.

### Key Insights from Mediation Analysis

- Sleep duration has a direct negative effect on stress levels (-6.48).
- While sleep duration positively influences physical activity (Step 1), the predicted activity also positively influences stress levels (Step 2). This suggests partial mediation, but physical activity does not fully account for the relationship between sleep duration and stress levels.


2. **Direct vs. Indirect Effects:**

This mediation analysis highlights the complexity of causal pathways and emphasizes the need to consider both direct and indirect effects when designing interventions.

#### Advanced Analysis
Interaction Effects: BMI and Sleep Duration on Stress

```python
# Include interaction term in regression
df['Sleep_BMI_Interaction'] = df['Sleep Duration'] * df['BMI Category'].factorize()[0]
X3 = df[['Sleep Duration', 'BMI Category', 'Sleep_BMI_Interaction']]
y3 = df['Stress Level']
X3 = sm.add_constant(X3)
model3 = sm.OLS(y3, X3).fit()

# Model summary
print(model3.summary())
```

The results of the Ordinary Least Squares (OLS) regression model highlight how Sleep Duration and BMI Category independently and interactively influence Stress Level. Below is a detailed breakdown of the findings:

1. **Model Performance:**

- The **R-squared value** is **0.682**, indicating that 68.2% of the variance in Stress Level is explained by the model. This is a strong indicator of the model's goodness-of-fit.
- The **Adjusted R-squared** of **0.679** accounts for the number of predictors, affirming the reliability of the model.
2. Key Predictors:

- **Intercept (const):** The baseline stress level (19.52) when all predictors are zero. This provides the foundation for the model’s predictions.
- **Sleep Duration:** The coefficient of **-1.96** implies that for every additional hour of sleep, Stress Level decreases by approximately 1.96 units, holding other factors constant. This finding is highly significant (p-value < 0.001) and aligns with the hypothesis that adequate sleep lowers stress.
- **BMI Category:** The coefficient of **-0.18** indicates that as BMI category improves (e.g., from overweight to normal), Stress Level decreases by 0.18 units, with a p-value < 0.001 confirming its statistical significance.
3. **Interaction Term: Sleep and BMI (Sleep_BMI_Interaction):**

- The interaction term, with a coefficient of 0.01, suggests that the combined effect of Sleep Duration and BMI on Stress Level is positive but not statistically significant (p-value = 0.383). This means that while there may be some interplay between these two factors, its impact on stress is not strong enough to be conclusive.
4. **Statistical Significance:**

- Both Sleep Duration and BMI Category are significant predictors of stress, as evidenced by their low p-values (< 0.001). The interaction term, however, does not significantly contribute to explaining Stress Level.
5. **Model Diagnostics:**

- The **Durbin-Watson statistic** (0.885) suggests potential autocorrelation in the residuals, which may need further exploration.
- The **Omnibus and Jarque-Bera** tests indicate a slight deviation from normality in the residuals, but this is unlikely to undermine the model's overall conclusions.

**Analytical Insights:**
The findings strongly support the hypothesis that both Sleep Duration and BMI independently impact Stress Level. Specifically:

- Sufficient sleep duration significantly reduces stress.
- Improved BMI categories (e.g., normal weight) are associated with lower stress levels.
- The interaction between Sleep Duration and BMI was not statistically significant, suggesting that their combined effect on stress is not as impactful as their individual contributions.

**Implications:**

From a practical standpoint, the results emphasize the importance of both adequate sleep and maintaining a healthy BMI to mitigate stress levels. While the interaction effect was insignificant, focusing on improving these variables individually could lead to better stress management strategies.

Future research could consider additional mediators, such as diet or physical activity, to explore other pathways through which sleep and BMI influence stress. Additionally, addressing residual autocorrelation and normality issues could further refine the model.
