## **CAUSALITY:**


Causality refers to the relationship between cause and effect, where a change in one variable influences the occurrence of another. It involves understanding the mechanisms behind observed phenomena and determining how alterations in one factor lead to changes in another. Causal relationships are fundamental for making predictions, guiding interventions, and optimizing strategies.


## **CAUSATION:**

1. Causation, on the other hand, is a relationship where one variable (the cause) directly influences another variable (the effect).
2. The classic definition of causation is the “counterfactual” definition, which states that X causes Y if and only if Y would not have occurred if X had not occurred.



### **FORMULAS USED IN CAUSAL INFERENCE:**

1. The causal effect of the treatment on the outcome can then be defined as the difference between the average potential outcomes under the treatment and control conditions:

   ```markdown
   ┌───────────────────┐
   │ E[Y(1) — Y(0)]    │
   └───────────────────┘


   Let's break down the equation:

   - Y(1): This represents the potential outcome if someone receives the treatment (in our previous example, it could be drinking milk).
   - Y(0): This represents the potential outcome if someone does not receive the treatment (in our example, it could be not drinking milk).

   So, E[Y(1) — Y(0)] tells us the difference between what happens when someone gets the treatment and what would happen if they didn't.

   - E[Y(1) — Y(0)]: This part of the equation finds the expected value, or the average, of the difference between Y(1) and Y(0) across all individuals or units in our study group.

In simpler terms, E[Y(1) — Y(0)] represents the average difference in outcomes between those who receive the treatment and those who do not. It helps us understand the average effect of the treatment on the outcome we're interested in.






### **CAUSAL INFERENCE**


Causal inference is a pivotal aspect of statistical analysis and research aimed at understanding cause-and-effect relationships between variables. It involves drawing conclusions about the impact of one variable on another based on observational or experimental data. Unlike correlation, which identifies statistical associations, causal inference goes a step further by attempting to establish a causal relationship.



### **KEY CONCEPTS**

- **Observational Data:**
   Causal inference often begins with observational data, where researchers observe variables without intervening. Analyzing such data requires careful consideration of potential confounding factors that might influence both the cause and the effect.


 **IGNORABILITY:**

- **Concept:**
  Ignorability emphasizes that, given observed variables, the assignment to treatment is independent of potential outcomes.

- **Example:**
  Imagine a study investigating the impact of a new educational program on students' academic performance. Ignorability assumes that, given students' prior academic records and socioeconomic status (observed variables), the decision to enroll in the program is independent of potential outcomes like improved grades. In other words, students with similar prior academic records and socioeconomic status have an equal chance of participating in the program, making the treatment assignment ignorable.



 **ASSOCIATION vs. CAUSATION:**

- **Concept:**
  - **Association:** Statistical connection between two variables but doesn't imply causation.
  - **Causation:** Indicates a direct cause-and-effect relationship between variables.

- **Example:**
  Consider the observation that ice cream sales and the number of drowning incidents both increase during the summer. There is an association between the two variables. However, it would be incorrect to conclude that buying more ice cream causes an increase in drowning incidents. The association is likely due to a confounding factor – the summer season.



 **CONFOUNDER:**

- **Concept:**
  A confounder is a variable associated with both the treatment and the outcome, potentially leading to spurious correlations.

- **Example:**
  Suppose a study aims to assess the impact of a fitness program on weight loss. If participants' dietary habits are not considered (a confounding variable), the observed weight loss might be erroneously attributed to the fitness program when, in reality, it could be influenced by changes in diet.



**RANDOMIZED CONTROLLED TRIALS (RCTs):**

- **Concept:**
  RCTs are experimental designs where participants are randomly assigned to treatment and control groups.

- **Example:**
  In a drug trial, participants are randomly assigned to either receive the new medication (treatment group) or a placebo (control group). Randomization helps ensure that any differences in outcomes can be attributed to the treatment, not pre-existing characteristics.



 **POTENTIAL OUTCOMES:**

- **Concept:**
  Potential outcomes represent different outcomes a unit (e.g., an individual) would have under various treatment conditions.

- **Example:**
  Consider a study evaluating the impact of a training program on job performance. The potential outcome for an individual under treatment might be improved performance, while the potential outcome under no treatment could be their regular job performance.



 **PROPENSITY SCORE MATCHING:**

- **Concept:**
  In observational studies, propensity score matching creates comparable treatment and control groups based on the likelihood of receiving treatment.

- **Example:**
  In a survey on the impact of online courses on academic achievement, researchers use propensity scores to match individuals with similar characteristics (age, prior GPA, etc.). This helps mimic the random assignment of a controlled experiment.



 **CHALLENGES AND ASSUMPTIONS:**

- **Concept:**
  Causal inference faces challenges, including unobserved confounding variables and the need for strong assumptions.

- **Example:**
  In a study investigating the effect of a new teaching method on student success, unobserved factors like teacher enthusiasm might influence both the teaching method chosen and student outcomes, posing a challenge to causal inference.



 **APPLICATIONS:**

- **Concept:**
  Causal inference plays a vital role in various fields, informing decision-making, policy development, and scientific knowledge.

- **Example:**
  In public health, causal inference helps determine the effectiveness of vaccination programs. By comparing health outcomes between vaccinated and unvaccinated groups, policymakers can make informed decisions about public health interventions.




Causal inference plays a vital role in various fields, including public health, economics, social sciences, and medicine. Understanding causal relationships is essential for informed decision-making, policy development, and the advancement of scientific knowledge.

In summary, causal inference provides a framework for moving beyond correlation, allowing researchers to make meaningful statements about the causal relationships between variables in a wide range of disciplines.

### **Conceptual Understanding of Causal Inference**


Causal inference is a crucial aspect of statistical analysis, aiming to understand cause-and-effect relationships between variables. Here, we delve into key concepts related to correlation patterns, causal pathways, and methods used in causal inference.


### **1. Correlation Patterns vs. Causal Pathways**

Correlation Patterns involve identifying statistical associations between variables, while Causal Pathways focus on understanding the mechanisms behind cause-and-effect relationships. It's essential to distinguish between the two for accurate causal inference.


### **Random Assignment**

**Pros:**
- **Gold Standard for Causal Effect:** Random assignment is considered the gold standard for isolating the causal effect of a treatment.
- **No Need to Identify Confounding Variables:** In a well-designed randomized experiment, there's no need to identify or measure all possible confounding variables.

**Cons:**
- **Impractical for Some Treatments:** Certain treatments may make randomized experiments impractical or cost-prohibitive.
- **Non-compliance:** Individuals assigned to a particular group may not comply with their assignment.



### **How EconML Helps with Random Assignment**

EconML aids in analyzing data from randomized experiments, revealing treatment impacts on different customer segments.



### **2. Confounders**

In the realm of Full Observability (Unconfoundedness), confounders are external factors that can influence both the treatment and the outcome being observed.

**Pros:**
- This method helps understand the real impact of the treatment on the outcome without a fancy experiment.

**Cons:**
- Requires identifying and measuring all potential confounding variables.



### **How EconML Helps with Observed Confounders**

EconML adapts and learns from data, using multiple advanced tools and techniques to understand complex relationships.



### **3. Instrumental Variables**

Instrumental variables are utilized in statistical modeling and econometrics to address potential endogeneity issues and establish causal relationships. These variables act as proxies or tools that are correlated with the treatment but are not directly related to the outcome. By using instrumental variables, researchers aim to isolate the causal impact of the treatment on the outcome, especially when there are concerns about unobserved or unmeasured factors influencing both the treatment and the outcome.

**Pros:**
- Works even when other unmeasured factors may influence both treatment and outcome.




**Example Clarification:**
Using instrumental variables, like distance to the store, helps separate the impact of a loyalty program on customer behavior.

These methods and tools provide insights into causal relationships, allowing researchers to make informed decisions in various fields.

### **DEFINITION OF TREATMENT AND OUTCOME**


---

- **TREATMENT:**
  - The treatment variable, denoted as \( T \), represents the intervention or action applied to individuals in a study or experiment.
  - In the context of the provided functions, the treatment \( T \) is determined based on the individual's income level and is influenced by probabilities (\( p \)) associated with different income thresholds:
    - If income < 1:
      - \( T = 0.9 \) with probability \( p = 0.3 \)
      - \( T = 0.8 \) with probability \( p = 0.5 \)
    - If income ≥ 1:
      - \( T = 1 \) with probability \( p = 0.7 \)
      - \( T = 0.9 \) with probability \( p = 0.2 \)



- **OUTCOME:**
  - The outcome variable, denoted as \( Y \), represents the observed response or result following the application of the treatment.
  - The outcome \( Y \) is determined by a combination of treatment effects (\( y(X) \cdot T \)) and a baseline function (\( B(X, W) \)) dependent on individual features:
    - \( y(X) \) represents the individual's response to the treatment.
    - \( B(X, W) \) is a function influenced by features \( X \) (e.g., income) and \( W \) (e.g., average hours, days visited).



### **EQUATION**:


\begin{array}{|c|}
\hline
Y = y(X) \cdot T + B(X, W) \\
\hline
\end{array}



- **EXPLANATION OF THE EQUATION:**
  - \( Y \): Outcome variable, representing the observed response.
  - \( y(X) \cdot T \): Treatment effect, where \( y(X) \) represents the individual's response to the treatment \( T \).
  - \( B(X, W) \): Baseline function, dependent on individual features \( X \) and \( W \), influencing the outcome \( Y \).
  - The equation combines the treatment effect and baseline function to determine the overall outcome observed in the study or experiment.

## **Get Causal Effects with EconML**

---

````python
# Define underlying treatment effect function given DGP
def gamma_fn(X):
    return -3 - 14 * (X["income"] < 1)

def beta_fn(X):
    return 20 + 0.5 * (X["avg_hours"]) + 5 * (X["days_visited"] > 4)

def demand_fn(data, T):
    Y = gamma_fn(data) * T + beta_fn(data)
    return Y

def true_te(x, n, stats):
    if x < 1:
        subdata = train_data[train_data["income"] < 1].sample(n=n, replace=True)
    else:
        subdata = train_data[train_data["income"] >= 1].sample(n=n, replace=True)
    te_array = subdata["price"] * gamma_fn(subdata) / (subdata["demand"])
    if stats == "mean":
        return np.mean(te_array)
    elif stats == "median":
        return np.median(te_array)
    elif isinstance(stats, int):
        return np.percentile(te_array, stats)
````

### **Explanation**
1. **Model Equation**:
    - The equation represents a model used in **EconML** to estimate causal effects.
    - It helps us understand how different factors (like discounts) affect outcomes (like revenue).



2. **Equation Components**:
    - The model has two parts:
        - **Outcome Model**:
            - Enclose the following equation in a box:


                \begin{array}{|c|}
                \hline
                \text{log(Y) = } \theta(X) \cdot \text{log(T)} + f(X, W) + \epsilon \\
                \hline
                \end{array}

                
                Here:
                - **Y**: Represents the outcome (e.g., revenue).
                - **X**: Factors we’re interested in (e.g., discount percentage).
                - **T**: Total sales (related to the outcome).
                - **θ(X)**: How X affects Y (elasticity).
                - **f(X, W)**: Other relevant factors (like W) affecting Y.
                - **ε**: Random variation (uncorrelated error).
        - **Treatment Model**:
            - Enclose the following equation in a box:
                
                \begin{array}{|c|}
                \hline
                \text{log(T) = g(X, W) + η} \\
                \hline
                \end{array}
                
                Here:
                - **g(X, W)**: How X and other factors affect T (total sales).
                - **η**: Another random variation.


3. **Interpretation**:
    - We’re trying to estimate how changes in X (like discounts) impact Y (revenue).
    - Despite not being a perfect match for real-world data, this model helps guide decisions.


````python
# Get the estimate and range of true treatment effect
truth_te_estimate = np.apply_along_axis(true_te, 1, X_test, 1000, "mean")  # estimate
truth_te_upper = np.apply_along_axis(true_te, 1, X_test, 1000, 95)  # upper level
truth_te_lower = np.apply_along_axis(true_te, 1, X_test, 1000, 5)  # lower level
````

- **Explanation**:
1. **Objective**:
    - We aim to estimate how changing factors, such as income, influence the number of songs users purchase.
2. **Estimation Method**:
    - We utilize the **`true_te`** function, which likely calculates the treatment effect of income on song purchases.
    - This function is applied to the test data (**`X_test`**), which contains different income levels.
3. **Calculation of Estimate**:
    - We calculate the treatment effect estimate by averaging across multiple trials (1000 times).
    - Each trial provides an estimate of the treatment effect for a specific income level.
4. **Determining the Range**:
    - We want to assess the uncertainty in our estimate, so we determine the upper and lower levels of the treatment effect range.
    - The upper level represents a high estimate (95th percentile), while the lower level represents a low estimate (5th percentile).
5. **Application to Dataset**:
    - These calculations are performed for each income level in the test data (**`X_test`**).
    - Thus, we obtain an estimate of the treatment effect and its corresponding range for various income levels.

In summary, the code helps us estimate how income influences song purchases, providing both an average estimate and a range of values to understand the uncertainty in our estimation process.

## **Make Policy Decision with EconML**
---

To maximize revenue rather than demand, we consider the following mathematical representation:


\begin{array}{|c|} \hline Rev =  Y.T\ \\ \hline \end{array}

\begin{array}{|c|} \hline Rev =  \exp(\log(Y)) \cdot T \\ \hline \end{array}

\begin{array}{|c|} \hline Rev =  \exp(f(x, W) \cdot T^{\theta(X) + 1} \\ \hline \end{array}





Here,
- \( Rev \): Revenue,
- \( Y \): Outcome variable,
- \( T \): Treatment variable,
- \( \log(r) \): Logarithm of revenue,
- \( \theta(X) \): CATE, varying based on covariates \( X \),
- \( I \): Income,
- \( f(X, W) \): Additional covariates,
- \( \epsilon \): Error term.

suggests that with a decrease in price, revenue increases only if \begin{array}{|c|} \hline  \theta(X)+1 < 0 \\ \hline \end{array}

 Thus, we set `sample_treatment_cast=-1` to identify the types of customers to give a small discount to maximize revenue.

The EconML library provides policy interpretability tools such as `SingleTreePolicyInterpreter`, which, given a treatment cost and treatment effects, learns simple rules about targeting profitable customers. In the figure below, the model recommends giving a discount to people with income less than 0.985 and providing the original price for others.


````python
intrp = SingleTreePolicyInterpreter(risk_level=0.05, max_depth=2, min_samples_leaf=1, min_impurity_decrease=0.001)
intrp.interpret(est_nonparam_dw, X_test, sample_treatment_costs=-1)
plt.figure(figsize=(25, 5))
intrp.plot(feature_names=["income"], treatment_names=["Discount", "No-Discount"], fontsize=12)
plt.show()
````

## **Revenue Policy Comparison**


---


````python
# define function to compute revenue
def revenue_fn(data, discount_level1, discount_level2, baseline_T, policy):
    policy_price = baseline_T * (1 - discount_level1) * policy + baseline_T * (1 - discount_level2) * (1 - policy)
    demand = demand_fn(data, policy_price)
    rev = demand * policy_price
    return rev
````

````python
policy_dic = {}
# our policy above
policy = intrp.treat(X)
policy_dic["New Policy"] = np.mean(revenue_fn(train_data, 0, 0.1, 1, policy))

## previous strategy
policy_dic["Previous Strategy"] = np.mean(train_data["price"] * train_data["demand"])

## give everyone discount
policy_dic["Give Everyone Discount"] = np.mean(revenue_fn(train_data, 0.1, 0, 1, np.ones(len(X))))

## don't give discount
policy_dic["Give No One Discount"] = np.mean(revenue_fn(train_data, 0, 0.1, 1, np.ones(len(X))))

## follow our policy, but give -10% discount for the group doesn't recommend to give discount
policy_dic["New Policy + Give Negative Discount for No-Discount Group"] = np.mean(revenue_fn(train_data, -0.1, 0.1, 1, policy))

## give everyone -10% discount
policy_dic["Give Everyone Negative Discount"] = np.mean(revenue_fn(train_data, -0.1, 0, 1, np.ones(len(X))))
````

````python
# get policy summary table
res = pd.DataFrame.from_dict(policy_dic, orient="index", columns=["Revenue"])
res["Rank"] = res["Revenue"].rank(ascending=False)
res
````
