### Correlation
---
Correlation measures the degree of association between two variables. It quantifies how one variable changes in response to the other but does not imply causation. For example:
- **Positive Correlation**: When one variable increases, the other also increases (e.g., height and weight).
- **Negative Correlation**: When one variable increases, the other decreases (e.g., exercise frequency and body fat percentage).
- **No Correlation**: Variables are unrelated (e.g., shoe size and intelligence).

#### Key Characteristics:
1. **Pearson's Correlation Coefficient**: Measures linear relationships (value ranges from -1 to +1).
2. **Spearman's Rank Correlation**: Suitable for non-linear relationships.

### Causal inference
---

**Causal inference** is a field of study that aims to establish a cause-and-effect relationship between variables based on observed data. Unlike correlation, which only measures the degree of association between two variables, causal inference seeks to determine whether a change in one variable directly leads to a change in another. This process is foundational in fields such as epidemiology, economics, and social sciences, where understanding causal relationships is critical for effective decision-making.

At its core, causal inference is concerned with answering questions like:

- **Does an increase in marketing spend lead to higher sales?**
- **Does a new drug reduce the risk of a disease?**
- **How does employee training influence productivity?**

By identifying and estimating causal effects, we can move beyond merely identifying associations between variables and begin to understand the underlying mechanisms that drive these associations.

#### Key Components of Causal Inference:

1. **Cause-and-Effect Relationships**:
   Causal inference focuses on uncovering the **causal relationship** between variables. For instance, if we hypothesize that variable A causes an effect on variable B, causal inference aims to determine the strength and direction of this relationship. This is particularly important in decision-making, where understanding the causal impact of interventions is crucial.

2. **Understanding Counterfactuals**:
   A crucial concept in causal inference is the idea of **counterfactual reasoning**. This refers to thinking about what would have happened in a different scenario (e.g., "What would the outcome have been if a different treatment had been applied?"). In causal inference, we try to estimate what the outcome would have been if a treatment or intervention had not occurred, which we call the **counterfactual outcome**.

3. **Identifying Causal Relationships in Observational Data**:
   Often, randomized controlled trials (RCTs) are not feasible, and researchers must rely on **observational data**. In such cases, causal inference techniques help identify potential causal relationships while controlling for confounders (other variables that might influence both the cause and effect). Some techniques for doing this include **propensity score matching**, **instrumental variables**, and **difference-in-differences (DiD)**.

4. **Prediction and Decision-Making**:
   Causal inference is a vital tool for **predicting outcomes** and making informed decisions. For example, if a company wants to know whether a change in product design will improve customer satisfaction, causal inference can help provide an estimate of how much customer satisfaction would improve with the change. This kind of predictive insight allows for better decision-making and resource allocation.

5. **Generalization Beyond the Sample**:
   One of the goals of causal inference is not only to establish causal effects for the study sample but also to generalize these findings to a larger population. This makes causal inference essential for **policy-making** and for understanding broader trends that can be applied across different settings or groups.

## Methods in Causal Inference:
---
To address these challenges, several methods are employed in causal inference:

- **Randomized Controlled Trials (RCTs)**: The gold standard for establishing causal relationships, where participants are randomly assigned to treatment and control groups. This ensures that any differences between the groups can be attributed to the treatment and not to confounders.
- **Natural Experiments**: These are situations where external factors or policies randomly assign treatments, allowing researchers to estimate causal effects without the need for randomized control.
- **Statistical Methods**: Techniques such as **propensity score matching**, **difference-in-differences**, and **instrumental variables** can help estimate causal effects in observational data by adjusting for confounders and biases.

## Causal Effect:
---
The causal effect refers to the impact of a cause (X) on an outcome (Y). It can be denoted as Y(X=1) - Y(X=0), where:

- Y(X=1): Outcome for units that received the cause (treatment group).
- Y(X=0): Outcome for units that didn't receive the cause (control group).

Challenges arise in determining the causal effect because we cannot directly observe both Y(X=1) and Y(X=0) for the same unit. For example, imagine giving someone a medication and then observing their health improve. ### Causal Effect Estimation:

**Goal:** The goal of causal effect estimation is to estimate the average causal effect for a population, approximating Y(X=1) - Y(X=0).

**Methods:** There are various methods for estimating causal effects, each with its strengths and weaknesses. Here are two common approaches:

1. **Randomized Controlled Trials (RCTs):**
   - **Description:** RCTs are considered the gold standard for estimating causal effects. In RCTs, units are randomly assigned to receive the cause (treatment) or not (control).
   - **Advantages:** Random assignment helps control for confounding variables, i.e., other factors affecting the outcome, resulting in unbiased causal effect estimates.
   - **Challenges:** RCTs may not always be feasible due to ethical or practical reasons. Additionally, they can be costly and time-consuming.

2. **Observational Studies:**
   - **Description:** Observational studies analyze existing data where units were not randomly assigned to a cause. These methods require careful consideration of confounding variables.
   - **Approaches:** Observational studies often rely on statistical adjustments or techniques like propensity score matching to estimate causal effects.
   - **Challenges:** Confounding variables are a major challenge in observational studies, and accurately accounting for them is crucial for obtaining reliable causal effect estimates.

Both RCTs and observational studies play important roles in causal inference, with RCTs providing rigorous control over confounding variables but often being limited in practicality, while observational studies allow for analysis of existing data but require sophisticated statistical techniques to address confounding.


## Confounding Variables in Causality
---
**Confounding variables** play a significant role in the analysis of causal relationships, as they can distort or mask the true relationship between the cause and effect. In causal inference, a **confounder** is a third variable that influences both the independent variable (cause) and the dependent variable (effect), creating a false or misleading impression of a causal relationship between them. Confounding can lead to biased or incorrect conclusions if not properly accounted for, making it one of the most important concepts to understand when working with observational data.

#### Key Points about Confounding Variables:

1. **Definition**:
   A **confounder** is a variable that:
   - Affects both the **exposure (independent variable)** and the **outcome (dependent variable)**.
   - Creates a spurious association between the exposure and the outcome, potentially leading to a misleading causal interpretation.
   
   For instance, if you're studying the relationship between exercise (cause) and weight loss (effect), a confounder could be **diet**. Diet affects both the likelihood of exercising and the amount of weight lost, so failing to account for diet could distort the observed relationship between exercise and weight loss.

2. **Example of Confounding**:
   Suppose we want to study the effect of **drinking coffee** on **heart disease**. We find that coffee drinkers appear to have a higher risk of heart disease. However, if we do not account for **smoking** as a confounder, the observed relationship between coffee drinking and heart disease could be misleading. Smoking is a confounder because:
   - Smokers are more likely to drink coffee (the exposure).
   - Smoking is a known risk factor for heart disease (the outcome).
   
   In this case, it’s not the coffee drinking itself that causes heart disease but the fact that smoking is related to both coffee consumption and heart disease. 

3. **Why Confounding is Problematic**:
   Confounding can lead to **biased estimates** of causal effects. If confounding is not addressed, the observed effect between the independent and dependent variables might be exaggerated, diminished, or even reversed. This leads to invalid conclusions about cause-and-effect relationships.

4. **Controlling for Confounding**:
   In causal inference, it's crucial to **control for confounders** to obtain valid and unbiased estimates of causal effects. Several methods are used to handle confounding:

   - **Randomization**: In randomized controlled trials (RCTs), participants are randomly assigned to treatment and control groups, ensuring that potential confounders are evenly distributed across both groups, thereby minimizing their impact.
   
   - **Statistical Control**: In observational studies, statistical methods such as **multivariable regression**, **propensity score matching**, and **instrumental variables** can be used to adjust for confounders and isolate the true causal effect. For instance, in a regression model, confounders can be included as additional variables to control for their influence.
   
   - **Stratification**: Stratifying the data by confounding variables involves splitting the data into subgroups (e.g., by age or gender) and analyzing the relationship between the exposure and outcome separately within each subgroup to control for the confounder’s effect.

5. **Identifying Confounders**:
   To identify potential confounders, researchers typically rely on **subject-matter knowledge** or **theoretical frameworks**. However, statistical methods can also provide insight into the role of confounders. For example, **causal graphs** or **Directed Acyclic Graphs (DAGs)** can help visualize the relationships between variables, making it easier to spot potential confounders.

6. **Confounding vs. Collapsibility**:
   It's important to distinguish between confounding and **collapsibility**. In some cases, controlling for a variable might not be necessary or could even introduce bias. This situation occurs when the effect of the independent variable on the outcome is **not truly causal**, or the relationship is more complex than initially understood. Collapsing refers to simplifying the analysis by ignoring certain variables when they don't need to be controlled for.

#### Visual Representation:
Confounders can often be represented in **Directed Acyclic Graphs (DAGs)**, where nodes represent variables, and directed edges represent causal relationships. In a DAG, a confounder is a variable that has arrows pointing to both the exposure and the outcome, creating a **backdoor path** that could lead to a spurious relationship between the two.

#### Example of Confounder in DAG:
Consider a situation where we study the effect of **exercise** (X) on **health** (Y). If **age** (Z) influences both exercise and health, it might confound the relationship between exercise and health, as shown in the DAG below:

```
  Age (Z) --> Exercise (X) --> Health (Y)
          \_________________________/
```

Here, age is a confounder because it affects both exercise and health, leading to a potential spurious relationship between exercise and health if not accounted for.

#### Dealing with Confounding in Data:
- **Randomized Control Trials (RCTs)**: The gold standard for addressing confounding. By randomly assigning treatments, confounding variables are expected to be distributed evenly across treatment groups.
- **Matching Methods**: Techniques like **propensity score matching** aim to match individuals from the treatment and control groups with similar confounding variable values, making it possible to estimate causal effects more accurately.
- **Instrumental Variables**: These are variables that are related to the treatment but not directly to the outcome, allowing for the isolation of the treatment’s effect by controlling for confounders.

### Judea Pearl's Ladder of Causality: A Framework for Understanding ---

In unraveling cause-and-effect relationships, Judea Pearl's "ladder of causality" serves as a guiding framework. It outlines three key steps that progressively deepen our understanding:  

- **Association (Seeing):** The starting point, where we observe patterns and correlations between events. For example, noticing if two variables change together suggests a potential link but doesn’t confirm causation.  
- **Intervention (Doing):** At this level, we actively test changes by altering one variable to observe its effect on another. This experimental approach helps establish cause-and-effect relationships, answering questions like, "If I do this, what will happen?"  
- **Causality (Imagining):** The final step explores hypothetical scenarios and counterfactuals, allowing us to consider "what if" questions. By imagining alternate outcomes, we gain deeper insights into complex causal dynamics.  

Pearl’s ladder guides us from simple observation to experimentation and finally to envisioning possibilities, offering a structured path to uncover the "why" behind events.

- The skeleton is then converted into a **Partially Directed Acyclic Graph (PDAG)** using the `skeleton_to_pdag()` method, followed by conversion into a **Directed Acyclic Graph (DAG)** with `pdag_to_dag()`. The edges of each graph are printed to show the structure of the causal relationships.
- Finally, warnings are restored to their default behavior using `warnings.filterwarnings('default')`.


````python
warnings.filterwarnings('ignore')
est = ConstraintBasedEstimator(df_disc)

skel, seperating_sets = est.estimate_skeleton(significance_level=0.01)
print("Undirected edges: ", skel.edges())

pdag = est.skeleton_to_pdag(skel, seperating_sets)
print("PDAG edges:", pdag.edges())

model = est.pdag_to_dag(pdag)
print("DAG edges:", model.edges())
warnings.filterwarnings('default')
````

### Generating Directed Graph of Variable Relationships


````python
np.set_printoptions(precision=3, suppress=True)
````

- This function `make_graph` visualizes a directed graph based on an adjacency matrix.
- The adjacency matrix is first filtered to include only elements with an absolute value greater than 0.5, creating a boolean index (`idx`).
- The `graphviz` library is used to generate a graph with nodes representing the labels (either provided or defaulting to `x0, x1, ...`), and directed edges between nodes are drawn for each non-zero coefficient in the adjacency matrix, with edge labels showing the coefficient value.


````python
def make_graph(adjacency_matrix, labels=None):
    idx = np.abs(adjacency_matrix) > 0.5
    dirs = np.where(idx)
    d = graphviz.Digraph(format='png', engine='dot')
    d.attr('node', fontname="MS Gothic", fontsize="10")
    names = labels if labels else [f'x{i}' for i in range(len(adjacency_matrix))]
    for to, from_, coef in zip(dirs[0], dirs[1], adjacency_matrix[idx]):
        d.edge(names[from_], names[to], label=f'{coef:.2f}', fontsize="10")
    return d
````

### Causal Inference and Visualization of Variable Relationships

Causal Inference: We apply the Direct LiNGAM model to uncover causal relationships between variables in the dataset.

Visualization: We visualize the variable relationships using a directed graph, providing a clear representation of causality between variables.


- This function creates a directed graph using NetworkX (`nx.DiGraph`), where nodes are labeled using the provided `labels` or the dataset's column names if no labels are provided.
- It iterates over the adjacency matrix, adding edges between nodes based on non-zero values, while also assigning weights to the edges.
- The positions of the nodes are determined using `nx.spring_layout`, which uses a force-directed algorithm for better visualization.
- The graph is drawn with nodes in skyblue color, and edges are drawn in gray with weights displayed on the edges.
- The graph is displayed using `matplotlib`, with labels for both the nodes and edges, and the axis is turned off for a clean view.ce.
les.

````python
def make_graph(adjacency_matrix, labels=None):
    G = nx.DiGraph()  # Directed graph
    # If labels are not provided, use the dataset's column names (assuming it's a DataFrame)
    if labels is None:
        labels = df_disc.columns.tolist()  # Adjust based on your dataset
    
    G.add_nodes_from(labels)
    
    # Add edges based on the adjacency matrix with weights
    for i in range(len(adjacency_matrix)):
        for j in range(len(adjacency_matrix)):
            if adjacency_matrix[i, j] != 0:
                G.add_edge(labels[j], labels[i], weight=adjacency_matrix[i, j])  # Add weight to edge
    
    pos = nx.spring_layout(G, seed=42)  # positions for all nodes, with a fixed seed for reproducibility
    plt.figure(figsize=(12, 8))
    # Nodes
    nx.draw_networkx_nodes(G, pos, node_size=700, node_color="skyblue")
    # Edges with weights as labels
    nx.draw_networkx_edges(G, pos, width=2, edge_color="gray", alpha=0.7)
    nx.draw_networkx_edge_labels(G, pos, edge_labels={(labels[j], labels[i]): f"{adjacency_matrix[i, j]:.2f}" for i in range(len(adjacency_matrix)) for j in range(len(adjacency_matrix)) if adjacency_matrix[i, j] != 0})
    # Labels
    nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif", font_weight="bold")

    plt.axis('off')
    plt.show()

# Example usage
make_graph(model.adjacency_matrix_, df_disc.columns.tolist())  # Pass column names as labels
````
