# üîç **Introduction to Causality**

Causality plays a critical role in data science, particularly when analyzing how one event impacts another. In the realm of stock markets, causality helps answer questions such as:  
- Did a specific event (e.g., Tesla‚Äôs Q4 earnings report) **cause** a stock price surge?  
- Or was the stock price movement simply **correlated** with broader market trends?  

Understanding causality goes beyond surface-level correlations. While correlation measures the strength of a relationship between two variables, it does not imply that one variable causes the other. Causality identifies the true **cause-and-effect relationship**, enabling better decision-making and predictions.

---

## üöÄ **Why Causality Matters in Financial Analysis?**

1. **Distinguishing Noise from Impact:**  
   Stock prices are influenced by numerous factors, such as economic policies, industry trends, and company-specific announcements. Without understanding causality, it's easy to misattribute random fluctuations to meaningful events.

2. **Actionable Insights for Investors:**  
   Identifying causal drivers of stock price changes enables investors to react appropriately and make informed decisions. For instance, was Tesla‚Äôs price surge due to its **earnings report** or external factors like **EV market growth**?

3. **Reliable Counterfactual Analysis:**  
   Causality allows analysts to answer **‚Äúwhat-if‚Äù scenarios**, such as:  
   - *What if Tesla hadn‚Äôt announced record profits?*  
   - *How would the stock price have behaved in the absence of the event?*

---

## üîë **Key Concepts in Causality**  

### 1Ô∏è‚É£ **Correlation vs. Causation**  
   - **Correlation:** Indicates a relationship between two variables. For example, Tesla's stock may correlate with Nasdaq trends.  
   - **Causation:** Demonstrates a direct cause-and-effect link, such as Tesla‚Äôs Q4 earnings directly boosting its stock price.  

   üõë **Common Pitfall:** Assuming correlation implies causation without robust statistical evidence.

---

### 2Ô∏è‚É£ **Approaches to Establish Causality**  
Various techniques are used to identify causality:  

1. **Randomized Controlled Trials (RCTs):**  
   - Considered the gold standard for causality.  
   - Not feasible in financial markets due to ethical and practical constraints (e.g., we can't manipulate stock prices).  

2. **Causal Inference Models:**  
   - Methods like **Google CausalImpact** or **Directed Acyclic Graphs (DAGs)** are used to estimate causal effects in observational data.  
   - Example: Assessing Tesla‚Äôs stock price post-earnings report using pre-event trends and related stock movements.  

3. **Counterfactual Reasoning:**  
   - Imagine alternative scenarios to evaluate what could have happened.  
   - Example: *What would Tesla‚Äôs stock price look like without its Q4 earnings report?*

---

### 3Ô∏è‚É£ **Challenges in Establishing Causality**  
   - **Confounding Variables:**  
     Factors like broader EV market trends or competitor announcements can confound analysis.  
     - *Solution:* Control for confounders using statistical models.  

   - **Reverse Causation:**  
     Did Tesla‚Äôs price rise due to its earnings report, or did rising market confidence drive both?  
     - *Solution:* Use temporal analysis to establish event sequences.  

   - **Ethical Constraints:**  
     Financial markets are observational by nature. Analysts must rely on robust econometric models instead of experimental setups.

---

# üõ†Ô∏è **Handling Data**

Data handling is a crucial step to prepare the dataset for analysis and modeling. It ensures data quality, consistency, and completeness by addressing common issues like missing values, duplicate entries, and formatting inconsistencies.

---

## 1Ô∏è‚É£ **Dealing with Missing Values**
Missing data is a common issue in stock market datasets, often due to non-trading days (e.g., weekends and holidays).  
**Approaches to handle missing values:**
- **Forward Fill (FFill)**: Propagates the last observed value forward to fill gaps.
- **Backward Fill (BFill)**: Uses the next observed value to fill preceding gaps.
- **Interpolation**: Estimates missing values based on surrounding data.

---

## 2Ô∏è‚É£ **Ensuring Continuous Time Series**
For time series modeling, the dataset must include a continuous sequence of dates. Missing dates are filled, and corresponding stock prices are handled appropriately using forward and backward filling.

---

## 3Ô∏è‚É£ **Adding New Features**
Enhance the dataset by introducing additional columns for better analysis:
- **Rolling Averages**: Smooths fluctuations and highlights trends (e.g., 7-day and 30-day moving averages).
- **Daily Returns**: Calculates percentage change in stock prices day-to-day.
- **Event Flags**: Marks key events (e.g., pre-event or post-event) to enable comparative analysis.

---

## 4Ô∏è‚É£ **Outlier Detection**
Detect and handle unusual stock price spikes or drops that could skew results. Outliers are identified using:
- Z-Score Method
- Interquartile Range (IQR)

---

## 5Ô∏è‚É£ **Feature Normalization**
Normalize numerical features (e.g., stock prices) to a standard scale to improve model performance. Techniques include:
- **Min-Max Scaling**: Scales values between 0 and 1.
- **Z-Score Normalization**: Transforms data to have a mean of 0 and standard deviation of 1.

---

## 6Ô∏è‚É£ **Final Data Validation**
Ensure the dataset is:
- Free of missing or invalid values.
- Properly indexed with continuous dates.
- Enhanced with meaningful features for analysis and modeling.

This comprehensive data handling process sets the stage for accurate and robust modeling of Tesla's stock prices.


```python
# 1Ô∏è‚É£ Handling Missing Values
# Check for missing values
missing_values = tesla_data.isnull().sum()
print("Missing Values:\n", missing_values)

# Fill missing values using forward and backward fill
tesla_data.fillna(method='ffill', inplace=True)
tesla_data.fillna(method='bfill', inplace=True)
```

```python
# 2Ô∏è‚É£ Ensuring Continuous Time Series
# Create a complete date range for the dataset
full_dates = pd.date_range(start=tesla_data['Date'].min(), end=tesla_data['Date'].max())
tesla_data = tesla_data.set_index('Date').reindex(full_dates).rename_axis('Date').reset_index()

# Fill missing data caused by non-trading days
tesla_data.fillna(method='ffill', inplace=True)
tesla_data.fillna(method='bfill', inplace=True)
```

```python
# Ensure 'Date' column is timezone-naive
tesla_data['Date'] = pd.to_datetime(tesla_data['Date']).dt.tz_localize(None)

# Specify a timezone-naive event date
event_date = pd.to_datetime('2022-01-26')

# Event Flags
tesla_data['Event'] = tesla_data['Date'].apply(lambda x: 'pre' if x < event_date else 'post')
```

```python
# Rename the column 'Adj Close TSLA' to 'Adj Close'
tesla_data.rename(columns={'Adj Close TSLA': 'Adj Close'}, inplace=True)

# Calculate Q1, Q3, and IQR
Q1 = tesla_data['Adj Close'].quantile(0.25)
Q3 = tesla_data['Adj Close'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = tesla_data[(tesla_data['Adj Close'] < lower_bound) | (tesla_data['Adj Close'] > upper_bound)]
print("Outliers:\n", outliers)

# Handle outliers by capping them within the bounds
tesla_data['Adj Close'] = tesla_data['Adj Close'].clip(lower=lower_bound, upper=upper_bound)
```

```python
# 5Ô∏è‚É£ Feature Normalization
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Min-Max Scaling
scaler = MinMaxScaler()
tesla_data['Adj Close Scaled'] = scaler.fit_transform(tesla_data[['Adj Close']])

# Z-Score Normalization
z_scaler = StandardScaler()
tesla_data['Adj Close Z-Score'] = z_scaler.fit_transform(tesla_data[['Adj Close']])
```

```python
# 6Ô∏è‚É£ Final Data Validation
# Check for missing values again
print("Final Missing Values:\n", tesla_data.isnull().sum())

# Preview the final dataset
tesla_data.head()
```

```python
# Recalculate rolling averages
tesla_data['7-Day MA'] = tesla_data['Adj Close'].rolling(window=7, min_periods=1).mean()
tesla_data['30-Day MA'] = tesla_data['Adj Close'].rolling(window=30, min_periods=1).mean()

# Calculate Daily Returns
tesla_data['Daily Returns'] = tesla_data['Adj Close'].pct_change()
tesla_data['Daily Returns'].fillna(0, inplace=True)

# Verify the data
print(tesla_data[['Adj Close', '7-Day MA', '30-Day MA', 'Daily Returns']].head())
```

- The **Fourth Quarter Earning Report** was released on **2022-01-26**.

```python
import numpy as np

# Parameters
last_price = tesla_data['Adj Close'].iloc[-1]
volatility = tesla_data['Adj Close'].pct_change().std()
n_simulations = 100
n_days = 30

# Simulations
simulated_paths = []
for _ in range(n_simulations):
    daily_returns = np.random.normal(0, volatility, n_days)
    price_path = [last_price]
    for return_ in daily_returns:
        price_path.append(price_path[-1] * (1 + return_))
    simulated_paths.append(price_path)

# Plot
plt.figure(figsize=(10, 6))
for path in simulated_paths:
    plt.plot(range(len(path)), path, color='grey', alpha=0.3)
plt.title("Monte Carlo Simulations for Tesla Stock Prices")
plt.xlabel("Days")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import numpy as np

# Normalize data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(tesla_data['Adj Close'].values.reshape(-1, 1))

# Prepare training data
look_back = 60
X_train, y_train = [], []
for i in range(look_back, len(scaled_data)):
    X_train.append(scaled_data[i-look_back:i, 0])
    y_train.append(scaled_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

# Build LSTM model
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    LSTM(units=50),
    Dense(units=1)
])
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Predict future prices
test_data = scaled_data[-look_back:]
X_test = [test_data]
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

predicted_prices = model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted_prices)

# Plot forecast
plt.figure(figsize=(10, 6))
plt.plot(tesla_data['Date'], tesla_data['Adj Close'], label='Observed', color='blue')
plt.plot(pd.date_range(start=tesla_data['Date'].iloc[-1], periods=1), predicted_prices, label='Forecast', color='green')
plt.title("Tesla Adjusted Close Price Forecast (LSTM)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

### üìò **Explanation of RCTs (Randomized Controlled Trials)**

Randomized Controlled Trials (RCTs) are considered the gold standard for identifying causal relationships. However, implementing RCTs directly in observational data like stock market data is infeasible due to the lack of random treatment allocation. Instead, we use an **RCT-like framework** in your code by simulating treatment and control groups based on predefined thresholds (e.g., high vs. low trading volume).

---

### **Steps to Simulate RCTs in Stock Data Analysis**

1. **Define Treatment and Control Groups**:
   - Use a metric (e.g., `Volume`) to divide the data into treatment (`Volume` above a threshold) and control groups (`Volume` below the threshold).

2. **Measure Outcomes**:
   - Analyze differences in the `Adj Close` price between treatment and control groups to simulate causal effects.

3. **Comparison Through Visuals**:

4. **Limitations**:

This approach is valuable for observational studies to estimate potential causal effects and guide further analysis.

### **Insights Derived**:
- Helps identify how certain "treatments" (e.g., high trading volumes) impact outcomes (`Adj Close` prices).

```python

# Define the treatment threshold as the 75th percentile of 'Volume TSLA'
volume_threshold = tesla_data['Volume TSLA'].quantile(0.75)

# Create treatment and control groups
tesla_data['Treatment'] = (tesla_data['Volume TSLA'] > volume_threshold).astype(int)

# Summarize average Adjusted Close prices for both groups
avg_prices = tesla_data.groupby('Treatment')['Adj Close'].mean()
print("Average Adjusted Close Prices:")
print(avg_prices)

# Visualize the effect of treatment using a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=tesla_data, x='Treatment', y='Adj Close', palette='coolwarm')
plt.title("Effect of High vs. Low Trading Volume on Tesla's Adjusted Close Price", fontsize=16)
plt.xlabel("Treatment Group (0 = Low Volume, 1 = High Volume)", fontsize=12)
plt.ylabel("Adjusted Close Price (USD)", fontsize=12)
plt.grid(alpha=0.5)
plt.show()

# Additional Analysis: Line Plot for Trends Over Time
plt.figure(figsize=(14, 7))
plt.plot(
    tesla_data[tesla_data['Treatment'] == 1]['Date'],
    tesla_data[tesla_data['Treatment'] == 1]['Adj Close'],
    label="Treatment Group (High Volume)", color="orange"
)
plt.plot(
    tesla_data[tesla_data['Treatment'] == 0]['Date'],
    tesla_data[tesla_data['Treatment'] == 0]['Adj Close'],
    label="Control Group (Low Volume)", color="blue"
)
plt.axvline(pd.to_datetime('2022-01-26'), color='red', linestyle='--', label='Earnings Report Date')
plt.title("Tesla Stock Prices by Treatment Group Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Adjusted Close Price (USD)", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```

### üìò **Explanation of Average Treatment Effect (ATE)**

The **Average Treatment Effect (ATE)** measures the causal effect of a treatment on an outcome by comparing the average outcomes between the treatment and control groups.

---

### **Steps to Calculate ATE**:
1. **Define Treatment and Control Groups**:
   - Divide data into groups based on a "treatment" variable (e.g., high vs. low trading volume).

2. **Compute Average Outcomes**:
   - Calculate the mean outcome (e.g., Adjusted Close price) for the treatment and control groups.

3. **Calculate ATE**:
   - Subtract the average outcome of the control group from the average outcome of the treatment group.
   - In simpler terms:
     - ATE = (Average outcome in the treatment group) - (Average outcome in the control group).

4. **Interpret ATE**:
   - A positive ATE indicates that the treatment has a beneficial effect on the outcome.
   - A negative ATE suggests that the treatment has a detrimental effect on the outcome.

---

### **Insights**:
- ATE quantifies the average impact of high trading volume on Tesla's adjusted close price.
- This helps determine whether high trading volume drives significant changes in stock prices.

---

### **Limitations**:
- Confounding factors may affect the validity of ATE unless controlled.
- Observational data may not perfectly simulate randomized experiments.


```python
# Define the treatment threshold as the 75th percentile of 'Volume TSLA'
volume_threshold = tesla_data['Volume TSLA'].quantile(0.75)

# Create treatment and control groups
tesla_data['Treatment'] = (tesla_data['Volume TSLA'] > volume_threshold).astype(int)

# Calculate average outcomes for treatment and control groups
treatment_mean = tesla_data[tesla_data['Treatment'] == 1]['Adj Close'].mean()
control_mean = tesla_data[tesla_data['Treatment'] == 0]['Adj Close'].mean()

# Calculate ATE
ate = treatment_mean - control_mean

# Display the results
print(f"Average Adjusted Close Price for Treatment Group (High Volume): ${treatment_mean:.2f}")
print(f"Average Adjusted Close Price for Control Group (Low Volume): ${control_mean:.2f}")
print(f"Average Treatment Effect (ATE): ${ate:.2f}")

# Optional: Visualize the ATE
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(['Control (Low Volume)', 'Treatment (High Volume)'], [control_mean, treatment_mean], color=['blue', 'orange'], alpha=0.7, edgecolor='black')
plt.title("Average Treatment Effect (ATE) on Tesla's Adjusted Close Price", fontsize=14)
plt.ylabel("Average Adjusted Close Price (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### üìò **Simpson's Paradox in Causality**

Simpson's Paradox refers to a phenomenon where a trend observed within multiple groups reverses or disappears when the groups are combined. This paradox often occurs when confounding variables are not properly accounted for, leading to misleading interpretations of data.

---

### **Key Concepts**:
1. **Confounding Variable**: A variable that affects both the treatment and the outcome, potentially distorting the causal relationship.
2. **Grouped vs. Combined Data**: Simpson's Paradox arises when aggregating data without considering group-level effects.
3. **Causal Interpretation**: To resolve the paradox, data must be stratified or adjusted based on the confounding variable.

---

### **Example in Stock Analysis**:
- **Scenario**: Tesla's stock price trends are analyzed for two different periods, `High Volume` and `Low Volume`.
- **Paradox**: Individually, high trading volume might appear beneficial, but when data is aggregated, it could show a negative impact, caused by an underlying variable (e.g., market conditions or specific events).

---

### **Steps to Demonstrate Simpson's Paradox**:
1. Create synthetic or real data for two groups (e.g., high and low trading volume).
2. Calculate trends for each group and the combined data.
3. Highlight how combining the groups can reverse the observed trend.
4. Adjust for the confounding variable to explain the paradox.

---

### **Insights**:
- Simpson's Paradox highlights the importance of stratifying data by confounding variables.
- Understanding and addressing this paradox is crucial for robust causal inference.

---


```python
# Step 1: Create synthetic data
np.random.seed(42)

# High Volume Group
high_volume = pd.DataFrame({
    "Volume": np.random.uniform(80, 100, 100),
    "Adjusted_Close": np.random.uniform(300, 400, 100),
    "Group": "High Volume"
})

# Low Volume Group
low_volume = pd.DataFrame({
    "Volume": np.random.uniform(20, 40, 100),
    "Adjusted_Close": np.random.uniform(200, 300, 100),
    "Group": "Low Volume"
})

# Step 2: Combine data
data = pd.concat([high_volume, low_volume])

# Add a confounding variable
data["Market_Condition"] = np.where(data["Group"] == "High Volume", "Bull Market", "Bear Market")

# Step 3: Compute trends
grouped_trend = data.groupby("Group")["Adjusted_Close"].mean()
overall_trend = data["Adjusted_Close"].mean()

# Step 4: Plot results
plt.figure(figsize=(10, 6))

# Plot individual group means
sns.barplot(data=data, x="Group", y="Adjusted_Close", ci=None, palette="coolwarm", alpha=0.7)
plt.axhline(overall_trend, color="green", linestyle="--", label="Combined Average")
plt.title("Simpson's Paradox: Grouped vs. Combined Data", fontsize=14)
plt.xlabel("Group")
plt.ylabel("Average Adjusted Close Price")
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Adjust for confounding variable
adjusted_trend = data.groupby(["Group", "Market_Condition"])["Adjusted_Close"].mean()

print("Grouped Trend:\n", grouped_trend)
print("\nOverall Trend (Combined):\n", overall_trend)
print("\nAdjusted Trend by Market Condition:\n", adjusted_trend)
```

### üìä **Sensitivity Analysis in Causal Inference**

**Objective**: This analysis evaluates the robustness of the observed Average Treatment Effect (ATE) under varying levels of unmeasured confounding. By introducing a synthetic confounder and adjusting its influence, we test how the treatment effect (high trading volume) on Tesla's adjusted closing price changes.

---

#### **Key Steps**:
1. **Treatment Definition**:
   - The treatment is defined based on Tesla's trading volume, where values above the 75th percentile are considered "high" (treatment group).

2. **Simulating Confounding**:
   - A synthetic confounder variable is added to simulate the impact of unobserved variables.

3. **Sensitivity Testing**:
   - The ATE is recalculated for a range of confounding strengths, visualizing how sensitive the causal relationship is to potential biases.

---

#### **Insights**:
- The sensitivity analysis helps quantify how robust the estimated causal relationship is to unmeasured confounding.
- A large range of stable ATEs suggests that the results are less sensitive to confounding.
- Significant variation in ATE implies that unmeasured confounding could heavily impact the conclusions.

---

#### **Limitations**:
- The synthetic confounder is hypothetical and does not capture real-world confounders.
- Assumptions about confounding strengths may oversimplify the true data-generating process.


```python
# Step 1: Estimate Naive ATE
avg_treated = tesla_data[tesla_data['Treatment'] == 1]['Adj Close'].mean()
avg_control = tesla_data[tesla_data['Treatment'] == 0]['Adj Close'].mean()
naive_ate = avg_treated - avg_control

print(f"Naive Average Treatment Effect (ATE): {naive_ate:.2f}")

# Step 2: Sensitivity Analysis
np.random.seed(42)  # Ensure reproducibility
tesla_data['Confounder'] = np.random.normal(0, 1, len(tesla_data))  # Simulate unmeasured confounder

# Define a range of unmeasured confounding strengths
confounding_strengths = np.linspace(-2, 2, 100)
sensitivity_results = []

for confounding_strength in confounding_strengths:
    # Adjust the adjusted close price to simulate unmeasured confounding
    tesla_data['Adjusted Outcome'] = tesla_data['Adj Close'] + confounding_strength * tesla_data['Confounder']
    avg_treated_adjusted = tesla_data[tesla_data['Treatment'] == 1]['Adjusted Outcome'].mean()
    avg_control_adjusted = tesla_data[tesla_data['Treatment'] == 0]['Adjusted Outcome'].mean()
    adjusted_ate = avg_treated_adjusted - avg_control_adjusted
    sensitivity_results.append(adjusted_ate)

# Step 3: Plot Sensitivity Analysis Results
plt.figure(figsize=(10, 6))
plt.plot(confounding_strengths, sensitivity_results, label="Adjusted ATE", color="blue")
plt.axhline(y=naive_ate, color='red', linestyle='--', label="Naive ATE")
plt.title("Sensitivity Analysis of Average Treatment Effect (ATE) on Tesla Stock")
plt.xlabel("Unmeasured Confounding Strength")
plt.ylabel("Adjusted ATE")
plt.legend()
plt.grid(alpha=0.5)
plt.show()

# Display ATE range
true_effect_range = (min(sensitivity_results), max(sensitivity_results))
print(f"Range of Adjusted ATE under Sensitivity Analysis: {true_effect_range[0]:.2f} to {true_effect_range[1]:.2f}")
```

### üìä **Temporal Precedence in Causality**

**Definition**:  
Temporal precedence refers to the principle that the cause must occur before the effect in time. In causal inference, establishing temporal precedence is critical to determining whether a variable (cause) influences another (effect).

---

### **Why Temporal Precedence Matters**:
1. **Direction of Causality**: Ensures that the observed relationship is causal and not reversed. For example, high trading volume (cause) must occur before changes in Tesla‚Äôs adjusted close price (effect).
2. **Avoiding Reverse Causation**: Without temporal precedence, there is a risk of concluding that the effect influences the cause.
3. **Establishing Time Dependencies**: Temporal precedence is foundational for time-series data where variables are analyzed over time.

---

### **Steps to Establish Temporal Precedence**:
1. **Identify Cause and Effect**:
   - Define the independent variable (e.g., trading volume) and the dependent variable (e.g., stock price).

2. **Verify Time Sequence**:
   - Ensure that the cause consistently occurs before the effect in the data.

3. **Lagged Variables**:
   - Introduce lagged variables (e.g., trading volume at time \( t-1 \)) to account for delays in the effect.

4. **Temporal Analysis**:
   - Use time-series models to test the causal relationship while respecting the chronological order of events.

---

### **Limitations**:
- Temporal precedence alone does not guarantee causality; confounders or external factors may still influence the relationship.
- Observational data may lack precision in time-stamping events, making it harder to establish clear precedence.

---

### **Application in Tesla Stock Analysis**:
- **Cause**: Trading volume (e.g., high volume as treatment).  
- **Effect**: Tesla's adjusted close price (e.g., changes in price driven by trading activity).  
- **Temporal Order**: Analyze the relationship between lagged trading volume (e.g., trading volume on day \( t-1 \)) and closing price on day \( t \) to establish temporal precedence.

This ensures that trading volume influences stock price changes, rather than the reverse.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import grangercausalitytests

# Use the correct column name for trading volume
volume_column = 'Volume TSLA'

# Create lagged variables
tesla_data['Lagged Volume'] = tesla_data[volume_column].shift(1)

# Drop rows with NaN values due to lagging
tesla_data = tesla_data.dropna()

# Step 1: Visualize Temporal Relationship
plt.figure(figsize=(14, 7))
plt.plot(tesla_data['Date'], tesla_data['Adj Close'], label='Adjusted Close Price', color='blue')
plt.plot(tesla_data['Date'], tesla_data['Lagged Volume'], label='Lagged Trading Volume', color='orange', alpha=0.7)
plt.title("Temporal Relationship Between Lagged Volume and Adjusted Close Price")
plt.xlabel("Date")
plt.ylabel("Values")
plt.legend()
plt.grid(alpha=0.5)
plt.show()

# Step 2: Granger Causality Test
# Prepare data for Granger causality
granger_data = tesla_data[['Adj Close', 'Lagged Volume']]
max_lag = 5  # Test up to 5-day lag

# Perform Granger Causality Test
granger_results = grangercausalitytests(granger_data, max_lag, verbose=True)

# Step 3: Interpret Results
for lag, test in granger_results.items():
    p_value = test[0]['ssr_ftest'][1]
    print(f"Lag {lag}: P-value = {p_value:.4f}")
    if p_value < 0.05:
        print(f"Significant Granger causality detected at lag {lag} (p < 0.05)")
```
