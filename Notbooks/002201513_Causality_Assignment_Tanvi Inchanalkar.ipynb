# Causality

Causality is the principle that explains the relationship between a cause and its effect, where one event (the cause) directly leads to the occurrence of another event (the effect). This concept is fundamental in fields like philosophy, science, and everyday reasoning, helping us understand how and why things happen.

In philosophy, causality has been widely debated, especially in relation to determinism, free will, and the essence of causation itself. Some key philosophical views include:

- Humean Causality: Proposed by David Hume, this theory suggests that causality is not something we directly observe but a mental habit formed by observing recurring patterns in events. According to Hume, we infer cause and effect from these regularities, not from the events themselves.

- Counterfactual Theories: These theories argue that causality is determined by considering counterfactual situations—if the cause had not occurred, the effect would not have happened. It frames causality as an “if-then” scenario to explore what would have been the outcome in different circumstances.

- Mechanistic Causality: Widely used in scientific contexts, this view suggests that causes are physical mechanisms or processes that produce effects through a series of steps. This perspective is focused on understanding causality through natural laws and physical interactions.

In the realm of science, causality is central to explaining natural phenomena and making predictions. Scientists use various methods, including experiments, observations, and statistical analysis, to establish causal links. However, determining causality in complex systems is not always straightforward, as correlation does not always imply causation. To accurately infer causal relationships, researchers often rely on techniques like randomized controlled trials, temporal precedence, and causal modeling.

# Why Should I Use Causality?

With the exponential growth of data over the past two decades, the scope of questions we can address using data has expanded significantly. Many of these questions are inherently causal:

- Why did metric X change this month?
- Which customers should we target for campaign Y?
- What would happen if we implemented policy change Z?

While correlation can tell us what has happened, it often falls short in explaining why or how something occurred. Here are some challenges causality helps address:

1. Understanding Associations: Correlation alone cannot establish causation. It's essential to determine whether an observed relationship is meaningful or just a coincidence.

2. Reversed Effects: In some cases, effects seem to be reversed. Causal inference helps clarify these complexities and corrects for misinterpretations.

Causal analysis is crucial for making informed, data-driven decisions, as it allows us to understand the root causes behind changes and outcomes.

# Fundamentals of Causality

- **Temporal Order**: This principle states that the cause must precede the effect in time. For example, when studying the impact of smoking on lung cancer, smoking must occur before the development of lung cancer for it to be considered a causal factor.

- **Covariation**: Covariation requires a consistent relationship between the cause and effect, meaning changes in the cause must lead to changes in the effect. This helps confirm the authenticity of the cause-effect relationship and ensures that the connection is not due to random chance.

- **Non-Spuriousness**: Non-spuriousness asserts that the association between the cause and effect should not be explained by an external variable. For instance, while smoking and lung cancer may appear linked, other factors like age might act as a confounding variable. To establish causality, it's essential to control for such spurious variables to ensure that the relationship is genuine and not misleading.

# What is Causal Inference?
Causal inference is the process of determining the cause-and-effect relationships between variables. Unlike correlation, which only identifies associations between variables, causal inference seeks to understand how one variable directly influences another. This is critical for understanding underlying mechanisms, predicting the impact of interventions, and making data-driven decisions across various domains.

# Methods/Techniques in Causal Inference for Wine Dataset
1. **Randomized Controlled Trials (RCTs):**
RCTs involve randomly assigning participants to treatment or control groups to assess the effectiveness of a treatment. Although RCTs are typically used in medical studies, a similar approach can be applied in wine research to test the impact of different production methods or ingredients on wine quality.

2. **Instrumental Variable Analysis:**
In wine studies, an instrumental variable (IV) could be something like climate factors, which influence grape quality but are not directly linked to the wine’s final quality. By using IVs, we can account for endogeneity and isolate the causal effect of production methods on wine quality.

3. **Regression Discontinuity:**
This method is useful when treatment is applied based on a threshold. For example, a wine production method might only be used for grapes above a certain sugar level. By analyzing how wine quality changes across this threshold, we can estimate the causal effect of the method.

4. **Difference-in-Differences (DID):**
The DID approach can be applied to compare the impact of different winemaking techniques over time. By examining the quality of wines produced before and after a change in technique, this method helps estimate the treatment effect while controlling for time-related trends.

5. **Propensity Score Matching:**
Propensity score matching can be used to create comparable groups of wines produced with different methods. By matching wines based on their characteristics (e.g., acidity, sugar content), we can estimate the causal effect of production methods on wine quality, reducing bias from confounding variables.

6. **LiNGAM (Linear Non-Gaussian Acyclic Model):**
LiNGAM can be applied to uncover causal relationships in the wine dataset. By assuming linear relationships and non-Gaussian distributions, LiNGAM helps discover how different variables (e.g., acidity, alcohol, and residual sugar) causally influence wine quality. It works effectively with observational data, making it suitable for real-world wine studies without needing experimental interventions.

# Causal Inference using LiNGAM
LiNGAM (Linear Non-Gaussian Acyclic Model) is a causal discovery algorithm that aims to uncover the causal relationships among multiple variables. Unlike traditional statistical methods, which primarily detect correlations, LiNGAM is designed to identify the actual causal structure by assuming non-Gaussian data distributions. It works by analyzing the dependencies between variables to determine the causal ordering, making it a powerful tool for discovering hidden cause-and-effect relationships in complex datasets.

```python
# Import Required Libraries for Causal Inference
from IPython.display import Image
import category_encoders as enc
import graphviz
import lingam
from lingam.utils import make_dot, make_prior_knowledge
```

```python
# Handle Missing Values
processed_data = wine_data.dropna()
```

```python
# Encode Categorical Features
encoder = enc.OrdinalEncoder(cols=processed_data.select_dtypes(include="object"))
encoded_data = encoder.fit_transform(processed_data)
```

```python
# Prepare Feature Mapping
feature_names = encoded_data.columns.tolist()
feature_mapping = {name: idx for idx, name in enumerate(feature_names)}
```

# Causal Diagrams (DAGs)

In causal inference, Directed Acyclic Graphs (DAGs) are used to visualize and understand the assumed causal relationships between variables. DAGs help identify confounders, mediation effects, and direct causal paths. In this case, we will use DAGs to map out the potential causal relationships within the wine dataset, particularly focusing on the influence of various physicochemical properties on wine quality.

**Components of a DAG**

- Nodes: Each node in the DAG represents a variable in the dataset. In our case, nodes represent variables like fixed acidity, volatile acidity, alcohol, and quality, among others.
- Edges (Arrows): Arrows represent assumed causal relationships. For example, an arrow from alcohol to quality would imply that alcohol content has a causal effect on the wine's quality.  
- Acyclic: A DAG is acyclic, meaning there are no loops. This ensures that the cause-and-effect relationships do not cycle back.

**Principles of DAGs in Causal Inference:**

- Causal Direction: Arrows indicate the direction of causality. For example, if fixed acidity influences quality, an arrow points from fixed acidity to quality.
- Conditional Independence: If two variables are conditionally independent, no direct arrow will connect them when other variables are conditioned upon.
d-separation: A criterion to assess whether a set of variables is independent of another set, given a third set, based on the structure of the DAG.
- Path Blocking: If a variable intercepts a path between two other variables, conditioning on it can block the influence between them, helping identify which variables need to be controlled for in causal analysis.
  
**Causal Inference with LiNGAM for the Wine Dataset**
  
To apply LiNGAM (Linear Non-Gaussian Acyclic Model) to discover causal relationships in the wine dataset, we first define the causal structure. LiNGAM works with non-Gaussian data to determine the causal ordering of variables based on statistical dependencies.

- Exogenous and Endogenous Variables in the Wine Dataset
Exogenous Variables: These variables are assumed to not be influenced by other variables in the model. In the wine dataset, we can consider variables like alcohol and fixed acidity to be exogenous, assuming their effects are not caused by other features in the dataset.

- Endogenous Variables: These are variables that are influenced by other variables in the system. For example, quality is likely influenced by variables such as alcohol, fixed acidity, and volatile acidity.

**Creating Prior Knowledge with make_prior_knowledge()**

To integrate domain knowledge into the LiNGAM algorithm, we use the make_prior_knowledge() function. This function allows us to specify which variables are exogenous and define known causal paths between variables. For instance, we can set the following assumptions for the wine dataset:

- Exogenous Variables: We assume that fixed acidity, volatile acidity, and alcohol are exogenous, meaning their values are determined outside the system.

- Causal Paths: We define specific causal relationships, such as fixed acidity influencing quality, volatile acidity influencing quality, and alcohol influencing quality.

Here's how we would implement this in the wine dataset:

```python
# Create Prior Knowledge
prior_knowledge = make_prior_knowledge(
    n_variables=len(feature_names),
    paths=[
        (feature_mapping["fixed acidity"], feature_mapping["quality"]),
        (feature_mapping["volatile acidity"], feature_mapping["quality"]),
        (feature_mapping["alcohol"], feature_mapping["quality"])
    ]
)
```

```python

# Fit DirectLiNGAM Model
causal_model = lingam.DirectLiNGAM(random_state=42, measure='pwling', prior_knowledge=prior_knowledge)
causal_model.fit(encoded_data.values)
```

```python
# Generate DAG
dag_visual = make_dot(causal_model.adjacency_matrix_, labels=feature_names)
dag_visual.format = 'png'
dag_visual.render('causal_dag')

Image("causal_dag.png")
```

**Key Insights:**
1. **Direct Influences on Quality:**

'Alcohol' has a direct positive path to 'quality', suggesting higher alcohol content may be associated with better perceived wine quality.
'pH' has a direct negative path to 'quality', indicating that lower pH (more acidic wine) may lead to lower quality.
- Role of Acidity:

While 'fixed acidity' does not directly affect 'quality', it is linked to 'citric acid' and 'density', which in turn influence 'quality'. This implies that 'fixed acidity' may affect quality indirectly through these variables.

2. **Sulphur Dioxide Relationship:**

Both 'free sulfur dioxide' and 'total sulfur dioxide' are directly related to each other, with 'free sulfur dioxide' also impacting 'residual sugar'. However, neither sulfur dioxide variable has a direct path to 'quality', suggesting their influence on wine quality may be indirect.
- Alcohol's Central Role:

'Alcohol' plays a central role in the graph, with multiple incoming paths and a direct link to 'quality', emphasizing its importance in determining wine quality.

3. **Inter-relationships:**

Several variables, including 'fixed acidity', 'citric acid', and 'density', are interconnected, indicating a complex relationship rather than isolated effects on 'quality'.

4. **Negative Connections:**

Negative correlations are observed between 'volatile acidity', 'citric acid', and 'quality', suggesting that increases in these variables may negatively affect the perceived quality of wine.

# Interpreting Results

```python
# Linear Regression for Causal Effects
from sklearn.linear_model import LinearRegression

X = encoded_data.drop("quality", axis=1)
y = encoded_data["quality"]

reg_model = LinearRegression()
reg_model.fit(X, y)
```

```python

# Generate DAG with Linear Coefficients
feature_indices = [idx for idx in range(X.shape[1])]
dag_with_weights = make_dot(
    causal_model.adjacency_matrix_,
    labels=feature_names,
    prediction_feature_indices=feature_indices,
    prediction_target_label="quality",
    prediction_coefs=reg_model.coef_
)
```

```python
dag_with_weights.format = 'png'
dag_with_weights.render('causal_dag_with_weights')

Image("causal_dag_with_weights.png")
```

**Key Insights:**

3. **Sulfur Dioxide’s Role in Sugar Content:**

The connection between 'free sulfur dioxide', 'total sulfur dioxide', and 'residual sugar' suggests that sulfur dioxide levels may be impacting the wine's sugar content, potentially affecting fermentation and sweetness.

4. **Interconnected Acidity Relationships:**

The links between 'fixed acidity', 'volatile acidity', and 'citric acid' show a strong interconnection, indicating that different types of acidity in the wine are closely related and influence each other, as well as potentially affecting other variables.
