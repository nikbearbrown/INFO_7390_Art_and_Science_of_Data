## 1. Introduction to Causality  
Causality is a cornerstone concept in both science and philosophy. It describes the relationship between a cause (an initiating event) and an effect (a resulting event), where the effect directly stems from the cause. Essentially, causality asserts that one event is a necessary precursor to the occurrence of another.  

In the realm of data science, understanding causal relationships between variables is a common objective. For instance, we may want to assess whether smoking leads to lung cancer or if a promotional campaign results in increased sales. However, determining causality from observational data can be quite complex due to the presence of confounding variables that may influence both the cause and the effect.  

---

## 2. Correlation vs. Causation  
Although often conflated, correlation and causation represent fundamentally different concepts.  

- **Correlation** indicates a statistical association between two variables. When two variables are correlated, their values change in a consistent pattern. For example, ice cream sales and outdoor temperatures often display a positive correlation—sales tend to rise as the temperature increases. Statistical tools like Pearson’s correlation coefficient or Spearman’s rank correlation are used to measure such relationships.  

- **Causation**, however, denotes a direct, mechanistic link where changes in one variable directly bring about changes in another. Establishing causation is far more challenging than identifying correlation, often requiring experimental or quasi-experimental methods that can account for potential confounders.  

A useful reminder of the distinction is the phrase, **“Correlation does not imply causation.”** Just because two variables are correlated does not mean one causes the other. Other variables may be influencing both, or the observed relationship could be coincidental.  

**Example:**  
Consider examining the relationship between ice cream sales and robbery incidents. Data collected over several months in a city might show a strong positive correlation between the two variables. As ice cream sales increase, robbery incidents appear to rise as well. However, concluding that ice cream sales cause robbery would be incorrect. Factors like higher temperatures or the time of year could influence both variables. Alternatively, the observed relationship could be entirely coincidental, with no causal connection between ice cream sales and robbery.

### **Core Concepts in Causal Inference**

#### **Counterfactuals**
Counterfactuals explore what would have happened if an event or intervention had not taken place or had been altered. In essence, they represent the hypothetical outcomes that might have been observed under different circumstances.

For instance, in a study examining the effectiveness of a new medication for lowering blood pressure, the participants who received the medication form the treatment group, while those who did not make up the control group. The key outcome here is the change in blood pressure over time.

The counterfactual scenario for a person in the treatment group is imagining what their outcome would have been had they not taken the medication. Similarly, for a person in the control group, the counterfactual outcome is imagining what would have happened had they been given the medication.

#### **Instrumental Variables**
Instrumental variables (IV) provide a method to estimate causal effects when unobserved confounding might distort the relationship between the treatment and the outcome. IVs rely on an instrument, a third variable that influences the treatment but does not directly affect the outcome, to derive the causal estimate.

Consider a situation where we wish to assess the causal impact of education on earnings. There may be confounding factors, such as inherent abilities or family background, that affect both education and earnings. An IV in this case could be the distance to the nearest college. While the proximity to a college may influence the likelihood of attending, it has no direct effect on an individual’s earnings. By using this distance as an instrument, we can estimate how education truly affects earnings.

The IV estimation equation is:

$$
Y = a + bX + e
$$
$$
X = c + zD + u
$$

Where:
- \( Y \) is the outcome variable (e.g., earnings),
- \( X \) is the treatment (e.g., education),
- \( D \) is the instrument (e.g., distance to the nearest college),
- \( a, b, c, \) and \( z \) are coefficients to estimate,
- \( e \) and \( u \) represent error terms.

#### **Randomized Control Trials (RCTs)**
RCTs are experimental studies in which individuals are randomly assigned to treatment or control groups, allowing researchers to isolate the causal effect of a specific intervention. This method is widely employed in fields such as medicine, psychology, and social science.

For example, in medical research, a pharmaceutical company may develop a new drug for treating hypertension. To test its efficacy, they randomly assign 500 patients to either a treatment group (receiving the drug) or a control group (receiving a placebo). After a set period, blood pressure measurements are compared between the groups. A significant drop in blood pressure in the treatment group relative to the control group would suggest the drug’s effectiveness.

RCTs offer the major benefit of mitigating confounding variables. Random assignment ensures that any external factors influencing the outcome are equally distributed across both groups, making it easier to attribute changes in the outcome to the treatment.

#### **Propensity Score Matching**
This technique is used in observational studies to address confounding variables and estimate treatment effects. The propensity score reflects the likelihood of an individual receiving a treatment, based on their characteristics. Participants who received the treatment are then matched with those who did not, but have similar propensity scores.

For example, suppose a study investigates a new drug’s effect on cholesterol levels. The treatment group may have more males, older participants, and individuals with a family history of high cholesterol. These factors can influence cholesterol levels, so propensity score matching can be used to adjust for these differences. By matching treated and untreated participants based on their propensity scores, researchers can estimate the drug’s effect more accurately.

#### **The Language of Causation**
The language of causality encompasses the tools and methods used to understand how variables relate and influence each other. Key concepts include causal models, which mathematically represent the mechanisms of variable interaction, and the study of causal effects, focusing on how interventions or treatments affect outcomes.

#### **Causal Models**
Causal models represent the causal connections between variables, often depicted using directed acyclic graphs (DAGs). In a DAG, variables are nodes, and arrows show the direction of influence between them.

For instance, to explore the causal relationship between smoking and lung cancer, one could create a DAG where smoking is the cause, and lung cancer is the effect, clearly illustrating the presumed directional influence.

### **Understanding Causal Models**

Causal models are powerful tools that allow us to predict the outcomes of various interventions on key variables. By simulating how different interventions impact a causal model, we can better understand the potential effects of different actions and make well-informed decisions.

#### **Causal Effects Explained**
Causal effects refer to the alterations in a variable resulting from a particular intervention. In simple terms, the causal effect measures the difference between the actual outcomes after an intervention and the hypothetical outcomes had the intervention not been implemented.

There are several categories of causal effects:
- **Average Causal Effects (ACE)**: These describe the overall impact of an intervention on a population.
- **Conditional Causal Effects (CCE)**: These focus on how an intervention affects a specific subgroup within the population.
- **Counterfactual Effects**: These measure the difference in outcomes between the actual intervention and a counterfactual scenario where a different intervention was applied.

#### **Model Validation**
To ensure that a causal model accurately reflects the data, a process called model checking is used. This process involves comparing the predicted results from the causal model with real-world data to assess their alignment.

For instance, if a model suggests that smoking causes lung cancer, we can validate this by comparing lung cancer rates among smokers and non-smokers. If smokers have a significantly higher rate of lung cancer, the data supports the model’s prediction.

However, various confounding factors, such as age, gender, and genetic predisposition, must be considered when validating causal models, as these factors could influence the observed relationship between smoking and lung cancer.

#### **Exploring Treatment Effects**
Treatment effects refer to the impact of a specific intervention or treatment on an outcome. Essentially, treatment effects measure how much a treatment changes an outcome when compared to other factors.

There are several types of treatment effects:

- **Average Treatment Effect (ATE)**:
  ATE represents the average difference in outcomes between individuals receiving the treatment and those in a control group. It provides an overall estimate of how effective the treatment is across the population.

  Example: A study may assess a new medication for lowering blood pressure. Participants are randomly assigned to either receive the medication or a placebo. The ATE is the average difference in blood pressure between the two groups.

$$
E ( Y_a - Y_a^* )
$$


- **Treatment Effect Heterogeneity**:
  This refers to variations in treatment effects across different subgroups of the population. Understanding this variation can highlight which groups benefit most from the treatment.

  Example: In the same blood pressure study, treatment effect heterogeneity can be explored by examining whether the medication works better for men or women, or for individuals with different baseline blood pressure levels.

- **Intention-to-Treat Effect (ITT)**:
  ITT measures the impact of being assigned to a treatment group, irrespective of whether the participant follows through with the treatment. This is important when there’s non-compliance or dropout in a study.

  Example: In a weight loss program study, even if some participants do not fully adhere to the program, ITT would include them in the treatment group when assessing the program’s effectiveness.

- **Causal Mediation Effect**:
  The causal mediation effect evaluates how much of the treatment effect can be attributed to changes in a mediator variable, which is influenced by the treatment and, in turn, impacts the outcome.

  Example: In a smoking cessation program, the mediator might be the number of cigarettes smoked per day. The causal mediation effect assesses how much of the program's success can be attributed to the reduction in smoking.

- **Average Treatment Effect on the Treated (ATT)**:
  ATT measures the average effect of treatment for those who actually received the treatment. This is calculated by comparing the outcomes of the treated group with those who did not receive the treatment, but only for those who were treated.

  Example: In a job training program study, we estimate the ATT by comparing the earnings of individuals who completed the program with those who didn’t. For instance, if the program completers saw an average salary increase of USD 5000, while non-completers saw an increase of only USD 2000, the ATT would be USD 3000.

$$
E[Y_{1i} - Y_{0i} | D_i = 1]
$$


- **Conditional Average Treatment Effect (CATE)**:
  CATE measures the treatment effect for specific subgroups within the treatment group based on particular characteristics. It provides insight into how different individuals within the group respond to the treatment.

  Example: In the study of class size and academic performance, CATE can help measure the impact of class size on subgroups such as low-income students or those with learning disabilities. For instance, class size might have a larger effect on low-income students, with an ATE increase of 10 points, while students without such status might see a smaller increase.

$$
\tau(x) = E[Y(1) - Y(0) | X = x]
$$


By understanding these various treatment effects, researchers can better assess how interventions impact different segments of the population.

#### **Challenges in Establishing Causality**
Determining causality is a complex and challenging aspect of data science and statistics. Several obstacles must be addressed, including:

- **Confounding Variables**:
  Confounders are variables that are associated with both the cause and the effect, which makes it hard to determine whether the cause truly leads to the effect.

  For example, if we aim to establish that smoking causes lung cancer, we may observe a correlation between the two. However, other confounding factors like age, gender, and exposure to environmental toxins could also explain this relationship. Without adjusting for these factors, we might incorrectly infer causality between smoking and lung cancer.




## Confounding Variables and Regression Analysis

To address the problem of confounding variables, we can employ **regression analysis**. This technique allows us to account for the effects of other variables, thereby isolating the effect of the presumed cause on the outcome. By controlling for confounding variables, regression analysis helps us draw more accurate inferences about causal relationships in observational data, where randomization is not possible.

Regression models, such as multiple linear regression, allow for the inclusion of multiple predictors, helping to estimate the true effect of a variable on the outcome while holding other variables constant. This approach reduces the risk of biased estimates caused by confounding, enabling a clearer understanding of causal impacts.

## Reverse Causality

Reverse causality refers to a situation where the assumed direction of causality is incorrect, and the presumed cause is actually the effect. This can lead to misleading conclusions about the relationships between variables.

For example, consider the relationship between exercise and weight loss. The typical assumption is that **exercise causes weight loss**. However, in some cases, this causal relationship may be reversed. Weight loss itself could lead to increased exercise. For instance, if someone experiences significant weight loss due to illness or other factors, they may feel healthier and more energetic, which could motivate them to exercise more. In this case, weight loss is the cause, and exercise is the effect, despite the initial assumption that exercise was the cause.

Recognizing reverse causality is essential when interpreting observational data, as it helps prevent drawing incorrect conclusions about causal relationships. To address reverse causality, techniques such as **Granger causality tests** or **instrumental variables** (IV) can be employed to establish the direction of causality more reliably.

### Selection Bias

**Selection bias** occurs when the sample or population being studied is not representative of the entire population, leading to skewed or inaccurate conclusions. It arises when participants are selected in a non-random way, causing the sample to differ systematically from the target population.

For example, consider a study aiming to evaluate the effectiveness of a new medication for a specific disease. If the study only recruits patients from a single hospital, particularly a specialized center that treats the disease in advanced stages, the sample may not represent the broader population with the disease. In this case, the findings could be misleading, as the hospital's patients are likely to have different characteristics compared to the general population, such as a more severe form of the disease or a higher likelihood of receiving specialized treatment.

The result is that the effectiveness of the medication may appear higher than it actually is because the sample is biased toward those who are already receiving the best possible care. This bias can lead to incorrect conclusions about the medication’s general effectiveness, as it may not work as well in the larger population of patients who do not have access to specialized care or who have different disease profiles.

#### Key Takeaways:
- **Non-random selection** can distort study results, making them unrepresentative of the wider population.
- **Selection bias** is especially problematic when findings are generalized to a larger, more diverse population.
- In order to minimize selection bias, researchers should aim for **random sampling** or use **statistical methods** to account for potential biases.

#### Methods to Mitigate Selection Bias:
1. **Random Sampling:** Ensuring that participants are randomly selected from the entire population to represent diverse characteristics.
2. **Stratified Sampling:** Dividing the population into subgroups and randomly sampling from each subgroup to ensure all relevant segments are represented.
3. **Statistical Adjustment:** Using techniques like propensity score matching or weighting to correct for selection bias in observational studies.

# Sample Size:

**Sample size** refers to the number of individuals or units included in a study or experiment. In statistical analysis, sample size is crucial for determining the reliability and generalizability of the findings. A small sample size can result in low statistical power, which increases the risk of Type II errors—where a true effect is not detected. This issue can significantly affect the ability to establish causal relationships.

In contrast, larger sample sizes generally provide greater statistical power and precision, allowing for more reliable conclusions. However, larger sample sizes also demand more resources and incur higher costs. Determining the appropriate sample size for a study depends on several factors, including the research question, study design, desired confidence level, and available resources. Researchers must balance statistical power with practical constraints to determine an optimal sample size.

# PyGAM and LinGAM Overview

## PyGAM

**PyGAM** is a Python package designed for building and estimating **Generalized Additive Models (GAMs)**. GAMs extend linear models by allowing non-linear relationships between the dependent variable and independent variables. This flexibility makes PyGAM a powerful tool for fitting complex models and exploring non-linear patterns in data. Key features of PyGAM include:

- **Non-linear Modeling**: Unlike traditional linear models, GAMs model non-linear relationships through smooth functions.
- **Variety of Smoothing Methods**: Supports multiple types of smoothing splines, including thin plate regression splines and cubic regression splines.
- **Flexible Response Specification**: PyGAM allows for a range of distribution families and link functions, enabling flexible modeling of the response variable’s distribution.

With a high-level interface, PyGAM is user-friendly, making it easier for researchers and data scientists to fit and interpret GAMs.

## LinGAM

**LinGAM** (Linear non-Gaussian Acyclic Models) is a statistical method used for inferring **causal relationships** from multivariate time series data. It is particularly effective for modeling non-Gaussian and non-linear relationships. The key principles of LinGAM are:

- **Directed Acyclic Graph (DAG)**: LinGAM models the causal relationships between variables using a directed acyclic graph. It identifies causal directions and estimates the strength of the relationships.
- **Non-Gaussian Errors**: The algorithm assumes that the errors in the model are non-Gaussian and linearly independent, which makes it suitable for real-world, non-normal data.
- **Causal Inference**: LinGAM can detect both **instantaneous** and **lagged** causal effects between variables, which is particularly useful in time series analysis.
- **High-Dimensional Data Handling**: LinGAM works effectively with datasets containing many variables, allowing for the exploration of complex, high-dimensional systems.

The algorithm fits a linear model to the data while identifying a DAG that best represents the causal relationships. The **prior_knowledge** parameter can be used to incorporate existing causal knowledge into the model, further refining the causal graph and improving the accuracy of the results.

## Conclusion

In conclusion, understanding **causality** is crucial in many fields, including **public health**, **economics**, and **social sciences**. While **correlation** and **causation** can often be confused, **causal inference** methods help us distinguish between the two. 

Key concepts include:
- **Causal Diagrams** and **Model Checking**: These help identify potential confounding factors and design experiments to infer causality.
- **Treatment Effects**: Estimating treatment effects is central to causal inference, and the type of treatment effect can vary depending on the research question.
  
However, there are several challenges in inferring causality, such as:
- **Selection Bias**
- **Reverse Causality**
- **Unobserved Confounding**

Despite these challenges, applying causal inference methods to real-world problems—such as **air pollution**—can help identify **interventions** that improve **public health**. By incorporating these concepts into research and decision-making, we can make more informed and impactful decisions.
