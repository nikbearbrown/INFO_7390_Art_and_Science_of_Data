**Abstract**

Causality refers to the relationship between cause and effect, where one event (the cause) brings about another event (the effect). It's the concept of understanding how changes in one variable lead to changes in another variable.

In various fields such as science, philosophy, and statistics, causality is a fundamental concept for understanding and explaining phenomena. It helps us understand the mechanisms underlying observed patterns and enables us to make predictions and interventions.

In statistics and data science, causal inference is concerned with identifying and estimating causal relationships between variables from observational or experimental data. This involves distinguishing between correlation (association between variables) and causation (the effect of one variable on another).

Understanding causality is crucial for making informed decisions, developing effective interventions, and advancing scientific knowledge in many domains

**Causality vs Correlation**

Causality and correlation are two fundamental concepts in statistics and research that describe the relationships between variables. Causality refers to the notion that one event or variable directly influences another, implying a cause-and-effect relationship. It suggests that changes in one variable lead to changes in another variable. Establishing causality often requires rigorous experimental design or observational studies that account for confounding factors to infer a direct influence. On the other hand, correlation measures the extent to which two variables change together. It quantifies the strength and direction of the linear relationship between variables but does not imply causation. Correlation can arise due to various reasons, including direct causation, indirect causation, coincidence, or the influence of confounding variables. Therefore, while correlation provides valuable insights into associations between variables, causality requires more rigorous investigation to determine the direction and nature of the relationship between variables.

**Common Causal Inference Models**

Causal inference models are statistical and machine learning techniques used to infer causal relationships between variables from observational or experimental data. These models aim to determine whether one variable causes changes in another variable, rather than just identifying correlations. Some common causal inference models include:

Propensity Score Matching: This technique is commonly used to estimate the causal effect of a treatment or intervention on an outcome variable by matching treated and untreated individuals based on their propensity scores, which represent the probability of receiving the treatment given observed covariates.

Difference-in-Differences (DID): DID is a quasi-experimental method used to estimate causal effects by comparing changes in an outcome variable over time between a treatment group and a control group, accounting for pre-existing differences between the groups.

Instrumental Variables (IV): IV methods are employed when there is endogeneity or omitted variable bias in estimating causal effects. IV models use instrumental variables that are correlated with the treatment variable but not directly associated with the outcome variable to identify causal effects.

Regression Discontinuity Design (RDD): RDD is a quasi-experimental design used to estimate causal effects by exploiting the discontinuity in the assignment of a treatment based on a continuous threshold variable.

Structural Equation Modeling (SEM): SEM is a statistical technique used to estimate causal relationships between variables by specifying a structural model that represents the relationships among observed and latent variables.

Causal Tree Models: These are decision tree-based models that recursively partition the data based on the treatment variable and other covariates to estimate causal effects within each partition.

Do-Calculus and Graphical Models: These methods use causal graphical models, such as Bayesian networks or causal directed acyclic graphs (DAGs), along with formal rules of do-calculus to infer causal effects from observational data.

Each causal inference model has its strengths and limitations, and the choice of model depends on the specific research question, the nature of the data, and the assumptions underlying each method.

Causality is a broad field of study that can be applied to various domains and datasets. Some common types of datasets where causality is frequently studied include:

Causality analysis in economics aims to understand the causal relationships between these variables, such as the impact of monetary policy on economic growth or the effect of education on income.

Causality analysis in healthcare aims to identify causal relationships between interventions (e.g., treatments, medications) and health outcomes (e.g., disease progression, mortality), as well as the impact of risk factors on disease development.

Social Science Datasets: Social science datasets encompass a wide range of topics, including education, sociology, psychology, and political science. Causality analysis in social science often investigates the causal effects of social policies, interventions, or cultural factors on human behavior, attitudes, and societal outcomes.

Environmental Datasets: Environmental datasets may include variables related to climate, pollution levels, ecological indicators, and human activities. Causality analysis in environmental science aims to understand the causal relationships between environmental factors and their impacts on ecosystems, biodiversity, and human health.

Business Datasets: Business datasets cover aspects such as sales, marketing campaigns, customer behavior, and operational metrics. Causality analysis in business often focuses on identifying the causal effects of business strategies, market conditions, and consumer preferences on business performance and outcomes.

These are just a few examples, and causality analysis can be applied to virtually any dataset where there is an interest in understanding cause-and-effect relationships between variables. The choice of dataset depends on the specific research question or domain of interest.

This code snippet performs the following tasks:

It imports the OrdinalEncoder from the category_encoders library.
It initializes an OrdinalEncoder object called oe.
It fits the OrdinalEncoder to the DataFrame df1 and transforms it, storing the result in a new DataFrame called df_en. The OrdinalEncoder is used to encode categorical columns in the DataFrame into ordinal integers.
It creates a list of column names from the DataFrame df1 and stores it in the variable df_list.

```python
#code by "https://github.com/KJMAN678/casual_reasoning_lignum/blob/master/LIGNAM.ipynb

oe = ce.ordinal.OrdinalEncoder(
    cols=df1.select_dtypes(include="object")
    )

df_en = oe.fit_transform(df1)
#code by "https://github.com/KJMAN678/casual_reasoning_lignum/blob/master/LIGNAM.ipynb

df_list = df1.columns.to_list()
df_dict = {}

for i, column in zip(range(len(df_list)), df_list):
    df_dict[column] = i

print(len(df_list))
print(df_dict)
```

```python
#code by "https://github.com/KJMAN678/casual_reasoning_lignum/blob/master/LIGNAM.ipynb

prior_knowledge = make_prior_knowledge(
    18,
    exogenous_variables = [df_dict["Type Water Body"], df_dict["State Name"]],
    paths = [
             [df_dict["pH (Max)"], df_dict["pH (Min)"],[df_dict['Temperature\n?C (Min)'],df_dict['Temperature\n?C (Max)'],df_dict['Dissolved Oxygen (mg/L) (Min)'],df_dict['Dissolved Oxygen (mg/L) (Max)']]],
             ],

)
```

 this makse_prior_knowledge function from the lingam.utils module to define prior knowledge about the causal relationships between variables. Here's a breakdown of what it does:

It calls the make_prior_knowledge function with the following arguments:
The number of variables in the dataset (18 in this case).
exogenous_variables: A list containing the indices of exogenous variables in the dataset. These are variables that are not influenced by other variables in the dataset.
paths: A list of paths representing the known causal relationships between variables. Each path is defined as a list of variable indices, where the last variable in the list is the target variable, and the preceding variables are its causes.
In this specific example:

The exogenous_variables list contains the indices of the "Type Water Body" and "State Name" variables, indicating that these variables are considered exogenous and not influenced by other variables in the dataset.
The paths list contains one path, representing the causal relationship between "pH (Max)" and "pH (Min)" variables. The last variable in this path is "pH (Min)", and its causes are "pH (Max)", "Temperature\n?C (Min)", "Temperature\n?C (Max)", "Dissolved Oxygen (mg/L) (Min)", and "Dissolved Oxygen (mg/L) (Max)".
This prior knowledge is used as input to causal discovery algorithms to guide the search for causal relationships between variables.
