# Types of Missing Data (Analyzing the Missing Data Problem)
Before we talk about imputation methods, let's first understand the types of missing data. There are three types:

1. **Missing Completely At Random (MCAR)**: This occurs when the probability of missing data is the same for all observations, and there is no relationship between the missing data and any other variables in the dataset. This means that the missing data is entirely random, and there is no pattern to it.

2. **Missing At Random (MAR)**: This occurs when the probability of missing data depends on other observed variables in the dataset, but not on the missing variable itself. This means that there is a pattern to the missing data, but it can be explained by other variables in the dataset.

3. **Missing Not At Random (MNAR)**: This occurs when the probability of missing data depends on the missing variable itself, and it cannot be explained by any other variables in the dataset. This means that the missing data is not random, and there is no pattern to it that can be explained by the other variables in the dataset.

image.png

The type of missing data present in a dataset can have important implications for the choice of imputation method. MCAR and MAR are generally easier to handle than MNAR because MNAR requires making assumptions about the relationship between the missing values and the unobserved variable.

# Imputation Methods for Missing Data (Theoretical Foundations / What is the theory behind it?)
Imputation is the process of estimating missing values in a dataset. There are several imputation methods for handling missing data, including:

image.png

*   **Mean/Mode/Median Imputation**

In this method, missing values are replaced with the mean, mode, or median value of the available data for a particular variable. This method is simple and straightforward to implement, but it can lead to biased estimates and reduce the variance of the data. Mean/mode/median imputation is best suited for MCAR data, where missing value is completely random.

*   **Hot-Deck Imputation**

In this method, missing values are imputed using the observed values of similar cases. Hot-deck imputation randomly selects a non-missing value from another observation that is similar to the observation with the missing value. This method preserves the correlations between variables and is less biased than mean imputation. Hot-deck imputation is best suited for MAR data, where missing value is related to some observed variables.

*   **Multiple Imputation**

Multiple imputation is a more sophisticated method that generates several plausible values for each missing value and creates multiple imputed datasets. These imputed datasets are analyzed separately, and the results are combined using statistical techniques to generate a final estimate. This method is based on the idea that the imputed values should vary within a plausible range to reflect the uncertainty in the imputation process. Multiple imputation is best suited for MNAR data, where missing value is related to unobserved variables.

In the single imputation method, the imputed values are treated as true values, ignoring the fact that no imputation method can provide accurate values. The single imputation method does not reflect the uncertainty in the prediction of missing values. The multiple imputation method was proposed by Rubin (1987). In this method, each missing observation yields D imputed values, so we obtain D complete data sets. From each complete dataset, estimates of the parameter of interest, θ, are obtained by using standard techniques, assuming that no nonresponse exists. This process yields valid statistical inferences that appropriately reflect the uncertainty due to missing values.

Let $\hat{\theta_d}$ and $\hat{\phi_d}$ be an estimator of $\theta$ and the estimator of the variance of $\hat{\theta_d}$ based on the dth complete data set. The combined estimator of θ under multiple imputations is given by

$\hat{\theta_D}$ = $\frac{1}{D} \sum_{d=1}^{D}\hat{\theta_d}$

The proposed variance estimator of $\hat{\theta_D}$ is given by

$\hat{\phi_D}$ = $\bar{\phi_d}$ + ( 1 + $\frac{1}{D}$ ) $B_d$

where $\hat{\phi_D}$ = $\frac{1}{D} \sum_{d=1}^{D}\hat{\phi_d}$  and $B_d$ = $\frac{1}{D-1}$ $\sum_{d=1}^{D}$ $(\hat{\theta_D} - \hat{\theta_d})^2$.

The first component $\bar{\phi_d}$ of Eq.(15.4.27) is the average within-imputation variance and the second component is the product of between-variance component $B_d$ and correction factor ( 1 + $\frac{1}{D}$ )

For large sample size the statistic

$\frac{\hat{\theta - \theta}}{\sqrt{\hat{\phi_D}}}$

follows t distribution with ν degrees of freedom where

ν = (D - 1)(1 + $\frac{1 \bar{\phi_D}^2}{D + 1 B_D}$ )

For smaller sample sizes, Barnard and Rubin (1999) proposed an improved variance formula.

The multiple imputation method is advantageous over the single imputation method because it uses several complete data sets and provides intra- and inter-imputation variability. The multiple imputation method facilitates simple formulas for variance estimation and interval estimation of the parameter of interest. Its only drawback is that analyzing the data is more tedious than the single imputation method. However, modern computational techniques such as computing do not pose any problem.

*   **Regression Imputation**

Regression imputation imputes missing values using a regression model that predicts the missing values from other variables in the dataset. This method is useful when the missing values are related to other variables in the dataset and can improve the accuracy of the imputation process. Regression imputation is best suited for MAR data, where missing value is related to some observed variables.

image.png

*   **K-Nearest Neighbors (KNN) Imputation**

KNN imputation imputes missing values using the values of the K nearest neighbors to the observation with the missing value. This method is useful when the missing values are related to the values of neighboring observations and can be used for both continuous and categorical data. KNN imputation is best suited for MAR data, where missing value is related to some observed variables.

*   **Expectation-Maximization (EM) Algorithm**

The EM algorithm is an iterative method that estimates the parameters of a statistical model with incomplete data. In the case of missing data, the EM algorithm imputes the missing values by estimating the parameters of the model that generate the observed data. This method is best suited for MNAR data, where missing value is related to unobserved variables.

# Choosing the Right Imputation Method
Choosing the right imputation method depends on the type and amount of missing data, the data distribution, and the research question. Here are some general guidelines:

* For MCAR data, mean/mode/median imputation or hot-deck imputation is suitable.
* For MAR data, regression imputation or KNN imputation is suitable.
* For MNAR data, multiple imputation or the EM algorithm is suitable.

It's important to note that imputation methods can introduce bias and affect the accuracy of the analysis. Therefore, it's recommended to compare the results obtained from different imputation methods and perform sensitivity analyses to assess the robustness of the results.

## Mean imputation
Mean imputation is a commonly used technique for handling missing data in statistical analyses, where missing values in a variable are replaced by the mean of the available cases. The approach is straightforward to apply and does not reduce the sample size. However, mean imputation has several limitations. When missing values are imputed using this method, the variability in the data is reduced, which results in underestimation of standard deviations and variance estimates. Moreover, by restricting the variability, mean imputation decreases the magnitude of the covariances and correlation, leading to biased estimates that are often observed irrespective of the underlying missing data mechanism. ## KNN imputation
K-Nearest Neighbors (KNN) imputation is a popular method for handling missing data in datasets. This technique is based on the assumption that similar data points are likely to have similar values. KNN imputation estimates missing values by identifying the k-nearest neighbors to the data point with missing features and using these neighbors to impute the missing values. The similarity between data points is typically measured using a distance metric such as Euclidean, Manhattan, or Minkowski distance.

image.png

Here's a step-by-step description of the KNN imputation process and the math behind it:

1. Determine the number of neighbors (k): Choose the value of k, which represents the number of nearest neighbors to consider for imputation. This parameter affects the performance of the imputation method and must be chosen carefully, often using cross-validation techniques.

2. Choose a distance metric: Select a distance metric to measure the similarity between data points. Common distance metrics include Euclidean, Manhattan, and Minkowski distance. Euclidean distance is the most widely used metric and is calculated as follows:

Euclidean Distance = √((x1 - x2)^2 + (y1 - y2)^2 + ... + (zn - zm)^2)

Here, (x1, y1, ..., zn) and (x2, y2, ..., zm) represent the feature values of two data points.

3. Calculate the distances: Compute the pairwise distances between the data point with missing values and all other data points in the dataset using the chosen distance metric.

4. Find the k-nearest neighbors: Sort the calculated distances in ascending order and select the k data points with the smallest distances. These are the k-nearest neighbors.

5. Impute the missing values: Depending on the nature of the data, different strategies can be used to impute the missing values:

    a. For continuous variables, use the mean or median of the k-nearest neighbors' values:

    Imputed Value = mean(y1, y2, ..., yk) or median(y1, y2, ..., yk)

    b. For categorical variables, use the mode (most frequent category) of the k-nearest neighbors' values:

Imputed Value = mode(y1, y2, ..., yk)

6. Iterate over all missing values: Perform steps 3-5 for each data point with missing values in the dataset.

KNN imputation has some advantages, such as being non-parametric and robust to noisy data. However, it can be computationally expensive for large datasets and may not perform well when the assumption of similar data points having similar values does not hold.

## MICE imputation
Multiple Imputation by Chained Equations (MICE), also known as Fully Conditional Specification (FCS), is a flexible and robust technique for handling missing data in datasets. MICE is an iterative method that generates multiple imputations for each missing value, creating a set of complete datasets. These complete datasets can then be analyzed separately, and the results are combined using Rubin's rules, which account for the uncertainty introduced by imputing the missing values.

image.png

Here's a step-by-step description of the MICE imputation process and the math behind it:

1. Initial Imputation: For each missing value, replace it with a simple imputation, such as the mean, median, or mode of the observed values for that variable. This initial imputation generates a complete dataset with no missing values.

2. Iterative Imputation: Perform the following steps for each variable containing missing values:

    a. Regression Model: Select a regression model suitable for the variable with missing values. If the variable is continuous, choose a linear regression model; if it's categorical, use a logistic regression model. Fit the regression model to the observed data, using the variable with missing values as the dependent variable and other variables as independent variables.

    b. Predict Missing Values: Using the fitted regression model, predict the missing values of the variable based on the values of the other variables in the dataset.

    c. Replace Missing Values: Replace the missing values with the predicted values obtained in the previous step.

    d. Iterate: Repeat steps a-c for each variable with missing values. This process is called one "cycle" of the MICE algorithm.

3. Convergence and Multiple Imputations: After a predefined number of cycles, check for convergence. If convergence is reached, store the imputed dataset as one complete dataset. If not, repeat step 2 until convergence is achieved. Repeat this process to generate multiple imputed datasets (typically 5-10).

4. Analysis and Pooling: Analyze each of the multiple imputed datasets separately using the statistical method of interest (e.g., regression, ANOVA, etc.). Then, combine the results using Rubin's rules, which account for the within-imputation and between-imputation variability:

    a. Point Estimates: Compute the average of the point estimates (e.g., regression coefficients) from each imputed dataset.

    b. Standard Errors: Calculate the average of the standard errors from each imputed dataset and adjust it using the between-imputation variance.

    c. Degrees of Freedom: Estimate the degrees of freedom for the t-distribution using a method such as Barnard-Rubin or the method of moments.

MICE is an attractive imputation method because it can handle various patterns of missing data and different types of variables (continuous, binary, ordinal, etc.). However, it can be computationally expensive for large datasets and relies on the assumption that the data are missing at random (MAR).

Since this question has a requirement for the percentage of missing data, we create a Function to calculate percentage of missing data in dataset and a Function to create missing values.

```python
# Function to calculate percentage of missing data in dataset
def get_percent_missing(dataframe):

    percent_missing = dataframe.isnull().sum() * 100 / len(dataframe)
    missing_value_df = pd.DataFrame(
        {"column_name": dataframe.columns, "percent_missing": percent_missing}
    )
    return missing_value_df


# Function to create missing values
def create_missing(dataframe, percent, col):
    dataframe.loc[dataframe.sample(frac=percent).index, col] = np.nan
```

## Step 9: Impute missing value using Mean, KNN, MICE
Here we use a total of three methods to fill the blank values:
1. Mean
2. KNN
3. MICE

Here we create a function that uses mean imputation to impute data and a function to calculates the degree of data recovery after performing the data imputation.

```python
# Function to impute data using mean imputation
def mean_impute(new_raw_data):
    mm_impute_train_df = new_raw_data.copy()  # Make a copy of dataframe for imputation
    mm_impute_train_df["Age_1_percent"] = mm_impute_train_df[
        "Age_1_percent"
    ].fillna(mm_impute_train_df["Age"].mean())
    mm_impute_train_df["Age_5_percent"] = mm_impute_train_df[
        "Age_5_percent"
    ].fillna(mm_impute_train_df["Age"].mean())
    mm_impute_train_df["Age_10_percent"] = mm_impute_train_df[
        "Age_10_percent"
    ].fillna(mm_impute_train_df["Age"].mean())
    return mm_impute_train_df


# Function to calculate how well the data has been recovered after performing data imputation
def percentage_change(l1, l2):
    percent_change = abs(l2 - l1) / (l1 + 0.000000001)
    avg_change = (percent_change.sum() / percent_change.count()) * 100
    return avg_change
```

```python
print(get_percent_missing(train_df_selected))
```

### 3. MICE Imputation
Finally we use MICE Imputation to do the data imputation. And we create mice_impute function to impute data using MICE method.

```python
# Function to impute data using MICE method
def mice_impute(train_df):
    mice_imputed_train_df = pd.DataFrame(
        MICE().fit_transform(train_df), columns=list(train_df.columns)
    )
    return mice_imputed_train_df
```
