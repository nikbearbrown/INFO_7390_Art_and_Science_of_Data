The dataset contains information from the American Community Survey (ACS) conducted in 2015. The dataset includes data from every census tract and county in the United States, including Washington D.C. and Puerto Rico. Census tracts are subdivisions of a county or equivalent political subdivision that are defined by the Census Bureau. They are designed to be relatively homogeneous in terms of population characteristics and are typically smaller in size than counties. A typical census tract has around 5000 residents. In contrast, counties are political subdivisions that have been established for centuries and can vary in size and population characteristics.

The census tract data file has an ID column named "GEOID", while the county data file has an ID column named "GEO.id2". This allows users to conduct analyses at both the census tract and county levels, depending on their research question.

Overall, this dataset provides a rich source of information on social, economic, and demographic characteristics at both the census tract and county levels for the entire United States, which can be used to inform various research questions and analyses.

# **Analyzing the Problem**

### **Constraints:**

**Data Quality and Availability:** The use of the KNN Impute Method suggests that the dataset might have missing values. The quality and quantity of the available data can significantly impact the accuracy of the prediction models.

**Dimensionality:** The application of Principal Component Analysis (PCA) indicates that the dataset likely has a high number of variables. PCA is typically used to reduce the dimensionality of the dataset while retaining most of the information. This reduction is crucial to improve model performance and interpretability but must be done without losing significant information.

**Noise and Outliers:** The use of DBSCAN, a clustering method known for its effectiveness in dealing with noise and outliers, implies that the data may contain such anomalies. These can distort the predictive model's performance if not appropriately handled.

### **Logic:**

**Imputation of Missing Data:** The KNN Impute Method is used to estimate and fill in missing values in the dataset. KNN works by finding the nearest neighbors of a data point with missing values and imputes them based on the values of these neighbors. This method assumes that similar data points are close to each other in the feature space.

**Dimensionality Reduction:** PCA is employed to reduce the number of variables. PCA transforms the original variables into a new set of variables (principal components), which are orthogonal and explain the maximum variance in the dataset. This step simplifies the model without losing critical information.

**Clustering and Outlier Detection:** DBSCAN is used for clustering and identifying outliers. DBSCAN groups data points that are closely packed together while marking points in low-density regions as outliers. This method is beneficial in understanding the inherent groupings in the data.

**Predictive Modeling:** The notebook then moves to predictive modeling using Linear Regression and Random Forest Regression. Linear Regression assumes a linear relationship between the dependent and independent variables, while Random Forest is a more complex, ensemble-based method that can capture non-linear relationships and interactions between variables.

### **Algorithmic Concepts:**

**K-Nearest Neighbors (KNN):** A simple, non-parametric algorithm used for imputation and classification, based on the principle that similar data points are near each other.

**Principal Component Analysis (PCA):** A statistical technique for dimensionality reduction, focusing on capturing the maximum variance in the dataset with fewer variables.

**DBSCAN:** A density-based clustering algorithm that differentiates regions of high density from regions of low density, useful in outlier detection and understanding the structure of complex datasets.

**Linear Regression:** A foundational statistical method used for predictive modeling, assuming a linear relationship between variables.

**Random Forest:** An ensemble learning method that operates by constructing multiple decision trees during training, providing improved accuracy and control over overfitting.

# **Principal Component Analysis (PCA)**

Principal Component Analysis (PCA) is a statistical method used to simplify complex datasets. It achieves this by transforming numerous variables into a smaller set of uncorrelated "principal components," which are ranked based on their ability to explain the variance in the original data. The first principal component accounts for the most variance, followed by the subsequent ones. This technique is widely applied in various fields like finance, engineering, and biology for purposes such as data compression, image analysis, and feature identification, making complex data more manageable and insightful.

![1665494667582](https://github.com/UtkarshaShirke/DataScience/assets/114371417/00fdf8a5-c6dd-4948-850f-5aaec4afc300)


```python
#Create instance of Principal Component Analysis in order to reduce dimensionality while maintaining variance
#n_components=3 will be chosen in order to visualize the data better
pca = PCA(n_components=3)

#fit to dataset
pca.fit(X_transformed)

#create dimentionality reduced dataset
X_reduced = pca.transform(X_transformed)

print('Dimensionality Reduced Data:')
print(X_reduced[0])
```

Using PCA, the dimensionality is reduced as seen in the above observation

# **Performing DBSCAN**

DBSCAN, short for Density-Based Spatial Clustering of Applications with Noise, is a machine learning algorithm used for clustering data based on spatial proximity and density. It groups data points into clusters based on their closeness and the density of their surrounding points, distinguishing core points (with many neighbors) from border points and noise points (isolated data points). Unlike some clustering methods, DBSCAN doesn't require pre-setting the number of clusters, making it adaptable to complex and irregular data structures. It's effective in identifying outliers as noise and is used in various fields like customer segmentation, image analysis, and anomaly detection. However, its performance may vary with datasets of varying densities or high dimensionality.

![image-58](https://github.com/UtkarshaShirke/DataScience/assets/114371417/4b4775f1-3e55-409a-9a83-14ae78fcea9f)


```python
#Create an instance of DBSCAN to create non spherical clusters based on data density
db = DBSCAN(eps=0.726, min_samples=26)

#fit to the dimentionality reduced dataset
db.fit(X_reduced)

#identify the clusters
clusters = db.labels_

#display metrics/sample
n_clusters_ = len(set(clusters)) - (1 if -1 in clusters else 0)
n_noise_ = list(clusters).count(-1)

print('Cluster Predictions')
print('-------------------------------')
print("Number of clusters: %d" % n_clusters_)
print("Number of noise points: %d" % n_noise_)
print('Number of points per cluster:')
for i in range(n_clusters_):
    print('Cluster', i, ':', len(clusters[clusters==i]))
```

```python
#plot the 3d dataset
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X_reduced.T[1],X_reduced.T[2],X_reduced.T[0],
           c=clusters)
ax.set_title("Household Clusters in 3 Dimensions")
ax.legend(['Outliers'])
plt.show()
```

```python
# Writing function to perform Linear Regression

def linearRegressionTest(input_df, target_df, feature_selection_method):

    params_count = len(input_df.columns)

    X_train,X_test,Y_train,Y_test = train_test_split(input_df,target_df,test_size = 0.2)

    model = LinearRegression()

    # Train the model, using training data set
    model.fit(X_train, Y_train)

    # Use trained model to predict on test dataset
    y_pred = model.predict(X_test)

    print("\n--------------------------------------------------\n")
    print("Score for model having \'",feature_selection_method, "\' using \"", params_count,"\" features\n" )

    model_score = model.score(input_df, target_df)
    mse = mean_squared_error(Y_test, y_pred)
    r2 = r2_score(Y_test, y_pred)
    rmse = math.sqrt(mse)
    mae = mean_absolute_error(Y_test, y_pred)
    evs = explained_variance_score(Y_test, y_pred)

    print("Model score is          = %.4f" % model_score)
    print("R2 score                = %.4f" % r2)
    print("Explained Variance score= %.4f" % evs)
    print("Mean absolute error     = %.4f" % mae)
    print("Mean squared error      = %.4f" % mse)
    print("Root Mean squared error = %.4f" % rmse)
    print("\n--------------------------------------------------\n")

    return model_score, r2, rmse, mae
```

# **Random Forest Regression Model**

```python

#Build the random forest regression model with input datasets based on the number of trees
def randomForestRegressionModel(X, y,num_of_trees, feature_selection_method):
    params_count = len(X.columns)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    #Build the random forest regressor model
    RFModel = RandomForestRegressor(n_estimators = num_of_trees, random_state = 42, verbose = 1)
    #Train the random forest regression model to dataset
    RFModel.fit(X_train,y_train)
    #Predict on test features
    y_pred = RFModel.predict(X_test)

    print("\n--------------------------------------------------\n")
    print("Score for model having \'",feature_selection_method, "\' using \"", params_count,
          "\" features with number of trees - ", num_of_trees)

    # evaluate model metrics
    model_score = RFModel.score(X, y)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    rmse = math.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    evs = explained_variance_score(y_test, y_pred)

    print("Model score is          = %.4f" % model_score)
    print("R2 score                = %.4f" % r2)
    print("Explained Variance score= %.4f" % evs)
    print("Mean absolute error     = %.4f" % mae)
    print("Mean squared error      = %.4f" % mse)
    print("Root Mean squared error = %.4f" % rmse)
    print("\n--------------------------------------------------\n")

    return model_score, r2, rmse, mae
```
