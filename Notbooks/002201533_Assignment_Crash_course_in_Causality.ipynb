# **Correlation vs. Causation**

One of the most important concepts in causality is understanding the difference between correlation and causation. These terms are often confused, but they represent fundamentally different ideas.

### **What is Correlation?**
Correlation refers to a statistical relationship between two variables. When two variables are correlated, their values tend to move together. For example:
- As outdoor temperature increases, ice cream sales typically rise.
- As education level increases, income tends to increase as well.

However, correlation does not tell us *why* the two variables are related. It only shows that a pattern exists.

### **What is Causation?**
Causation implies that one variable directly affects another. If we say that X causes Y, it means:
- A change in X leads to a change in Y.
- The relationship is not coincidental but based on a direct cause-and-effect mechanism.

For example:
- Exposure to sunlight causes an increase in vitamin D production.
- Smoking causes an increased risk of lung cancer.

---

### **Why Correlation Does Not Imply Causation**

Just because two variables are correlated does not mean one causes the other. There are several reasons why correlation alone cannot establish causation:

1. **Reverse Causation**:  
   The direction of the relationship might be misunderstood. For instance:
   - Imagine a study finds that people who carry lighters are more likely to get lung cancer. This might suggest that carrying a lighter causes cancer, but in reality, the true cause is smoking (which is linked to carrying lighters).

2. **Confounding Variables (Lurking Variables)**:  
   A third variable might be influencing both observed variables, creating a false impression of causation.  
   - Example: Ice cream sales and drowning incidents tend to rise at the same time. The confounding variable here is warm weather, which increases both ice cream consumption and swimming activity.

3. **Coincidence**:  
   Some correlations are purely coincidental and have no meaningful relationship.  
   - Example: There is a well-known, spurious correlation between the number of films Nicolas Cage appears in and the number of people who drown in swimming pools. This is a random coincidence and not a causal relationship.

4. **Bidirectional Relationships**:  
   Sometimes, two variables may influence each other, making it hard to establish which is the cause and which is the effect.  
   - Example: Exercise and mood are positively correlated. Exercising can improve mood, but being in a good mood might also encourage exercise.

---

### **Why is the Distinction Important?**

Misinterpreting correlation as causation can lead to incorrect conclusions and poor decision-making. For example:
- A company might notice that ice cream sales and sunburns are correlated and mistakenly invest in promoting ice cream to reduce sunburns. The real cause, however, is sunny weather, which independently drives both ice cream sales and sunburns.

Understanding this distinction is critical for drawing accurate inferences and making informed decisions.

---

### **How to Establish Causation**

While correlation can hint at a potential relationship, causation must be established through additional methods:

1. **Controlled Experiments**:  
   - Experiments involve actively manipulating one variable while holding others constant to observe the effect.  
   - Example: To test whether a new drug reduces blood pressure, researchers administer the drug to one group and a placebo to another while keeping other conditions identical.

2. **Causal Diagrams (Graphical Models)**:  
   - Tools like Directed Acyclic Graphs (DAGs) visually represent causal relationships between variables. These models help identify potential confounders and suggest paths to analyze.

3. **Statistical Techniques**:  
   - Methods like instrumental variables, difference-in-differences, or propensity score matching are used to infer causation from observational data when experiments are not feasible.

4. **Temporal Information**:  
   - Causation often requires that the cause precedes the effect in time. By analyzing the sequence of events, we can better determine the likely direction of influence.

5. **Domain Knowledge**:  
   - Understanding the context and mechanisms of the system you're analyzing is essential. For instance, knowledge of biology helps establish that smoking causes lung cancer, not the other way around.

---

### **Key Takeaways**

- Correlation is a statistical relationship, while causation is a direct cause-and-effect link.
- Not all correlations indicate causation; they might result from confounding variables, reverse causation, or coincidence.
- Establishing causation requires careful analysis using experiments, causal diagrams, and statistical methods.

By understanding and applying these principles, you can move beyond surface-level patterns in data to uncover the true drivers of outcomes, enabling better decision-making and insights.

# **Validating Causation**

Establishing causation is a critical process in understanding the direct cause-and-effect relationships between variables. Unlike correlation, which only measures how variables move together, causation provides the evidence that one variable's change directly drives a change in another. Demonstrating causation requires rigorous analysis, careful study design, and often empirical evidence from controlled experiments or observational studies. For example, understanding the causal link between smoking and lung cancer cannot rely solely on observing a correlation. Instead, it demands robust studies and biological evidence illustrating how smoking leads to cancer development.

---

## **1. Understanding Causality**

Causality lies at the core of scientific exploration, aiming to answer questions like:
- What factors lead to a particular outcome?
- How does manipulating one variable influence another?

Understanding causality allows us to go beyond statistical relationships to uncover the mechanisms driving observed behaviors or patterns. It focuses on identifying direct connections where a cause (X) produces an effect (Y).

### **Causal Inference:**
Causal inference is the process of using data to determine whether one variable directly influences another. It becomes especially important when experiments are not feasible or ethical. For example, in medical research, it might not be possible to test a treatment by withholding it from a control group, so researchers must rely on observational data and advanced inference techniques.

### **Example Scenario:**
Imagine a new drug is introduced to treat a disease. A natural question arises: does the drug cause an improvement in patient outcomes? However, in an observational study, the effects of the drug may be confounded by other factors, such as patients' age, lifestyle, or pre-existing conditions. This complexity makes isolating the drug's effect challenging and highlights the need for advanced causal inference methods.

### **Challenges in Establishing Causality:**
1. **Confounding Variables**: Factors that influence both the cause and the effect can distort the true relationship.
2. **Reverse Causation**: It may be unclear whether X causes Y or Y causes X.
3. **Unmeasured Variables**: Hidden variables might obscure the true causal relationship.

Understanding and addressing these challenges is essential to accurately determine causality.

---

## **2. Graphical Models for Causality**

Graphical models provide a visual and intuitive way to represent and analyze causal relationships. These models help map out the relationships between variables, highlighting potential confounders, mediators, and direct causal paths. Among the most common models are **Directed Acyclic Graphs (DAGs)**.

### **Directed Acyclic Graphs (DAGs):**
- **Nodes**: Represent variables (e.g., age, income, health status).
- **Directed Edges (Arrows)**: Indicate causal relationships (e.g., X → Y implies X causes Y).
- **Acyclic**: There are no feedback loops (i.e., no variable can cause itself).

DAGs are powerful tools for identifying the structure of a causal system. They help visualize:
1. **Direct Causes**: Variables with direct arrows pointing to the outcome.
2. **Confounders**: Variables that affect both the cause and the effect.
3. **Mediators**: Variables that lie on the causal pathway between the cause and effect.

### **Bayesian Networks:**
Bayesian Networks extend DAGs by incorporating probabilistic dependencies between variables. They allow for probabilistic reasoning, enabling us to compute the likelihood of an effect given certain causes.

### **Example in Practice:**
Consider the graphical representation of the relationship between sunny weather, ice cream sales, and sunburns:
- **Sunny weather → Ice cream sales** (sunny weather causes more people to buy ice cream).
- **Sunny weather → Sunburns** (sunny weather increases exposure to UV rays).
Here, sunny weather is a confounder that explains the correlation between ice cream sales and sunburns. This is illustrated in the image.

---

## **3. Causal Inference Methods**

To identify causal relationships from data, researchers use a variety of techniques. Two foundational approaches are the **Back-Door Criterion** and the **Front-Door Criterion**.

### **Back-Door Criterion:**
The back-door criterion focuses on blocking "back-door paths" between the cause and effect. Back-door paths are spurious associations introduced by confounders. By conditioning (adjusting) for these confounders, we can isolate the true causal effect.

#### **Steps:**
1. Identify all back-door paths from the cause to the effect in the DAG.
2. Select variables that, when conditioned on, block these paths.
3. Use statistical techniques (e.g., regression or stratification) to adjust for these variables.

#### **Example:**
In the ice cream and sunburn scenario, conditioning on sunny weather eliminates the spurious correlation between ice cream sales and sunburns.

### **Front-Door Criterion:**
The front-door criterion addresses cases where there are unmeasured confounders between the cause and effect. It involves identifying a mediator variable that lies on the causal pathway and using it to estimate the causal effect.

#### **Steps:**
1. Identify a mediator variable that is affected by the cause and also affects the effect.
2. Establish that there is no confounding between the cause and the mediator.
3. Use the mediator to estimate the effect of the cause on the outcome.

#### **Example:**
If the relationship between exercise (cause) and weight loss (effect) is confounded by genetics, tracking calorie burn as a mediator can help estimate the causal effect of exercise.

# **Causal Inference: Understanding Cause-and-Effect Relationships**

Causal inference is a discipline within mathematics and statistics dedicated to uncovering cause-and-effect relationships using observed data. It allows us to investigate how changes in one variable or event influence another, offering insights into the mechanisms driving observed patterns and behaviors. This field employs a variety of theoretical frameworks, statistical methods, and tools to analyze and interpret causal relationships.

---

### **1. Graphical Models**

**Bayesian Networks**:  
Bayesian networks are probabilistic graphical models that use directed acyclic graphs (DAGs) to represent relationships between variables. Each node represents a variable, while the edges illustrate causal dependencies. By incorporating Bayesian inference, these models capture both causal relationships and uncertainties within a system, enabling us to better understand how variables influence each other.

---

### **2. Structural Equation Modeling (SEM)**

Structural Equation Modeling (SEM) is a comprehensive statistical method for analyzing relationships between observed variables and unobserved (latent) variables. SEM provides a framework to:
- Model complex systems.
- Understand direct and indirect effects.
- Estimate causal effects between variables.

SEM is widely used in fields like psychology, economics, and social sciences, where variables often interact in intricate ways.

---

### **3. Counterfactuals and Potential Outcomes**

**Potential Outcome Framework**:  
The potential outcome framework is a foundational concept in causal inference. It compares what happens under different conditions or treatments to estimate causal effects. By analyzing the difference between observed and potential outcomes, this approach helps quantify the causal impact of interventions.

**Counterfactuals**:  
Counterfactuals are hypothetical scenarios that imagine alternative outcomes had different decisions or actions been taken. For example, “What if a patient had received a different treatment?” Counterfactuals are a powerful tool for reasoning about causality, as they provide a way to compare the actual outcome to a possible alternative.

---

### **4. Causal Inference Methods**

Causal inference methods help address challenges like confounding variables and selection bias, making it possible to extract meaningful causal insights from observational data.

- **Instrumental Variables (IVs)**:  
   IVs are used to estimate causal effects when unobserved variables might confound the relationship between the cause and effect. This technique is particularly common in econometrics and epidemiology.

- **Propensity Score Matching**:  
   This method reduces bias in observational studies by matching treated and untreated subjects with similar characteristics. It helps ensure that comparisons between groups are fair and not influenced by confounding factors.

- **Structural Causal Models (SCMs)**:  
   SCMs formalize the causal relationships between variables, allowing us to represent, simulate, and estimate the effects of interventions. They provide a framework for understanding the mechanisms underlying a system.

---

### **5. Information Theory**

**Granger Causality**:  
Granger causality is a statistical technique used in time-series analysis to determine whether past values of one variable can predict future values of another. While it does not establish true causation, it provides a practical way to infer directional relationships between variables in temporal datasets. This method is particularly useful in economics and neuroscience.

---

### **6. Probability and Statistics**

Probability and statistics form the foundation of causal inference, providing the tools needed to quantify uncertainty, model relationships, and test hypotheses.

#### a. **Do-Calculus**:  
Do-calculus is a mathematical framework used to estimate causal effects in complex systems. #### c. **Statistical Methods**:  
Statistical techniques, such as regression analysis, hypothesis testing, and machine learning models, are essential for identifying correlations and inferring causal relationships. By combining statistical tools with domain knowledge, researchers can differentiate between spurious associations and meaningful causal effects.

# **Understanding Propensity Score Matching**

## **Introduction**
Propensity Score Matching (PSM) is a statistical technique used in research to estimate the causal effect of a treatment or intervention when random assignment is not possible. It aims to create a fair comparison by pairing individuals who received the treatment with similar individuals who did not, based on their likelihood of receiving the treatment. Think of it as matching participants based on their "propensity" or probability of receiving the treatment, making the groups as comparable as possible.

---

## **Why Use Propensity Score Matching?**

In observational studies, individuals often self-select into treatment or control groups. This self-selection introduces bias because the decision to undergo treatment might depend on factors such as age, health status, or socioeconomic background. Without accounting for these factors, it’s challenging to determine whether observed differences in outcomes are due to the treatment or other confounding variables.

Propensity Score Matching helps address this issue by:
1. **Reducing Selection Bias**: Ensures that treated and untreated groups are similar in terms of observed characteristics.
2. **Facilitating Causal Inference**: Provides a way to estimate treatment effects when randomization is not feasible.

---

## **How Does Propensity Score Matching Work?**

### **Step 1: Estimate Propensity Scores**
- The propensity score is the probability of an individual receiving the treatment based on their observed characteristics (e.g., age, gender, income, health status).
- This score is typically calculated using a logistic regression or other machine learning models.





### **Step 2: Match Treated and Untreated Individuals**
- Individuals who received the treatment are matched with those who did not based on similar propensity scores.
- Matching methods include:
  - **Nearest Neighbor Matching**: Pair each treated individual with the closest untreated individual in terms of propensity score.
  - **Caliper Matching**: Matches are made only if the difference in propensity scores is within a predefined threshold.

### **Step 3: Assess Balance**
- After matching, evaluate whether the treated and untreated groups are similar in terms of baseline characteristics. Tools like standardized mean differences or visualizations (e.g., histograms of propensity scores) can help assess balance.

---

## **Estimating the Treatment Effect**

Once matched pairs are established, the treatment effect can be estimated by comparing the outcomes between the treated and untreated groups. This can be done by calculating the **average treatment effect on the treated (ATT)**, which measures the average impact of the treatment for those who received it.



---

## **Pros and Cons**

### **Advantages**
- **Reduces Bias**: Balances observed characteristics between treated and untreated groups, improving the validity of causal estimates.
- **Useful for Non-Randomized Studies**: Enables causal inference when random assignment of treatments is not possible.
- **Straightforward Interpretation**: Provides a clear framework for estimating treatment effects.

### **Limitations**
- **Dependence on Observed Variables**: Matching is only as good as the quality and completeness of the observed variables. Unmeasured confounders can still bias results.
- **Loss of Data**: Some unmatched individuals may be excluded from analysis, reducing the sample size.
- **Imperfect Matching**: Matching quality depends on the method used, and balance assessment is critical to ensure validity.

---

## **Key Takeaways**

Propensity Score Matching is a valuable tool for causal inference in observational studies, allowing researchers to create balanced comparisons between treated and untreated groups. By accounting for observed differences, it helps isolate the true effect of a treatment or intervention. However, like any statistical method, its reliability depends on the quality of data, the comprehensiveness of observed variables, and careful implementation.

By matching individuals with similar characteristics, PSM serves as a practical alternative to randomized controlled trials, enabling researchers to gain meaningful insights into cause-and-effect relationships in complex, real-world settings.

# **Exploratory Data Analysis: Investigating Survival on the Titanic**

In this example, we explore the famous Titanic dataset from Kaggle to examine whether having a cabin influenced passengers' chances of survival during the tragedy. By applying exploratory data analysis (EDA) and statistical techniques such as Propensity Score Matching (PSM), we aim to uncover any potential connections between cabin allocation and survival rates.

---

## **Understanding the Dataset**

The Titanic dataset contains information about the passengers, such as their age, fare paid, and whether they had a cabin. These features provide valuable insights into factors that may have influenced survival. Our goal is to investigate if having a cabin (an indicator of class or privilege) played a role in determining survival outcomes.

### **Key Variables**
1. **Features (X):**  
   These are the observable characteristics of passengers that might influence their chances of survival, including:
   - Age
   - Fare paid
   - Sex
   - Passenger class (Pclass)
   - Embarkation port

2. **Treatment (T):**  
   This is a binary variable indicating whether a passenger had a cabin or not:
   - **T=1:** Passenger had a cabin (treated group).  
   - **T=0:** Passenger did not have a cabin (untreated group).  

3. **Outcome (Y):**  
   This is the result of interest:
   - **Y=1:** Passenger survived.  
   - **Y=0:** Passenger did not survive.  

---

## **Using Propensity Score Matching (PSM)**

Propensity Score Matching is used to ensure a fair comparison between passengers with cabins and those without. The goal is to adjust for observable differences between the two groups, so we can isolate the impact of having a cabin on survival.

### **Steps in PSM**
1. **Calculate Propensity Scores:**  
   We estimate the likelihood (propensity score) of a passenger having a cabin based on their characteristics, such as age, fare, and class. Logistic regression is commonly used for this.

2. **Match Passengers:**  
   Passengers with similar propensity scores are paired. This creates balanced groups, where the only key difference is whether or not they had a cabin.

3. **Assess Balance:**  
   After matching, we check if the treated (cabin) and untreated (no cabin) groups are similar in terms of their characteristics. If they are well-balanced, any difference in survival can be more confidently attributed to having a cabin.

4. **Estimate Treatment Effect:**  
   Finally, we compare survival rates between the matched groups to estimate the causal effect of having a cabin on survival.

---

## **Challenges with Observational Data**

Unlike controlled experiments, observational data comes with inherent challenges:
- **Selection Bias:** Passengers with cabins may differ significantly from those without cabins in terms of wealth, social class, or access to resources.
- **Confounding Variables:** Factors like passenger class, age, and gender might influence both survival chances and cabin allocation, making it difficult to isolate the effect of having a cabin.
- **Unmeasured Variables:** Not all factors influencing survival may be captured in the dataset (e.g., physical strength, swimming ability).

PSM helps mitigate these challenges by balancing observed characteristics between the groups, though it cannot address unmeasured confounders.

---

## **Interpreting the Results**

By analyzing the Titanic dataset using Propensity Score Matching, we aim to determine if having a cabin played a significant role in survival. This involves:
1. **Checking Survival Rates:** Comparing survival outcomes between passengers with and without cabins.
2. **Assessing Balance:** Ensuring the matched groups are comparable in terms of observed features.
3. **Drawing Conclusions:** Determining if any observed difference in survival rates is likely due to having a cabin.

````python
def perform_match_exact(row, df , *args):
    # row is the the item that we want to match
    # df is the source Pandas dataframe that we want to match it with other items
    # print('Start matching')
    sub_set = df

    for arg in args:
        sub_set = sub_set.loc[sub_set[arg] == row[arg]]
        # print(sub_set)

    return sub_set.index


def logit(p):
    logit_value = math.log(p / (1-p))
    return logit_value

def hasCabin(x):
    if pd.isna(x):
        return 0
    else:
        return 1

def cohenD (tmp, metricName):
    treated_metric = tmp[tmp.treatment == 1][metricName]
    untreated_metric = tmp[tmp.treatment == 0][metricName]

    d = ( treated_metric.mean() - untreated_metric.mean() ) / math.sqrt(((treated_metric.count()-1)*treated_metric.std()**2 + (untreated_metric.count()-1)*untreated_metric.std()**2) / (treated_metric.count() + untreated_metric.count()-2))
    return d
````

### **One-Hot Encoding: Making Categorical Data Usable for Machine Learning**

One-hot encoding is a method to convert categorical data into a numerical format, making it suitable for machine learning algorithms that operate on numerical inputs. It transforms each unique category in a categorical feature into separate binary columns.

---

### **How One-Hot Encoding Works:**
1. **Create Binary Columns**:  
   For each unique category in a categorical feature (e.g., "male" and "female" in the `Sex` column), a separate binary column is created.
   
   - For example, the `Sex` column with categories `male` and `female` will be transformed into two new columns: `Sex_male` and `Sex_female`.

2. **Assign Binary Values**:  
   - **1** is assigned to the column representing the category a data point belongs to.  
   - **0** is assigned to all other columns for that data point.  
   Example:
```
   | Original `Sex` | `Sex_male` | `Sex_female` |
   |----------------|------------|--------------|
   | Male           | 1          | 0            |
   | Female         | 0          | 1            |
```
---

### **Why One-Hot Encoding is Important:**
Machine learning algorithms cannot directly interpret categorical data. Converting these categories into binary numerical values allows the algorithms to:
1. **Utilize Categorical Information**: Algorithms can now process categorical features as distinct inputs.  
2. **Avoid Ordinal Misinterpretation**: Unlike label encoding, one-hot encoding avoids introducing any unintended ordinal relationships between categories.

# **Balancing the Scales: Understanding Propensity Scores**

Imagine you're studying the effects of a new medication. Ideally, you'd test it on everyone to measure the results. In reality, you might only test it on a specific group. Propensity scores help address this challenge by enabling fair comparisons between treatment and non-treatment groups.

---

### **What Are Propensity Scores?**
Propensity scores estimate the likelihood of someone receiving a treatment based on their characteristics (e.g., age, health history). They predict who might have received the treatment even if they weren’t part of the test group.

---

### **Why Are Propensity Scores Important?**

1. **Fair Comparisons**:  
   - Without adjustment, treatment and non-treatment groups may differ significantly (e.g., younger, healthier individuals receiving the medication).  
   - Propensity scores account for these differences, enabling more accurate comparisons.

2. **Reducing Bias**:  
   - By balancing groups based on propensity scores, we can isolate the treatment’s effect, reducing the influence of confounding factors.

---

### **Visualizing Propensity Scores**

1. **Distribution Plots**:  
   - Plots show the spread of propensity scores for both treatment and non-treatment groups.  
   - Overlap in distributions indicates better comparability between groups.

2. **Logit Transformation**:  
   - Raw scores often cluster near 0 or 1.  
   - The logit transformation stretches these scores, making it easier to visualize differences and overlap.

````python
def perfom_matching_v2(row, indexes, df_data):
    current_index = int(row['index']) # Obtain value from index-named column, not the actual DF index.
    prop_score_logit = row['propensity_score_logit']
    for idx in indexes[current_index,:]:
        if (current_index != idx) and (row.treatment == 1) and (df_data.loc[idx].treatment == 0):
            return int(idx)

df_data['matched_element'] = df_data.reset_index().apply(perfom_matching_v2, axis = 1, args = (indexes, df_data))
````

# **Average Treatment Effect (ATT)**

ATT estimates the causal effect of a treatment by comparing outcomes for the treatment group to what their outcomes would have been without the treatment (counterfactual outcomes). Let’s explore this concept with an example:

---

### **Example: Vitamin C and Cold Prevention**
- **Scenario**: You test if taking vitamin C daily prevents colds.
- **Problem**: Health-conscious individuals may self-select into taking vitamin C, and they might be less prone to colds even without the treatment. This is a confounding variable.
- **Solution**: ATT aims to estimate the true effect of vitamin C by accounting for confounding variables.

---

### **How ATT Works**
1. **Treatment Group**: Individuals who receive the treatment (e.g., take vitamin C).
2. **Control Group**: Individuals who do not receive the treatment.
3. **Propensity Score Matching**:
   - Matches individuals in the treatment group with similar individuals in the control group based on characteristics (e.g., age, health status).
   - Reduces the influence of confounding factors.
4. **Calculate ATT**: The difference in average outcomes between the matched groups estimates the treatment effect.

# **Estimating the Causal Effect of Lifeboats on Titanic Survival Using Propensity Score Matching (PSM)**

This study investigates the causal effect of being in a lifeboat on the survival rate of passengers aboard the Titanic. While randomized controlled trials provide the most robust causal estimates, they are neither feasible nor ethical in this historical context. Therefore, **Propensity Score Matching (PSM)** is employed to analyze observational data and reduce bias introduced by confounding variables.

---

## **Steps in Applying PSM**

### **1. Identifying Confounding Variables**
Confounding variables influence both the likelihood of being assigned to a lifeboat (treatment) and survival probability (outcome). For Titanic passengers, potential confounders include:
- **Age**
- **Gender**
- **Socioeconomic status** (approximated by travel class)

### **2. Estimating Propensity Scores**
Propensity scores represent the likelihood of being assigned to a lifeboat based on observed characteristics. These scores are calculated using **logistic regression** and quantify the "similarity" of passengers in the treatment (lifeboat) and control (no lifeboat) groups, excluding treatment assignment.

### **3. Matching Treatment and Control Groups**
Passengers in the control group are matched with passengers in the treatment group based on similar propensity scores. This matching creates comparable groups, balancing measured characteristics and isolating the treatment effect.

### **4. Assessing Balance**
After matching, **visual tools** such as density plots and jitter plots are used to check if the distributions of confounders (e.g., age, gender, class) are balanced across the treatment and control groups. Effective matching ensures that both groups are comparable, apart from treatment assignment.

### **5. Estimating the Average Treatment Effect (ATT)**
The **ATT** is calculated as the average difference in survival rates between the treatment and control groups after matching. This provides an estimate of the effect of being in a lifeboat on survival probability, adjusted for confounding variables.

---

## **Key Insights**

1. **ATT Estimate**:  
   The ATT quantifies the average increase (or decrease) in survival probability due to being in a lifeboat. For instance, an ATT of **0.1189** indicates that being in a lifeboat increased a passenger's survival probability by approximately **11.89%**.

2. **Limitations**:
   - **Unobserved Confounding**: PSM cannot account for variables not measured in the dataset (e.g., physical strength, proximity to lifeboats).  
   - **Average Effect**: ATT represents the average effect across all passengers, and individual impacts may vary.

---

## **Conclusion**
This study demonstrates how **Propensity Score Matching** can be applied to estimate the causal effect of lifeboats on Titanic survival rates. By mitigating the impact of confounding variables, PSM provides a more reliable estimate than unadjusted analyses. However, researchers should recognize the limitations of observational studies and supplement PSM with additional statistical methods to strengthen causal inferences.
