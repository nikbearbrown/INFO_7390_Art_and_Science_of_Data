## Methods for Investigating Causality: A Deep Dive into Randomized Controlled Trials (RCTs)

### Understanding RCTs: The Gold Standard of Causal Research

Randomized Controlled Trials (RCTs) represent a pinnacle of scientific methodology for establishing causal relationships. They are particularly powerful in fields like medicine, psychology, and social sciences, offering a rigorous approach to understanding how interventions truly impact outcomes. However, RCTs are often impractical for large-scale ecological datasets due to ethical and logistical constraints. Instead, observational methods like **difference-in-differences**, **propensity score matching**, or **instrumental variable analysis** are more applicable in such contexts.

### Core Components of RCTs

#### 1. Randomization: The Great Equalizer

Randomization is the cornerstone of RCTs, serving as a powerful tool to neutralize potential biases. By randomly assigning participants to different groups, researchers ensure that:
- Every participant has an equal chance of being in any group
- Individual characteristics are likely to be distributed evenly
- Selection bias is minimized

#### 2. Control Group: The Critical Benchmark

The control group acts as a crucial reference point, allowing researchers to:
- Compare outcomes between treated and untreated populations
- Isolate the specific effects of an intervention
- Distinguish between true causal effects and random variations

#### 3. Blinding: Protecting Against Psychological Influences

Blinding techniques help prevent:
- Participant expectations from influencing results
- Researcher bias in interpretation
- Placebo effects from skewing outcomes

### Strengths of RCTs

**Unparalleled Causal Insight**  
- Provides the strongest evidence for cause-and-effect relationships
- Minimizes the risk of spurious correlations
- Offers a systematic approach to understanding interventions

### Limitations and Challenges

While powerful, RCTs are not without constraints:

1. **Resource Intensive**  
   - Expensive to design and implement
   - Require significant time and financial investment

2. **Practical Constraints**  
   - Not always feasible or ethical to randomize
   - Potential limitations in real-world generalizability

3. **Ethical Considerations**  
   - Some interventions cannot be ethically tested
   - Potential risks to participants must be carefully weighed

### Context-Specific Causal Methods for Ecological Data

Given the challenges of using RCTs for large ecological datasets, more context-specific causal inference methods are typically employed:

- **Difference-in-Differences (DiD)**: This method compares the changes in outcomes over time between a group that is exposed to a treatment (e.g., a policy change) and a group that is not. It helps isolate the causal effect of interventions when randomization is not feasible. DiD is particularly useful in policy evaluation, where the goal is to determine the effect of a policy intervention on a specific outcome.

- **Propensity Score Matching (PSM)**: PSM helps reduce selection bias by matching treated and untreated observations with similar characteristics. This is particularly useful in observational studies where groups differ in systematic ways. By creating a more balanced dataset, PSM enhances the reliability of causal estimates, especially when randomized controlled trials (RCTs) are not possible.

- **Instrumental Variable Analysis**: This approach uses instruments—variables that are correlated with the treatment but not directly with the outcome except through the treatment—to address endogeneity issues. It is useful for dealing with unobserved confounders that may bias causal estimates, making it valuable when certain variables are difficult to measure or control direc#tly.

## Distinguishing Correlation from Causation

Understanding the distinction between correlation and causation is fundamental to causal inference. Correlation indicates that two variables move together, but it does not imply that one variable causes the other. Causation, on the other hand, implies a direct influence of one variable on another. Establishing causation requires careful consideration of confounding factors, proper study design, and the use of causal inference techniques like those discussed above. #  Handling Missing Data

### This session focuses on visualizing and imputing missing data to ensure the dataset is complete and suitable for analysis.

- **Missing Data Visualization**:
  - A heatmap is used to visualize the missing data in the dataset. This visualization provides an overview of which columns contain missing values and helps identify patterns in missingness.

- **Imputation Techniques**:
  - **Mean/Median Imputation for Numerical Columns**: Missing values in numerical columns such as `crop_land`, `grazing_land`, `forest_land`, `carbon`, and `total` are imputed using either the mean or median. This helps ensure that the dataset remains complete without dropping any records.
  - **Country-Specific Imputation**: For `crop_land` and `grazing_land`, missing values are imputed based on the median value for each country. This ensures that the imputation is more context-specific, preserving country-level variations.
  - **Temporal Interpolation for Time-Series Continuity**: The dataset is sorted by `country` and `year` and then interpolated to fill in gaps, ensuring continuity in the time series data.

These steps help ensure that the dataset is free from missing values, which is crucial for reliable data analysis and modeling.

```python
# Handling Missing Data

# Missing Data Visualization
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Data')
plt.show()

# Imputation Techniques
# Mean/Median Imputation for numerical columns
for column in ['crop_land', 'grazing_land', 'forest_land', 'carbon', 'total']:
    if df[column].notna().sum() > 0:  # Only impute if there are non-NaN values
        if column in ['carbon', 'total']:
            df[column] = df[column].fillna(df[column].mean())
        else:
            df[column] = df[column].fillna(df[column].median())

# Country-Specific Imputation
for column in ['crop_land', 'grazing_land']:
    if df[column].notna().sum() > 0:
        df[column] = df.groupby('country')[column].transform(lambda x: x.fillna(x.median()))

# Temporal Interpolation for time-series continuity
df.sort_values(by=['country', 'year'], inplace=True)
df.interpolate(method='linear', inplace=True)
```

# Time Series Analysis

### This session conducts time series analysis to understand trends and forecast future ecological metrics.

- **Country-Level Analysis**:
  - The data is filtered to include specific countries (`United States`, `Canada`, `Mexico`) for analysis. The data is then grouped by `year` to calculate the mean values for relevant numerical features.

- **Plotting Trends**:
  - Line plots are created to visualize trends in `total_land_use` and `carbon` over the years. This helps to observe how these metrics have changed over time for the aggregated countries.

- **Seasonal Decomposition**:
  - Seasonal decomposition is performed on `total_land_use` to separate it into trend, seasonality, and residual components. This analysis helps in understanding the underlying patterns in the data.

- **Forecasting using ARIMA**:
  - The `ARIMA` model is used for forecasting future values of `total_land_use`. The data is first transformed into log form to stabilize the variance, and then ARIMA is applied to forecast the next 10 years.
  - The forecasted values are plotted alongside the historical data to visualize potential future trends in land use.

These steps help in understanding the temporal trends and patterns in the data and provide a basis for forecasting future ecological metrics, which can be useful for planning and decision-making.


```python
# Time Series Analysis

# Country-Level Analysis
# Analyzing trends in biocapacity and ecological footprint over time
countries_to_aggregate = ['United States', 'Canada', 'Mexico']
df_aggregated = df[df['country'].isin(countries_to_aggregate)]

# Grouping by 'year' and calculating the mean for only numeric columns
df_aggregated_numeric = df_aggregated.groupby('year')[df_aggregated.select_dtypes(include=['number']).columns].mean().reset_index()

# Plotting trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_aggregated_numeric, x='year', y='total_land_use', label='Total Land Use')
sns.lineplot(data=df_aggregated_numeric, x='year', y='carbon', label='Carbon Footprint')
plt.title('Trends in Ecological Footprints for Aggregated Countries')
plt.xlabel('Year')
plt.ylabel('Footprint')
plt.legend()
plt.show()

# Seasonal Decomposition
aggregated_data = df_aggregated_numeric.set_index('year')['total_land_use']
decomposition = seasonal_decompose(aggregated_data, model='additive', period=5)
decomposition.plot()
plt.show()

# Forecasting using ARIMA
if not df_aggregated_numeric.empty:
    # Ensure 'year' is in datetime format and set the frequency explicitly
    df_aggregated_numeric.set_index('year', inplace=True)
    df_aggregated_numeric.index = pd.to_datetime(df_aggregated_numeric.index, format='%Y')
    df_aggregated_numeric = df_aggregated_numeric.asfreq('YE')  # Set frequency to yearly (year-end)
    
    non_null_obs = df_aggregated_numeric['total_land_use'].notna().sum()

    if non_null_obs >= 20:
        # Log-transform and fit the ARIMA model
        df_aggregated_numeric['total_log'] = np.log1p(df_aggregated_numeric['total_land_use'])
        arima_model = ARIMA(df_aggregated_numeric['total_log'], order=(0, 1, 0))
        arima_result = arima_model.fit()
        forecast_log = arima_result.forecast(steps=10)
        forecast = np.expm1(forecast_log)

        # Generating a date range for the forecast with YE frequency
        last_year = df_aggregated_numeric.index[-1]
        forecast_years = pd.date_range(start=last_year + pd.DateOffset(years=1), periods=10, freq='YE')

        # Plotting the forecast
        plt.figure(figsize=(10, 6))
        plt.plot(df_aggregated_numeric.index, df_aggregated_numeric['total_land_use'], label='Historical Total Land Use')
        plt.plot(forecast_years, forecast, linestyle='--', label='Forecast')
        plt.title('Forecast of Total Land Use for Aggregated Countries using ARIMA')
        plt.xlabel('Year')
        plt.ylabel('Total Land Use')
        plt.legend()
        plt.show()
```

# Causal Analysis Using Linear Regression

In this notebook, we are performing a causal analysis to understand the relationships between variables such as `footprint_production`, `biocapacity`, and `footprint_consumption`. We use different econometric methods, including linear regression, Difference-in-Differences (DiD) analysis, and counterfactual analysis to estimate these relationships.

### Setting Up the Regression Model to Estimate Causal Effect

We start by using a linear regression model to estimate the causal effect of production footprint and biocapacity on consumption footprint:

- **Independent Variables (`X`)**: `footprint_production` and `biocapacity`
- **Dependent Variable (`Y`)**: `footprint_consumption`
- **Constant Term**: Added to the predictors to improve the model by capturing any baseline shift.

The model estimates the causal effect of `footprint_production` and `biocapacity` on `footprint_consumption`. The `OLS` (Ordinary Least Squares) method is used to fit the model, and the results provide estimates of coefficients, standard errors, and statistical significance, which can help us infer the relationship between thes Difference-in-Differences analysis.


```python
# Causal Analysis Using Linear Regression

# Setting up the regression model to estimate causal effect
X = df_pivot[['footprint_production', 'biocapacity']]  # Independent variables
X = sm.add_constant(X)  # Adds a constant term to the predictor
Y = df_pivot['footprint_consumption']  # Dependent variable

model = sm.OLS(Y, X).fit()  # Ordinary Least Squares regression
print("\nCausal Estimate of Footprint Production and Biocapacity on Footprint Consumption:")
print(model.summary())
```

### Difference-in-Differences (DiD) Analysis

To evaluate the impact of a policy intervention, such as a renewable energy policy, we use a Difference-in-Differences (DiD) approach.

- We assume that there was a policy intervention in the year 2015.
- We create a variable `policy_intervention` to differentiate between pre-policy and post-policy periods.
- We then compute the mean difference in ecological footprint before and after the policy to estimate its impact.

The Difference-in-Differences estimate helps in understanding the effect of the policy on `footprint_consumption`. However, we also ensure that there are enough observations in both pre and post-policy periods for a reliable estimation.

```python
# Difference-in-Differences (DiD) Analysis
# Assuming we have data before and after a policy intervention (e.g., renewable energy policy)
df_pivot['policy_intervention'] = np.where(df_pivot['year'].dt.year >= 2015, 1, 0)  # Example policy change in 2015
pre_policy = df_pivot[df_pivot['policy_intervention'] == 0]
post_policy = df_pivot[df_pivot['policy_intervention'] == 1]

# Checking if there are enough observations in both periods
if len(pre_policy) > 0 and len(post_policy) > 0:
    # Calculating the mean difference in ecological footprint before and after the policy
    did_estimate = post_policy['footprint_consumption'].mean() - pre_policy['footprint_consumption'].mean()
    print("\nDifference-in-Differences Estimate for Policy Impact on Footprint Consumption:")
    print(did_estimate)
else:
    print("\nNot enough observations for Difference-in-Differences analysis.")
```

### Counterfactual Analysis

In this analysis, we estimate what would have happened without a particular policy intervention. Here, we simulate a hypothetical scenario where biocapacity increases by 10% (e.g., due to afforestation efforts) and observe its effect on `footprint_consumption`.

- We create a new variable `afforestation_increase` that simulates a 10% increase in `biocapacity`.
- We then fit a regression model with this counterfactual variable as one of the predictors.

The counterfactual model allows us to estimate the potential impact on `footprint_consumption` if biocapacity was increased by 10%. This provides insights into the effectiveness of environmental policies that could lead to increased biocapacity.

```python
# Counterfactual Analysis
# Estimating what would have happened without a policy (e.g., afforestation increased by 10%)
df_pivot['afforestation_increase'] = df_pivot['biocapacity'] * 1.10  # Simulate a 10% increase in biocapacity
X_cf = df_pivot[['afforestation_increase', 'footprint_production']]
X_cf = sm.add_constant(X_cf)

model_cf = sm.OLS(Y, X_cf).fit()
print("\nCounterfactual Estimate for 10% Increase in Biocapacity on Footprint Consumption:")
print(model_cf.summary())

# Ensure sufficient observations for Difference-in-Differences
if len(pre_policy) < 10 or len(post_policy) < 10:
    print("\nWarning: Not enough observations for a robust Difference-in-Differences analysis. Consider increasing data availability or choosing a different policy intervention year.")
```

### Ensuring Robust Difference-in-Differences Analysis

To ensure the validity of our Difference-in-Differences estimate, we check if there are enough observations for both pre-policy and post-policy periods. If there are less than 10 observations in either period, a warning is printed indicating that the analysis might not be robust.

This step is important because having insufficient data can lead to misleading conclusions from the Difference-in-Differences analysis.

```python
# Country-Specific Reports: Generate reports for specific countries
countries_to_report = ['United States', 'China', 'India']
valid_countries = []

# First, gather the valid countries that have data and print their reports
for country in countries_to_report:
    country_data = df_pivot[df_pivot['country'] == country]
    if not country_data.empty:
        valid_countries.append(country)
        # Print the statistical report for the country
        print(f"\nReport for {country}:")
        print(country_data[['year', 'footprint_consumption', 'footprint_production', 'biocapacity', 'ecological_deficit']].describe())
    else:
        print(f"\nWarning: No data available for {country}.")

# Create subplots based on the number of valid countries
num_valid = len(valid_countries)
if num_valid == 0:
    print("No data available for any of the requested countries.")
else:
    nrows = (num_valid + 1) // 2  # Calculate the number of rows needed
    ncols = 2 if num_valid > 1 else 1

    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(14, nrows * 5))
    axes = axes.flatten() if num_valid > 1 else [axes]

    # Plotting for each valid country
    for i, country in enumerate(valid_countries):
        country_data = df_pivot[df_pivot['country'] == country]
        ax = axes[i]

        sns.lineplot(data=country_data, x='year', y='footprint_consumption', label='Footprint Consumption', ax=ax)
        sns.lineplot(data=country_data, x='year', y='footprint_production', label='Footprint Production', ax=ax)
        sns.lineplot(data=country_data, x='year', y='biocapacity', label='Biocapacity', ax=ax)

        ax.set_title(f'Trends in Ecological Metrics for {country}')
        ax.set_xlabel('Year')
        ax.set_ylabel('Value (GHA)')
        ax.legend()

    plt.tight_layout()
    plt.show()
```
