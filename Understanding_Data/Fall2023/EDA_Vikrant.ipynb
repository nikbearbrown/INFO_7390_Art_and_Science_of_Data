#Exploratory Data Analysis

EDA in data science stands for "Exploratory Data Analysis." It is a crucial initial step in the data analysis process where analysts or data scientists examine and summarize key characteristics of a dataset. EDA involves various techniques and approaches to understand the data's structure, distribution, patterns, and relationships between variables. The primary goal of EDA is to gain insights and identify potential issues or areas for further analysis before proceeding with more advanced modeling or hypothesis testing. EDA includes tasks such as data visualization, summary statistics, data cleaning, and detection of outliers and missing values. It helps in shaping research questions, selecting appropriate modeling techniques, and making informed decisions throughout the data analysis pipeline.

**Basics of Algorithms and Problem Design in EDA:**


1. Algorithm Design: EDA is not typically associated with a specific algorithm in the way that machine learning or optimization problems are. Instead, EDA encompasses a collection of techniques and methods that aim to uncover patterns, relationships, and insights in data. These methods include statistical summaries, data visualization, and data transformation techniques. The "algorithm" in EDA is more of a workflow or process that guides data exploration and analysis.

2. Problem Formulation: EDA begins with formulating the problem of understanding a dataset. This involves defining the objectives, research questions, and hypotheses that guide the analysis. The problem formulation phase in EDA is more about setting the context for exploration rather than specifying a mathematical or computational problem.

### Remove Outliers from Dataset

Shape (number of rows and columns) of Dataset before removing Outliers

```python
old_shape = wine_df.shape
print("Old shape of dataset : ",old_shape)
```

Defining a function that contains a list to remove Outliers
- Using inter quantile range method to remove outliers from upper and lower quantile
- Removing values above and below the inter quantile range

```python
def remove_outliers(wine_df,column):
    q1 = wine_df[column].quantile(0.25)
    q3 = wine_df[column].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 *iqr
    outliers_list = wine_df.index[(wine_df[column] < lower) | (wine_df[column] > upper)]

    return outliers_list
```

iterating over the columns and storing the outlier values in an empty list

```python
list_res =[]
for i in ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality_level']:
    list_res.extend(remove_outliers(wine_df,i))
```

Total number of outliers present -

```python
len(list_res)
```

Creating a function to return final cleaned dataset after removing outliers

```python
def cleaned_dataframe(wine_df,lis):
    lis = sorted(set(lis))
    wine_df = wine_df.drop(lis)
    return wine_df
```

Call the above function

```python
clean_wine_df = cleaned_dataframe(wine_df,list_res)
```
