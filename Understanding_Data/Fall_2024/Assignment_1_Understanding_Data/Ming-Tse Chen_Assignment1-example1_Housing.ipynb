# üöÄ Exploring Exploratory Data Analysis (EDA)

Since we now recognize the importance of EDA in evaluating and refining data quality, let's take a deeper look. We will now explore EDA in detail to understand how it uncovers patterns, identifies trends, and extracts valuable insights to enhance data-driven decision-making.

## üìä What is Exploratory Data Analysis (EDA)?

Exploratory Data Analysis, or **EDA**, is a key step in understanding a dataset before applying complex models or statistical tests. It involves using **visualizations, summary statistics, and hypothesis testing** to uncover insights and detect potential issues in the data.

## üîç Goals of EDA:

1Ô∏è‚É£ **Summarize the dataset** to get an overview of its structure.

2Ô∏è‚É£ **Identify patterns and relationships** between different variables.

3Ô∏è‚É£ **Spot outliers, missing data, and class imbalances** to ensure data quality.

4Ô∏è‚É£ **Connect the data with business goals**, ensuring relevant insights are extracted for decision-making.

By performing EDA, we can make informed decisions about data preparation, feature selection, and modeling techniques, leading to more effective and reliable analyses.

## üìå Key Techniques in EDA

EDA combines both **graphical and statistical** techniques to explore and understand data effectively:

1Ô∏è‚É£ **Descriptive Statistics** - Summarizes key features of data using measures like mean, median, mode, range, and standard deviation.

Example: Calculating average monthly sales to understand business trends.

2Ô∏è‚É£ **Data Visualization** - Uses charts such as histograms, box plots, and scatter plots to reveal patterns and relationships.

Example: A histogram displaying customer purchase frequency over time.

3Ô∏è‚É£ **Identifying Outliers** - Detects extreme values that could indicate errors or unique insights.

Example: Spotting an unusually high transaction amount that might suggest fraud.

4Ô∏è‚É£ **Examining Relationships** - Measures correlations between variables using statistical techniques like correlation coefficients and cross-tabulations.

Example: Analyzing if higher marketing spend correlates with increased sales.

5Ô∏è‚É£ **Group Comparisons** - Compares data across different groups to find patterns.

Example: Comparing average exam scores across multiple schools to evaluate educational performance.



### üìñ Understanding the Theory Behind EDA

Exploratory Data Analysis (EDA) was introduced by John W. Tukey as a flexible and interactive approach to exploring data before applying complex statistical models. EDA focuses on summarizing the main features of a dataset, uncovering patterns, detecting anomalies, and verifying assumptions to ensure accurate analysis. 



Using descriptive statistics, visualizations, and correlation analysis, EDA helps in understanding data relationships, selecting important variables, cleaning data, and recognizing meaningful patterns. 



This exploratory phase plays a crucial role in helping analysts gain valuable insights and make data-driven decisions before proceeding with formal statistical modeling or machine learning techniques.

## üî¢ Basics of Algorithms and Problem Design in EDA

## 1Ô∏è‚É£ Algorithm Design

Unlike machine learning models, EDA does not rely on a specific algorithm. Instead, it consists of **various techniques and workflows** used to explore and understand data. These techniques include **data transformation tools, visualization methods, and statistical summaries**. In EDA, an "algorithm" is better described as a **structured approach** that guides data exploration and analysis.

## 2Ô∏è‚É£ Problem Formulation

The first step in EDA is to **define the problem** that needs exploration. This involves setting **clear objectives, research questions, and hypotheses** that will guide the analysis. Unlike traditional mathematical problem-solving, EDA focuses on **contextual exploration** rather than defining rigid computational problems.

By following these principles, EDA ensures that **data-driven insights are meaningful, reliable, and aligned with business or research objectives.**

# üìä Box Plots: Identifying Outliers and Spread of Data

### **Definition:**  
A **box plot** (also called a **box-and-whisker plot**) summarizes numerical data using five key metrics:
- **Minimum** (Lowest value excluding outliers)
- **First Quartile (Q1)** (25th percentile)
- **Median (Q2)** (50th percentile, middle value)
- **Third Quartile (Q3)** (75th percentile)
- **Maximum** (Highest value excluding outliers)

It also highlights **outliers** as individual points outside the whiskers.

---

### **üìå Why Use Box Plots?**
‚úî **Quickly identifies outliers**.  
‚úî **Shows the spread and skewness** of the data.  
‚úî **Helps compare distributions** across different variables.

---

### **üìå Example: Box Plot of House Prices**
We visualize the **spread and outliers** in house prices using a box plot.


```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the box plot for house prices to identify outliers
plt.figure(figsize=(8, 6))
sns.boxplot(y=housing_df['price'], color="lightgreen")
plt.title("Box Plot of House Prices")
plt.ylabel("Price (INR)")
plt.show()
```

## üîç Insights from the Box Plot

‚úî **Most House Prices Fall Between ‚Çπ30 Lakhs to ‚Çπ70 Lakhs (INR)**  
   - The **Interquartile Range (IQR)** (middle 50% of the data) represents **affordable housing**.

‚úî **Presence of High-Value Outliers**  
   - There are **luxury properties exceeding ‚Çπ1 crore (INR)**.  
   - These **outliers could impact predictive modeling** and might need handling.

‚úî **Potential Data Cleaning Step**  
   - **Outlier treatment** (e.g., removing or capping extreme values) could **improve data quality**.  
   - A **log transformation on price** may help **normalize the data**.

# üö® Outlier Detection Using Z-Score Analysis

### **Definition:**

The **Z-score method** detects outliers by measuring how many **standard deviations a data point is from the mean**.

---

### **üìå Why Use It?**
‚úî **Detects extreme values** that may skew analysis.  
‚úî **Prevents biased results** in machine learning models.  
‚úî **Ensures data consistency and reliability** before further processing.

### **üìå Example: Detecting Outliers in House Prices**
We use **Z-score analysis to identify extreme house prices** that might distort insights.


```python
# Import necessary libraries
from scipy import stats
import numpy as np

# Compute Z-scores for house prices
z_scores = np.abs(stats.zscore(housing_df['price']))

# Identify outliers as values with a Z-score greater than 3
outliers_zscore = housing_df[z_scores > 3]

# Count number of outliers
num_outliers = len(outliers_zscore)

# Display number of detected outliers
print(f"Number of outliers detected using Z-Score method: {num_outliers}")
```

# üö® IQR (Interquartile Range) Method for Outlier Detection

### **Definition:**  
The **IQR method** identifies outliers by measuring the **spread of the middle 50% of data**.  
It flags values that are **significantly higher or lower** than the normal range.

---

### **üìå Why Use It?**
‚úî **Detects extreme values** that might distort analysis.  
‚úî **More robust to skewed data** compared to standard deviation-based methods.  
‚úî **Essential for cleaning datasets** before modeling.

---

### **üìå Example: Detecting Outliers in House Prices**
We use the **Interquartile Range (IQR) method** to identify **extreme house prices**.


```python
# Import necessary libraries
import pandas as pd

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for house prices
Q1 = housing_df['price'].quantile(0.25)
Q3 = housing_df['price'].quantile(0.75)
IQR = Q3 - Q1  # Interquartile Range

# Identify outliers using the IQR method
outliers_iqr = housing_df[(housing_df['price'] < Q1 - 1.5 * IQR) | (housing_df['price'] > Q3 + 1.5 * IQR)]

# Count number of detected outliers
num_outliers_iqr = len(outliers_iqr)

# Display number of detected outliers
print(f"Number of outliers detected using IQR method: {num_outliers_iqr}")
```

## üîç Insights from IQR Outlier Detection

‚úî **15 House Prices Were Identified as Outliers**  
   - These houses have prices that are **significantly different from the rest of the dataset**.  
   - This suggests the presence of **luxury homes or extreme pricing anomalies**.

‚úî **Impact of Outliers on Analysis**  
   - Outliers can **skew price averages and affect model predictions**.  
   - Some extreme values may **represent valid high-value properties**, while others might be **data errors**.

‚úî **How to Handle These Outliers?**  
   - **Option 1: Remove** them if they distort analysis.

# üîç Principal Component Analysis (PCA)

### **Definition:**  
PCA is a **dimensionality reduction technique** that transforms correlated features into a set of **uncorrelated principal components** while retaining as much variance as possible.

---

### **üìå Why Use It?**
‚úî **Reduces dataset complexity** while preserving important information.  
‚úî **Helps visualize high-dimensional data** in a lower-dimensional space.  
‚úî **Removes redundant features**, improving **model efficiency and performance**.  

---

### **üìå Example: Applying PCA to the Housing Dataset**
We apply **PCA to reduce dimensionality** while keeping essential data variance.


```python
# Import necessary libraries
from sklearn.decomposition import PCA

# Select only numeric columns
housing_df_numeric = housing_df.select_dtypes(include=["float64", "int64"])

# Handle missing values by filling with column mean
housing_df_numeric = housing_df_numeric.fillna(housing_df_numeric.mean())

# Initialize PCA with 2 components
pca = PCA(n_components=2)

# Fit PCA to the dataset, excluding the target variable ('price')
pca_result = pca.fit_transform(housing_df_numeric.drop('price', axis=1))

# Display transformed dataset
print("First 5 rows of PCA transformed data:\n", pca_result[:5])
```

## üîç Insights from PCA

‚úî **The two principal components capture most of the dataset's variability.**  
   - The **original dataset had multiple features**, but PCA compresses it into just **two uncorrelated components**.  

‚úî **This transformation helps in visualizing patterns and reducing computation time.**  
   - Instead of working with high-dimensional data, we can **interpret key relationships in a lower-dimensional space**.

‚úî **Explained Variance Analysis**  
   - PCA can be **further analyzed using explained variance ratios** to determine **how much information each component retains**.  
   - This helps decide whether **more or fewer components** are needed.

# üõ† Handling Missing Values (Imputation)

## **Strategy:**
We apply different imputation methods based on the data type:

- **Numerical Data** (e.g., `price`, `area`, `bedrooms`, `bathrooms`):  
  ‚úÖ Use **median** imputation to handle missing values while avoiding the influence of outliers.

- **Categorical Data** (e.g., `furnishingstatus`, `mainroad`, `airconditioning`):  
  ‚úÖ Use **mode** imputation to replace missing values with the most common category.

---

## **4.1 Filling Missing Values**
### **üìú Code Implementation**

```python
# Fill numerical columns with median
numerical_cols = housing_df.select_dtypes(include=['float64', 'int64']).columns
housing_df[numerical_cols] = housing_df[numerical_cols].fillna(housing_df[numerical_cols].median())

# Fill categorical columns with mode
categorical_cols = housing_df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    housing_df[col] = housing_df[col].fillna(housing_df[col].mode()[0])

# Verify that missing values have been handled
housing_df.isnull().sum()
```

## üîç Insights from Handling Missing Values (Imputation)

‚úî **All missing values have been successfully handled.**  
   - No missing values remain in the dataset, ensuring data consistency.  

‚úî **Numerical Data: Median Imputation**  
   - The **median** was used instead of the mean to **reduce the impact of extreme values (outliers)**.  
   - This approach ensures a **stable and representative** dataset.  

‚úî **Categorical Data: Mode Imputation**  
   - The **most frequent category** was used to fill missing values in categorical columns.  
   - This prevents potential inconsistencies while maintaining data integrity.  


# üìä Detecting and Handling Outliers

Outliers can distort analysis. We use the **Interquartile Range (IQR)** to detect extreme values in house prices.

---

## **5.1 Outlier Detection Using IQR**
We identify outliers based on the **IQR method**, which captures values that fall outside **1.5 times the interquartile range**.

```python
# Define Q1 (25th percentile) and Q3 (75th percentile)
Q1 = housing_df['price'].quantile(0.25)
Q3 = housing_df['price'].quantile(0.75)

# Compute IQR
IQR = Q3 - Q1

# Identify outliers as values outside 1.5 times the IQR range
outliers = housing_df[(housing_df['price'] < (Q1 - 1.5 * IQR)) | (housing_df['price'] > (Q3 + 1.5 * IQR))]

# Display number of outliers detected
print(f"Number of outliers detected: {len(outliers)}")
```

## 3. Handling Skewness Using Log Transformation in Housing Data

Highly skewed variables like **Sale Price** and **Lot Size** can distort analysis. Log transformation helps normalize these distributions, making them more suitable for modeling.


```python
# Import necessary library
import numpy as np

# Function for log transformation
def log_transform(df, cols):
    for col in cols:
        if (df[col] == 0).all():  # Avoid log(0) issue
            continue
        else:
            df[col + '_log'] = np.log(df[col] + 1)  # Log transformation with offset to avoid log(0)
    return df

# Applying log transformation to skewed housing variables
df = log_transform(housing_df, ['price', 'area'])

# Display the first few rows to confirm the transformation
df[['price', 'price_log', 'area', 'area_log']].head()
```

### Insights:
- **Sale Price** and **Lot Size** distributions become more **normal** after log transformation.
- **Helps reduce the impact of extreme values** and maintain consistency for better model accuracy.
- **Log transformation improves predictive modeling** by stabilizing variance in highly skewed data.
- By normalizing data distributions, we enhance **real estate market analysis** and pricing predictions! üìäüè°

## 7. Handling Outliers in Housing Data

Outliers can distort analysis and need to be removed using Z-score or IQR (Interquartile Range) methods to improve model accuracy.

### Z-score Removal

Z-scores identify outliers by determining how many standard deviations a data point is from the mean.



```python
# Import necessary library
import numpy as np
from scipy import stats

# Function to remove outliers using Z-score
def handle_outliers_zscore(df, column):
    z_scores = np.abs(stats.zscore(df[column]))  # Compute Z-score for the column
    return df[z_scores < 3]  # Keep only rows where Z-score is less than 3

# Applying Z-score method to detect and remove extreme values for Sale_Price and Area
df = handle_outliers_zscore(housing_df, 'price')
df = handle_outliers_zscore(df, 'area')

# IQR-Based Outlier Removal
# Function to remove outliers using the IQR method
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[column] >= (Q1 - 1.5 * IQR)) & (df[column] <= (Q3 + 1.5 * IQR))]

# Applying IQR method for Area outliers detection and removal
df = handle_outliers_iqr(df, 'area')

# Display the cleaned dataframe
df.head()
```

### Insights:
- **Removing outliers** improves model accuracy by reducing noise in the dataset.
- **Sale Price** and **Area** had extreme values that required capping to prevent distortion in analysis.
- Helps ensure better trend analysis, accurate predictions, and reliable housing market insights.

By handling outliers effectively, we refine data quality for more precise real estate analysis! üìäüè°
