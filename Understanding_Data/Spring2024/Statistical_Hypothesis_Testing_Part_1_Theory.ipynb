Hypothesis testing is a fundamental concept in statistics used to make informed decisions about population parameters based on sample data. It involves formulating and testing hypotheses about these parameters. This process helps researchers draw conclusions and make inferences about the characteristics of a population. In this Jupyter Notebook, we will explore the theoretical background of hypothesis testing and provide practical examples.

### A Brief History of Hypothesis Testing

Formal hypothesis testing is perhaps the most prominent and widely-employed form of statistical analysis. It is sometimes seen as the most rigorous and definitive part of a statistical analysis, but it is also the source of many statistical controversies. The currently-prevalent approach to hypothesis testing dates to developments that took place between 1925 and 1940, especially the work of Ronald Fisher, Jerzy Neyman, and Egon Pearson.

### Key Terms in Hypothesis Testing

<i><b>Statistical hypothesis</b></i>: a claim about the value of a parameter or population characteristic. <br>
Examples: <br>
• H: µ = 75 cents, where µ is the true population average of daily per-student candy+soda expenses in US high schools <br>
• H: p < .10, where p is the population proportion of defective helmets for a given manufacturer <br>
• If µ1 and µ2 denote the true average breaking strengths of two different types of twine, one hypothesis might be the assertion that µ1 – µ2 = 0, or another is the statement µ1 – µ2 > 5 

A <i>falsifiable hypothesis</i> is a statement, or hypothesis, that can be contradicted with evidence. In empirical (data-driven) research, this evidence will always be obtained through the data. <br>
In <i>statistical hypothesis testing</i>, the hypothesis that we formally test is called the <i>**null hypothesis (H<sub>0</sub>)**</i>. The <i>**alternative hypothesis(H<sub>1</sub> or H<sub>A</sub>)**</i> is a second hypothesis that is our proposed explanation for what happens if the null hypothesis is wrong.

Suppose for example that we are interested in whether newly-licensed drivers have a higher risk of being in a serious accident than drivers who have substantial driving experience. <br>
We would state as our null hypothesis that “newly-licensed drivers and experienced drivers are equally likely to be in a severe accident”. The alternative hypothesis might be that “newly-licensed drivers are more likely to be in a severe accident than experienced drivers”.

### Components of a Hypothesis Test

**1. Formulate the hypothesis to be tested.**

   Is there strong evidence for the alternative?
   - The burden of proof is placed on those who believe in the alternative claim.
   - This initially favored claim (H0) will not be rejected in favor of the alternative claim (Ha or H1) unless the sample evidence provides significant support for the alternative assertion.
   - If the sample does not strongly contradict H0, we will continue to believe in the plausibility of the null hypothesis. <br>

   The two possible conclusions: 1) Reject H0. 2) Fail to reject H0.

   **2. Determine the appropriate test statistic and calculate it using the sample data.**
   - A test statistic is a rule, based on sample data, for deciding whether to reject H0 <br>
       Which test statistic is “best”??
   - Choice of a particular test procedure must be based on the probability the test will produce incorrect results. <br>
     
     
   _Errors in Hypothesis Testing_ - <br>
     Definition : <br>
        - A **type I error** is when the null hypothesis is rejected, but it is true (False Negative) <br>
        - A **type II error** is not rejecting H0 when H0 is false (False Positive) <br>
        
   _Statistical power_, the power of a hypothesis test, is the probability that the test correctly rejects the null hypothesis. That is, the probability of a true positive result if there is such an effect to detect.

**3. Comparison of test statistic to critical region to draw initial conclusions.**
  
 - Specify the largest value of α that can be tolerated, and then find a rejection region with that α.
 - The resulting value of α is often referred to as the <i>significance level</i> of the test.
 - Traditional levels of significance are .10, .05, and .01, though the level in any particular problem will depend on the        seriousness of a type I error— the more serious the type I error, the smaller the significance level should be. 
 - We can also obtain a smaller value of α -- the probability that the null will be incorrectly rejected – by decreasing          the size of the <i>rejection region</i>. 

Let's understand this with help of an example, and some math:

  **4. Calculation of p-value.**
  - The p-value measures the “extremeness” of the sample.
  - **Definition:** The p-value is the probability we would get the sample we have or something more extreme if the null hypothesis were true.
  - So, the smaller the P-value, the more evidence there is in the sample data against the null hypothesis and for the alternative hypothesis. 
  - The p-value is between 0 and 1.

   **Common Misconceptions-<br>**
  - Misinterpretation of p-values:<br>
    P-value is not the probability of the null hypothesis being true.<br>
    Small p-value doesn't prove a hypothesis; it indicates evidence against the null.

  - Statistical vs. Practical Significance:<br>
    Statistical significance doesn't imply practical importance.<br>
    It's crucial to consider the size of the effect in addition to statistical results.

  **Interpretation of p-value-**
  - A small p-value (typically <= 0.05, 0.05 is a commonly used threshold, the threshold is often denoted as  α) indicates **strong evidence** against the null hypothesis, so we reject the null hypothesis. 
  - A large p-value (> 0.05) indicates **weak evidence** against the null hypothesis, so we fail to reject the null hypothesis.   -Although p-value is still in our favor, we cannot conclusively say that it was not due to random noise.
  - p-values very close to the cutoff (0.05) are considered to be **marginal**.

### Types of Hypothesis Tests

Different tests are suited for different types of data and hypotheses. <br>
These tests are powerful tools for drawing statistical inferences based on different types of data and research questions.

1. **t-test:**
 - **Purpose:** The t-test is used to compare the means of two groups and determine if there is a significant difference between them.
- **Use Case:** For example, you might use a t-test to compare the average scores of students who attended extra classes with those who didn't.
- **Assumptions:** Assumes that the data is approximately normally distributed, and the variances of the two groups are roughly equal.
- **Interpretation:**
   - If the p-value is less than the chosen significance level (commonly 0.05), you would reject the null hypothesis.
   - In the example, if the p-value is less than 0.05, it would indicate a significant difference in the average scores between the two groups (students who attended extra classes vs. those who didn't).
   - A lower p-value suggests stronger evidence against the null hypothesis.

```python
from scipy import stats

# Example data
group1_scores = [78, 82, 88, 95, 79]
group2_scores = [65, 70, 75, 80, 72]

# Performing t-test
t_statistic, p_value = stats.ttest_ind(group1_scores, group2_scores)

# Interpretation
if p_value < 0.05:
    interpretation = "The t-test result is statistically significant. There is a significant difference in average scores between the two groups."
else:
    interpretation = "The t-test result is not statistically significant. There is not enough evidence to conclude a significant difference."

print(interpretation)
```

2. Chi-square test:
- **Purpose:** The chi-square test is used for categorical data to determine if there is an association between two categorical variables.
- **Use Case:** It's commonly used to assess if there is a significant relationship between gender (male/female) and voting preference (yes/no).
- **Assumptions:** Assumes that the data is categorical and the observations are independent.
- **Interpretation:**
    - If the p-value is less than the chosen significance level, you would reject the null hypothesis.
    - In the example, a small p-value would indicate a significant association between gender and voting preference.
    - A larger p-value suggests weaker evidence against the null hypothesis.

```python
from scipy.stats import chi2_contingency

# Example data
observed_values = [[30, 20], [15, 25]]

# Performing chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(observed_values)

# Interpretation
if p_value < 0.05:
    interpretation = "The chi-square test result is statistically significant. There is a significant association between gender and voting preference."
else:
    interpretation = "The chi-square test result is not statistically significant. There is not enough evidence to conclude a significant association."

print(interpretation)
```

3. ANOVA (Analysis of Variance):
   - **Purpose:** ANOVA is used to compare means of more than two groups to assess if there are any statistically significant differences among them.
   - **Use Case:** For example, you might use ANOVA to compare the average scores of students who attended extra classes, regular classes, and private tutoring.
   - **Assumptions:** Assumes that the data is normally distributed, and the variances of the groups are roughly equal.
   - **Interpretation:**<br>
       - If the p-value is less than the chosen significance level, you would reject the null hypothesis.
       - In the example, a small p-value would suggest that there is at least one group with a different average score.
       - Post-hoc tests may be conducted to identify which specific groups differ if the overall ANOVA is significant.
       -  A larger p-value suggests that there is no significant difference in the means of the groups.

```python
from scipy.stats import f_oneway

# Example data
group1_scores = [75, 80, 85, 90, 78]
group2_scores = [70, 75, 80, 85, 72]
group3_scores = [85, 90, 95, 98, 88]

# Performing ANOVA
f_statistic, p_value = f_oneway(group1_scores, group2_scores, group3_scores)

# Interpretation
if p_value < 0.05:
    interpretation = "The ANOVA result is statistically significant. There is a significant difference in average scores among the groups."
else:
    interpretation = "The ANOVA result is not statistically significant. There is not enough evidence to conclude a significant difference in means."

print(interpretation)
```

### Worked Example 

**Research Question:<br>**
"Do students who attend extra classes perform significantly better than those who don't?"

**Why is it Interesting? <br>**
Understanding whether attending extra classes has a significant impact on students' exam scores is crucial for educational institutions and policymakers. It can inform decisions related to resource allocation, educational interventions, and strategies to enhance student performance.

**Theory Behind it <br>**
The hypothesis being tested is based on the assumption that extra classes provide additional learning opportunities, leading to improved academic performance. The theory is that there is a positive relationship between attending extra classes and higher exam scores.

**Hypotheses:<br>**
Null Hypothesis (H0): There is no significant difference in exam scores between students who attend extra classes and those who don't.<br>
Alternative Hypothesis (H1): Students who attend extra classes have significantly higher exam scores.

**Hypothesis Test:<br>**
Perform a t-test to compare the mean exam scores of the two groups.

```python
# Importing necessary libraries
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Creating Dataset
np.random.seed(42)  # For reproducibility
extra_classes_scores = np.random.normal(75, 10, 100)  # Scores of students who attend extra classes
regular_classes_scores = np.random.normal(70, 12, 100)  # Scores of students who don't attend extra classes
```

```python
# Research Question
# "Do students who attend extra classes perform significantly better than those who don't?"

# Hypotheses
# Null Hypothesis (H0): There is no significant difference in exam scores between the two groups.
# Alternative Hypothesis (H1): Students who attend extra classes have significantly higher exam scores.


# Hypothesis Test (t-test)
t_statistic, p_value = stats.ttest_ind(extra_classes_scores, regular_classes_scores)
```

```python
# Data Analysis - Visualizations
plt.hist(extra_classes_scores, alpha=0.5, label='Extra Classes')
plt.hist(regular_classes_scores, alpha=0.5, label='Regular Classes')
plt.legend(loc='upper right')
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.show()
```

```python
# Statistical Analysis
mean_extra_classes = np.mean(extra_classes_scores)
std_extra_classes = np.std(extra_classes_scores)
mean_regular_classes = np.mean(regular_classes_scores)
std_regular_classes = np.std(regular_classes_scores)
print(mean_extra_classes)
print(std_extra_classes)
print(mean_regular_classes)
print(std_regular_classes)
```

```python
# Conclusion - Summarize results
if p_value < 0.05:
    conclusion = "Reject the null hypothesis. There is significant evidence that students who attend extra classes perform better."
else:
    conclusion = "Fail to reject the null hypothesis. There is not enough evidence to conclude a significant difference."

print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")
print(conclusion)
```
