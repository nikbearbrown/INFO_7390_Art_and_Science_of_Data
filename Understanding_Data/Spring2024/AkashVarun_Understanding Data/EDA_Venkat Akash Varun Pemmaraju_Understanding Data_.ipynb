#### What is Exploratory Data Analysis (EDA) ? 

Exploratory Data Analysis, commonly known as EDA, is an approach in data analysis that employs a variety of techniques to understand and summarize the main characteristics of a dataset, often visually. It's a critical step before formal modeling or hypothesis testing.

Basically , The goal of EDA is to leverage visualization tools, summary tables, and hypothesis testing to:

1. **Provide summary level insight into a dataset.**

2. **Uncover underlying patterns and structures in you data.**

3. **Identify outliers, missing data, class balance, and other data-related issues.**

4. **Relate the available data to the business opportunity.**



#### Key Techniques in EDA

EDA combines both graphical and quantitative techniques to delve into data:

1. **Descriptive Statistics**: Involves summarizing main features of data, typically through measures like mean, median, mode, range, and standard deviation. For instance, analyzing average sales per month.
   
2. **Data Visualization**: Uses graphical representations like histograms, box plots, and scatter plots to understand distributions and relationships within the data.
   
3. **Identifying Outliers**: Detecting unusual data points that may indicate variability in measurements, experimental errors, or a novelty.

4. **Examining Relationships**: Exploring correlations or associations between variables using methods like correlation coefficients and cross-tabulation.

5. **Group Comparisons**: Comparing metrics across different groups. For example, comparing average exam scores across different schools.



#### What is the theory behind it?

The theory behind Exploratory Data Analysis (EDA) is rooted in the work of John W. Tukey, advocating a comprehensive and open-ended approach to data exploration. EDA emphasizes summarizing key dataset characteristics, identifying patterns, detecting outliers, and checking assumptions. 

By employing **descriptive statistics, visualization techniques, and correlation analysis**, EDA allows for a nuanced understanding of data relationships, aiding in variable selection, data cleaning, and pattern recognition. 

This informal yet essential stage in data analysis helps analysts glean insights and make informed decisions before applying more formal statistical methods or modeling approaches.

#### Basics of Algorithms and Problem Design in EDA:

#### 1. Algorithm Design: 

Unlike machine learning or optimization challenges, EDA is usually not linked to a particular algorithm. Rather, EDA is a set of approaches and procedures meant to find links, patterns, and insights inside data. These techniques include data transformation tools, data visualization, and statistical summaries. In EDA, a "algorithm" is more accurately described as a workflow or procedure that directs data analysis and exploration.

#### 2. Problem Formulation:

The first step in EDA is to formulate the problem of comprehending a dataset. To do this, the goals, research questions, and hypotheses that direct the analysis must be defined. Defining a mathematical or computational problem is not as important as providing the context for exploration during the problem formulation phase of EDA.

#### Descriptive Statistics:
#### Mean, Median, Mode:

**Definition**: Measures of central tendency.

**Useful for**: Providing a measure of central tendency.

**Insights**: Mean represents the average, median is the middle value, and mode is the most frequently occurring value. They give an idea of the dataset's central tendency.

### Statistical Measures:
When describing statistical measures such as **mean, median, and mode**, it's helpful to provide mathematical formulas to explain how they are calculated. Here are the formulas:

**Mean (μ)**: The mean is calculated by summing up all the values in the dataset and dividing by the total number of values. The formula is:
$$
μ = (Σ xi) / n
$$
Where:
**μ** is the **mean**,

**Σxi** represents the **sum of all values** in the dataset, and

**n** is the **total number of values**.

**Median**: The median is the middle value of a dataset when it is ordered from least to greatest. If the dataset has an odd number of values, the median is the middle number. If the dataset has an even number of values, the median is the average of the two middle numbers. There isn't a specific formula for the median, but it can be described as the value that divides the dataset into two equal halves.

**Median (Odd Number of Terms)**: If n (number of terms) is odd:
$$
{Median} = \left[ \frac{{(n + 1)}}{2} \right] \text{th observation}
$$
**Median (Even Number of Terms)**: If n (number of terms) is even:
$$
{Median} = \frac{{\left( \frac{n}{2} \right)\text{th term} + \left( \frac{n}{2} + 1 \right)\text{th term}}}{2}
$$
**Mode**: The mode is the value that appears most frequently in the dataset. If all values appear with the same frequency, the dataset is said to be multimodal. There isn't a specific formula for the mode, but it can be determined by counting the frequency of each value and identifying the one with the highest frequency.

#### Example:
Calculating the mean, median, and mode of house prices.

```python
# Code for Mean House Price
mean_price = boston_df['PRICE'].mean()

# Code for Median House Price
median_price = boston_df['PRICE'].median()

# Code for Mode of House Price
mode_price = boston_df['PRICE'].mode()[0]

# mean_price, median_price, mode_price

print("Mean House Price: ", mean_price)
print("Median House Price: ", median_price)
print("Mode of House Price: ", mode_price)
```

#### Standard Deviation and Variance

**Definition:** Standard deviation and variance are measures of dispersion, indicating how spread out the data points are from the mean.

**Useful for:** These measures are valuable for quantifying the spread or dispersion of data.

**Insights:** A high standard deviation or variance suggests greater variability in the data. They complement the mean by providing information on how individual data points deviate from the mean value.


**Standard Deviation (σ):**
  The standard deviation is the square root of the variance.
  $$
  \sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (x_i - \bar{x})^2}
  $$
  Where:
  - \( N \) is the total number of data points.
  - \( x_i \) represents each individual data point.
  - \( bar{x} \) is the mean of the data points.

- **Variance (σ^2):**
  The variance is the average of the squared differences from the Mean.
  $$
  \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \bar{x})^2
  $$
  Where:
  - \( N \) is the total number of data points.
  - \( x_i \) represents each individual data point.
  - \( bar{x} \) is the mean of the data points.


#### Example:
Calculating the standard deviation and variance of house prices.

```python
# Code for Standard Deviation of House Prices
std_price = boston_df['PRICE'].std()

# Code for Variance of House Prices
var_price = boston_df['PRICE'].var()

# std_price, var_price


print("Standard Deviation of House Prices: ", std_price)
print("Variance of House Prices: ", var_price)
```

#### Correlation Analysis:

**Correlation Coefficients (e.g., Pearson, Spearman):**
*Definition:* Measure the strength and direction of linear relationships between two variables.

**Formulae:**

1.**Pearson Correlation Coefficient**:
$$
r_{xy} = \frac{{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}}{{\sqrt{\sum{(X_i - \bar{X})^2}\sum{(Y_i - \bar{Y})^2}}}}
$$
**Explanation**: Pearson correlation measures the linear relationship between two variables. It is calculated by dividing the covariance of the variables by the product of their standard deviations.

This coefficient measures the strength and direction of the linear relationship between two variables. It indicates how much the variables change together. If r_xy is close to 1, it suggests a strong positive linear relationship, while if it's close to -1, it suggests a strong negative linear relationship. A value around 0 indicates no linear relationship between the variables.

2. **Spearman Rank Correlation Coefficient**:

$$ \rho = 1 - \frac{{6\sum{d^2}}}{{n(n^2 - 1)}} $$

**Explanation**: Spearman correlation assesses the strength and direction of association between two ranked variables. It is based on the ranks of the data rather than the raw data itself, making it suitable for ordinal data or when the relationship between variables is not linear.

Unlike Pearson correlation, which is based on the actual values of the variables, Spearman correlation is based on the ranks of the data. It assesses the strength and direction of association between two ranked variables. It is particularly useful when the relationship between variables is not linear or when dealing with ordinal data. If ρ is close to 1, it indicates a strong positive association between ranks, while if it's close to -1, it suggests a strong negative association. A value around 0 indicates no association between the ranks of the variables.

#### Outlier Detection:


#### Z-score approach 

The **Z-Score** measures how many standard deviations a data point is from the mean of the dataset. It helps identify outliers by quantifying the deviation of each data point from the mean in terms of standard deviations.
Mathematical Explanation:
The Z-Score of a data point x in a dataset with mean μ and standard deviation σ is calculated using the following formula:
$$
Z = \frac{x - \mu}{\sigma}
$$

**Step-by-Step Explanation**:

**Calculate the Mean (μ) and Standard Deviation (σ)**: Compute the mean and standard deviation of the dataset.

**Compute the Z-Score for Each Data Point**: For each data point x, calculate its Z-Score using the formula above.

**Identify Outliers**: Data points with Z-Scores beyond a certain threshold (commonly 2 or 3 standard deviations) are considered outliers.

**Handle Outliers**: Depending on the analysis, outliers can be removed, transformed, or treated separately.


#### Example:
Detecting outliers in house prices using Z-scores.

```python
# Code for Z-Score Analysis
z_scores = (boston_df['PRICE'] - boston_df['PRICE'].mean()) / boston_df['PRICE'].std()
outliers_zscore = boston_df[abs(z_scores) > 3]
print("Number of outliers detected using Z-Score method: ", len(outliers_zscore))
```

#### IQR (Interquartile Range) Approach for Handling Outliers

The **IQR** is a measure of **statistical dispersion, representing the range between the first quartile (25th percentile) and the third quartile (75th percentile)** of the dataset. Outliers are identified based on this range.

**Mathematical Explanation**:

The Interquartile Range (IQR) is calculated as follows:

**IQR = Q3 - Q1**

Where:

**Q1 is the first quartile (25th percentile)**.

**Q3 is the third quartile (75th percentile)**.

**Step-by-Step Explanation**:

**Calculate Quartiles**: Compute the **first quartile (Q1) and third quartile (Q3)** of the dataset.

**Compute IQR**: Calculate the Interquartile Range (IQR) using the formula above.

**Define Outlier Bounds**: Define the **lower bound (Q1−1.5×IQR)** and **upper bound (Q3+1.5×IQR**).

**Identify Outliers**: Data points outside the defined bounds are considered outliers.

**Handle Outliers**: Similar to Z-Score Analysis, outliers can be managed based on the specific analysis requirements.


#### Example:
Detecting outliers in house prices using the IQR method.

```python
# Code for IQR (Interquartile Range) Method
Q1 = boston_df['PRICE'].quantile(0.25)
Q3 = boston_df['PRICE'].quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = boston_df[(boston_df['PRICE'] < Q1 - 1.5 * IQR) | (boston_df['PRICE'] > Q3 + 1.5 * IQR)]
print("Number of outliers detected using IQR method: ", len(outliers_iqr))
```

To address this issue, we can apply a normalization technique to these variables. **Normalization** is a process that changes the range of data to make it more suitable for the algorithms we want to apply. It can help to reduce the impact of outliers and bring the data into a range that is more manageable for our algorithms.

One common method of normalization is log transformation. **Log transformation** is a powerful tool for handling skewed data and after its application, the distribution becomes more approximate to normal. In addition to reducing the impact of outliers, log transformation also compresses the scale on which the data is presented, while maintaining the order of values. This can make patterns in the data more visible and easier to interpret.

In this case, applying a log transformation to the 'price' and 'kilometers driven' variables can help to normalize the data and make it more suitable for further analysis and modeling.

```python
# Function for log transformation of the column
def log_transform(df,col):
    for colname in col:
        if (df[colname] == 1.0).all():
            df[colname + '_log'] = np.log(df[colname]+1)
        else:
            df[colname + '_log'] = np.log(df[colname])
    df.info()
```

#### Handling Outliers

##### Let's look at 2 commonly used approaches for handling outliers :
1. Z-score Approach
2. IQR Approach


#### Z-score approach 

The **Z-Score** measures how many standard deviations a data point is from the mean of the dataset. It helps identify outliers by quantifying the deviation of each data point from the mean in terms of standard deviations.
Mathematical Explanation:
The Z-Score of a data point x in a dataset with mean μ and standard deviation σ is calculated using the following formula:
$$
Z = \frac{x - \mu}{\sigma}
$$

**Step-by-Step Explanation**:

**Calculate the Mean (μ) and Standard Deviation (σ)**: Compute the mean and standard deviation of the dataset.

**Compute the Z-Score for Each Data Point**: For each data point x, calculate its Z-Score using the formula above.

**Identify Outliers**: Data points with Z-Scores beyond a certain threshold (commonly 2 or 3 standard deviations) are considered outliers.

**Handle Outliers**: Depending on the analysis, outliers can be removed, transformed, or treated separately.


```python
# This the Function to detect and handle outliers using z-scores
def handle_outliers_zscore(df, column):
    z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
    df = df[(z_scores < 3)]  # Keep only data within 3 standard deviations
    return df

# Handling outliers for 'Kilometers_Driven' using z-score
df = handle_outliers_zscore(df, 'Kilometers_Driven')

# Handling outliers for 'Price' using z-score
df = handle_outliers_zscore(df, 'Price')

# Visualizing the distribution after handling outliers
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Kilometers_Driven'])
plt.title('Boxplot - Kilometers Driven (Outliers Removed)')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Price'])
plt.title('Boxplot - Price (Outliers Removed)')

plt.show()
```

#### IQR (Interquartile Range) Approach for Handling Outliers

The **IQR** is a measure of **statistical dispersion, representing the range between the first quartile (25th percentile) and the third quartile (75th percentile)** of the dataset. Outliers are identified based on this range.

**Mathematical Explanation**:

The Interquartile Range (IQR) is calculated as follows:

**IQR = Q3 - Q1**

Where:

**Q1 is the first quartile (25th percentile)**.

**Q3 is the third quartile (75th percentile)**.

**Step-by-Step Explanation**:

**Calculate Quartiles**: Compute the **first quartile (Q1) and third quartile (Q3)** of the dataset.

**Compute IQR**: Calculate the Interquartile Range (IQR) using the formula above.

**Define Outlier Bounds**: Define the **lower bound (Q1−1.5×IQR)** and **upper bound (Q3+1.5×IQR**).

**Identify Outliers**: Data points outside the defined bounds are considered outliers.

**Handle Outliers**: Similar to Z-Score Analysis, outliers can be managed based on the specific analysis requirements.




```python
# This is the function to detect and handle outliers using IQR (percentile-based)
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Handling outliers for 'Kilometers_Driven' using IQR
df = handle_outliers_iqr(df, 'Kilometers_Driven')

# Handling outliers for 'Price' using IQR
df = handle_outliers_iqr(df, 'Price')

# Visualizing the distribution after handling outliers
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Kilometers_Driven'])
plt.title('Boxplot - Kilometers Driven (Outliers Removed)')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Price'])
plt.title('Boxplot - Price (Outliers Removed)')

plt.show()
```
